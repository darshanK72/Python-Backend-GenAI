# Introduction to Fine-Tuning and Alignment

Modern LLMs are not trained once and deployed. They pass through **multiple stages** — broad pre-training, instruction tuning, and alignment — each shaping a different aspect of behavior. Understanding this pipeline helps you decide when prompting or RAG is enough, and when adapting model weights is justified.

## 1. Why Training Is Multi-Stage

A single training objective cannot capture every requirement we care about in deployed systems. During **pre-training**, the model learns broad language structure and world patterns from massive text corpora, but it is not yet explicitly optimized for instruction-following, helpfulness, or policy constraints.

Later stages progressively shape behavior: supervised instruction data teaches response format, preference data teaches style and quality trade-offs, and alignment methods constrain unsafe or unhelpful behavior. Thinking in stages helps separate **capability building** from **behavior steering**.

---

## 2. Pre-Training: Base Capability

Pre-training uses **next-token prediction** over diverse corpora. The model learns grammar, idioms, reasoning patterns, and broad factual priors. This stage dominates compute cost and parameter learning.

Even strong base models can produce brittle instruction behavior because they optimize likelihood, not task success against human intent. Pre-training is **necessary but not sufficient** for reliable assistant behavior.

---

## 3. Supervised Fine-Tuning (SFT)

**Supervised fine-tuning** teaches the model to map prompts to desirable responses using curated demonstration pairs. It improves instruction-following, tone consistency, and response structure.

SFT data **quality** usually matters more than raw quantity — the model quickly absorbs annotation style. Poor or narrow SFT data can overfit formatting habits or inject dataset biases.

---

## 4. Alignment: RLHF and DPO (Overview)

After SFT, teams often apply **preference optimization** so outputs better match human expectations.

| Method | Core idea | Trade-off |
|--------|-----------|----------|
| **RLHF** | Train a reward model from human rankings, then optimize the policy against it | Can improve helpfulness/refusal behavior; depends on reward-model quality |
| **DPO** | Learn directly from preferred vs rejected response pairs | Often simpler to operate; still depends on preference data coverage |

Both methods shape **how** the model responds — tone, safety, helpfulness — rather than injecting new factual knowledge. That distinction matters when choosing between alignment and RAG.

---

## 5. Parameter-Efficient Fine-Tuning (Concept)

Full fine-tuning updates every model weight — expensive in GPU memory and storage. **Parameter-efficient fine-tuning (PEFT)** methods such as **LoRA** train only small adapter layers while keeping the base model frozen.

This makes it practical to maintain multiple domain variants of one base model by swapping lightweight adapters instead of storing full model copies. Hands-on PEFT/LoRA/QLoRA implementation is covered in **09. Open Source Models → 07. Fine-Tuning Open Models**.

---

## 6. Training Stages Pipeline

```text
Raw Text + Code Corpora
          |
          v
   [Pre-Training]
  Learn broad language priors
          |
          v
 [Supervised Fine-Tuning]
 Learn instruction formatting and task behavior
          |
          +-----------------------------+
          |                             |
          v                             v
      [RLHF]                         [DPO]
 Preference-driven shaping        Pairwise preference shaping
          \                             /
           \                           /
            v                         v
         [Aligned Assistant for Deployment]
```

In practice this pipeline is **iterative**: deployment feedback often leads to additional SFT or preference data and periodic refreshes.

---

## 7. Prompt vs RAG vs Fine-Tuning

Before reaching for fine-tuning, most teams work through a simpler ladder:

1. **Prompting** — when behavior can be specified with instructions and examples
2. **RAG** — when failures are mainly about missing or stale **knowledge**
3. **Fine-tuning** — when failures are persistent **behavior**, tone, or output-format gaps that prompting cannot fix

| Dominant failure mode | Start here |
|-----------------------|------------|
| Unclear instructions | Improve prompts and examples |
| Missing or outdated facts | RAG with a maintained knowledge base |
| Consistent style, schema, or domain language mismatch | Fine-tuning (often PEFT first) |

Hybrid strategies are common: light fine-tuning for format and style, plus RAG for fresh factual grounding.

````python
import numpy as np

# Toy scoring rubric: which lever fits which scenario?
scenarios = {
    "FAQ assistant with evolving policies": {"knowledge_gap": 0.9, "behavior_gap": 0.3},
    "Strict JSON extraction at scale": {"knowledge_gap": 0.2, "behavior_gap": 0.9},
    "General brainstorming helper": {"knowledge_gap": 0.3, "behavior_gap": 0.2},
}

for name, gaps in scenarios.items():
    kg, bg = gaps["knowledge_gap"], gaps["behavior_gap"]
    scores = {
        "prompt": (1 - bg) * 0.8 + (1 - kg) * 0.4,
        "rag": kg * 1.0 + (1 - bg) * 0.2,
        "fine_tune": bg * 1.0 + 0.2 * (1 - kg),
    }
    best = max(scores, key=scores.get)
    print(f"{name}: -> {best}")
````

---

## 8. Risks at a Glance

- **Fine-tuning** can improve consistency but may reduce generality if domain data is narrow
- **Alignment** can improve UX while masking factual errors behind confident style
- **PEFT** reduces cost but does not fix base-model knowledge gaps — pair with RAG when facts matter

Treat alignment as an **ongoing process**, not a one-time step before launch.

---

## 9. Summary and Where to Go Next

**Pre-training** builds broad capability. **SFT** teaches instruction behavior. **RLHF/DPO** shape preferences and safety. **PEFT** offers a cost-efficient path to domain adaptation.

For most applications, start with prompting, add RAG for knowledge gaps, and fine-tune only when measured failures justify the operational cost.
