# Google Gemini

**Google Gemini** is Google's family of multimodal large language models, available to developers through **Google AI Studio** (API keys) and **Vertex AI** (enterprise on Google Cloud).

Gemini excels at **native multimodal input** (text, images, audio, video) and **long context**. This notebook covers model tiers, API design, and differences from OpenAI's chat completions pattern.


---

## 1. Two Ways to Access Gemini

| Path | Audience | Authentication | Best for |
|------|----------|----------------|----------|
| **Google AI Studio** | Developers, prototypes | API key (`GOOGLE_API_KEY`) | Learning, side projects |
| **Vertex AI** | Enterprises on GCP | IAM, service accounts | Production, VPC, SLAs |

This course uses **Google AI Studio** via the `google-genai` Python SDK.


---

## 2. Model Families

| Model | Typical use | Notes |
|-------|-------------|-------|
| **gemini-2.0-flash** | Fast chat, multimodal | Low latency, cost-effective |
| **gemini-2.5-pro** | Complex tasks, long docs | Higher capability |
| **Embedding models** | Semantic search, RAG | Separate from chat |

### Strengths

- **Multimodal by design** — text, images, audio, video in one API
- **Long context** on Pro-tier models
- **Google ecosystem** — GCP, Workspace, Android integration
- **Competitive Flash-tier pricing** for high volume


---

## 3. API Design

| Item | Value |
|------|-------|
| **Package** | `google-genai` |
| **Client** | `genai.Client(api_key=...)` |
| **Main call** | `client.models.generate_content()` |
| **Streaming** | `generate_content_stream()` |

```python
from google import genai

client = genai.Client()  # GOOGLE_API_KEY from env
response = client.models.generate_content(
    model="gemini-2.0-flash",
    contents="Explain APIs in one sentence.",
    config={"temperature": 0.4, "max_output_tokens": 200},
)
print(response.text)
```

Multi-turn uses `contents` with `user` and `model` roles and `parts` arrays instead of OpenAI's `messages`.


---

## 4. OpenAI vs Gemini

| Aspect | OpenAI | Gemini |
|--------|--------|--------|
| Format | `messages: [{role, content}]` | `contents: [{role, parts}]` |
| Assistant role | `assistant` | `model` |
| System prompt | `role: system` | `system_instruction` in config |
| Max output | `max_tokens` | `max_output_tokens` |
| Streaming | `stream=True` | `generate_content_stream()` |


---

## 5. Multimodal Input

Gemini `parts` can include text, inline image bytes, audio, and video depending on model. This enables visual Q&A, document OCR, and video summarization without separate vision pipelines.


---

## 6. Rate Limits and Safety

Google applies quotas and **safety filters**. Blocked prompts may return empty or filtered output. Verify `GOOGLE_API_KEY` and monitor quotas in AI Studio.
