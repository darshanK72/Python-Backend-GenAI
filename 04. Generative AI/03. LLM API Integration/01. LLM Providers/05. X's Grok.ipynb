# X's Grok (xAI)

**Grok** is the large language model family from **xAI**, the company founded by Elon Musk. Grok is integrated with the **X** (formerly Twitter) platform and positioned as a model with access to real-time information and a distinct conversational style.

This notebook covers what makes Grok different, how developers access it via the **xAI API**, and when it may fit your use case. Grok is not integrated in course scripts yet, but the API follows patterns familiar from OpenAI.

## 1. What Is Grok?

Grok models are general-purpose LLMs trained by xAI. Public positioning emphasizes:

- **Real-time awareness** — connection to X platform data for current events (capabilities vary by product tier)
- **Direct, conversational tone** — less filtered style in some configurations
- **Competition with frontier models** — Grok-2, Grok-3 class models on benchmarks

Grok is available to consumers through X subscriptions and to developers through the **xAI API** at [console.x.ai](https://console.x.ai).

## 3. xAI API Design

The xAI API is largely **OpenAI-compatible**, which lowers integration cost if you already use the `openai` Python SDK:

| Item | Value |
|------|-------|
| **Base URL** | `https://api.x.ai/v1` |
| **Auth** | `Authorization: Bearer <XAI_API_KEY>` |
| **Endpoint** | `POST /v1/chat/completions` |
| **SDK** | `openai` package with custom `base_url` |

```python
from openai import OpenAI

client = OpenAI(
    api_key="xai-...",  # or XAI_API_KEY env
    base_url="https://api.x.ai/v1",
)

response = client.chat.completions.create(
    model="grok-2-latest",  # verify current model id
    messages=[
        {"role": "user", "content": "Summarize today's top AI news."},
    ],
)
print(response.choices[0].message.content)
```

Because the shape matches OpenAI, concepts from `02. OpenAI.ipynb` — messages, streaming, temperature — apply directly.

## 4. Strengths and Limitations

### Strengths

| Area | Notes |

### Limitations

| Area | Notes |
|------|-------|
| **Ecosystem size** | Fewer tutorials and third-party tools than OpenAI |
| **Enterprise features** | Less mature than Azure/Bedrock/Vertex paths |
| **Policy and availability** | Regional and product restrictions may apply |
