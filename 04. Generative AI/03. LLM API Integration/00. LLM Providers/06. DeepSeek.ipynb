# DeepSeek

**DeepSeek** is a Chinese AI research lab that gained global attention for **high-quality open-weight models** and **cost-efficient API access**. Models like **DeepSeek-V3** and the reasoning-focused **DeepSeek-R1** compete with frontier closed models on many benchmarks at lower cost.

This notebook explains DeepSeek's model lineup, API access, open-weight vs hosted tradeoffs, and when DeepSeek fits your stack.


---

## 1. DeepSeek in the Market

DeepSeek disrupted pricing expectations in the LLM market by demonstrating that strong models could be trained and served at **fractional cost** of some Western frontier APIs. For developers this means:

- Inexpensive **API access** for high-volume workloads
- **Open-weight releases** you can self-host (Ollama, vLLM)
- Strong **coding and reasoning** performance (especially R1 family)

DeepSeek is relevant both as an **API provider** and as an **open-weight model source**.


---

## 2. Model Families

| Model | Type | Notes |
|-------|------|-------|
| **DeepSeek-V3** | General chat / code | Strong MoE architecture; cost-efficient API |
| **DeepSeek-R1** | Reasoning | Chain-of-thought style reasoning; higher latency |
| **DeepSeek-Coder** | Code-focused | Specialized for programming tasks |
| **Distilled R1 variants** | Smaller reasoning | Run locally on consumer hardware |

Open-weight checkpoints appear on Hugging Face; API model names appear in [DeepSeek API docs](https://api-docs.deepseek.com).


---

## 3. DeepSeek API

DeepSeek's hosted API is **OpenAI-compatible**:

| Item | Value |
|------|-------|
| **Base URL** | `https://api.deepseek.com` |
| **Auth** | `Authorization: Bearer <DEEPSEEK_API_KEY>` |
| **Endpoint** | `POST /v1/chat/completions` |
| **SDK** | `openai` Python package with custom `base_url` |

```python
from openai import OpenAI

client = OpenAI(
    api_key="sk-...",  # DEEPSEEK_API_KEY
    base_url="https://api.deepseek.com",
)

response = client.chat.completions.create(
    model="deepseek-chat",
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "Explain mixture-of-experts in simple terms."},
    ],
)
print(response.choices[0].message.content)
```

For reasoning models, use the model ID specified in DeepSeek docs (e.g. `deepseek-reasoner`) and expect longer responses with visible reasoning traces in some configurations.

## 4. Open-Weight vs API

| Approach | Pros | Cons |
|----------|------|------|
| **DeepSeek API** | No GPU ops; pay per token; fast to start | Data leaves your network; subject to provider terms |
| **Self-host open weights** | Privacy, fixed infra cost at scale | Need GPUs; model ops and updates are your job |
| **Ollama locally** | Free dev/testing on laptop | Limited model size vs cloud |

Many teams prototype on the **API** and move high-volume stable workloads to **self-hosted** open weights.


---

## 5. Strengths and Considerations

### Strengths

- **Cost efficiency** — attractive for high token volume
- **Coding performance** — competitive on programming benchmarks
- **R1 reasoning** — strong on math and logic tasks
- **Open weights** — auditability and self-hosting option

### Considerations

- **Data residency and compliance** — evaluate policies for regulated industries
- **Latency** — reasoning models (R1) can be slower than flat chat models
- **Regional access** — verify availability in your jurisdiction
- **Not in course scripts yet** — add `DEEPSEEK_API_KEY` to `.env` to experiment


---

## 6. When to Choose DeepSeek

| Priority | DeepSeek fit |
|----------|--------------|
| Minimize API cost at scale | Strong |
| Hard reasoning / math | Consider R1 |
| Air-gapped / on-prem | Use open-weight self-host |
| Largest third-party ecosystem | OpenAI may still win |
| Enterprise Azure/Bedrock IAM | Use marketplace paths instead |
