# Tracing with Langfuse

**Langfuse** is an open-source LLM observability platform for **traces**, **spans**, **generations**, **scores**, and **evaluations**. For multi-step agents, it answers:

- What path did the agent take?
- Which tool or retrieval step failed?
- How many tokens did each LLM call cost?
- Can we replay a bad trace for debugging?

```
  Trace: shopco-support-session-42
  ├── Span: route_intent
  ├── Span: retrieve_policy
  │     └── Generation: embed_query (offline mock)
  ├── Span: lookup_order
  └── Generation: compose_answer
```

Langfuse Python SDK v3 is **OpenTelemetry-based** — spans nest automatically via context propagation. The `@observe` decorator and `start_as_current_observation()` are the primary instrumentation APIs.

This notebook builds **ShopCo Support** with:

1. Offline `ShopCoLangfuse` tracer (mirrors Langfuse semantics — no server required)
2. `@observe` decorator and nested spans
3. Generations with token/cost metadata
4. `propagate_attributes` for user_id / session_id
5. Scores, error capture, and LangGraph callback pattern

Offline by default; optional live Langfuse cloud at the end.

## 1. Langfuse Data Model

| Object | Purpose | Agent example |
|--------|---------|---------------|
| **Trace** | One end-user request | Full support ticket handling |
| **Span** | A step in the workflow | `route_intent`, `retrieve_policy` |
| **Generation** | LLM call | `compose_answer` with model + tokens |
| **Event** | Point-in-time marker | `human_escalation_triggered` |
| **Score** | Quality / feedback | `task_complete=1`, user thumbs-up |

Traces group spans in a tree. Generations are specialized spans with model metadata.

## 2. Offline ShopCoLangfuse — Mirrors SDK Semantics

````python
ObservationType = Literal["span", "generation", "event"]

_current_observation: ContextVar[str | None] = ContextVar("current_observation", default=None)
_trace_attrs: ContextVar[dict[str, Any]] = ContextVar("trace_attrs", default={})


@dataclass
class Observation:
    id: str
    trace_id: str
    parent_id: str | None
    name: str
    obs_type: ObservationType
    input: Any = None
    output: Any = None
    metadata: dict[str, Any] = field(default_factory=dict)
    start_time: str = ""
    end_time: str = ""
    latency_ms: float = 0.0
    level: str = "DEFAULT"
    status_message: str = ""
    model: str = ""
    usage: dict[str, int] = field(default_factory=dict)


@dataclass
class TraceRecord:
    id: str
    name: str
    user_id: str = ""
    session_id: str = ""
    tags: list[str] = field(default_factory=list)
    metadata: dict[str, Any] = field(default_factory=dict)
    scores: list[dict[str, Any]] = field(default_factory=list)


class ShopCoLangfuse:
    """In-memory Langfuse-compatible tracer for offline demos."""

    def __init__(self) -> None:
        self.traces: dict[str, TraceRecord] = {}
        self.observations: dict[str, Observation] = {}
        self._root_trace_id: str | None = None

    def _ensure_trace(self, name: str) -> str:
        if self._root_trace_id and self._root_trace_id in self.traces:
            return self._root_trace_id
        tid = f"trace-{uuid.uuid4().hex[:12]}"
        attrs = _trace_attrs.get()
        self.traces[tid] = TraceRecord(
            id=tid, name=attrs.get("trace_name", name),
            user_id=attrs.get("user_id", ""), session_id=attrs.get("session_id", ""),
            tags=list(attrs.get("tags", [])),
        )
        self._root_trace_id = tid
        return tid

    @contextmanager
    def start_as_current_observation(
        self,
        *,
        as_type: ObservationType = "span",
        name: str,
        model: str = "",
        input: Any = None,
        capture_input: bool = True,
    ) -> Generator[Observation, None, None]:
        tid = self._ensure_trace(name)
        parent = _current_observation.get()
        oid = f"obs-{uuid.uuid4().hex[:10]}"
        t0 = time.perf_counter()
        obs = Observation(
            id=oid, trace_id=tid, parent_id=parent, name=name, obs_type=as_type,
            input=input if capture_input else "[redacted]",
            start_time=utc_now(), model=model,
        )
        self.observations[oid] = obs
        token = _current_observation.set(oid)
        try:
            yield obs
        except Exception as exc:
            obs.level = "ERROR"
            obs.status_message = str(exc)
            raise
        finally:
            obs.end_time = utc_now()
            obs.latency_ms = (time.perf_counter() - t0) * 1000
            _current_observation.reset(token)

    def update_current_observation(self, **kwargs: Any) -> None:
        oid = _current_observation.get()
        if oid and oid in self.observations:
            obs = self.observations[oid]
            for k, v in kwargs.items():
                if hasattr(obs, k):
                    setattr(obs, k, v)

    def score(self, *, name: str, value: float, trace_id: str | None = None, comment: str = "") -> None:
        tid = trace_id or self._root_trace_id
        if tid and tid in self.traces:
            self.traces[tid].scores.append({"name": name, "value": value, "comment": comment})

    def flush(self) -> int:
        """Simulate export — returns observation count."""
        count = len(self.observations)
        return count

    def reset(self) -> None:
        self.traces.clear()
        self.observations.clear()
        self._root_trace_id = None

    def print_trace_tree(self, trace_id: str | None = None) -> None:
        tid = trace_id or self._root_trace_id
        if not tid:
            print("(no trace)")
            return
        trace = self.traces[tid]
        print(f"Trace {tid} name={trace.name} user={trace.user_id} session={trace.session_id}")
        obs_list = [o for o in self.observations.values() if o.trace_id == tid]

        def render(parent_id: str | None, indent: int = 0) -> None:
            children = sorted([o for o in obs_list if o.parent_id == parent_id], key=lambda x: x.start_time)
            for o in children:
                mark = "⚠" if o.level == "ERROR" else " "
                extra = f" model={o.model}" if o.model else ""
                print(f"{'  '*indent}{mark} [{o.obs_type}] {o.name} ({o.latency_ms:.1f}ms){extra}")
                render(o.id, indent + 1)

        render(None)
        if trace.scores:
            print("Scores:", trace.scores)


langfuse = ShopCoLangfuse()
print("ShopCoLangfuse tracer ready.")
````

## 3. `@observe` Decorator

````python
def observe(
    func: Callable | None = None,
    *,
    name: str | None = None,
    as_type: ObservationType = "span",
    capture_input: bool = True,
    capture_output: bool = True,
) -> Callable:
    """Offline mirror of langfuse.observe decorator."""

    def decorator(fn: Callable) -> Callable:
        obs_name = name or fn.__name__

        @wraps(fn)
        def wrapper(*args: Any, **kwargs: Any) -> Any:
            inp = {"args": args, "kwargs": kwargs} if capture_input else "[redacted]"
            with langfuse.start_as_current_observation(as_type=as_type, name=obs_name, input=inp) as obs:
                result = fn(*args, **kwargs)
                obs.output = result if capture_output else "[redacted]"
                return result

        return wrapper

    if func is not None:
        return decorator(func)
    return decorator


@contextmanager
def propagate_attributes(
    *,
    user_id: str = "",
    session_id: str = "",
    trace_name: str = "",
    tags: list[str] | None = None,
) -> Generator[None, None, None]:
    """Mirror of langfuse.propagate_attributes — sets trace-level context."""
    token = _trace_attrs.set({
        "user_id": user_id, "session_id": session_id,
        "trace_name": trace_name, "tags": tags or [],
    })
    try:
        yield
    finally:
        _trace_attrs.reset(token)


@observe(name="add_numbers")
def add_numbers(a: int, b: int) -> int:
    return a + b


with propagate_attributes(user_id="alice", session_id="sess-1", trace_name="demo"):
    add_numbers(2, 3)
langfuse.print_trace_tree()
````

## 4. ShopCo Support — Traced Workflow

````python
langfuse.reset()


@observe(name="route_intent")
def route_intent(question: str) -> str:
    q = question.lower()
    if "return" in q or "refund" in q:
        return "policy"
    if "order" in q or re.search(r"ORD-[0-9]{4}", question.upper()):
        return "order"
    return "general"


@observe(name="retrieve_policy")
def retrieve_policy(question: str) -> dict[str, str]:
    q = question.lower()
    if "return" in q:
        return {"id": "pol-returns-01", "text": POLICY_DOCS["returns"]}
    if "shipping" in q:
        return {"id": "pol-shipping-02", "text": POLICY_DOCS["shipping"]}
    return {"id": "none", "text": "No policy found."}


@observe(name="lookup_order")
def lookup_order(question: str) -> dict[str, str]:
    m = re.search(r"ORD-[0-9]{4}", question.upper())
    if not m:
        return {"order_id": "", "status": "unknown"}
    oid = m.group(0)
    return ORDERS_DB.get(oid, {"order_id": oid, "status": "not_found"})


@observe(name="compose_answer", as_type="generation")
def compose_answer(question: str, intent: str, context: dict[str, Any]) -> str:
    langfuse.update_current_observation(
        model="gpt-4o-mini-offline",
        usage={"input_tokens": 120, "output_tokens": 45, "total_tokens": 165},
    )
    if intent == "policy":
        doc = context.get("policy", {})
        return f"{doc.get('text', 'No policy')}"
    if intent == "order":
        row = context.get("order", {})
        return f"Order {row.get('order_id', '?')} status: {row.get('status', 'unknown')}."
    return "How can I help with your ShopCo order or policy?"


@observe(name="handle_support_request")
def handle_support_request(user_id: str, session_id: str, question: str) -> str:
    with propagate_attributes(user_id=user_id, session_id=session_id, trace_name="shopco-support", tags=["support"]):
        intent = route_intent(question)
        ctx: dict[str, Any] = {}
        if intent == "policy":
            ctx["policy"] = retrieve_policy(question)
        elif intent == "order":
            ctx["order"] = lookup_order(question)
        answer = compose_answer(question, intent, ctx)
        langfuse.score(name="task_complete", value=1.0 if "unknown" not in answer.lower() else 0.5)
        return answer


ans = handle_support_request("alice", "chat-42", "What is the return policy for ORD-1001?")
print(ans)
print()
langfuse.print_trace_tree()
````

## 5. Nested Spans — Automatic Parent/Child

````python
def count_children(trace_id: str, parent_name: str) -> int:
    parent = next((o for o in langfuse.observations.values() if o.trace_id == trace_id and o.name == parent_name), None)
    if not parent:
        return 0
    return sum(1 for o in langfuse.observations.values() if o.parent_id == parent.id)


tid = langfuse._root_trace_id
root_children = count_children(tid, "handle_support_request") if tid else 0
print(f"Root span 'handle_support_request' has {root_children} direct child spans")
print("Nested tree printed above — route → retrieve/lookup → compose")
````

## 6. Error Capture in Spans

````python
langfuse.reset()


@observe(name="failing_tool")
def failing_tool(order_id: str) -> str:
    if order_id == "ORD-9999":
        raise ValueError("Order not found in warehouse API")
    return "ok"


@observe(name="handle_with_error")
def handle_with_error(question: str) -> str:
    with propagate_attributes(user_id="bob", session_id="err-1", trace_name="error-demo"):
        try:
            failing_tool("ORD-9999")
        except ValueError:
            return "Sorry, I could not look up that order."
    return "ok"


print(handle_with_error("Where is ORD-9999?"))
langfuse.print_trace_tree()

error_obs = [o for o in langfuse.observations.values() if o.level == "ERROR"]
print(f"\nError observations: {[(o.name, o.status_message) for o in error_obs]}")
````

## 8. ReleaseFlow — Traced Deploy Pipeline

````python
langfuse.reset()


@observe(name="run_tests")
def run_tests(service: str) -> dict[str, Any]:
    return {"service": service, "passed": 42, "failed": 0}


@observe(name="deploy_canary")
def deploy_canary(service: str, version: str) -> str:
    return f"{service}@{version} at 10% traffic"


@observe(name="check_slo", as_type="generation")
def check_slo(service: str) -> dict[str, float]:
    langfuse.update_current_observation(model="metrics-api", usage={"input_tokens": 0, "output_tokens": 0})
    return {"error_rate": 0.001, "p99_ms": 120.0}


@observe(name="releaseflow_deploy")
def releaseflow_deploy(service: str, version: str) -> str:
    with propagate_attributes(user_id="ci-bot", session_id="rel-2026-07-10", trace_name="releaseflow", tags=["deploy"]):
        tests = run_tests(service)
        if tests["failed"] > 0:
            langfuse.score(name="deploy_ok", value=0.0, comment="tests failed")
            return "BLOCKED: tests failed"
        deploy_canary(service, version)
        slo = check_slo(service)
        ok = slo["error_rate"] < 0.01
        langfuse.score(name="deploy_ok", value=1.0 if ok else 0.0)
        return f"Canary live. error_rate={slo['error_rate']}"


print(releaseflow_deploy("shopco-api", "v2.4.1"))
langfuse.print_trace_tree()
````

## 9. LangGraph Callback Pattern (Conceptual)

````python
class OfflineLangfuseCallback:
    """Mirrors langfuse.langchain.CallbackHandler — logs LangGraph node runs."""

    def __init__(self, tracer: ShopCoLangfuse) -> None:
        self.tracer = tracer

    def on_chain_start(self, name: str, inputs: dict[str, Any]) -> str:
        cm = self.tracer.start_as_current_observation(name=f"lg:{name}", input=inputs)
        cm.__enter__()
        return name

    def on_chain_end(self, name: str, output: Any) -> None:
        self.tracer.update_current_observation(output=output)
        oid = _current_observation.get()
        if oid:
            _current_observation.set(None)


def simulate_langgraph_run(question: str) -> str:
    langfuse.reset()
    cb = OfflineLangfuseCallback(langfuse)
    with propagate_attributes(user_id="alice", session_id="lg-1", trace_name="langgraph-support"):
        for node_name, fn in [("load_memory", lambda: {}), ("respond", lambda: handle_support_request("alice", "lg-1", question))]:
            cb.on_chain_start(node_name, {"question": question})
            out = fn()
            cb.on_chain_end(node_name, out)
    return out if isinstance(out, str) else str(out)


print(simulate_langgraph_run("Return policy?")[:80])
langfuse.print_trace_tree()
````

## 10. Trace Analytics — Latency and Cost

````python
def trace_analytics(tracer: ShopCoLangfuse, trace_id: str | None = None) -> dict[str, Any]:
    tid = trace_id or tracer._root_trace_id
    obs = [o for o in tracer.observations.values() if o.trace_id == tid]
    total_ms = sum(o.latency_ms for o in obs)
    gens = [o for o in obs if o.obs_type == "generation"]
    tokens = sum(g.usage.get("total_tokens", 0) for g in gens)
    errors = [o.name for o in obs if o.level == "ERROR"]
    return {
        "trace_id": tid,
        "observation_count": len(obs),
        "total_latency_ms": round(total_ms, 1),
        "generation_count": len(gens),
        "total_tokens": tokens,
        "errors": errors,
        "scores": tracer.traces[tid].scores if tid in tracer.traces else [],
    }


handle_support_request("alice", "analytics-1", "ORD-1001 status")
print(pretty(trace_analytics(langfuse)))
````

## 11. Langfuse vs Raw OpenTelemetry

| Feature | OpenTelemetry | Langfuse |
|---------|---------------|----------|
| Standard spans | Yes | Yes (OTEL-based v3) |
| LLM generations | Manual attributes | First-class `generation` type |
| Prompt/completion UI | Generic backends | Langfuse dashboard |
| Scores / evals | Custom | Built-in on traces |
| LangChain/LangGraph | Via OTEL bridges | Native `CallbackHandler` |

Use **OTEL** for infra-wide telemetry; **Langfuse** when LLM-specific trace UX and eval loops matter.

## 13. Common Mistakes

| Mistake | Symptom | Fix |
|---------|---------|-----|
| No `flush()` before exit | Missing traces in cloud | Call `langfuse.flush()` |
| Flat logs instead of spans | Can't see hierarchy | Use `@observe` nesting |
| Logging PII in span input | Compliance violation | `capture_input=False` |
| One trace per LLM call | Can't debug full agent | Root span per user request |
| No scores on traces | Can't filter bad runs | `score()` task_complete, quality |
| Ignoring error levels | Silent tool failures | Let exceptions propagate into spans |

## 15. Optional Live Langfuse

````python
def demo_live_langfuse() -> None:
    if not USE_LIVE_LANGFUSE:
        print("Set USE_LIVE_LANGFUSE = True and configure LANGFUSE_* env vars.")
        print(f"Host: {LANGFUSE_HOST}")
        return

    from langfuse import Langfuse, observe as lf_observe, propagate_attributes as lf_propagate

    client = Langfuse(
        public_key=LANGFUSE_PUBLIC_KEY,
        secret_key=LANGFUSE_SECRET_KEY,
        host=LANGFUSE_HOST,
    )

    @lf_observe(name="live_support")
    def live_support(question: str) -> str:
        with lf_propagate(user_id="alice", session_id="live-1", trace_name="shopco-live"):
            return handle_support_request("alice", "live-1", question)

    result = live_support("What is the return policy?")
    client.flush()
    print("Live trace sent:", result[:80])


demo_live_langfuse()
````

## 17. Summary

**Langfuse tracing** gives agent workflows production-grade observability:

1. **Traces** group a full user request; **spans** nest workflow steps
2. **`@observe`** auto-captures input, output, latency, errors
3. **Generations** track LLM calls with model and token usage
4. **`propagate_attributes`** tags user_id, session_id, trace name
5. **Scores** attach quality signals for eval loops

ShopCo Support and ReleaseFlow deploy pipelines demonstrate nested spans, error capture, and analytics — offline first, Langfuse cloud when `USE_LIVE_LANGFUSE=True`.
