# Introduction to Agent Observability and Evaluation

A **single LLM call** fails loudly — wrong JSON, timeout, 500 error. An **agent** can fail quietly: wrong tool chosen, infinite loop, plausible hallucination, or runaway cost across ten steps. You need **two complementary disciplines**:

| Discipline | Question it answers | When |
|------------|---------------------|------|
| **Observability** | What happened on this run? | Every request in prod |
| **Evaluation** | Is quality good enough to ship? | CI, pre-release, regression |

```
  Agent run
     │
     ├──► TRACES (spans: LLM, tools, router)
     ├──► METRICS (latency, tokens, $, success rate)
     └──► LOGS (errors, policy blocks)
              │
              ▼
     EVAL HARNESS (golden tasks, rubrics, pass/fail gates)
```

This notebook introduces observability primitives and a minimal **evaluation harness** for **ShopCo Support Hub** and **ReleaseFlow** — plain Python, no external tracing SaaS required.

````python
"""
Agent observability + evaluation — introduction lab.
"""

import json
import os
import re
import time
import uuid
from dataclasses import dataclass, field
from datetime import datetime, timezone
from enum import Enum
from typing import Any, Callable

OPENAI_API_KEY = os.environ.get("OPENAI_API_KEY", "sk-proj-placeholder-replace-with-your-real-key")
os.environ.setdefault("OPENAI_API_KEY", OPENAI_API_KEY)

USE_LIVE_LLM = False
TOKEN_COST_PER_1K = 0.002  # illustrative USD


def pretty(obj: Any) -> str:
    return json.dumps(obj, indent=2, default=str)


def utc_now() -> str:
    return datetime.now(timezone.utc).isoformat()


POLICY_SNIPPETS = {
    "returns": "Returns within 30 days with receipt. [pol-returns-01]",
    "shipping": "Free shipping on orders over $50. [pol-shipping-02]",
    "hours": "Support hours Mon–Fri 9am–6pm ET. [faq-hours-05]",
}

ORDERS_DB = {
    "ORD-1001": {"order_id": "ORD-1001", "status": "shipped"},
    "ORD-1002": {"order_id": "ORD-1002", "status": "processing"},
}

RELEASE_CHECKS = {
    "unit_tests": {"status": "pass"},
    "security_scan": {"status": "warn"},
}

print("Observability lab ready.")
````

## 1. Why Agents Need More Than API Monitoring

| Traditional API | Agent system |
|-----------------|--------------|
| One request → one response | Many internal steps |
| Deterministic status codes | Probabilistic outputs |
| Latency = one hop | Latency = sum of LLM + tools |
| Unit tests on functions | Eval on **task completion** |

Observability tells you **what path** the agent took. Evaluation tells you **whether that path was good enough**.

## 2. The Three Pillars — Logs, Traces, Metrics

````python
class SpanKind(str, Enum):
    AGENT = "agent"
    LLM = "llm"
    TOOL = "tool"
    ROUTER = "router"


@dataclass
class Span:
    span_id: str
    trace_id: str
    name: str
    kind: SpanKind
    start_ts: str
    end_ts: str | None = None
    attributes: dict[str, Any] = field(default_factory=dict)
    status: str = "ok"

    @property
    def duration_ms(self) -> int:
        if not self.end_ts:
            return 0
        t0 = datetime.fromisoformat(self.start_ts)
        t1 = datetime.fromisoformat(self.end_ts)
        return int((t1 - t0).total_seconds() * 1000)


@dataclass
class Trace:
    trace_id: str
    service: str
    spans: list[Span] = field(default_factory=list)

    def add_span(self, span: Span) -> None:
        self.spans.append(span)

    def total_duration_ms(self) -> int:
        return sum(s.duration_ms for s in self.spans)


trace = Trace(trace_id=f"tr-{uuid.uuid4().hex[:8]}", service="shopco-support")
print("Trace model ready.")
````

---

## 3. Tracer — Instrument Agent Steps

````python
class AgentTracer:
    def __init__(self, service: str) -> None:
        self.trace = Trace(trace_id=f"tr-{uuid.uuid4().hex[:8]}", service=service)
        self.logs: list[dict[str, Any]] = []
        self.metrics: dict[str, float] = {
            "llm_calls": 0,
            "tool_calls": 0,
            "tokens_in": 0,
            "tokens_out": 0,
        }

    def start_span(self, name: str, kind: SpanKind, **attrs: Any) -> Span:
        return Span(
            span_id=f"sp-{uuid.uuid4().hex[:8]}",
            trace_id=self.trace.trace_id,
            name=name,
            kind=kind,
            start_ts=utc_now(),
            attributes=dict(attrs),
        )

    def end_span(self, span: Span, status: str = "ok", **attrs: Any) -> None:
        span.end_ts = utc_now()
        span.status = status
        span.attributes.update(attrs)
        self.trace.add_span(span)

    def log(self, level: str, message: str, **extra: Any) -> None:
        self.logs.append({"ts": utc_now(), "level": level, "message": message, **extra})

    def record_llm(self, tokens_in: int, tokens_out: int) -> None:
        self.metrics["llm_calls"] += 1
        self.metrics["tokens_in"] += tokens_in
        self.metrics["tokens_out"] += tokens_out

    def record_tool(self) -> None:
        self.metrics["tool_calls"] += 1

    def estimated_cost_usd(self) -> float:
        total_tokens = self.metrics["tokens_in"] + self.metrics["tokens_out"]
        return round(total_tokens / 1000 * TOKEN_COST_PER_1K, 4)


tracer = AgentTracer("shopco-support")
print("Tracer ready.")
````

## 4. ShopCo Agent — Instrumented Run

````python
def get_order(order_id: str) -> dict[str, Any]:
    return ORDERS_DB.get(order_id.upper(), {"status": "not_found"})


def search_policy(query: str) -> str:
    return POLICY_SNIPPETS["returns"] if "return" in query.lower() else POLICY_SNIPPETS["shipping"]


def shopco_agent(message: str, tracer: AgentTracer) -> str:
    root = tracer.start_span("shopco_agent", SpanKind.AGENT, message=message[:80])

    rspan = tracer.start_span("route_intent", SpanKind.ROUTER)
    intent = "return" if "return" in message.lower() else "order" if "order" in message.lower() else "faq"
    tracer.end_span(rspan, intent=intent)
    tracer.record_llm(120, 15)

    parts: list[str] = []
    m = re.search(r"ORD-[0-9]{4}", message.upper())
    if m:
        tspan = tracer.start_span("get_order", SpanKind.TOOL, order_id=m.group(0))
        rec = get_order(m.group(0))
        tracer.record_tool()
        tracer.end_span(tspan, status=rec.get("status", "unknown"))
        parts.append(f"Order status: {rec.get('status')}")

    if intent == "return":
        tspan = tracer.start_span("search_policy", SpanKind.TOOL)
        pol = search_policy(message)
        tracer.record_tool()
        tracer.end_span(tspan, policy_id="pol-returns-01")
        parts.append(pol)
        tracer.record_llm(80, 40)

    if intent == "faq" and "hour" in message.lower():
        parts.append(POLICY_SNIPPETS["hours"])

    answer = " ".join(parts) if parts else "How can ShopCo help?"
    tracer.end_span(root, answer_len=len(answer))
    return answer


shopco_tracer = AgentTracer("shopco-support")
answer = shopco_agent("Can I return ORD-1001?", shopco_tracer)
print("Answer:", answer)
print("Spans:", [s.name for s in shopco_tracer.trace.spans])
````

## 5. ReleaseFlow — Instrumented Pipeline

````python
def releaseflow_pipeline(version: str, tracer: AgentTracer) -> dict[str, Any]:
    root = tracer.start_span("release_pipeline", SpanKind.AGENT, version=version)
    check_results = []
    for check in RELEASE_CHECKS:
        cspan = tracer.start_span(f"check_{check}", SpanKind.TOOL)
        status = RELEASE_CHECKS[check]["status"]
        tracer.record_tool()
        tracer.end_span(cspan, status=status)
        check_results.append({"check": check, "status": status})
    gate = "proceed" if all(c["status"] != "fail" for c in check_results) else "block"
    tracer.log("info", f"release gate={gate}", version=version)
    tracer.end_span(root, gate=gate)
    return {"version": version, "gate": gate, "checks": check_results}


rf_tracer = AgentTracer("releaseflow")
rf_result = releaseflow_pipeline("2.4.0", rf_tracer)
print(pretty(rf_result))
````

## 6. Metrics Summary — Per Trace Dashboard

````python
@dataclass
class TraceSummary:
    trace_id: str
    service: str
    duration_ms: int
    span_count: int
    llm_calls: int
    tool_calls: int
    tokens_total: int
    cost_usd: float
    error_spans: int


def summarize_trace(tracer: AgentTracer) -> TraceSummary:
    errors = sum(1 for s in tracer.trace.spans if s.status not in ("ok", "pass", "warn"))
    return TraceSummary(
        trace_id=tracer.trace.trace_id,
        service=tracer.trace.service,
        duration_ms=tracer.trace.total_duration_ms(),
        span_count=len(tracer.trace.spans),
        llm_calls=int(tracer.metrics["llm_calls"]),
        tool_calls=int(tracer.metrics["tool_calls"]),
        tokens_total=int(tracer.metrics["tokens_in"] + tracer.metrics["tokens_out"]),
        cost_usd=tracer.estimated_cost_usd(),
        error_spans=errors,
    )


print(pretty(summarize_trace(shopco_tracer)))
````

## 7. Evaluation — Golden Tasks and Assertions

Observability is **per-run**. Evaluation is **across many runs** against expected behavior.

````python
@dataclass
class EvalCase:
    case_id: str
    input_message: str
    must_contain: list[str] = field(default_factory=list)
    must_cite_policy: bool = False
    max_tool_calls: int | None = None
    max_cost_usd: float | None = None


@dataclass
class EvalResult:
    case_id: str
    passed: bool
    failures: list[str]
    answer: str
    summary: TraceSummary | None = None


SHOPCO_EVAL_SET: list[EvalCase] = [
    EvalCase("ec-1", "Can I return ORD-1001?", must_contain=["shipped", "pol-returns-01"], must_cite_policy=True, max_tool_calls=3),
    EvalCase("ec-2", "What are support hours?", must_contain=["faq-hours-05"], max_tool_calls=1),
    EvalCase("ec-3", "ORD-1002 status", must_contain=["processing"]),
]

print(f"Eval set: {len(SHOPCO_EVAL_SET)} cases")
````

---

## 8. Eval Runner — Assert + Trace

````python
def run_eval_case(case: EvalCase, agent_fn: Callable[[str, AgentTracer], str]) -> EvalResult:
    tracer = AgentTracer("shopco-support")
    answer = agent_fn(case.input_message, tracer)
    summary = summarize_trace(tracer)
    failures: list[str] = []
    low = answer.lower()
    for token in case.must_contain:
        if token.lower() not in low:
            failures.append(f"missing '{token}'")
    if case.must_cite_policy and "[pol-" not in answer:
        failures.append("missing policy citation")
    if case.max_tool_calls is not None and summary.tool_calls > case.max_tool_calls:
        failures.append(f"tool_calls {summary.tool_calls} > {case.max_tool_calls}")
    if case.max_cost_usd is not None and summary.cost_usd > case.max_cost_usd:
        failures.append(f"cost ${summary.cost_usd} > ${case.max_cost_usd}")
    return EvalResult(case.case_id, len(failures) == 0, failures, answer, summary)


def run_eval_suite(cases: list[EvalCase]) -> dict[str, Any]:
    results = [run_eval_case(c, shopco_agent) for c in cases]
    passed = sum(1 for r in results if r.passed)
    return {
        "total": len(results),
        "passed": passed,
        "pass_rate": round(passed / len(results), 2) if results else 0,
        "results": [{"case_id": r.case_id, "passed": r.passed, "failures": r.failures} for r in results],
    }


print(pretty(run_eval_suite(SHOPCO_EVAL_SET)))
````

## 9. Quality Dimensions for Agents

| Dimension | ShopCo example | How to measure |
|-----------|----------------|----------------|
| **Task completion** | Answered the question | `must_contain` assertions |
| **Grounding** | Policy citation present | `must_cite_policy` |
| **Efficiency** | Not 10 tool calls for FAQ | `max_tool_calls`, cost cap |
| **Safety** | No refund without policy | Red-team cases (later notebooks) |
| **Reliability** | Same input → stable path | Trace span fingerprint |

## 10. Offline vs Online (Preview)

| Mode | Data | Purpose |
|------|------|---------|
| **Offline eval** | Golden dataset in CI | Regression before deploy |
| **Online monitoring** | Production traces | Drift, cost spikes, error rate |

Use the **same metrics** (`pass_rate`, `tool_calls`, `cost_usd`) in both — compare offline baseline to live dashboards.

````python
OFFLINE_BASELINE = run_eval_suite(SHOPCO_EVAL_SET)

ONLINE_WINDOW = {
    "pass_rate": 0.67,
    "avg_tool_calls": 2.1,
    "avg_cost_usd": 0.004,
    "alerts": ["pass_rate below offline baseline 1.0"],
}

print("Offline baseline pass_rate:", OFFLINE_BASELINE["pass_rate"])
print("Online window:", pretty(ONLINE_WINDOW))
````

## 11. Failure Modes Observability Catches

| Failure | Trace signal | Metric signal |
|---------|--------------|---------------|
| **Tool loop** | Same tool span repeated | `tool_calls` >> baseline |
| **Cost overrun** | Many LLM spans | `cost_usd` spike |
| **Wrong route** | Router span `intent=faq` on return Q | Eval `missing pol-*` |
| **Silent hallucination** | No tool spans but factual answer | Eval grounding fail |

````python
def detect_loop(tracer: AgentTracer, threshold: int = 3) -> bool:
    tool_names = [s.name for s in tracer.trace.spans if s.kind == SpanKind.TOOL]
    if not tool_names:
        return False
    from collections import Counter
    counts = Counter(tool_names)
    return any(c >= threshold for c in counts.values())


loop_tracer = AgentTracer("loop-demo")
for _ in range(4):
    s = loop_tracer.start_span("search_policy", SpanKind.TOOL)
    loop_tracer.record_tool()
    loop_tracer.end_span(s)
print("Loop detected:", detect_loop(loop_tracer))
````

## 13. Export Trace for Debugging

````python
def export_trace(tracer: AgentTracer) -> dict[str, Any]:
    return {
        "trace_id": tracer.trace.trace_id,
        "service": tracer.trace.service,
        "spans": [
            {"name": s.name, "kind": s.kind.value, "status": s.status, "ms": s.duration_ms, "attrs": s.attributes}
            for s in tracer.trace.spans
        ],
        "metrics": tracer.metrics,
        "logs": tracer.logs,
        "summary": summarize_trace(tracer).__dict__,
    }


print(pretty(export_trace(shopco_tracer)))
````

## 14. Optional Live LLM Judge (Single Case)

Set `USE_LIVE_LLM = True` to score one answer with `gpt-4o-mini`.

````python
def llm_judge(question: str, answer: str) -> dict[str, Any]:
    from langchain_core.messages import HumanMessage, SystemMessage
    from langchain_openai import ChatOpenAI

    llm = ChatOpenAI(model="gpt-4o-mini", api_key=OPENAI_API_KEY, temperature=0)
    resp = llm.invoke([
        SystemMessage(content="Score 1-5 for helpfulness and grounding. Reply JSON {score, reason}."),
        HumanMessage(content=f"Q: {question}\nA: {answer}"),
    ])
    try:
        return json.loads(str(resp.content))
    except json.JSONDecodeError:
        return {"score": 0, "reason": "parse error"}


q, a = "Can I return ORD-1001?", shopco_agent("Can I return ORD-1001?", AgentTracer("judge"))
if USE_LIVE_LLM:
    print(llm_judge(q, a))
else:
    print("Offline eval:", run_eval_case(SHOPCO_EVAL_SET[0], shopco_agent))
````

## 15. Common Mistakes

| Mistake | Symptom | Fix |
|---------|---------|-----|
| Logs only, no traces | Can't see tool path | Span per LLM/tool |
| Eval only in prod | Users find regressions | Offline golden set in CI |
| No cost metrics | Bill shock | Token + $ on every trace |
| Subjective "looks fine" | Drift undetected | Assertions + pass_rate |
| One metric only | Good citations, wrong answer | Multi-dimension rubric |
