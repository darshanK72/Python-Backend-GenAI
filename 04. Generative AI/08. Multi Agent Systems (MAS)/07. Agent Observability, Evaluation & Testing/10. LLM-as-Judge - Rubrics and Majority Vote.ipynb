# LLM-as-Judge — Rubrics and Majority Vote

When rule-based checks are not enough — tone, helpfulness, completeness — teams use an **LLM as judge** to score agent outputs against a **rubric**. A single judge can be noisy or biased, so production systems often run **multiple judge passes** and take a **majority vote** or median score.

```
  Agent output + context
           |
     +-----+-----+-----+
     v     v     v     v
  Judge1 Judge2 Judge3 ...   (same rubric, varied prompts/temps)
     |     |     |
     +-----+-----+
           v
    Majority vote / median score
           v
    Pass / fail + feedback
```

This notebook implements LLM-as-judge evaluation for **ShopCo Support Hub** and **ReleaseFlow** in plain Python — rubric definitions, structured judge responses, panel voting, disagreement analysis, and batch benchmarks.

````python
"""
LLM-as-Judge — rubrics and majority vote — ShopCo + ReleaseFlow.
"""

import json
import re
import statistics
import uuid
from dataclasses import dataclass, field
from datetime import datetime, timezone
from enum import Enum
from typing import Any, Callable

USE_LIVE_LLM = False
DEFAULT_PASS_SCORE = 0.75
JUDGE_PANEL_SIZE = 3


def pretty(obj: Any) -> str:
    return json.dumps(obj, indent=2, default=str)


def utcnow() -> str:
    return datetime.now(timezone.utc).isoformat()


ORDERS: dict[str, dict[str, Any]] = {
    "ORD-1001": {"order_id": "ORD-1001", "customer": "alice@shop.com", "status": "shipped", "total_usd": 1299.0},
    "ORD-1002": {"order_id": "ORD-1002", "customer": "bob@shop.com", "status": "processing", "total_usd": 449.0},
}

POLICIES: dict[str, str] = {
    "returns": "30-day returns on unused items. [pol-returns-01]",
    "refunds": "Cite policy before promising refunds. Escalate over $1000. [pol-refunds-03]",
}

RELEASE_CHECKS: dict[str, dict[str, str]] = {
    "unit_tests": {"status": "pass", "detail": "412 tests passed"},
    "security_scan": {"status": "warn", "detail": "1 medium CVE in dev dependency"},
    "lint": {"status": "pass", "detail": "no issues"},
}

JUDGE_TRACE: list[dict[str, Any]] = []

print("LLM-as-judge lab ready")
````

---

## 1. When Rule-Based Metrics Are Not Enough

| Approach | Strength | Weakness |
|----------|----------|----------|
| **Exact match** | Fast, deterministic | Misses paraphrases |
| **Regex / keyword** | Cheap at scale | Cannot judge tone |
| **LLM-as-judge** | Nuanced, rubric-driven | Cost, variance, bias |

Use judges for **subjective quality dimensions** — empathy, clarity, policy alignment — while keeping hard checks (order ID present, citation ID) as rule-based gates.

---

## 2. Rubric Model — Criteria and Weights

````python
@dataclass
class RubricCriterion:
    name: str
    description: str
    weight: float = 1.0
    pass_threshold: float = 0.7


@dataclass
class Rubric:
    rubric_id: str
    name: str
    criteria: list[RubricCriterion]
    overall_pass_threshold: float = DEFAULT_PASS_SCORE

    def criterion_names(self) -> list[str]:
        return [c.name for c in self.criteria]


SHOPCO_REPLY_RUBRIC = Rubric(
    "shopco-reply-v1",
    "ShopCo support reply quality",
    [
        RubricCriterion("accuracy", "Facts match order/policy context", 1.5),
        RubricCriterion("policy_citation", "Includes required policy reference for refunds", 1.2),
        RubricCriterion("empathy", "Professional and empathetic tone", 1.0),
        RubricCriterion("actionability", "Clear next steps for the customer", 0.8),
    ],
)

RELEASE_NOTES_RUBRIC = Rubric(
    "release-notes-v1",
    "ReleaseFlow ship-readiness summary",
    [
        RubricCriterion("completeness", "Covers tests, lint, and security status", 1.2),
        RubricCriterion("risk_disclosure", "Security warnings clearly stated", 1.5),
        RubricCriterion("decision_clarity", "Explicit GO or NO-GO recommendation", 1.0),
    ],
)

print(f"Rubrics: {SHOPCO_REPLY_RUBRIC.name}, {RELEASE_NOTES_RUBRIC.name}")
````

---

## 3. Judge Response Schema

Judges return **structured JSON** — one score per criterion plus rationale — so downstream code can aggregate reliably.

````python
@dataclass
class CriterionJudgment:
    name: str
    score: float
    passed: bool
    rationale: str


@dataclass
class JudgeVerdict:
    judge_id: str
    rubric_id: str
    criterion_scores: list[CriterionJudgment]
    overall_score: float
    passed: bool
    summary: str

    def to_dict(self) -> dict[str, Any]:
        return {
            "judge_id": self.judge_id,
            "overall_score": round(self.overall_score, 3),
            "passed": self.passed,
            "criteria": [{"name": c.name, "score": c.score, "passed": c.passed} for c in self.criterion_scores],
            "summary": self.summary,
        }


def weighted_overall(judgments: list[CriterionJudgment], rubric: Rubric) -> float:
    weights = {c.name: c.weight for c in rubric.criteria}
    total_w = sum(weights.get(j.name, 1.0) for j in judgments)
    if total_w == 0:
        return 0.0
    return sum(j.score * weights.get(j.name, 1.0) for j in judgments) / total_w


print("Judge response schema ready")
````

## 4. Offline ShopCo Judge — Simulating LLM Scoring

Production judges call an LLM with the rubric in the system prompt. Here a **deterministic scorer** mimics that behavior for offline demos.

````python
SAMPLE_REPLIES = {
    "good_refund": (
        "We understand your frustration. ORD-1001 is shipped ($1299). "
        "Per our refunds policy: Cite policy before promising refunds. Escalate over $1000. [pol-refunds-03]. "
        "A specialist will follow up within 24 hours."
    ),
    "bad_refund": "Sure, refund approved! Money back tomorrow.",
    "ok_status": "ORD-1002 is currently processing.",
}


class ShopCoJudge:
    """Offline judge — simulates LLM rubric scoring."""

    def __init__(self, judge_id: str, rubric: Rubric, *, strictness: float = 1.0):
        self.judge_id = judge_id
        self.rubric = rubric
        self.strictness = strictness  # 1.0 = normal; >1 stricter

    def judge(self, reply: str, context: dict[str, Any]) -> JudgeVerdict:
        text = reply.lower()
        oid = context.get("order_id", "")
        is_refund = context.get("intent") == "refund"
        judgments: list[CriterionJudgment] = []

        # accuracy
        acc = 1.0
        if oid and oid.lower() not in text:
            acc = 0.3
        if oid and oid in ORDERS and ORDERS[oid]["status"] not in reply.lower():
            acc = min(acc, 0.6)
        judgments.append(CriterionJudgment("accuracy", acc * self.strictness, acc >= 0.7, "Order facts present." if acc >= 0.7 else "Missing or wrong order facts."))

        # policy citation
        has_pol = "[pol-" in reply or "pol-refunds" in text
        pol = 1.0 if (not is_refund or has_pol) else 0.2
        judgments.append(CriterionJudgment("policy_citation", pol, pol >= 0.7, "Policy cited." if has_pol else "No policy citation for refund."))

        # empathy
        empathetic = any(w in text for w in ("understand", "sorry", "frustration", "apologize"))
        angry = context.get("customer_upset", False)
        emp = 1.0 if (empathetic or not angry) else 0.4
        judgments.append(CriterionJudgment("empathy", emp, emp >= 0.7, "Tone OK." if emp >= 0.7 else "Needs empathy."))

        # actionability
        actionable = any(w in text for w in ("follow up", "next step", "within", "contact", "specialist"))
        act = 1.0 if actionable or not is_refund else 0.5
        judgments.append(CriterionJudgment("actionability", act, act >= 0.7, "Clear next steps." if act >= 0.7 else "Vague resolution."))

        overall = weighted_overall(judgments, self.rubric)
        passed = overall >= self.rubric.overall_pass_threshold
        return JudgeVerdict(self.judge_id, self.rubric.rubric_id, judgments, overall, passed, f"Overall {overall:.2f}")


judge_a = ShopCoJudge("judge-a", SHOPCO_REPLY_RUBRIC)
v_good = judge_a.judge(SAMPLE_REPLIES["good_refund"], {"order_id": "ORD-1001", "intent": "refund", "customer_upset": True})
v_bad = judge_a.judge(SAMPLE_REPLIES["bad_refund"], {"order_id": "ORD-1001", "intent": "refund", "customer_upset": True})
print(f"Good reply: {v_good.overall_score:.2f} pass={v_good.passed}")
print(f"Bad reply:  {v_bad.overall_score:.2f} pass={v_bad.passed}")
````

## 5. ReleaseFlow Judge

````python
RELEASE_SAMPLES = {
    "good": "NO-GO: security scan warn — 1 medium CVE in dev dependency. Tests: 412 passed. Lint: pass.",
    "bad": "Ship it! Everything looks fine.",
}


class ReleaseFlowJudge:
    def __init__(self, judge_id: str, rubric: Rubric):
        self.judge_id = judge_id
        self.rubric = rubric

    def judge(self, summary: str, context: dict[str, Any]) -> JudgeVerdict:
        text = summary.lower()
        judgments: list[CriterionJudgment] = []

        has_tests = "test" in text or "412" in text
        has_lint = "lint" in text
        has_sec = "security" in text or "cve" in text
        comp = (int(has_tests) + int(has_lint) + int(has_sec)) / 3
        judgments.append(CriterionJudgment("completeness", comp, comp >= 0.7, f"Coverage {comp:.0%}."))

        sec_warn = RELEASE_CHECKS["security_scan"]["status"] == "warn"
        disclosed = "warn" in text or "cve" in text or "no-go" in text
        risk = 1.0 if (not sec_warn or disclosed) else 0.2
        judgments.append(CriterionJudgment("risk_disclosure", risk, risk >= 0.7, "Risk disclosed." if disclosed else "Missing security warning."))

        has_decision = "go" in text or "no-go" in text
        dec = 1.0 if has_decision else 0.3
        judgments.append(CriterionJudgment("decision_clarity", dec, dec >= 0.7, "Decision stated." if has_decision else "No GO/NO-GO."))

        overall = weighted_overall(judgments, self.rubric)
        passed = overall >= self.rubric.overall_pass_threshold
        return JudgeVerdict(self.judge_id, self.rubric.rubric_id, judgments, overall, passed, f"Release summary score {overall:.2f}")


rf_judge = ReleaseFlowJudge("rf-judge-1", RELEASE_NOTES_RUBRIC)
rf_v = rf_judge.judge(RELEASE_SAMPLES["good"], {})
print(f"Release good: {rf_v.overall_score:.2f} pass={rf_v.passed}")
````

## 6. Judge Panel — Majority Vote on Pass/Fail

Run multiple judges with slightly different **strictness** or prompts, then aggregate.

````python
@dataclass
class PanelResult:
    panel_id: str
    verdicts: list[JudgeVerdict]
    majority_pass: bool
    pass_votes: int
    fail_votes: int
    median_score: float
    mean_score: float
    disagreement: bool


def majority_vote(verdicts: list[JudgeVerdict]) -> PanelResult:
    passes = [v.passed for v in verdicts]
    pass_votes = sum(passes)
    fail_votes = len(passes) - pass_votes
    majority_pass = pass_votes > fail_votes
    scores = [v.overall_score for v in verdicts]
    disagreement = len(set(passes)) > 1
    return PanelResult(
        str(uuid.uuid4())[:8], verdicts, majority_pass,
        pass_votes, fail_votes,
        statistics.median(scores), statistics.mean(scores), disagreement,
    )


def build_shopco_panel() -> list[ShopCoJudge]:
    return [
        ShopCoJudge("strict", SHOPCO_REPLY_RUBRIC, strictness=0.95),
        ShopCoJudge("neutral", SHOPCO_REPLY_RUBRIC, strictness=1.0),
        ShopCoJudge("lenient", SHOPCO_REPLY_RUBRIC, strictness=1.05),
    ]


ctx = {"order_id": "ORD-1001", "intent": "refund", "customer_upset": True}
borderline = "ORD-1001 is shipped. Refund may be possible. [pol-refunds-03]"
panel_verdicts = [j.judge(borderline, ctx) for j in build_shopco_panel()]
panel = majority_vote(panel_verdicts)
print(f"Panel: {panel.pass_votes} pass / {panel.fail_votes} fail -> majority_pass={panel.majority_pass}")
print(f"Scores: {[round(v.overall_score, 2) for v in panel.verdicts]} disagreement={panel.disagreement}")
````

## 7. Per-Criterion Majority Vote

Beyond overall pass/fail, vote **per criterion** to see which dimension judges disagree on.

````python
def criterion_majority(verdicts: list[JudgeVerdict], rubric: Rubric) -> dict[str, dict[str, Any]]:
    result: dict[str, dict[str, Any]] = {}
    for name in rubric.criterion_names():
        scores = []
        passes = []
        for v in verdicts:
            cj = next((c for c in v.criterion_scores if c.name == name), None)
            if cj:
                scores.append(cj.score)
                passes.append(cj.passed)
        pass_votes = sum(passes)
        result[name] = {
            "median_score": statistics.median(scores) if scores else 0,
            "majority_pass": pass_votes > len(passes) / 2,
            "pass_votes": pass_votes,
            "total_votes": len(passes),
            "disagreement": len(set(passes)) > 1 if passes else False,
        }
    return result


crit_vote = criterion_majority(panel.verdicts, SHOPCO_REPLY_RUBRIC)
for name, info in crit_vote.items():
    print(f"{name}: median={info['median_score']:.2f} majority_pass={info['majority_pass']} disagree={info['disagreement']}")
````

## 8. Anchor Examples — Calibrating Judges

**Anchor examples** are pre-scored samples included in the judge prompt so all judges calibrate to the same bar.

````python
@dataclass
class AnchorExample:
    example_id: str
    reply: str
    context: dict[str, Any]
    expected_pass: bool
    expected_min_score: float


SHOPCO_ANCHORS: list[AnchorExample] = [
    AnchorExample("anchor-pass", SAMPLE_REPLIES["good_refund"], {"order_id": "ORD-1001", "intent": "refund", "customer_upset": True}, True, 0.85),
    AnchorExample("anchor-fail", SAMPLE_REPLIES["bad_refund"], {"order_id": "ORD-1001", "intent": "refund", "customer_upset": True}, False, 0.50),
]


def calibrate_judge(judge: ShopCoJudge, anchors: list[AnchorExample]) -> dict[str, Any]:
    results = []
    for a in anchors:
        v = judge.judge(a.reply, a.context)
        aligned = (v.passed == a.expected_pass) and (v.overall_score >= a.expected_min_score if a.expected_pass else v.overall_score < a.expected_min_score)
        results.append({"id": a.example_id, "score": v.overall_score, "passed": v.passed, "aligned": aligned})
    alignment_rate = sum(1 for r in results if r["aligned"]) / len(results)
    return {"judge_id": judge.judge_id, "alignment_rate": alignment_rate, "details": results}


for j in build_shopco_panel():
    cal = calibrate_judge(j, SHOPCO_ANCHORS)
    print(f"{cal['judge_id']}: alignment={cal['alignment_rate']:.0%}")
````

## 9. Batch Evaluation Suite

````python
@dataclass
class EvalCase:
    case_id: str
    reply: str
    context: dict[str, Any]
    expected_majority_pass: bool


SHOPCO_EVAL_CASES: list[EvalCase] = [
    EvalCase("case-1", SAMPLE_REPLIES["good_refund"], {"order_id": "ORD-1001", "intent": "refund", "customer_upset": True}, True),
    EvalCase("case-2", SAMPLE_REPLIES["bad_refund"], {"order_id": "ORD-1001", "intent": "refund", "customer_upset": True}, False),
    EvalCase("case-3", SAMPLE_REPLIES["ok_status"], {"order_id": "ORD-1002", "intent": "status"}, True),
    EvalCase("case-4", borderline, {"order_id": "ORD-1001", "intent": "refund", "customer_upset": True}, False),
]


def run_batch(cases: list[EvalCase], panel: list[ShopCoJudge]) -> list[dict[str, Any]]:
    rows = []
    for case in cases:
        verdicts = [j.judge(case.reply, case.context) for j in panel]
        pr = majority_vote(verdicts)
        aligned = pr.majority_pass == case.expected_majority_pass
        rows.append({
            "case_id": case.case_id,
            "majority_pass": pr.majority_pass,
            "expected": case.expected_majority_pass,
            "aligned": aligned,
            "median_score": round(pr.median_score, 3),
            "disagreement": pr.disagreement,
        })
    return rows


batch = run_batch(SHOPCO_EVAL_CASES, build_shopco_panel())
for row in batch:
    print(f"{row['case_id']}: pass={row['majority_pass']} expected={row['expected']} aligned={row['aligned']}")
alignment = sum(1 for r in batch if r["aligned"]) / len(batch)
print(f"Batch alignment rate: {alignment:.0%}")
````

## 10. Handling Judge Disagreement

When judges split, escalate to human review or require **unanimous pass** for high-risk categories.

````python
class AggregationPolicy(str, Enum):
    MAJORITY = "majority"
    UNANIMOUS = "unanimous"
    MEDIAN_SCORE = "median_score"


def aggregate_panel(verdicts: list[JudgeVerdict], policy: AggregationPolicy, *, score_threshold: float = DEFAULT_PASS_SCORE) -> dict[str, Any]:
    pr = majority_vote(verdicts)
    if policy == AggregationPolicy.MAJORITY:
        final_pass = pr.majority_pass
    elif policy == AggregationPolicy.UNANIMOUS:
        final_pass = all(v.passed for v in verdicts)
    else:
        final_pass = pr.median_score >= score_threshold
    return {
        "policy": policy.value,
        "final_pass": final_pass,
        "majority_pass": pr.majority_pass,
        "unanimous_pass": all(v.passed for v in verdicts),
        "median_score": pr.median_score,
        "needs_human_review": pr.disagreement and policy == AggregationPolicy.MAJORITY,
    }


for pol in AggregationPolicy:
    agg = aggregate_panel(panel.verdicts, pol)
    print(f"{pol.value}: final_pass={agg['final_pass']} human_review={agg.get('needs_human_review')}")
````

## 11. Judge Prompt Template (Structured Output)

````python
def build_judge_prompt(rubric: Rubric, reply: str, context: dict[str, Any], anchors: list[AnchorExample] | None = None) -> str:
    criteria_block = "\n".join(f"- {c.name} (weight {c.weight}): {c.description}" for c in rubric.criteria)
    anchor_block = ""
    if anchors:
        anchor_block = "\n\nCalibration examples:\n" + "\n".join(
            f"[{a.example_id}] pass={a.expected_pass}: {a.reply[:80]}..." for a in anchors
        )
    return (
        f"You are an evaluator for {rubric.name}.\n"
        f"Score each criterion 0.0-1.0. Return JSON with criteria scores and rationale.\n\n"
        f"Rubric:\n{criteria_block}\n"
        f"{anchor_block}\n\n"
        f"Context: {json.dumps(context)}\n"
        f"Reply to evaluate:\n{reply}"
    )


prompt = build_judge_prompt(SHOPCO_REPLY_RUBRIC, SAMPLE_REPLIES["good_refund"], ctx, SHOPCO_ANCHORS)
print(prompt[:400] + "...")
````

## 12. Trace and Audit Log

````python
def log_panel_evaluation(case_id: str, panel: PanelResult) -> None:
    JUDGE_TRACE.append({
        "case_id": case_id,
        "panel_id": panel.panel_id,
        "majority_pass": panel.majority_pass,
        "median_score": round(panel.median_score, 3),
        "disagreement": panel.disagreement,
        "verdicts": [v.to_dict() for v in panel.verdicts],
        "ts": utcnow(),
    })


for case in SHOPCO_EVAL_CASES:
    verdicts = [j.judge(case.reply, case.context) for j in build_shopco_panel()]
    log_panel_evaluation(case.case_id, majority_vote(verdicts))

print(f"Logged {len(JUDGE_TRACE)} panel evaluations")
print(f"Disagreements: {sum(1 for t in JUDGE_TRACE if t['disagreement'])}")
````

## 13. When to Use LLM-as-Judge

| Good fit | Poor fit |
|----------|----------|
| Tone, empathy, clarity | Exact numeric answers |
| Rubric with 3-6 criteria | Binary API success/fail |
| Offline eval suites | Real-time per-request (cost) |
| Borderline quality bands | Safety-critical without human review |

---

## 14. Anti-Patterns

| Anti-pattern | Problem | Fix |
|--------------|---------|-----|
| Single judge | High variance | Panel + majority vote |
| Vague rubric ("be helpful") | Inconsistent scores | Named criteria with examples |
| Judge sees ground truth leak | Inflated scores | Blind evaluation — context only |
| Same model gen + judge | Self-preference bias | Separate models or rule gates |
| Ignoring disagreement | Wrong borderline calls | Flag for human review |
| No anchor calibration | Score drift over time | Anchor set in every prompt |

## 16. Optional — Live LLM Judge Panel

Set `USE_LIVE_LLM = True` to run three judge calls and aggregate with majority vote.

````python
if USE_LIVE_LLM:
    import os
    from openai import OpenAI

    client = OpenAI(api_key=os.environ.get("OPENAI_API_KEY"))
    prompt = build_judge_prompt(SHOPCO_REPLY_RUBRIC, SAMPLE_REPLIES["good_refund"], ctx, SHOPCO_ANCHORS)
    live_verdicts = []
    for jid in ("j1", "j2", "j3"):
        resp = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=[{"role": "user", "content": prompt}],
            response_format={"type": "json_object"},
            temperature=0.3 if jid != "j2" else 0.7,
        )
        data = json.loads(resp.choices[0].message.content or "{}")
        score = float(data.get("overall_score", data.get("score", 0)))
        live_verdicts.append(JudgeVerdict(jid, SHOPCO_REPLY_RUBRIC.rubric_id, [], score, score >= DEFAULT_PASS_SCORE, str(data)))
    live_panel = majority_vote(live_verdicts)
    print(f"Live panel: majority_pass={live_panel.majority_pass} median={live_panel.median_score:.2f}")
else:
    print("Offline mode — rubric judges with majority vote for ShopCo and ReleaseFlow.")
````

## 17. Summary

LLM-as-judge adds **nuanced quality scoring** when rules fall short:

- **Rubrics** — weighted criteria with pass thresholds
- **Structured verdicts** — per-criterion scores and rationale
- **Majority vote** — panel of judges reduces variance
- **Anchors** — calibrate judges to consistent standards
- **Disagreement handling** — human review or unanimous policy for high-risk cases

Combine judges with rule-based gates; log every verdict for audit and regression analysis.
