# Reasoning, Tool Selection, and Task Completion Metrics

Traditional API monitoring tracks latency and error rates. **Agent evaluation** needs metrics on **how** the agent thinks and acts:

| Metric family | What it measures | Example question |
|---------------|------------------|------------------|
| **Reasoning** | Quality and efficiency of intermediate steps | Did thoughts reference real observations? |
| **Tool selection** | Correct tools, order, and redundancy | Was `lookup_order` called before `check_billing`? |
| **Task completion** | End-to-end success and cost | Did the ticket resolve with policy citation? |

```
  Agent trace
       |
  +----+----+----+
  |    |    |    |
  v    v    v    v
Reason Tool Task Cost
 score select done  $
```

This notebook builds an **offline evaluation harness** for **ShopCo Support Hub** and **ReleaseFlow** — golden tasks, trace capture, metric calculators, and a benchmark report.

````python
"""
Reasoning, tool selection, and task completion metrics — ShopCo + ReleaseFlow.
"""

import json
import re
import uuid
from dataclasses import dataclass, field
from datetime import datetime, timezone
from enum import Enum
from typing import Any, Callable

USE_LIVE_LLM = False


def pretty(obj: Any) -> str:
    return json.dumps(obj, indent=2, default=str)


def utcnow() -> str:
    return datetime.now(timezone.utc).isoformat()


ORDERS: dict[str, dict[str, Any]] = {
    "ORD-1001": {"order_id": "ORD-1001", "customer": "alice@shop.com", "status": "shipped", "total_usd": 1299.0},
    "ORD-1002": {"order_id": "ORD-1002", "customer": "bob@shop.com", "status": "processing", "total_usd": 449.0},
}

POLICIES: dict[str, str] = {
    "returns": "30-day returns. [pol-returns-01]",
    "refunds": "Cite policy before promising refunds. Escalate over $1000. [pol-refunds-03]",
}

BILLING: dict[str, dict[str, Any]] = {
    "ORD-1001": {"refund_eligible": True},
    "ORD-1002": {"refund_eligible": True},
}

RELEASE_CHECKS: dict[str, dict[str, str]] = {
    "unit_tests": {"status": "pass", "detail": "412 tests passed"},
    "security_scan": {"status": "warn", "detail": "1 medium CVE"},
    "lint": {"status": "pass", "detail": "no issues"},
}

METRIC_TRACE: list[dict[str, Any]] = []

print("Agent metrics lab ready")
````

---

## 1. Why Three Metric Families?

A correct final answer can hide bad process: wrong tools retried, hallucinated reasoning, or 12 steps for a 2-step task. Measuring **process** catches regressions before users notice.

- **Reasoning metrics** — grounding, coherence, step efficiency
- **Tool selection metrics** — precision, recall, order correctness
- **Task completion metrics** — success rate, latency, token cost, partial credit

---

## 2. Trace Model — Capturing Agent Runs

````python
class StepKind(str, Enum):
    THOUGHT = "thought"
    ACTION = "action"
    OBSERVATION = "observation"
    FINAL = "final"


@dataclass
class TraceStep:
    kind: StepKind
    content: str
    tool_name: str | None = None
    tool_args: dict[str, Any] = field(default_factory=dict)
    tokens: int = 0


@dataclass
class AgentTrace:
    trace_id: str
    task_id: str
    question: str
    steps: list[TraceStep] = field(default_factory=list)
    final_answer: str = ""
    latency_ms: float = 0.0
    total_tokens: int = 0

    def tools_used(self) -> list[str]:
        return [s.tool_name for s in self.steps if s.kind == StepKind.ACTION and s.tool_name]

    def thoughts(self) -> list[str]:
        return [s.content for s in self.steps if s.kind == StepKind.THOUGHT]

    def observations(self) -> list[str]:
        return [s.content for s in self.steps if s.kind == StepKind.OBSERVATION]


print("Trace model ready")
````

---

## 3. Golden Tasks — Expected Tools and Outcomes

````python
@dataclass
class GoldenTask:
    task_id: str
    question: str
    expected_tools: list[str]
    required_in_answer: list[str]
    forbidden_in_answer: list[str] = field(default_factory=list)
    max_steps: int = 8
    domain: str = "shopco"


SHOPCO_GOLDEN: list[GoldenTask] = [
    GoldenTask(
        "sc-001", "Where is ORD-1002?",
        ["lookup_order"], ["ORD-1002", "processing"],
    ),
    GoldenTask(
        "sc-002", "I want a refund for ORD-1001",
        ["lookup_order", "get_policy", "check_billing"],
        ["ORD-1001", "[pol-refunds-03]"],
        max_steps=6,
    ),
    GoldenTask(
        "sc-003", "Refund ORD-1001 — get me a manager",
        ["lookup_order", "get_policy", "check_billing", "create_escalation"],
        ["ORD-1001", "escalation"],
    ),
]

RELEASE_GOLDEN: list[GoldenTask] = [
    GoldenTask(
        "rf-001", "Can we ship v2.4.1?",
        ["run_tests", "run_lint", "run_security"],
        ["NO-GO", "security"],
        domain="releaseflow",
    ),
]

print(f"Golden tasks: {len(SHOPCO_GOLDEN)} ShopCo, {len(RELEASE_GOLDEN)} ReleaseFlow")
````

---

## 4. Simulated Agent — Produces Traces for Evaluation

````python
def extract_order_id(text: str) -> str | None:
    m = re.search(r"ORD-\d+", text.upper())
    return m.group() if m else None


SHOPCO_TOOL_IMPL = {
    "lookup_order": lambda args: f"{args['order_id']} status={ORDERS[args['order_id']]['status']}, total=${ORDERS[args['order_id']]['total_usd']:.0f}",
    "get_policy": lambda args: POLICIES.get(args.get("topic", "returns"), ""),
    "check_billing": lambda args: f"refund_eligible={BILLING[args['order_id']]['refund_eligible']}",
    "create_escalation": lambda args: f"escalation ticket ESC-{uuid.uuid4().hex[:4].upper()} created",
}

RF_TOOL_IMPL = {
    "run_tests": lambda _: RELEASE_CHECKS["unit_tests"]["detail"],
    "run_lint": lambda _: RELEASE_CHECKS["lint"]["detail"],
    "run_security": lambda _: f"security {RELEASE_CHECKS['security_scan']['status']}: {RELEASE_CHECKS['security_scan']['detail']}",
}


class ShopCoAgent:
    """Deterministic agent for offline metric demos."""

    def run(self, question: str, task_id: str = "") -> AgentTrace:
        trace = AgentTrace(str(uuid.uuid4())[:8], task_id or str(uuid.uuid4())[:8], question)
        oid = extract_order_id(question)
        is_refund = "refund" in question.lower()
        wants_manager = "manager" in question.lower() or "human" in question.lower()
        tools_done: set[str] = set()

        def step(thought: str, tool: str, args: dict[str, Any]) -> None:
            trace.steps.append(TraceStep(StepKind.THOUGHT, thought, tokens=20))
            trace.steps.append(TraceStep(StepKind.ACTION, "", tool_name=tool, tool_args=args, tokens=15))
            obs = SHOPCO_TOOL_IMPL[tool](args)
            trace.steps.append(TraceStep(StepKind.OBSERVATION, obs, tokens=25))
            tools_done.add(tool)

        if oid:
            step(f"Look up {oid}.", "lookup_order", {"order_id": oid})
        if is_refund:
            step("Fetch refunds policy.", "get_policy", {"topic": "refunds"})
            if oid:
                step("Check billing eligibility.", "check_billing", {"order_id": oid})
        if wants_manager or (oid == "ORD-1001" and is_refund):
            step("Open escalation.", "create_escalation", {})

        parts = [s.content for s in trace.steps if s.kind == StepKind.OBSERVATION]
        if is_refund and "get_policy" in tools_done:
            parts.append(POLICIES["refunds"])
        trace.final_answer = " ".join(parts)
        trace.steps.append(TraceStep(StepKind.FINAL, trace.final_answer, tokens=30))
        trace.total_tokens = sum(s.tokens for s in trace.steps)
        trace.latency_ms = len(trace.tools_used()) * 45.0
        return trace


class ReleaseFlowAgent:
    def run(self, question: str, task_id: str = "") -> AgentTrace:
        trace = AgentTrace(str(uuid.uuid4())[:8], task_id or str(uuid.uuid4())[:8], question)
        for tool in ("run_tests", "run_lint", "run_security"):
            trace.steps.append(TraceStep(StepKind.THOUGHT, f"Run {tool}.", tokens=15))
            trace.steps.append(TraceStep(StepKind.ACTION, "", tool_name=tool, tool_args={}, tokens=10))
            trace.steps.append(TraceStep(StepKind.OBSERVATION, RF_TOOL_IMPL[tool]({}), tokens=20))
        sec = RELEASE_CHECKS["security_scan"]["status"]
        trace.final_answer = "NO-GO: security warn — resolve CVE." if sec == "warn" else "GO: all checks passed."
        trace.steps.append(TraceStep(StepKind.FINAL, trace.final_answer, tokens=25))
        trace.total_tokens = sum(s.tokens for s in trace.steps)
        trace.latency_ms = 135.0
        return trace


shopco_agent = ShopCoAgent()
sample_trace = shopco_agent.run(SHOPCO_GOLDEN[1].question, SHOPCO_GOLDEN[1].task_id)
print(f"Sample trace tools: {sample_trace.tools_used()}")
````

## 5. Tool Selection Metrics

````python
@dataclass
class ToolSelectionMetrics:
    precision: float
    recall: float
    f1: float
    order_correct: bool
    redundant_calls: int
    missing_tools: list[str]
    extra_tools: list[str]


def tool_selection_metrics(trace: AgentTrace, golden: GoldenTask) -> ToolSelectionMetrics:
    expected = golden.expected_tools
    actual = trace.tools_used()
    expected_set = set(expected)
    actual_set = set(actual)

    tp = len(expected_set & actual_set)
    precision = tp / len(actual_set) if actual_set else (1.0 if not expected_set else 0.0)
    recall = tp / len(expected_set) if expected_set else 1.0
    f1 = 2 * precision * recall / (precision + recall) if (precision + recall) else 0.0

    # Order: filter actual to expected tools only, compare sequence
    actual_filtered = [t for t in actual if t in expected_set]
    order_correct = actual_filtered == expected

    redundant = len(actual) - len(actual_set)
    missing = [t for t in expected if t not in actual_set]
    extra = [t for t in actual_set if t not in expected_set]

    return ToolSelectionMetrics(precision, recall, f1, order_correct, redundant, missing, extra)


tsm = tool_selection_metrics(sample_trace, SHOPCO_GOLDEN[1])
print(f"Tool F1={tsm.f1:.2f} order_ok={tsm.order_correct} missing={tsm.missing_tools}")
````

---

## 6. Reasoning Metrics

````python
@dataclass
class ReasoningMetrics:
    grounding_score: float
    coherence_score: float
    step_efficiency: float
    hallucination_flags: list[str]


def reasoning_metrics(trace: AgentTrace, golden: GoldenTask) -> ReasoningMetrics:
    obs_text = " ".join(trace.observations()).lower()
    thoughts = trace.thoughts()
    flags: list[str] = []

    # Grounding: thoughts should not claim facts absent from observations
    grounded = 0
    for t in thoughts:
        tl = t.lower()
        if "ord-9999" in tl and "ord-9999" not in obs_text:
            flags.append(f"hallucinated order in thought: {t}")
        else:
            grounded += 1
    grounding = grounded / len(thoughts) if thoughts else 1.0

    # Coherence: each action should follow a thought in sequence
    action_count = sum(1 for s in trace.steps if s.kind == StepKind.ACTION)
    thought_count = len(thoughts)
    coherence = min(action_count, thought_count) / max(action_count, thought_count, 1)

    # Step efficiency: actual steps vs golden max
    action_steps = len(trace.tools_used())
    ideal = len(golden.expected_tools)
    efficiency = min(ideal / action_steps, 1.0) if action_steps else 0.0

    return ReasoningMetrics(grounding, coherence, efficiency, flags)


rm = reasoning_metrics(sample_trace, SHOPCO_GOLDEN[1])
print(f"Grounding={rm.grounding_score:.2f} coherence={rm.coherence_score:.2f} efficiency={rm.step_efficiency:.2f}")
````

---

## 7. Task Completion Metrics

````python
@dataclass
class TaskCompletionMetrics:
    success: bool
    partial_score: float
    required_hits: int
    required_total: int
    forbidden_violations: list[str]
    within_step_budget: bool
    latency_ms: float
    total_tokens: int


def task_completion_metrics(trace: AgentTrace, golden: GoldenTask) -> TaskCompletionMetrics:
    answer = trace.final_answer.lower()
    hits = sum(1 for req in golden.required_in_answer if req.lower() in answer)
    partial = hits / len(golden.required_in_answer) if golden.required_in_answer else 1.0
    violations = [f for f in golden.forbidden_in_answer if f.lower() in answer]
    success = partial == 1.0 and not violations
    steps = len(trace.tools_used())
    within_budget = steps <= golden.max_steps

    return TaskCompletionMetrics(
        success, partial, hits, len(golden.required_in_answer),
        violations, within_budget, trace.latency_ms, trace.total_tokens,
    )


tcm = task_completion_metrics(sample_trace, SHOPCO_GOLDEN[1])
print(f"Success={tcm.success} partial={tcm.partial_score:.2f} tokens={tcm.total_tokens}")
````

---

## 8. Combined Evaluation Result

````python
@dataclass
class EvalResult:
    task_id: str
    trace_id: str
    reasoning: ReasoningMetrics
    tool_selection: ToolSelectionMetrics
    task_completion: TaskCompletionMetrics

    @property
    def overall_pass(self) -> bool:
        return (
            self.task_completion.success
            and self.tool_selection.recall >= 1.0
            and self.reasoning.grounding_score >= 0.9
            and not self.reasoning.hallucination_flags
        )


def evaluate_trace(trace: AgentTrace, golden: GoldenTask) -> EvalResult:
    return EvalResult(
        golden.task_id, trace.trace_id,
        reasoning_metrics(trace, golden),
        tool_selection_metrics(trace, golden),
        task_completion_metrics(trace, golden),
    )


result = evaluate_trace(sample_trace, SHOPCO_GOLDEN[1])
print(f"Task {result.task_id}: overall_pass={result.overall_pass}")
````

---

## 9. Benchmark Suite — Run All Golden Tasks

````python
def run_benchmark(
    golden_tasks: list[GoldenTask],
    agent_fn: Callable[[str, str], AgentTrace],
) -> list[EvalResult]:
    results: list[EvalResult] = []
    for g in golden_tasks:
        trace = agent_fn(g.question, g.task_id)
        results.append(evaluate_trace(trace, g))
    return results


shopco_results = run_benchmark(SHOPCO_GOLDEN, shopco_agent.run)
rf_results = run_benchmark(RELEASE_GOLDEN, ReleaseFlowAgent().run)
all_results = shopco_results + rf_results

for r in all_results:
    print(
        f"{r.task_id}: pass={r.overall_pass} "
        f"tool_f1={r.tool_selection.f1:.2f} "
        f"task_ok={r.task_completion.success}"
    )
````

---

## 10. Aggregate Dashboard Metrics

````python
def aggregate_metrics(results: list[EvalResult]) -> dict[str, Any]:
    n = len(results) or 1
    return {
        "task_count": len(results),
        "pass_rate": sum(1 for r in results if r.overall_pass) / n,
        "avg_tool_f1": sum(r.tool_selection.f1 for r in results) / n,
        "avg_tool_recall": sum(r.tool_selection.recall for r in results) / n,
        "order_correct_rate": sum(1 for r in results if r.tool_selection.order_correct) / n,
        "avg_grounding": sum(r.reasoning.grounding_score for r in results) / n,
        "avg_step_efficiency": sum(r.reasoning.step_efficiency for r in results) / n,
        "avg_partial_score": sum(r.task_completion.partial_score for r in results) / n,
        "avg_tokens": sum(r.task_completion.total_tokens for r in results) / n,
        "avg_latency_ms": sum(r.task_completion.latency_ms for r in results) / n,
        "hallucination_count": sum(len(r.reasoning.hallucination_flags) for r in results),
    }


dashboard = aggregate_metrics(all_results)
print("Benchmark dashboard:")
print(pretty(dashboard))
````

## 11. Bad Agent — Detecting Regressions

````python
class BadShopCoAgent:
    """Skips policy lookup and hallucinates order facts."""

    def run(self, question: str, task_id: str = "") -> AgentTrace:
        trace = AgentTrace(str(uuid.uuid4())[:8], task_id, question)
        trace.steps.append(TraceStep(StepKind.THOUGHT, "ORD-9999 is delivered.", tokens=20))
        trace.steps.append(TraceStep(StepKind.ACTION, "", tool_name="lookup_order", tool_args={"order_id": "ORD-1001"}, tokens=15))
        trace.steps.append(TraceStep(StepKind.OBSERVATION, "ORD-1001 status=shipped", tokens=25))
        trace.steps.append(TraceStep(StepKind.ACTION, "", tool_name="lookup_order", tool_args={"order_id": "ORD-1001"}, tokens=15))
        trace.steps.append(TraceStep(StepKind.OBSERVATION, "ORD-1001 status=shipped", tokens=25))
        trace.final_answer = "Refund approved immediately."
        trace.steps.append(TraceStep(StepKind.FINAL, trace.final_answer, tokens=30))
        trace.total_tokens = sum(s.tokens for s in trace.steps)
        trace.latency_ms = 90.0
        return trace


bad_results = run_benchmark([SHOPCO_GOLDEN[1]], BadShopCoAgent().run)
bad = bad_results[0]
print(f"Bad agent pass={bad.overall_pass}")
print(f"  tool recall={bad.tool_selection.recall:.2f} missing={bad.tool_selection.missing_tools}")
print(f"  hallucinations={bad.reasoning.hallucination_flags}")
print(f"  task success={bad.task_completion.success} partial={bad.task_completion.partial_score:.2f}")
````

## 12. Metric Thresholds and Alerts

````python
@dataclass
class MetricThresholds:
    min_pass_rate: float = 0.90
    min_tool_recall: float = 0.95
    min_grounding: float = 0.90
    max_avg_tokens: float = 300.0


def check_alerts(dashboard: dict[str, Any], thresholds: MetricThresholds) -> list[str]:
    alerts: list[str] = []
    if dashboard["pass_rate"] < thresholds.min_pass_rate:
        alerts.append(f"pass_rate {dashboard['pass_rate']:.2f} below {thresholds.min_pass_rate}")
    if dashboard["avg_tool_recall"] < thresholds.min_tool_recall:
        alerts.append(f"tool_recall {dashboard['avg_tool_recall']:.2f} below {thresholds.min_tool_recall}")
    if dashboard["avg_grounding"] < thresholds.min_grounding:
        alerts.append(f"grounding {dashboard['avg_grounding']:.2f} below {thresholds.min_grounding}")
    if dashboard["avg_tokens"] > thresholds.max_avg_tokens:
        alerts.append(f"avg_tokens {dashboard['avg_tokens']:.0f} above {thresholds.max_avg_tokens}")
    if dashboard["hallucination_count"] > 0:
        alerts.append(f"hallucination_count={dashboard['hallucination_count']}")
    return alerts


good_alerts = check_alerts(dashboard, MetricThresholds())
bad_dash = aggregate_metrics(bad_results)
bad_alerts = check_alerts(bad_dash, MetricThresholds())
print(f"Good agent alerts: {good_alerts or ['none']}")
print(f"Bad agent alerts: {bad_alerts}")
````

## 13. Per-Metric Interpretation Guide

| Metric | High value means | Low value signals |
|--------|------------------|-------------------|
| **Tool recall** | Required tools were invoked | Skipped policy/billing lookups |
| **Tool precision** | No unnecessary tools | Wasted latency and tokens |
| **Order correct** | Tools ran in logical sequence | Billing before order lookup |
| **Grounding** | Thoughts match observations | Hallucinated facts |
| **Step efficiency** | Minimal steps for task | Bloated ReAct loops |
| **Partial score** | Answer contains required facts | Missing citations or IDs |
| **Pass rate** | End-to-end golden task success | Production regression |

---

## 14. Anti-Patterns

| Anti-pattern | Problem | Fix |
|--------------|---------|-----|
| Final-answer-only eval | Bad tool use goes unnoticed | Add tool + reasoning metrics |
| No golden tool list | Cannot measure recall | Define expected_tools per task |
| Single aggregate score | Masks which dimension broke | Report family metrics separately |
| Ignoring order | Wrong sequencing passes | Check tool order correctness |
| No bad-agent baselines | Alerts never tested | Include regression fixtures |
| Online-only measurement | Cannot reproduce failures | Offline benchmark suite |

## 17. Summary

Agent quality requires **process metrics**, not just final answers:

- **Reasoning** — grounding, coherence, step efficiency, hallucination flags
- **Tool selection** — precision, recall, order, redundancy
- **Task completion** — success, partial credit, step budget, cost

Golden tasks plus trace evaluation produce benchmark dashboards and alerts. Run the suite on every agent change; include bad-agent fixtures to validate detectors.
