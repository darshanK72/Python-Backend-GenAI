# Why Agent Observability Is Different

Traditional **microservice observability** assumes: one HTTP request, deterministic handler, `200` or `5xx`, done. **Agent observability** breaks every assumption:

- The same user message can produce **different tool paths** on each run.
- The HTTP response may be **`200 OK`** while the answer is **wrong** (semantic failure).
- Cost and latency are **unbounded** unless you trace every LLM and tool hop.
- Debugging requires **prompts, tool args, and observations** — not just status codes.

```
  REST API trace                 Agent trace
  ─────────────                  ─────────────────────────────
  GET /orders/1001               user message
       │                              │
       ▼                              ├─► router (LLM)
  200 {status: shipped}                ├─► get_order (tool)
                                       ├─► search_policy (tool)
                                       └─► final answer (LLM)
```

This notebook contrasts **API-style monitoring** with **agent-grade telemetry** on **ShopCo Support Hub** and **ReleaseFlow**, and shows what you miss if you only log the outer request.

````python
"""
Why agent observability is different — ShopCo + ReleaseFlow.
"""

import json
import os
import random
import re
import uuid
from dataclasses import dataclass, field
from datetime import datetime, timezone
from enum import Enum
from typing import Any

OPENAI_API_KEY = os.environ.get("OPENAI_API_KEY", "sk-proj-placeholder-replace-with-your-real-key")
os.environ.setdefault("OPENAI_API_KEY", OPENAI_API_KEY)

USE_LIVE_LLM = False
random.seed(42)


def pretty(obj: Any) -> str:
    return json.dumps(obj, indent=2, default=str)


def utc_now() -> str:
    return datetime.now(timezone.utc).isoformat()


POLICY_SNIPPETS = {
    "returns": "Returns within 30 days with receipt. [pol-returns-01]",
    "shipping": "Free shipping over $50. [pol-shipping-02]",
}

ORDERS_DB = {
    "ORD-1001": {"order_id": "ORD-1001", "status": "shipped"},
    "ORD-1002": {"order_id": "ORD-1002", "status": "processing"},
}

print("Lab ready.")
````

## 1. Six Ways Agent Telemetry Differs

| # | Microservice default | Agent reality |
|---|---------------------|---------------|
| 1 | Deterministic path | Stochastic routing / tool choice |
| 2 | Binary success (`2xx`) | Semantic success (grounded answer) |
| 3 | One span per request | Tree of LLM + tool spans |
| 4 | Latency ≈ DB + logic | Latency = Σ model calls |
| 5 | Cost ≈ compute ms | Cost = **tokens** × price |
| 6 | Logs = stack traces | Logs = **prompts + observations** |

## 2. Simulated REST API — What Normal APM Sees

````python
@dataclass
class ApiRequestLog:
    method: str
    path: str
    status_code: int
    latency_ms: int
    response_body: dict[str, Any]


def rest_get_order(order_id: str) -> ApiRequestLog:
    rec = ORDERS_DB.get(order_id.upper(), {"status": "not_found"})
    return ApiRequestLog(
        method="GET",
        path=f"/api/orders/{order_id}",
        status_code=200,
        latency_ms=45,
        response_body=rec,
    )


api_log = rest_get_order("ORD-1001")
print(pretty(api_log))
````

## 3. ShopCo Agent — Same Question, Richer Trace

````python
class SpanKind(str, Enum):
    AGENT = "agent"
    LLM = "llm"
    TOOL = "tool"
    ROUTER = "router"


@dataclass
class Span:
    name: str
    kind: SpanKind
    status: str
    attributes: dict[str, Any] = field(default_factory=dict)


@dataclass
class AgentTrace:
    trace_id: str
    user_message: str
    spans: list[Span] = field(default_factory=list)
    final_answer: str = ""
    http_status: int = 200
    tokens: int = 0
    semantic_ok: bool = True

    def path_fingerprint(self) -> str:
        return "→".join(s.name for s in self.spans)


def shopco_agent_traced(message: str, variant: int = 0) -> AgentTrace:
    trace = AgentTrace(trace_id=f"tr-{uuid.uuid4().hex[:6]}", user_message=message)
    intent = "return" if "return" in message.lower() else "order"
    trace.spans.append(Span("route_intent", SpanKind.ROUTER, "ok", {"intent": intent}))
    trace.tokens += 135

    m = re.search(r"ORD-[0-9]{4}", message.upper())
    if m and variant != 1:  # variant 1 skips order lookup (bad path)
        oid = m.group(0)
        rec = ORDERS_DB[oid]
        trace.spans.append(Span("get_order", SpanKind.TOOL, "ok", {"order_id": oid, "status": rec["status"]}))

    if intent == "return":
        trace.spans.append(Span("search_policy", SpanKind.TOOL, "ok", {"policy": "pol-returns-01"}))
        trace.tokens += 120
        if variant == 1:
            trace.final_answer = "Returns are sometimes allowed."  # no citation — semantic fail
            trace.semantic_ok = False
        else:
            trace.final_answer = f"Order shipped. {POLICY_SNIPPETS['returns']}"
    else:
        trace.final_answer = "Please provide more detail."

    trace.spans.append(Span("compose_answer", SpanKind.LLM, "ok"))
    return trace


good_trace = shopco_agent_traced("Can I return ORD-1001?", variant=0)
bad_trace = shopco_agent_traced("Can I return ORD-1001?", variant=1)
print("Good path:", good_trace.path_fingerprint())
print("Bad path:", bad_trace.path_fingerprint(), "semantic_ok:", bad_trace.semantic_ok)
````

## 4. Silent Failure — HTTP 200, Wrong Answer

APM dashboards show green. Customers still complain.

````python
@dataclass
class MonitoringView:
    source: str
    status_ok: bool
    detects_semantic_failure: bool
    sees_tool_path: bool
    sees_token_cost: bool


def api_monitoring_view(trace: AgentTrace) -> MonitoringView:
    return MonitoringView(
        source="api_gateway_only",
        status_ok=trace.http_status == 200,
        detects_semantic_failure=False,
        sees_tool_path=False,
        sees_token_cost=False,
    )


def agent_monitoring_view(trace: AgentTrace) -> MonitoringView:
    return MonitoringView(
        source="agent_trace",
        status_ok=trace.http_status == 200 and trace.semantic_ok,
        detects_semantic_failure=not trace.semantic_ok,
        sees_tool_path=len(trace.spans) > 1,
        sees_token_cost=trace.tokens > 0,
    )


print("API view on BAD trace:", pretty(api_monitoring_view(bad_trace)))
print("Agent view on BAD trace:", pretty(agent_monitoring_view(bad_trace)))
````

## 5. Non-Determinism — Path Variance on Identical Input

````python
def shopco_stochastic(message: str) -> AgentTrace:
    """Simulates model occasionally skipping policy search."""
    trace = AgentTrace(trace_id=f"tr-{uuid.uuid4().hex[:6]}", user_message=message)
    trace.spans.append(Span("route_intent", SpanKind.ROUTER, "ok"))
    m = re.search(r"ORD-[0-9]{4}", message.upper())
    if m:
        trace.spans.append(Span("get_order", SpanKind.TOOL, "ok", {"order_id": m.group(0)}))
    if random.random() > 0.3:
        trace.spans.append(Span("search_policy", SpanKind.TOOL, "ok"))
        trace.final_answer = POLICY_SNIPPETS["returns"]
        trace.semantic_ok = "[pol-" in trace.final_answer
    else:
        trace.final_answer = "You might be able to return it."
        trace.semantic_ok = False
    trace.spans.append(Span("compose_answer", SpanKind.LLM, "ok"))
    return trace


runs = [shopco_stochastic("Return ORD-1001?") for _ in range(8)]
fingerprints = [r.path_fingerprint() for r in runs]
semantic_pass = sum(1 for r in runs if r.semantic_ok)
print("Unique paths:", len(set(fingerprints)))
print("Semantic pass rate:", semantic_pass, "/", len(runs))
print("HTTP 200 on all:", all(r.http_status == 200 for r in runs))
````

## 6. Cost Is Invisible to Request-Only Metrics

````python
@dataclass
class CostProfile:
    pattern: str
    llm_calls: int
    tool_calls: int
    tokens: int
    est_usd: float


def profile_trace(trace: AgentTrace) -> CostProfile:
    llm = sum(1 for s in trace.spans if s.kind in (SpanKind.LLM, SpanKind.ROUTER))
    tools = sum(1 for s in trace.spans if s.kind == SpanKind.TOOL)
    return CostProfile("agent", llm, tools, trace.tokens, round(trace.tokens / 1000 * 0.002, 4))


def profile_api() -> CostProfile:
    return CostProfile("rest_api", 0, 0, 0, 0.0001)


print("REST:", profile_api())
print("Agent:", profile_trace(good_trace))
````

## 7. ReleaseFlow — Pipeline vs Agent Loop

````python
@dataclass
class ReleaseRun:
    mode: str
    steps_visible: list[str]
    can_explain_block_reason: bool


def release_ci_only_logging() -> ReleaseRun:
    return ReleaseRun(
        mode="jenkins_console_only",
        steps_visible=["pipeline finished"],
        can_explain_block_reason=False,
    )


def release_agent_traced() -> ReleaseRun:
    return ReleaseRun(
        mode="agent_trace",
        steps_visible=["run_unit_tests", "run_security_scan", "gatekeeper_llm", "deploy_staging"],
        can_explain_block_reason=True,
    )


print(pretty(release_ci_only_logging()))
print(pretty(release_agent_traced()))
````

## 9. Session Continuity — Thread-Level Traces

Agents are **multi-turn**. Correlating by `trace_id` per message loses the ticket story — use **`session_id`** or **`thread_id`** across turns.

````python
@dataclass
class SessionTrace:
    session_id: str
    turns: list[dict[str, Any]] = field(default_factory=list)

    def add_turn(self, user: str, trace: AgentTrace) -> None:
        self.turns.append({
            "user": user,
            "trace_id": trace.trace_id,
            "path": trace.path_fingerprint(),
            "semantic_ok": trace.semantic_ok,
        })


session = SessionTrace(session_id="sess-alice-001")
session.add_turn("Hi, ORD-1001", shopco_agent_traced("Hi, ORD-1001"))
session.add_turn("Can I return it?", shopco_agent_traced("Can I return ORD-1001?"))
print(pretty(session))
````

## 10. Human-in-the-Loop Gaps

When agents pause for approval, **wall-clock latency** spans hours but **compute time** is minutes. Track `awaiting_human` spans separately.

````python
@dataclass
class HitlSpan:
    name: str
    status: str
    wait_minutes: int


hitl_example = [
    Span("security_scan", SpanKind.TOOL, "warn"),
    HitlSpan("awaiting_manager_approval", "pending", 180),
    Span("deploy_production", SpanKind.TOOL, "ok"),
]
print("Compute spans vs HITL wait:", [type(x).__name__ for x in hitl_example])
````

## 12. Side-by-Side — Same Ticket, Two Monitoring Stacks

````python
def compare_monitoring(trace: AgentTrace) -> dict[str, Any]:
    return {
        "user_message": trace.user_message,
        "api_monitoring": {
            "healthy": trace.http_status == 200,
            "alert_fired": False,
        },
        "agent_monitoring": {
            "healthy": trace.semantic_ok,
            "path": trace.path_fingerprint(),
            "missing_policy_citation": "[pol-" not in trace.final_answer,
            "tokens": trace.tokens,
            "alert_fired": not trace.semantic_ok,
        },
    }


print(pretty(compare_monitoring(bad_trace)))
````

## 13. Design Implications — What to Instrument

1. **Span per LLM call** — model, tokens, latency
2. **Span per tool** — name, args (redacted), observation status
3. **Span per router / planner** — chosen intent or plan id
4. **Session id** on all spans in a chat
5. **Semantic eval hooks** — citation check, task assertions
6. **Cost accumulator** on trace root

## 14. Optional Live LLM — Same Prompt, Two Runs

Set `USE_LIVE_LLM = True` to show output variance (illustrates non-determinism).

````python
def live_two_runs(prompt: str) -> list[str]:
    from langchain_core.messages import HumanMessage
    from langchain_openai import ChatOpenAI

    llm = ChatOpenAI(model="gpt-4o-mini", api_key=OPENAI_API_KEY, temperature=0.7)
    return [str(llm.invoke([HumanMessage(content=prompt)]).content)[:80] for _ in range(2)]


if USE_LIVE_LLM:
    outs = live_two_runs("One sentence ShopCo return policy.")
    print("Run A:", outs[0])
    print("Run B:", outs[1])
    print("Identical:", outs[0] == outs[1])
else:
    print("Offline: stochastic demo showed", len(set(fingerprints)), "unique paths on identical input")
````

## 15. Common Mistakes

| Mistake | Symptom | Fix |
|---------|---------|-----|
| Reuse API dashboards only | Miss semantic failures | Agent spans + eval |
| One span for whole agent | Can't find slow tool | Child spans |
| Ignore token metrics | Cost surprises | Per-trace cost |
| Trace per message not session | Lost multi-turn context | `session_id` |
| No path fingerprint | Can't detect drift | Compare span sequences |

## 17. Summary

**Agent observability is different** because success is **semantic**, paths are **stochastic**, and cost scales with **tokens and steps** — not one deterministic handler.

ShopCo showed silent failures with green HTTP status. ReleaseFlow showed why CI console logs are not enough for agent pipelines. Instrument **LLM + tool spans**, **sessions**, **cost**, and **semantic eval** — not just the outer request.

Treat agent telemetry as **debugging UX for operators**, not as optional logging.
