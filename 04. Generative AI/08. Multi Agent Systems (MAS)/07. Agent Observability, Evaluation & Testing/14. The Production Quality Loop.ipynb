# The Production Quality Loop

Shipping an agent once is easy. **Keeping it good in production** is the hard part. Models drift, policies change, users ask new questions, and retrieval indexes rot. A **production quality loop** is the operating rhythm that connects **observability** (what happened) with **evaluation** (was it good enough) in a repeatable cycle.

```
  ┌─────────┐    ┌──────────────┐    ┌─────────────┐    ┌──────────────┐
  │  BUILD  │───►│ OFFLINE EVAL │───►│ CANARY DEPLOY│───►│ ONLINE METRICS│
  └─────────┘    └──────────────┘    └─────────────┘    └──────┬───────┘
       ▲                                                         │
       │              ┌──────────────┐    ┌─────────────┐        │
       └──────────────│ FIX + RETEST │◄───│ ALERT/MINE  │◄───────┘
                      └──────────────┘    └─────────────┘
```

| Phase | Purpose | Artifact |
|-------|---------|----------|
| **Build** | Change prompt, tools, RAG | Version tag |
| **Offline eval** | Regression before deploy | Golden suite pass rate |
| **Canary** | Limit blast radius | 5–10% traffic slice |
| **Online metrics** | Real user behavior | Traces + aggregates |
| **Alert / mine** | Find failures worth fixing | New eval cases |
| **Fix** | Patch root cause | Changelog entry |

This notebook implements the full loop in plain Python for **ShopCo Support Hub** (RAG support agent) and **ReleaseFlow** (release checklist agent). You will collect production traces, run CI gates, simulate a canary, detect regressions, grow a golden dataset from failures, and iterate.

````python
"""
Production quality loop — ShopCo + ReleaseFlow.
"""

import json
import os
import re
import uuid
from collections import Counter, defaultdict
from dataclasses import dataclass, field
from datetime import datetime, timezone
from enum import Enum
from typing import Any, Callable

OPENAI_API_KEY = os.environ.get("OPENAI_API_KEY", "sk-proj-placeholder-replace-with-your-real-key")
os.environ.setdefault("OPENAI_API_KEY", OPENAI_API_KEY)

USE_LIVE_LLM = False
TOKEN_COST_PER_1K = 0.002


def pretty(obj: Any) -> str:
    return json.dumps(obj, indent=2, default=str)


def utc_now() -> str:
    return datetime.now(timezone.utc).isoformat()


POLICY_CORPUS: list[dict[str, str]] = [
    {"id": "pol-returns-01", "title": "Returns", "text": "Customers may return items within 30 days of delivery with receipt. Final-sale items cannot be returned. [pol-returns-01]"},
    {"id": "pol-shipping-02", "title": "Shipping", "text": "Free standard shipping on orders over $50. Express available at checkout. [pol-shipping-02]"},
    {"id": "pol-billing-03", "title": "Billing", "text": "Charges appear as SHOPCO* on statements. Partial refunds may take two billing cycles. [pol-billing-03]"},
    {"id": "faq-hours-05", "title": "Support Hours", "text": "Support hours Mon–Fri 9am–6pm ET. [faq-hours-05]"},
]

ORDERS_DB = {
    "ORD-1001": {"order_id": "ORD-1001", "status": "shipped", "days_since_delivery": 12},
    "ORD-1002": {"order_id": "ORD-1002", "status": "processing", "days_since_delivery": 0},
}

RELEASE_CHECKS = {
    "unit_tests": {"status": "pass", "owner": "backend"},
    "security_scan": {"status": "warn", "owner": "security"},
    "changelog": {"status": "pass", "owner": "pm"},
}

print("Quality loop lab ready.")
````

## 1. Why a Loop — Not a One-Time Eval

A single pre-launch eval answers: *"Is v1 acceptable?"* Production answers different questions every week:

- Did the new embedding model hurt retrieval?
- Are users asking about a policy we never indexed?
- Is the agent looping on `lookup_order`?
- Did token usage spike after a prompt change?

Without a loop, teams **ship fixes blindly** or **wait for support tickets**. The loop makes quality **measurable, gated, and cumulative** — each production failure can become tomorrow's regression test.

---

## 2. Loop Phases and Ownership

| Phase | Owner | Exit criteria |
|-------|-------|---------------|
| Offline eval | ML / agent engineer | `pass_rate ≥ threshold` |
| Canary | SRE + product | Online metrics within bounds for N hours |
| Full rollout | Release manager | Canary stable, no P0 alerts |
| Mining | On-call + QA | Failure traces triaged weekly |
| Dataset growth | Eval owner | New cases reviewed and merged |

The code below models each phase as a **state machine** you can run in notebooks, CI, or a small orchestrator.

````python
class LoopPhase(str, Enum):
    BUILD = "build"
    OFFLINE_EVAL = "offline_eval"
    CANARY = "canary"
    ONLINE = "online"
    ALERT = "alert"
    FIX = "fix"


@dataclass
class QualityLoopState:
    version: str
    phase: LoopPhase = LoopPhase.BUILD
    offline_pass_rate: float = 0.0
    canary_pass_rate: float = 0.0
    alerts: list[str] = field(default_factory=list)
    golden_case_count: int = 0

    def advance(self, next_phase: LoopPhase) -> None:
        self.phase = next_phase


state = QualityLoopState(version="shopco-v1.0.0")
print(f"Starting loop at {state.phase.value}")
````

## 3. Observability Primitives — Traces for Every Run

The loop starts with **instrumentation**. Every agent invocation produces a trace: spans for retrieval, LLM, tools, plus token and cost metrics.

````python
class SpanKind(str, Enum):
    AGENT = "agent"
    LLM = "llm"
    TOOL = "tool"
    RETRIEVER = "retriever"


@dataclass
class Span:
    span_id: str
    trace_id: str
    name: str
    kind: SpanKind
    start_ts: str
    end_ts: str | None = None
    attributes: dict[str, Any] = field(default_factory=dict)
    status: str = "ok"

    @property
    def duration_ms(self) -> int:
        if not self.end_ts:
            return 0
        t0 = datetime.fromisoformat(self.start_ts)
        t1 = datetime.fromisoformat(self.end_ts)
        return int((t1 - t0).total_seconds() * 1000)


@dataclass
class ProductionTrace:
    trace_id: str
    service: str
    version: str
    user_question: str
    final_answer: str
    spans: list[Span] = field(default_factory=list)
    metrics: dict[str, float] = field(default_factory=dict)
    eval_scores: dict[str, float] = field(default_factory=dict)
    passed: bool | None = None
    ts: str = field(default_factory=utc_now)

    def add_span(self, span: Span) -> None:
        self.spans.append(span)

    def estimated_cost_usd(self) -> float:
        tokens = self.metrics.get("tokens_in", 0) + self.metrics.get("tokens_out", 0)
        return round(tokens / 1000 * TOKEN_COST_PER_1K, 4)
````

## 4. AgentTracer — Instrument ShopCo Runs

````python
class AgentTracer:
    def __init__(self, service: str, version: str) -> None:
        self.trace = ProductionTrace(
            trace_id=f"tr-{uuid.uuid4().hex[:8]}",
            service=service,
            version=version,
            user_question="",
            final_answer="",
        )
        self.metrics: dict[str, float] = {"llm_calls": 0, "tool_calls": 0, "tokens_in": 0, "tokens_out": 0}

    def start_span(self, name: str, kind: SpanKind, **attrs: Any) -> Span:
        return Span(
            span_id=f"sp-{uuid.uuid4().hex[:8]}",
            trace_id=self.trace.trace_id,
            name=name,
            kind=kind,
            start_ts=utc_now(),
            attributes=dict(attrs),
        )

    def end_span(self, span: Span, status: str = "ok", **attrs: Any) -> None:
        span.end_ts = utc_now()
        span.status = status
        span.attributes.update(attrs)
        self.trace.add_span(span)

    def record_llm(self, tokens_in: int = 120, tokens_out: int = 80) -> None:
        self.metrics["llm_calls"] += 1
        self.metrics["tokens_in"] += tokens_in
        self.metrics["tokens_out"] += tokens_out

    def record_tool(self) -> None:
        self.metrics["tool_calls"] += 1

    def finalize(self, question: str, answer: str) -> ProductionTrace:
        self.trace.user_question = question
        self.trace.final_answer = answer
        self.trace.metrics = dict(self.metrics)
        return self.trace
````

## 5. ShopCo Support Agent — Retrieve, Tool, Compose

````python
AGENT_VERSION = "shopco-v1.0.0"


def tokenize(text: str) -> list[str]:
    return [t for t in re.split(r"\W+", text.lower()) if len(t) > 2]


def retrieve(query: str, corpus: list[dict[str, str]], top_k: int = 2) -> list[str]:
    terms = tokenize(query)
    scored: list[tuple[float, str]] = []
    for doc in corpus:
        hay = f"{doc['title']} {doc['text']}".lower()
        score = sum(1 for t in terms if t in hay) if terms else 0
        if score > 0:
            scored.append((score, doc["text"]))
    scored.sort(key=lambda x: x[0], reverse=True)
    return [text for _, text in scored[:top_k]] or [corpus[0]["text"]]


def lookup_order(order_id: str) -> dict[str, Any]:
    return ORDERS_DB.get(order_id, {"error": "not_found"})


def compose_answer(question: str, contexts: list[str], order_info: dict[str, Any] | None) -> str:
    q = question.lower()
    parts: list[str] = []
    if "return" in q:
        parts.append("Returns are allowed within 30 days with receipt. [pol-returns-01]")
    if "shipping" in q or "free" in q:
        parts.append("Free standard shipping applies to orders over $50. [pol-shipping-02]")
    if "hour" in q or "support" in q:
        parts.append("Support is available Mon–Fri 9am–6pm ET. [faq-hours-05]")
    if order_info and "order_id" in order_info:
        parts.append(f"Order {order_info['order_id']} status: {order_info['status']}.")
    if not parts and contexts:
        parts.append(contexts[0][:120])
    return " ".join(parts) if parts else "I could not find a policy match."


def shopco_agent(question: str, tracer: AgentTracer) -> str:
    root = tracer.start_span("shopco_support", SpanKind.AGENT, question=question)

    ret_span = tracer.start_span("retrieve_policy", SpanKind.RETRIEVER)
    contexts = retrieve(question, POLICY_CORPUS)
    tracer.end_span(ret_span, chunk_count=len(contexts))

    order_info = None
    m = re.search(r"ORD-\d+", question.upper())
    if m:
        tool_span = tracer.start_span("lookup_order", SpanKind.TOOL, order_id=m.group())
        tracer.record_tool()
        order_info = lookup_order(m.group())
        tracer.end_span(tool_span, result=order_info.get("status", "error"))

    llm_span = tracer.start_span("compose_answer", SpanKind.LLM)
    tracer.record_llm()
    answer = compose_answer(question, contexts, order_info)
    tracer.end_span(llm_span, answer_len=len(answer))

    tracer.end_span(root, answer_preview=answer[:60])
    return answer


demo_tracer = AgentTracer("shopco-support", AGENT_VERSION)
demo_answer = shopco_agent("Can I return ORD-1001?", demo_tracer)
demo_trace = demo_tracer.finalize("Can I return ORD-1001?", demo_answer)
print(demo_answer)
print(f"Trace {demo_trace.trace_id}: {len(demo_trace.spans)} spans, ${demo_trace.estimated_cost_usd()}")
````

## 6. EvalCase — Offline Assertions

Offline eval uses **golden cases**: input, expected substrings, optional rubric scores, and max cost.

````python
@dataclass
class EvalCase:
    case_id: str
    question: str
    must_contain: list[str] = field(default_factory=list)
    must_not_contain: list[str] = field(default_factory=list)
    max_cost_usd: float = 0.05
    source: str = "golden"  # golden | mined | synthetic


@dataclass
class EvalResult:
    case_id: str
    passed: bool
    answer: str
    failures: list[str] = field(default_factory=list)
    trace_id: str = ""
    cost_usd: float = 0.0


def run_eval_case(case: EvalCase, agent_fn: Callable[[str, AgentTracer], str], version: str) -> EvalResult:
    tracer = AgentTracer("shopco-support", version)
    answer = agent_fn(case.question, tracer)
    trace = tracer.finalize(case.question, answer)
    failures: list[str] = []
    lower = answer.lower()
    for needle in case.must_contain:
        if needle.lower() not in lower:
            failures.append(f"missing:{needle}")
    for bad in case.must_not_contain:
        if bad.lower() in lower:
            failures.append(f"forbidden:{bad}")
    if trace.estimated_cost_usd() > case.max_cost_usd:
        failures.append(f"cost:{trace.estimated_cost_usd()}")
    return EvalResult(case.case_id, len(failures) == 0, answer, failures, trace.trace_id, trace.estimated_cost_usd())


GOLDEN_SUITE: list[EvalCase] = [
    EvalCase("g1", "What is the return window?", ["30 days", "pol-returns-01"]),
    EvalCase("g2", "When is free shipping available?", ["$50", "pol-shipping-02"]),
    EvalCase("g3", "What are support hours?", ["9am", "faq-hours-05"]),
    EvalCase("g4", "Can I return ORD-1001?", ["30 days", "shipped"]),
]

print(f"Golden suite: {len(GOLDEN_SUITE)} cases")
````

## 7. Offline Eval Harness — CI Gate Before Deploy

````python
OFFLINE_GATES = {"pass_rate_min": 0.75, "max_avg_cost_usd": 0.03}


def run_eval_suite(cases: list[EvalCase], agent_fn: Callable, version: str) -> dict[str, Any]:
    results = [run_eval_case(c, agent_fn, version) for c in cases]
    passed = sum(1 for r in results if r.passed)
    avg_cost = round(sum(r.cost_usd for r in results) / len(results), 4) if results else 0.0
    return {
        "version": version,
        "total": len(results),
        "passed": passed,
        "pass_rate": round(passed / len(results), 2) if results else 0.0,
        "avg_cost_usd": avg_cost,
        "failures": [{"case": r.case_id, "reasons": r.failures} for r in results if not r.passed],
    }


def offline_ci_gate(suite: dict[str, Any]) -> dict[str, Any]:
    checks = {
        "pass_rate": suite["pass_rate"] >= OFFLINE_GATES["pass_rate_min"],
        "avg_cost": suite["avg_cost_usd"] <= OFFLINE_GATES["max_avg_cost_usd"],
    }
    return {"deploy_ok": all(checks.values()), "checks": checks, "suite": suite}


offline_suite = run_eval_suite(GOLDEN_SUITE, shopco_agent, AGENT_VERSION)
offline_gate = offline_ci_gate(offline_suite)
state.offline_pass_rate = offline_suite["pass_rate"]
state.golden_case_count = len(GOLDEN_SUITE)
state.advance(LoopPhase.OFFLINE_EVAL)
print(pretty(offline_gate))
````

## 8. Production Trace Store — Simulate Live Traffic

After deploy, traces accumulate in a store. We simulate a day of ShopCo traffic — mostly good, a few failures.

````python
LIVE_QUESTIONS = [
    "What is the return window?",
    "When is free shipping available?",
    "What are support hours?",
    "Can I return ORD-1001?",
    "Can I return ORD-1001?",
    "billing charge name on statement?",
    "return policy for final sale items?",
    "What is the return window?",
    "How do I contact support?",
    "Is express shipping free?",
]


class TraceStore:
    def __init__(self) -> None:
        self.traces: list[ProductionTrace] = []

    def ingest(self, trace: ProductionTrace) -> None:
        self.traces.append(trace)

    def for_version(self, version: str) -> list[ProductionTrace]:
        return [t for t in self.traces if t.version == version]

    def failures(self) -> list[ProductionTrace]:
        return [t for t in self.traces if t.passed is False]


def simulate_production_day(store: TraceStore, version: str, questions: list[str]) -> None:
    for q in questions:
        tracer = AgentTracer("shopco-support", version)
        answer = shopco_agent(q, tracer)
        trace = tracer.finalize(q, answer)
        matched = next((c for c in GOLDEN_SUITE if c.question.lower() == q.lower()), None)
        if matched:
            result = run_eval_case(matched, shopco_agent, version)
            trace.passed = result.passed
            trace.eval_scores = {"assertions": 1.0 if result.passed else 0.0}
        else:
            trace.passed = "pol-" in answer or "faq-" in answer
            trace.eval_scores = {"heuristic": 1.0 if trace.passed else 0.0}
        store.ingest(trace)


prod_store = TraceStore()
simulate_production_day(prod_store, AGENT_VERSION, LIVE_QUESTIONS)
print(f"Ingested {len(prod_store.traces)} production traces")
````

## 9. Canary Deploy — Compare Versions on a Traffic Slice

````python
CANARY_VERSION = "shopco-v1.1.0"  # simulated regression: worse retrieval
CANARY_FRACTION = 0.3


def bad_retrieve(query: str, corpus: list[dict[str, str]], top_k: int = 2) -> list[str]:
    return [corpus[2]["text"]]  # always billing — simulates broken index


def shopco_agent_v11(question: str, tracer: AgentTracer) -> str:
    root = tracer.start_span("shopco_support", SpanKind.AGENT, question=question)
    ret_span = tracer.start_span("retrieve_policy", SpanKind.RETRIEVER)
    contexts = bad_retrieve(question, POLICY_CORPUS)
    tracer.end_span(ret_span, chunk_count=len(contexts))
    llm_span = tracer.start_span("compose_answer", SpanKind.LLM)
    tracer.record_llm(tokens_in=180, tokens_out=120)
    answer = compose_answer(question, contexts, None)
    tracer.end_span(llm_span, answer_len=len(answer))
    tracer.end_span(root)
    return answer


@dataclass
class CanaryReport:
    baseline_version: str
    canary_version: str
    baseline_pass_rate: float
    canary_pass_rate: float
    promote: bool
    reason: str


def run_canary(
    cases: list[EvalCase],
    baseline_fn: Callable,
    canary_fn: Callable,
    baseline_ver: str,
    canary_ver: str,
    min_delta: float = -0.10,
) -> CanaryReport:
    base_suite = run_eval_suite(cases, baseline_fn, baseline_ver)
    canary_suite = run_eval_suite(cases, canary_fn, canary_ver)
    delta = canary_suite["pass_rate"] - base_suite["pass_rate"]
    promote = delta >= min_delta and canary_suite["pass_rate"] >= OFFLINE_GATES["pass_rate_min"]
    reason = "canary within bounds" if promote else f"regression delta={delta:.2f}"
    return CanaryReport(baseline_ver, canary_ver, base_suite["pass_rate"], canary_suite["pass_rate"], promote, reason)


canary_report = run_canary(GOLDEN_SUITE, shopco_agent, shopco_agent_v11, AGENT_VERSION, CANARY_VERSION)
state.canary_pass_rate = canary_report.canary_pass_rate
state.advance(LoopPhase.CANARY)
print(pretty(canary_report.__dict__))
````

## 10. Online Metrics Aggregator

Roll up traces into **SLIs** (Service Level Indicators): pass rate, P95 latency proxy, cost per request, tool-loop rate.

````python
@dataclass
class OnlineMetrics:
    window: str
    request_count: int
    pass_rate: float
    avg_cost_usd: float
    p95_duration_ms: int
    loop_rate: float


def detect_tool_loop(trace: ProductionTrace, threshold: int = 3) -> bool:
    names = [s.name for s in trace.spans if s.kind == SpanKind.TOOL]
    if not names:
        return False
    return any(c >= threshold for c in Counter(names).values())


def aggregate_online(traces: list[ProductionTrace], window: str = "24h") -> OnlineMetrics:
    if not traces:
        return OnlineMetrics(window, 0, 0.0, 0.0, 0, 0.0)
    scored = [t for t in traces if t.passed is not None]
    pass_rate = sum(1 for t in scored if t.passed) / len(scored) if scored else 0.0
    durations = [sum(s.duration_ms for s in t.spans) for t in traces]
    durations.sort()
    p95 = durations[int(0.95 * (len(durations) - 1))] if durations else 0
    avg_cost = sum(t.estimated_cost_usd() for t in traces) / len(traces)
    loop_rate = sum(1 for t in traces if detect_tool_loop(t)) / len(traces)
    return OnlineMetrics(
        window=window,
        request_count=len(traces),
        pass_rate=round(pass_rate, 2),
        avg_cost_usd=round(avg_cost, 4),
        p95_duration_ms=p95,
        loop_rate=round(loop_rate, 2),
    )


online_metrics = aggregate_online(prod_store.traces)
state.advance(LoopPhase.ONLINE)
print(pretty(online_metrics.__dict__))
````

## 11. Alert Rules — When to Stop the Rollout

Alerts turn metrics into action. Define thresholds once; evaluate on every aggregation window.

````python
ALERT_THRESHOLDS = {
    "pass_rate_min": 0.70,
    "avg_cost_max_usd": 0.04,
    "loop_rate_max": 0.05,
    "canary_regression_max": -0.10,
}


@dataclass
class Alert:
    alert_id: str
    severity: str
    message: str
    metric: str
    value: float


def evaluate_alerts(metrics: OnlineMetrics, canary: CanaryReport) -> list[Alert]:
    alerts: list[Alert] = []
    if metrics.pass_rate < ALERT_THRESHOLDS["pass_rate_min"]:
        alerts.append(Alert("a1", "P1", "pass rate below floor", "pass_rate", metrics.pass_rate))
    if metrics.avg_cost_usd > ALERT_THRESHOLDS["avg_cost_max_usd"]:
        alerts.append(Alert("a2", "P2", "cost per request elevated", "avg_cost_usd", metrics.avg_cost_usd))
    if metrics.loop_rate > ALERT_THRESHOLDS["loop_rate_max"]:
        alerts.append(Alert("a3", "P1", "tool loop rate high", "loop_rate", metrics.loop_rate))
    delta = canary.canary_pass_rate - canary.baseline_pass_rate
    if delta < ALERT_THRESHOLDS["canary_regression_max"]:
        alerts.append(Alert("a4", "P0", "canary regression detected", "canary_delta", delta))
    if not canary.promote:
        alerts.append(Alert("a5", "P0", canary.reason, "canary_promote", 0.0))
    return alerts


alerts = evaluate_alerts(online_metrics, canary_report)
state.alerts = [a.message for a in alerts]
state.advance(LoopPhase.ALERT)
print(pretty([a.__dict__ for a in alerts]))
````

## 12. Mine Failures — Grow the Golden Dataset

Production failures are **free eval cases** if you capture question + answer + trace. Human review prevents polluting the suite.

````python
@dataclass
class MinedCandidate:
    trace_id: str
    question: str
    answer: str
    suggested_must_contain: list[str]
    status: str = "pending_review"


class GoldenDataset:
    def __init__(self, cases: list[EvalCase] | None = None) -> None:
        self.cases: list[EvalCase] = list(cases or [])
        self.mined: list[MinedCandidate] = []

    def mine_from_store(self, store: TraceStore) -> list[MinedCandidate]:
        candidates: list[MinedCandidate] = []
        for trace in store.failures():
            if any(c.question.lower() == trace.user_question.lower() for c in self.cases):
                continue
            suggestion = ["pol-", "faq-"] if "return" in trace.user_question.lower() else ["support"]
            cand = MinedCandidate(trace.trace_id, trace.user_question, trace.final_answer, suggestion)
            candidates.append(cand)
            self.mined.append(cand)
        return candidates

    def approve(self, trace_id: str, must_contain: list[str]) -> EvalCase | None:
        cand = next((m for m in self.mined if m.trace_id == trace_id), None)
        if not cand:
            return None
        case = EvalCase(f"mined-{len(self.cases)+1}", cand.question, must_contain, source="mined")
        cand.status = "approved"
        self.cases.append(case)
        return case


golden_ds = GoldenDataset(GOLDEN_SUITE)

canary_store = TraceStore()
for q in LIVE_QUESTIONS[:6]:
    tracer = AgentTracer("shopco-support", CANARY_VERSION)
    answer = shopco_agent_v11(q, tracer)
    trace = tracer.finalize(q, answer)
    trace.passed = "pol-returns" in answer or "pol-shipping" in answer or "faq-hours" in answer
    canary_store.ingest(trace)

mined = golden_ds.mine_from_store(canary_store)
print(f"Mined {len(mined)} new candidates")
if mined:
    approved = golden_ds.approve(mined[0].trace_id, ["30 days", "pol-returns-01"])
    print("Approved case:", approved.case_id if approved else None)
print(f"Golden dataset size: {len(golden_ds.cases)}")
````

## 13. Fix Cycle — Patch, Version Bump, Re-Eval

When alerts fire, fix the root cause, bump the version, and re-run offline + canary before retrying deploy.

````python
FIXED_VERSION = "shopco-v1.1.1"


def shopco_agent_fixed(question: str, tracer: AgentTracer) -> str:
    """v1.1.1 restores correct retrieval after bad v1.1.0 deploy."""
    return shopco_agent(question, tracer)


@dataclass
class ChangeLogEntry:
    version: str
    change_type: str
    description: str
    eval_before: float
    eval_after: float
    ts: str = field(default_factory=utc_now)


class QualityChangelog:
    def __init__(self) -> None:
        self.entries: list[ChangeLogEntry] = []

    def record(self, entry: ChangeLogEntry) -> None:
        self.entries.append(entry)


def fix_and_retest(cases: list[EvalCase], broken_ver: str, fixed_ver: str, broken_fn: Callable, fixed_fn: Callable) -> dict[str, Any]:
    before = run_eval_suite(cases, broken_fn, broken_ver)
    after = run_eval_suite(cases, fixed_fn, fixed_ver)
    gate = offline_ci_gate(after)
    return {"before": before, "after": after, "gate": gate, "recovered": after["pass_rate"] > before["pass_rate"]}


changelog = QualityChangelog()
fix_result = fix_and_retest(golden_ds.cases, CANARY_VERSION, FIXED_VERSION, shopco_agent_v11, shopco_agent_fixed)
changelog.record(ChangeLogEntry(
    FIXED_VERSION, "retrieval_fix",
    "Restored keyword retriever after bad embedding rollout",
    fix_result["before"]["pass_rate"], fix_result["after"]["pass_rate"],
))
state.advance(LoopPhase.FIX)
print(pretty(fix_result))
````

## 14. ReleaseFlow — Same Loop for Pipeline Agents

**ReleaseFlow** checks release gates before promoting a build. The quality loop applies: offline checklist eval → canary environment → online pass rate.

````python
@dataclass
class ReleaseEvalCase:
    case_id: str
    release_id: str
    required_checks: list[str]
    block_on_warn: list[str] = field(default_factory=list)


def releaseflow_agent(release_id: str, checks_db: dict[str, dict[str, str]]) -> dict[str, Any]:
    results = {name: checks_db.get(name, {"status": "missing"}) for name in checks_db}
    blockers = [n for n, r in results.items() if r.get("status") == "fail"]
    warns = [n for n, r in results.items() if r.get("status") == "warn"]
    return {
        "release_id": release_id,
        "checks": results,
        "blockers": blockers,
        "warnings": warns,
        "promote_ok": len(blockers) == 0,
    }


def eval_release(case: ReleaseEvalCase, checks_db: dict[str, dict[str, str]]) -> dict[str, Any]:
    report = releaseflow_agent(case.release_id, checks_db)
    missing = [c for c in case.required_checks if c not in report["checks"]]
    blocked_warn = [w for w in report["warnings"] if w in case.block_on_warn]
    passed = report["promote_ok"] and not missing and not blocked_warn
    return {"case_id": case.case_id, "passed": passed, "report": report, "missing": missing, "blocked_warn": blocked_warn}


RELEASE_SUITE = [
    ReleaseEvalCase("rf1", "rel-2025.07.10", ["unit_tests", "security_scan", "changelog"], block_on_warn=["security_scan"]),
]

release_eval = eval_release(RELEASE_SUITE[0], RELEASE_CHECKS)
print(pretty(release_eval))
print("Release blocked until security_scan warn cleared — same loop: fix → re-eval → promote")
````

## 16. Run Full Loop — One Function

Tie all phases into a single orchestrator you can call from CI or a nightly job.

````python
def run_quality_loop(
    cases: list[EvalCase],
    baseline_fn: Callable,
    candidate_fn: Callable,
    baseline_ver: str,
    candidate_ver: str,
) -> dict[str, Any]:
    loop_state = QualityLoopState(version=baseline_ver)

    offline = run_eval_suite(cases, baseline_fn, baseline_ver)
    gate = offline_ci_gate(offline)
    if not gate["deploy_ok"]:
        return {"status": "blocked_offline", "gate": gate}

    loop_state.advance(LoopPhase.OFFLINE_EVAL)
    canary = run_canary(cases, baseline_fn, candidate_fn, baseline_ver, candidate_ver)
    loop_state.advance(LoopPhase.CANARY)

    store = TraceStore()
    simulate_production_day(store, candidate_ver, LIVE_QUESTIONS[:5])
    online = aggregate_online(store.traces)
    loop_state.advance(LoopPhase.ONLINE)

    alert_list = evaluate_alerts(online, canary)
    loop_state.advance(LoopPhase.ALERT)

    return {
        "status": "promote" if canary.promote and not alert_list else "hold",
        "offline": offline,
        "canary": canary.__dict__,
        "online": online.__dict__,
        "alerts": [a.__dict__ for a in alert_list],
    }


loop_outcome = run_quality_loop(GOLDEN_SUITE, shopco_agent, shopco_agent_v11, AGENT_VERSION, CANARY_VERSION)
print(pretty(loop_outcome))
````

## 17. Human-in-the-Loop Review Queue

Not every failure should auto-enter the golden set. Queue mined cases for human approval with trace context.

````python
@dataclass
class ReviewItem:
    item_id: str
    trace_id: str
    question: str
    answer: str
    reviewer: str | None = None
    decision: str = "pending"  # pending | approve | reject


class ReviewQueue:
    def __init__(self) -> None:
        self.items: list[ReviewItem] = []

    def enqueue_mined(self, candidates: list[MinedCandidate]) -> None:
        for c in candidates:
            self.items.append(ReviewItem(f"rev-{len(self.items)+1}", c.trace_id, c.question, c.answer))

    def approve(self, item_id: str, reviewer: str) -> ReviewItem | None:
        item = next((i for i in self.items if i.item_id == item_id), None)
        if item:
            item.reviewer = reviewer
            item.decision = "approve"
        return item

    def pending(self) -> list[ReviewItem]:
        return [i for i in self.items if i.decision == "pending"]


queue = ReviewQueue()
queue.enqueue_mined(golden_ds.mined)
if queue.items:
    queue.approve(queue.items[0].item_id, "qa-lead")
print(f"Pending reviews: {len(queue.pending())}")
````

## 18. Optional Live LLM — Online Judge Sample

Set `USE_LIVE_LLM = True` to score a single production answer with `gpt-4o-mini`. Use sparingly — sample 1–5% of traffic, not every request.

````python
def llm_judge(question: str, answer: str) -> dict[str, Any]:
    from langchain_core.messages import HumanMessage, SystemMessage
    from langchain_openai import ChatOpenAI

    llm = ChatOpenAI(model="gpt-4o-mini", api_key=OPENAI_API_KEY, temperature=0)
    resp = llm.invoke([
        SystemMessage(content="Score helpfulness and grounding 1-5. Reply JSON {helpfulness, grounding, reason}."),
        HumanMessage(content=f"Q: {question}\nA: {answer}"),
    ])
    try:
        return json.loads(str(resp.content))
    except json.JSONDecodeError:
        return {"helpfulness": 0, "grounding": 0, "reason": "parse error"}


sample_q = "What is the return window?"
sample_a = shopco_agent(sample_q, AgentTracer("shopco-support", AGENT_VERSION))
if USE_LIVE_LLM:
    print(llm_judge(sample_q, sample_a))
else:
    print("Offline loop complete. Enable USE_LIVE_LLM for sampled online judging.")
````

## 19. Common Mistakes

| Mistake | Symptom | Fix |
|---------|---------|-----|
| Eval only at launch | Regressions reach users | Nightly offline suite in CI |
| No canary | Bad deploy hits 100% traffic | 5–10% slice with rollback |
| Traces without versions | Can't compare v1.1 vs v1.0 | Tag every trace with `version` |
| Auto-add all failures to golden set | Noisy, flaky tests | Human review queue |
| Online = offline metrics | Alert fatigue or blind spots | Align definitions; allow drift bands |
| Fix without changelog | Same bug returns | Record eval before/after per version |
| Ignore cost SLIs | Bill spikes unnoticed | `avg_cost_usd` in every window |

## 21. Summary

Agent quality is not a checkpoint — it is a **loop**. Instrument every run with versioned traces, gate deploys with offline eval, limit risk with canaries, watch online SLIs, alert on regression, mine failures into reviewed golden cases, fix with changelog discipline, and repeat.

This notebook implemented that loop end-to-end for **ShopCo Support Hub** and **ReleaseFlow** in plain Python: `AgentTracer`, `run_eval_suite`, `run_canary`, `aggregate_online`, `evaluate_alerts`, `GoldenDataset.mine_from_store`, and `run_quality_loop`.

The teams that win treat eval datasets as **living assets** and production traces as **continuous feedback** — not as separate tools used once before launch.
