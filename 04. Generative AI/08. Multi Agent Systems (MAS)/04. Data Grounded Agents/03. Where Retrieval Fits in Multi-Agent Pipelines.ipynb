# Where Retrieval Fits in Multi-Agent Pipelines

In a **multi-agent pipeline**, retrieval is not a single step you bolt on at the end — it is a **placement decision** that affects cost, grounding quality, and how agents hand off evidence.

Common placements:

| Placement | Pattern | Best when |
|-----------|---------|----------|
| **Pre-pipeline** | Fetch shared context before any agent runs | Every ticket needs same policy baseline |
| **Dedicated retriever agent** | Specialist stage owns search quality | Complex corpora, query rewriting |
| **Tool inside agent loop** | Model calls `search_policy` when needed | Variable intent, fewer wasted retrievals |
| **Supervisor-routed** | Router sends "policy" vs "order" branches | Mixed intents in one inbox |
| **Per-stage refresh** | Re-retrieve before validate/publish | Long pipelines, stale mid-flight risk |

This notebook implements **ShopCo Support Pipeline** — intake → retrieve → compose → validate — in plain Python, compares placement patterns on the same tickets, and shows anti-patterns to avoid.

Everything is self-contained. No frameworks or prior notebooks are required.

````python
"""
ShopCo Support Pipeline — retrieval placement lab.
"""

import json
import os
import re
import uuid
from dataclasses import dataclass, field
from datetime import datetime, timezone
from enum import Enum
from typing import Any, Callable, Literal

OPENAI_API_KEY = os.environ.get("OPENAI_API_KEY", "sk-proj-placeholder-replace-with-your-real-key")
os.environ.setdefault("OPENAI_API_KEY", OPENAI_API_KEY)

USE_LIVE_LLM = False

RetrievalPlacement = Literal[
    "pre_pipeline", "retriever_agent", "tool_in_loop", "supervisor_routed"
]


def pretty(obj: Any) -> str:
    return json.dumps(obj, indent=2, default=str)


POLICY_DOCS: list[dict[str, str]] = [
    {
        "id": "pol-returns-01",
        "title": "Returns Policy",
        "text": "Customers may return items within 30 days of delivery. Items must be unused and in original packaging. Refunds in 5-7 business days. Final-sale items cannot be returned.",
    },
    {
        "id": "pol-shipping-02",
        "title": "Shipping Policy",
        "text": "Free standard shipping on orders over $50. Express shipping at checkout. Tracking emailed at carrier scan.",
    },
    {
        "id": "pol-billing-03",
        "title": "Billing Policy",
        "text": "Charges appear as SHOPCO* on statements. Partial refunds may take two billing cycles. Disputes require order ID.",
    },
    {
        "id": "faq-tracking-04",
        "title": "Tracking FAQ",
        "text": "Shipped orders have tracking numbers. Processing orders do not yet have tracking.",
    },
]

ORDERS_DB: dict[str, dict[str, Any]] = {
    "ORD-1001": {"order_id": "ORD-1001", "status": "shipped", "tracking": "1Z999AA10123456784", "total_usd": 89.50},
    "ORD-1002": {"order_id": "ORD-1002", "status": "processing", "tracking": None, "total_usd": 42.00},
}

print(f"Corpus: {len(POLICY_DOCS)} docs | Orders: {len(ORDERS_DB)}")
````

## 1. Pipeline Topology with Retrieval

```
                    ┌─────────────────────────────────────────┐
  ticket ──────────►│           ShopCo Support Pipeline        │
                    │                                         │
                    │  Intake ──► Retrieve ──► Compose ──► Validate │
                    │     │          ▲                           │

                    │     └──► order lookup (structured, not RAG)│

```

**Retrieval** feeds **unstructured policy prose**. **Structured order lookup** is a separate path — do not force SQL rows through a vector index.

## 2. Shared Pipeline State and Evidence Bundle

Multi-agent pipelines pass an **evidence bundle** forward — retrieved chunks, query used, retrieval timestamp — not just the final answer text.

````python
@dataclass
class EvidenceChunk:
    doc_id: str
    title: str
    text: str
    score: float
    retrieved_by: str
    retrieved_at: str = field(default_factory=lambda: datetime.now(timezone.utc).isoformat())


@dataclass
class PipelineState:
    run_id: str
    ticket_id: str
    question: str
    intent: str = ""
    order_id: str | None = None
    evidence: list[EvidenceChunk] = field(default_factory=list)
    order_facts: dict[str, Any] | None = None
    draft_answer: str = ""
    final_answer: str = ""
    status: str = "running"
    retrieval_count: int = 0
    trace: list[str] = field(default_factory=list)
    errors: list[str] = field(default_factory=list)

    def log(self, step: str) -> None:
        self.trace.append(step)

    def evidence_ids(self) -> list[str]:
        return [e.doc_id for e in self.evidence]


STEP_CONTRACTS: dict[str, dict[str, list[str]]] = {
    "intake": {"reads": ["question"], "writes": ["intent", "order_id"]},
    "retrieve": {"reads": ["question", "intent"], "writes": ["evidence", "retrieval_count"]},
    "order_lookup": {"reads": ["order_id"], "writes": ["order_facts"]},
    "compose": {"reads": ["question", "evidence", "order_facts"], "writes": ["draft_answer"]},
    "validate": {"reads": ["draft_answer", "evidence"], "writes": ["final_answer", "status"]},
}


def validate_contract(stage: str, state: PipelineState) -> list[str]:
    contract = STEP_CONTRACTS.get(stage, {})
    issues: list[str] = []
    for req in contract.get("reads", []):
        if req == "question" and not state.question:
            issues.append(f"{stage}: missing question")
        if req == "intent" and not state.intent:
            issues.append(f"{stage}: missing intent")
        if req == "evidence" and not state.evidence:
            issues.append(f"{stage}: missing evidence")
        if req == "order_id" and not state.order_id:
            issues.append(f"{stage}: missing order_id")
        if req == "order_facts" and state.order_facts is None and state.order_id:
            issues.append(f"{stage}: order lookup required but order_facts empty")
        if req == "draft_answer" and not state.draft_answer:
            issues.append(f"{stage}: missing draft_answer")
    return issues


print("PipelineState + contracts ready.")
````

---

## 3. Core Retriever (Shared by All Placements)

````python
def retrieve_docs(query: str, top_k: int = 2) -> list[EvidenceChunk]:
    terms = [t for t in re.split(r"\W+", query.lower()) if len(t) > 2]
    scored: list[tuple[float, dict[str, str]]] = []
    for doc in POLICY_DOCS:
        haystack = f"{doc['title']} {doc['text']}".lower()
        score = sum(1 for term in terms if term in haystack) if terms else 0.0
        if score > 0:
            scored.append((float(score), doc))
    scored.sort(key=lambda x: (-x[0], x[1]["id"]))
    return [
        EvidenceChunk(
            doc_id=d["id"], title=d["title"], text=d["text"],
            score=s, retrieved_by="keyword_retriever",
        )
        for s, d in scored[:top_k]
    ]


demo_hits = retrieve_docs("return policy refund")
print("Retriever hits:", [h.doc_id for h in demo_hits])
````

## 4. Pipeline Stage Implementations

````python
def intake_stage(state: PipelineState) -> PipelineState:
    q = state.question.lower()
    has_order = bool(re.search(r"ord-[0-9]{4}", q))
    policy_terms = ("return", "refund", "shipping", "billing", "track", "policy")
    has_policy = any(t in q for t in policy_terms)

    if has_order and has_policy:
        state.intent = "mixed"
    elif has_order:
        state.intent = "order_status"
    elif has_policy:
        state.intent = "policy"
    else:
        state.intent = "general"

    m = re.search(r"ORD-[0-9]{4}", state.question.upper())
    state.order_id = m.group(0) if m else None
    state.log(f"intake:intent={state.intent},order={state.order_id}")
    return state


def retriever_agent_stage(state: PipelineState) -> PipelineState:
    issues = validate_contract("retrieve", state)
    if issues and state.intent == "order_status":
        state.log("retrieve:skipped(order_only)")
        return state

    query = state.question
    if state.intent == "mixed":
        query = "return refund policy"  # retriever agent may rewrite query

    chunks = retrieve_docs(query)
    state.evidence.extend(chunks)
    state.retrieval_count += 1
    state.log(f"retrieve:hits={len(chunks)},query={query[:40]}")
    return state


def order_lookup_stage(state: PipelineState) -> PipelineState:
    if not state.order_id:
        state.log("order_lookup:skipped")
        return state
    state.order_facts = ORDERS_DB.get(state.order_id)
    state.log(f"order_lookup:found={state.order_facts is not None}")
    return state


def compose_stage(state: PipelineState) -> PipelineState:
    parts: list[str] = []
    if state.evidence:
        top = state.evidence[0]
        parts.append(f"Policy [{top.doc_id}]: {top.text[:100]}...")
    if state.order_facts:
        parts.append(
            f"Order {state.order_facts['order_id']} status: {state.order_facts['status']}."
        )
        if state.order_facts.get("tracking"):
            parts.append(f"Tracking: {state.order_facts['tracking']}.")
    if not parts:
        state.draft_answer = "I need more details about your ShopCo order or policy question."
    else:
        state.draft_answer = " ".join(parts)
    state.log("compose:draft_ready")
    return state


def validate_stage(state: PipelineState) -> PipelineState:
    issues: list[str] = []
    if state.intent in ("policy", "mixed") and not state.evidence:
        issues.append("policy intent but no retrieved evidence")
    if state.intent in ("order_status", "mixed") and state.order_id and not state.order_facts:
        issues.append("order id provided but lookup failed")

    cite_ids = set(state.evidence_ids())
    for chunk in state.evidence:
        if chunk.doc_id not in cite_ids:
            issues.append(f"orphan evidence {chunk.doc_id}")

    if issues:
        state.status = "failed_validation"
        state.errors.extend(issues)
        state.final_answer = "Unable to verify a complete answer. Escalating to human support."
    else:
        state.status = "completed"
        cites = ", ".join(f"[{c.doc_id}]" for c in state.evidence) if state.evidence else ""
        state.final_answer = state.draft_answer + (f" {cites}" if cites else "")
    state.log(f"validate:{state.status}")
    return state
````

## 5. Pattern A — Pre-Pipeline Retrieval

Fetch baseline policies **before** intake classifies the ticket. Cheap when every conversation needs the same compliance context.

````python
def run_pre_pipeline(ticket_id: str, question: str) -> PipelineState:
    state = PipelineState(run_id=str(uuid.uuid4()), ticket_id=ticket_id, question=question)

    # PRE-PIPELINE: global prefetch
    baseline = retrieve_docs("returns shipping billing", top_k=2)
    state.evidence.extend(baseline)
    state.retrieval_count += 1
    state.log(f"pre_pipeline:prefetch={len(baseline)}")

    state = intake_stage(state)
    state = order_lookup_stage(state)
    # Skip dedicated retriever — already prefetched
    state = compose_stage(state)
    state = validate_stage(state)
    return state


pre = run_pre_pipeline("T-101", "What is your return policy?")
print("Pre-pipeline:", pre.final_answer[:80], "...")
print("Retrievals:", pre.retrieval_count, "| Evidence:", pre.evidence_ids())
````

---

## 6. Pattern B — Dedicated Retriever Agent (Sequential Stage)

A **retriever agent** sits after intake so queries can be **rewritten** from intent. This is the default for auditable support pipelines.

````python
def run_retriever_agent_pipeline(ticket_id: str, question: str) -> PipelineState:
    state = PipelineState(run_id=str(uuid.uuid4()), ticket_id=ticket_id, question=question)
    state = intake_stage(state)
    state = retriever_agent_stage(state)
    state = order_lookup_stage(state)
    state = compose_stage(state)
    state = validate_stage(state)
    return state


seq = run_retriever_agent_pipeline("T-102", "Return ORD-1001 — changed my mind")
print("Retriever-agent pipeline:", seq.final_answer[:100], "...")
print("Trace:", seq.trace)
````

## 7. Pattern C — Retrieval as a Tool Inside an Agent Loop

Instead of a fixed pipeline stage, the **composer agent** calls `search_policy` only when its plan requires it — fewer retrievals for pure order-status tickets.

````python
def search_policy_tool(query: str, state: PipelineState) -> list[EvidenceChunk]:
    chunks = retrieve_docs(query)
    state.retrieval_count += 1
    state.log(f"tool:search_policy:hits={len(chunks)}")
    return chunks


def run_tool_in_loop_pipeline(ticket_id: str, question: str) -> PipelineState:
    state = PipelineState(run_id=str(uuid.uuid4()), ticket_id=ticket_id, question=question)
    state = intake_stage(state)
    state = order_lookup_stage(state)

    # Agent loop: decide whether to call retrieval tool
    if state.intent in ("policy", "mixed", "general"):
        chunks = search_policy_tool(state.question, state)
        state.evidence.extend(chunks)
    else:
        state.log("tool_loop:skip_retrieval(order_only)")

    state = compose_stage(state)
    state = validate_stage(state)
    return state


tool_run = run_tool_in_loop_pipeline("T-103", "Where is order ORD-1001?")
print("Tool-in-loop retrievals:", tool_run.retrieval_count)
print("Answer:", tool_run.final_answer)
````

## 8. Pattern D — Supervisor-Routed Retrieval

A **supervisor** routes to retrieval branch vs order branch vs both — useful for mixed inboxes with heterogeneous intents.

````python
def supervisor_route(state: PipelineState) -> list[str]:
    routes: list[str] = []
    if state.intent in ("policy", "mixed", "general"):
        routes.append("retrieve")
    if state.intent in ("order_status", "mixed") and state.order_id:
        routes.append("order_lookup")
    if not routes:
        routes.append("compose_only")
    state.log(f"supervisor:routes={routes}")
    return routes


def run_supervisor_pipeline(ticket_id: str, question: str) -> PipelineState:
    state = PipelineState(run_id=str(uuid.uuid4()), ticket_id=ticket_id, question=question)
    state = intake_stage(state)
    for route in supervisor_route(state):
        if route == "retrieve":
            state = retriever_agent_stage(state)
        elif route == "order_lookup":
            state = order_lookup_stage(state)
    state = compose_stage(state)
    state = validate_stage(state)
    return state


sup = run_supervisor_pipeline("T-104", "Return policy for ORD-1002 and its status")
print("Supervisor routes in trace:", [t for t in sup.trace if "supervisor" in t])
print("Evidence + order:", sup.evidence_ids(), sup.order_facts)
````

## 10. When to Place Retrieval — Decision Guide

| Signal | Recommended placement |
|--------|----------------------|
| Every answer needs compliance footer | **Pre-pipeline** prefetch |
| Query rewriting / hyDE / multi-hop search | **Dedicated retriever agent** |
| Intent varies widely; many order-only tickets | **Tool in loop** or **supervisor skip** |
| Mixed policy + order in one inbox | **Supervisor-routed** parallel branches |
| Validator keeps failing grounding | **Per-stage refresh** before validate |
| Long pipeline (>5 min) | Re-retrieve if evidence TTL expired |

## 11. Anti-Pattern — Every Agent Retrieves Independently

Duplicated retrieval wastes latency, costs, and can return **inconsistent chunks** to composer vs validator.

````python
def bad_every_agent_retrieves(state: PipelineState) -> PipelineState:
    """Anti-pattern: intake, compose, and validate each call retriever separately."""
    for stage in ("intake", "compose", "validate"):
        hits = retrieve_docs(state.question)
        state.evidence.extend(hits)
        state.retrieval_count += 1
        state.log(f"BAD:{stage}:redundant_retrieve={len(hits)}")
    return state


bad_state = PipelineState(run_id="x", ticket_id="BAD-1", question="return policy")
bad_state = bad_every_agent_retrieves(bad_state)
print(f"Redundant retrievals: {bad_state.retrieval_count}")
print(f"Duplicate evidence entries: {len(bad_state.evidence)} (should be 1 call)")
````

---

## 12. Anti-Pattern — Retrieve After Compose

Validation cannot fix hallucinations already baked into the draft.

````python
def bad_late_retrieval(state: PipelineState) -> PipelineState:
    state = intake_stage(state)
    state.draft_answer = "You can return items within 60 days for a full refund."  # hallucinated
    state.log("compose:unguarded_draft")
    # Retrieve too late — draft already wrong
    state.evidence = retrieve_docs(state.question)
    state.retrieval_count += 1
    state.log("retrieve:too_late")
    state = validate_stage(state)
    return state


late = PipelineState(run_id="y", ticket_id="BAD-2", question="return policy")
late.intent = "policy"
late = bad_late_retrieval(late)
print("Draft still wrong:", "60 days" in late.draft_answer)
print("Evidence says 30 days:", any("30 days" in e.text for e in late.evidence))
````

## 13. Good Pattern — Single Evidence Bundle, Shared Forward

Retrieve once (or via supervisor), store in `state.evidence`, pass through compose **and** validate.

````python
def good_shared_evidence_pipeline(ticket_id: str, question: str) -> PipelineState:
    state = PipelineState(run_id=str(uuid.uuid4()), ticket_id=ticket_id, question=question)
    state = intake_stage(state)
    state = retriever_agent_stage(state)
    state = order_lookup_stage(state)
    state = compose_stage(state)
    # Validator reads SAME evidence bundle
    assert all(c.doc_id in state.evidence_ids() for c in state.evidence)
    state = validate_stage(state)
    return state


good = good_shared_evidence_pipeline("T-200", "return policy")
print("Single retrieval:", good.retrieval_count)
print("Validated:", good.status)
````

## 14. Retrieval in Multi-Agent Topologies

| Topology | Typical retrieval placement |
|----------|----------------------------|
| **Sequential** | Dedicated stage after intake |
| **Supervisor** | Router triggers retrieve branch |
| **Group chat** | One "researcher" role owns retrieval |
| **Nested** | Sub-pipeline retriever inside research subgraph |
| **ReAct single agent** | Retrieval as tool in loop |

The rule is consistent: **one owner** for evidence quality, **one bundle** passed forward.

## 15. Observability — Retrieval Metrics per Run

````python
@dataclass
class RetrievalMetrics:
    retrieval_calls: int
    unique_docs: int
    placement: str
    latency_ms_simulated: int
    grounded: bool


def metrics_from_state(state: PipelineState, placement: str) -> RetrievalMetrics:
    return RetrievalMetrics(
        retrieval_calls=state.retrieval_count,
        unique_docs=len(set(state.evidence_ids())),
        placement=placement,
        latency_ms_simulated=state.retrieval_count * 45,
        grounded=bool(state.evidence) or state.intent == "order_status",
    )


for name, runner in PLACEMENT_RUNNERS.items():
    s = runner("M-1", "return ORD-1001")
    m = metrics_from_state(s, name)
    print(f"{m.placement:<18} calls={m.retrieval_calls} docs={m.unique_docs} ~{m.latency_ms_simulated}ms")
````

## 16. Golden-Set Eval — Grounding by Placement

````python
GOLDEN = [
    {"q": "What is the return window?", "expect_doc": "pol-returns", "expect_fact": "30"},
    {"q": "Where is ORD-1001?", "expect_fact": "shipped", "needs_order": True},
    {"q": "Return policy for ORD-1002", "expect_doc": "pol-returns", "expect_fact": "processing"},
]


def eval_placement(runner) -> float:
    hits = 0
    for i, row in enumerate(GOLDEN):
        s = runner(f"G-{i}", row["q"])
        doc_ok = True
        if "expect_doc" in row:
            doc_ok = any(row["expect_doc"] in eid for eid in s.evidence_ids())
        fact_ok = row["expect_fact"].lower() in s.final_answer.lower() or (
            s.order_facts and row["expect_fact"].lower() in pretty(s.order_facts).lower()
        )
        if doc_ok and fact_ok and s.status == "completed":
            hits += 1
    return hits / len(GOLDEN)


print("Grounding scores by placement:")
for name, runner in PLACEMENT_RUNNERS.items():
    print(f"  {name:<18} {eval_placement(runner):.0%}")
````

## 17. Common Mistakes

| Mistake | Symptom | Fix |
|---------|---------|-----|
| Retrieve in every agent | 3× latency, inconsistent chunks | Single retriever owner |
| Retrieve after draft | Validator can't un-hallucinate | Retrieve before compose |
| No evidence in state | Can't audit compose step | `EvidenceChunk` in `PipelineState` |
| RAG for order rows | Wrong status from prose | Structured lookup stage |
| Skip retrieval on mixed intent | Policy part hallucinated | Supervisor routes both branches |
| No retrieval metrics | Silent regressions | Log calls, docs, scores per run |

## 18. Optional Live LLM Compose Stage

Set `USE_LIVE_LLM = True` to synthesize `compose_stage` answers from the evidence bundle via `gpt-4o-mini`. Retrieval placement stays identical.

````python
def live_compose(state: PipelineState) -> str:
    from langchain_core.messages import HumanMessage, SystemMessage
    from langchain_openai import ChatOpenAI

    evidence = "\n".join(f"[{e.doc_id}] {e.text}" for e in state.evidence)
    order = pretty(state.order_facts) if state.order_facts else "(none)"
    llm = ChatOpenAI(model="gpt-4o-mini", api_key=OPENAI_API_KEY, temperature=0)
    resp = llm.invoke([
        SystemMessage(content="Answer from evidence only. Cite doc ids."),
        HumanMessage(content=f"Q: {state.question}\nEvidence:\n{evidence}\nOrder:\n{order}"),
    ])
    return str(resp.content)


if USE_LIVE_LLM:
    s = run_retriever_agent_pipeline("LIVE-1", "What is the return policy for ORD-1001?")
    s.draft_answer = live_compose(s)
    s = validate_stage(s)
    print(s.final_answer)
else:
    s = run_retriever_agent_pipeline("DEMO", "What is the return window?")
    print("Offline pipeline answer:", s.final_answer)
    print("Set USE_LIVE_LLM = True for LLM compose from evidence bundle.")
````

## 20. Summary

Retrieval placement is an **orchestration choice**, not a retrieval-algorithm choice:

- **Pre-pipeline** — shared baseline context
- **Retriever agent** — query rewriting, auditable stage
- **Tool in loop** — skip retrieval when intent is order-only
- **Supervisor-routed** — mixed intents get parallel evidence paths

ShopCo Support Pipeline showed intake → retrieve → compose → validate with a **single evidence bundle**, contracts per stage, and metrics per placement. Keep one retrieval owner, retrieve **before** synthesis, and separate **RAG** from **structured order lookup**.
