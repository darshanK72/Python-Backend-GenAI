# RAG vs Fine-Tuning vs Long Context for Agents

When you build a **data-grounded agent**, you must choose how private knowledge reaches the model. Three common strategies dominate production discussions:

| Strategy | Core idea | Best at |
|----------|-----------|--------|
| **RAG** | Retrieve relevant chunks at query time | Fresh facts, citations, large corpora |
| **Fine-tuning** | Bake patterns into model weights | Tone, format, domain jargon |
| **Long context** | Put documents directly in the prompt | Small stable corpora, single-shot Q&A |

They are **not mutually exclusive**. Teams often fine-tune for *how* the agent speaks, RAG for *what* it knows today, and long context for *one-off* large attachments.

This notebook compares all three on **ShopCo Support** — same questions, same corpus, different grounding strategies — with a cost model, a policy-update scenario, and a decision scorecard. Everything runs offline in plain Python.

````python
"""
ShopCo Support — RAG vs fine-tuning vs long-context comparison lab.
"""

import json
import os
import re
from dataclasses import dataclass, field
from datetime import datetime, timezone
from enum import Enum
from typing import Any, Literal

OPENAI_API_KEY = os.environ.get("OPENAI_API_KEY", "sk-proj-placeholder-replace-with-your-real-key")
os.environ.setdefault("OPENAI_API_KEY", OPENAI_API_KEY)

USE_LIVE_LLM = False

StrategyName = Literal["rag", "fine_tuning", "long_context", "hybrid"]


def pretty(obj: Any) -> str:
    return json.dumps(obj, indent=2, default=str)


def estimate_tokens(text: str) -> int:
    """Rough token estimate (~4 chars per token) for cost modeling."""
    return max(1, len(text) // 4)


# --- Live knowledge corpus (mutable — simulates CMS / policy store) ---

POLICY_DOCS: list[dict[str, str]] = [
    {
        "id": "pol-returns-01",
        "title": "Returns Policy",
        "text": "Customers may return items within 30 days of delivery. Items must be unused and in original packaging. Refunds are processed within 5-7 business days after warehouse receipt. Final-sale items cannot be returned.",
    },
    {
        "id": "pol-shipping-02",
        "title": "Shipping Policy",
        "text": "Standard shipping is free on orders over $50. Express shipping is available at checkout. Tracking numbers are emailed when the carrier scans the package.",
    },
    {
        "id": "pol-billing-03",
        "title": "Billing Policy",
        "text": "Charges appear as SHOPCO* on credit card statements. Partial refunds may take up to two billing cycles to appear. Contact support with order ID for billing disputes.",
    },
    {
        "id": "faq-tracking-04",
        "title": "Tracking FAQ",
        "text": "If order status is shipped, use the tracking number from your confirmation email. Processing orders do not yet have tracking numbers.",
    },
]

# --- Simulated fine-tuned weights (frozen at training time) ---

FINE_TUNED_KNOWLEDGE: dict[str, str] = {
    "return_window_days": "30",  # stale after policy update demo
    "refund_days": "5-7 business days",
    "free_shipping_threshold": "50",
    "tone": "warm, concise, always greet by first name if known",
    "format": "bullet points for policy answers; include disclaimer on final-sale items",
}

FINE_TUNED_TRAINED_AT = "2026-01-15"

LONG_CONTEXT_LIMIT_TOKENS = 800  # simulated model window slice for corpus

print(f"Corpus: {len(POLICY_DOCS)} docs, ~{sum(estimate_tokens(d['text']) for d in POLICY_DOCS)} tokens")
````

## 1. The Three Strategies — Mental Models

```
  RAG                         Fine-tuning                  Long context
  ───                         ───────────                  ────────────
  Query ──► Retriever         Weights already              Query + FULL corpus
              │                 contain patterns               in one prompt
              ▼                      │                              │
         top-k chunks               ▼                              ▼
              │                 Model answers                 Model answers
              ▼                 (no live index)               (no retriever)
         Model + chunks
```

| Question | RAG | Fine-tuning | Long context |
|----------|-----|-------------|--------------|
| Where do facts live? | External index | Model weights | Prompt text |
| Updates when policy changes? | Re-index docs | Retrain / redeploy | Edit prompt corpus |
| Citations possible? | Yes (chunk IDs) | No native provenance | Yes (if you tag sections) |
| Scales to 10M docs? | Yes (with good retrieval) | No | No |

---

## 2. Comparison Dimensions for Agents

Agents add requirements beyond single-shot Q&A:

| Dimension | RAG | Fine-tuning | Long context |
|-----------|-----|-------------|--------------|
| **Freshness** | High (index refresh) | Low (retrain lag) | Medium (prompt update) |
| **Per-turn cost** | Retrieval + smaller prompt | Smallest prompt | Largest prompt |
| **Latency** | +retriever RTT | Lowest | High with big corpus |
| **Auditability** | Strong (chunk cites) | Weak | Medium |
| **Multi-turn agents** | Re-retrieve each turn | Style persists in weights | Context grows fast |
| **Tool loops** | Natural (retrieve tool) | Orthogonal | Competes for window |
| **Failure mode** | Wrong chunks retrieved | Stale/wrong baked facts | Context rot, truncation |

## 3. Shared Result Type

````python
@dataclass
class StrategyResult:
    strategy: StrategyName
    answer: str
    citations: list[str] = field(default_factory=list)
    prompt_tokens: int = 0
    grounded: bool = False
    stale_risk: bool = False
    notes: list[str] = field(default_factory=list)


print("StrategyResult ready.")
````

---

## 4. Strategy A — RAG (Retrieval-Augmented Generation)

At query time, search an index, inject top-k chunks, answer with citations. Production agents often expose retrieval as a **tool** inside a ReAct loop.

````python
def retrieve_docs(query: str, corpus: list[dict[str, str]], top_k: int = 2) -> list[dict[str, str]]:
    terms = [t for t in re.split(r"\W+", query.lower()) if len(t) > 2]
    scored: list[tuple[int, dict[str, str]]] = []
    for doc in corpus:
        haystack = f"{doc['title']} {doc['text']}".lower()
        score = sum(1 for term in terms if term in haystack) if terms else 0
        if score > 0:
            scored.append((score, doc))
    scored.sort(key=lambda x: (-x[0], x[1]["id"]))
    return [{"id": d["id"], "title": d["title"], "text": d["text"]} for _, d in scored[:top_k]]


def rag_strategy(question: str, corpus: list[dict[str, str]] | None = None) -> StrategyResult:
    corpus = corpus or POLICY_DOCS
    hits = retrieve_docs(question, corpus)
    if not hits:
        return StrategyResult(
            strategy="rag",
            answer="I could not find matching ShopCo policy documents.",
            grounded=False,
            notes=["retrieval_miss"],
        )

    citations = [h["id"] for h in hits]
    evidence = "\n".join(f"[{h['id']}] {h['text']}" for h in hits)
    prompt = f"System: Answer from evidence only.\nEvidence:\n{evidence}\nUser: {question}"

    q = question.lower()
    if "return" in q or "refund" in q:
        days = re.search(r"(\d+) days", hits[0]["text"])
        window = days.group(1) if days else "30"
        answer = f"Returns are allowed within {window} days of delivery. [{hits[0]['id']}]"
    elif "shipping" in q or "free" in q:
        answer = f"Free standard shipping applies on orders over $50. [{hits[0]['id']}]"
    else:
        answer = f"Per ShopCo policy: {hits[0]['text'][:100]}... [{hits[0]['id']}]"

    return StrategyResult(
        strategy="rag",
        answer=answer,
        citations=citations,
        prompt_tokens=estimate_tokens(prompt),
        grounded=True,
        notes=[f"retrieved {len(hits)} chunks"],
    )


demo_rag = rag_strategy("What is the return window?")
print(demo_rag.answer)
print("Citations:", demo_rag.citations, "| Tokens:", demo_rag.prompt_tokens)
````

## 5. Strategy B — Fine-Tuning (Simulated)

Fine-tuning updates **weights** on example conversations. It excels at **style and structure**, not live order databases.

We simulate a fine-tuned model as a responder that reads `FINE_TUNED_KNOWLEDGE` — frozen at `FINE_TUNED_TRAINED_AT`. When policies change, answers become **stale** until you retrain.

````python
def fine_tuning_strategy(question: str) -> StrategyResult:
    q = question.lower()
    tone = FINE_TUNED_KNOWLEDGE["tone"]
    notes = [f"weights from {FINE_TUNED_TRAINED_AT}", f"tone: {tone}"]

    if "return" in q or "refund" in q:
        days = FINE_TUNED_KNOWLEDGE["return_window_days"]
        refund = FINE_TUNED_KNOWLEDGE["refund_days"]
        answer = (
            f"Hi there! • Return window: {days} days • Refunds in {refund} "
            f"• Final-sale items excluded (per fine-tuned policy memory)"
        )
    elif "shipping" in q:
        threshold = FINE_TUNED_KNOWLEDGE["free_shipping_threshold"]
        answer = f"Hi there! • Free shipping over ${threshold} • Express available at checkout"
    else:
        answer = "Hi there! I'm happy to help with ShopCo orders and policies."

    # Small prompt — knowledge is in weights, not context
    prompt = f"System: ShopCo fine-tuned assistant.\nUser: {question}"

    return StrategyResult(
        strategy="fine_tuning",
        answer=answer,
        citations=[],  # no native provenance
        prompt_tokens=estimate_tokens(prompt),
        grounded=True,  # confident but not citeable
        stale_risk=True,
        notes=notes,
    )


demo_ft = fine_tuning_strategy("What is the return window?")
print(demo_ft.answer)
print("Citations:", demo_ft.citations or "none", "| Stale risk:", demo_ft.stale_risk)
````

## 6. Strategy C — Long Context (Full Corpus in Prompt)

Stuff documents into the context window — no retriever. Works when the corpus is **small and stable**. Cost and attention quality degrade as the corpus grows.

````python
def build_full_corpus_block(corpus: list[dict[str, str]]) -> str:
    return "\n\n".join(f"## {d['id']} — {d['title']}\n{d['text']}" for d in corpus)


def long_context_strategy(
    question: str,
    corpus: list[dict[str, str]] | None = None,
    token_limit: int = LONG_CONTEXT_LIMIT_TOKENS,
) -> StrategyResult:
    corpus = corpus or POLICY_DOCS
    block = build_full_corpus_block(corpus)
    corpus_tokens = estimate_tokens(block)
    truncated = corpus_tokens > token_limit

    if truncated:
        # simulate truncation — later docs dropped (context rot risk)
        kept: list[dict[str, str]] = []
        running = 0
        for doc in corpus:
            t = estimate_tokens(doc["text"])
            if running + t > token_limit:
                break
            kept.append(doc)
            running += t
        block = build_full_corpus_block(kept)
        notes = [f"truncated corpus {corpus_tokens}→{estimate_tokens(block)} tokens", f"kept {len(kept)}/{len(corpus)} docs"]
    else:
        notes = [f"full corpus {corpus_tokens} tokens in prompt"]

    prompt = f"System: Answer only from documents below.\n{block}\nUser: {question}"
    q = question.lower()

    if "return" in q:
        for doc in corpus:
            if "return" in doc["title"].lower():
                m = re.search(r"(\d+) days", doc["text"])
                days = m.group(1) if m else "?"
                answer = f"Return window is {days} days. [{doc['id']}]"
                break
        else:
            answer = "Return policy not found in provided documents (possibly truncated)."
    elif "billing" in q and truncated:
        answer = "Billing policy may be missing — it was truncated from the context window."
    else:
        answer = f"Based on the full ShopCo document pack in context: see policies above."

    return StrategyResult(
        strategy="long_context",
        answer=answer,
        citations=[d["id"] for d in corpus if d["id"] in block],
        prompt_tokens=estimate_tokens(prompt),
        grounded=not truncated or "missing" not in answer.lower(),
        stale_risk=False,
        notes=notes,
    )


demo_lc = long_context_strategy("What is the return window?")
print(demo_lc.answer)
print("Tokens:", demo_lc.prompt_tokens, "| Notes:", demo_lc.notes)
````

## 8. Policy Update Scenario — Freshness Test

ShopCo legal **shortens returns from 30 → 14 days**. Which strategies pick up the change immediately?

````python
UPDATED_CORPUS = [dict(d) for d in POLICY_DOCS]
for doc in UPDATED_CORPUS:
    if doc["id"] == "pol-returns-01":
        doc["text"] = doc["text"].replace("30 days", "14 days")

q = "What is the return window?"

after_rag = rag_strategy(q, corpus=UPDATED_CORPUS)
after_ft = fine_tuning_strategy(q)  # weights still say 30
after_lc = long_context_strategy(q, corpus=UPDATED_CORPUS)

print("After policy change (30 → 14 days):\n")
for label, r in [("RAG", after_rag), ("Fine-tuning", after_ft), ("Long context", after_lc)]:
    has_14 = "14" in r.answer
    has_30 = "30" in r.answer
    fresh = "✓ fresh" if has_14 and not has_30 else ("✗ stale" if has_30 else "? unclear")
    print(f"  {label:<14} {fresh:<10} {r.answer[:55]}...")
````

## 9. Cost Model — Per-Ticket Estimates

Simplified cost: `tokens × price_per_1k`. Retrieval infrastructure adds fixed cost omitted here.

````python
PRICE_PER_1K_INPUT = 0.00015  # illustrative gpt-4o-mini-class pricing


@dataclass
class CostEstimate:
    strategy: str
    prompt_tokens: int
    cost_usd: float
    amortized_training: float = 0.0


def estimate_cost(result: StrategyResult, tickets_per_month: int = 10_000) -> CostEstimate:
    inference = (result.prompt_tokens / 1000) * PRICE_PER_1K_INPUT
    training = 0.0
    if result.strategy == "fine_tuning":
        # illustrative: $200 training amortized over monthly tickets
        training = 200.0 / tickets_per_month
    return CostEstimate(result.strategy, result.prompt_tokens, inference, training)


sample_q = "What is the return window?"
print(f"{'Strategy':<14} {'Tokens':>7} {'Inference':>10} {'Train/ticket':>12}")
print("-" * 48)
for fn in STRATEGIES.values():
    r = fn(sample_q)
    c = estimate_cost(r)
    print(f"{c.strategy:<14} {c.prompt_tokens:>7} ${c.cost_usd:>8.5f} ${c.amortized_training:>10.5f}")
````

## 10. Corpus Scale — When Long Context Breaks

Simulate adding FAQ documents until the corpus exceeds the token budget.

````python
def generate_faq_corpus(n: int) -> list[dict[str, str]]:
    base = list(POLICY_DOCS)
    for i in range(n):
        base.append({
            "id": f"faq-extra-{i:03d}",
            "title": f"FAQ Topic {i}",
            "text": f"Synthetic FAQ entry {i}. " * 20,
        })
    return base


for n in [0, 5, 20, 50]:
    corpus = generate_faq_corpus(n)
    r = long_context_strategy("What is the billing policy?", corpus=corpus)
    total_tokens = sum(estimate_tokens(d["text"]) for d in corpus)
    print(f"+{n} extra FAQs | corpus ~{total_tokens} tokens | prompt {r.prompt_tokens} | grounded={r.grounded}")
    if r.notes:
        print(f"  note: {r.notes[0]}")
````

## 11. Hybrid Pattern — Fine-Tune Style + RAG Facts

Production sweet spot for many support agents:

- **Fine-tune** (or system prompt) for tone, bullet format, escalation phrasing
- **RAG** for policy facts that change quarterly
- **Tools/SQL** for live order rows (not covered by any of the three alone)

````python
def hybrid_strategy(question: str, corpus: list[dict[str, str]] | None = None) -> StrategyResult:
    rag = rag_strategy(question, corpus)
    tone_prefix = "Hi there! "
    format_suffix = " (Final-sale exclusions apply.)"
    answer = tone_prefix + rag.answer
    if "return" in question.lower():
        answer += format_suffix
    return StrategyResult(
        strategy="hybrid",
        answer=answer,
        citations=rag.citations,
        prompt_tokens=rag.prompt_tokens + 20,
        grounded=rag.grounded,
        notes=["RAG facts + fine-tuned tone/format"],
    )


hybrid = hybrid_strategy("What is the return window?")
print(hybrid.answer)
print("Notes:", hybrid.notes)
````

## 12. Agent-Specific Considerations

| Agent pattern | RAG | Fine-tuning | Long context |
|---------------|-----|-------------|--------------|
| **ReAct tool loop** | `search_policy` tool per turn | Style in weights; facts via tools anyway | Eats window each turn |
| **Multi-agent pipeline** | Researcher agent owns retriever | Writer agent inherits tone from FT model | Hard to share 200k ctx across agents |
| **Checkpoint resume** | Re-fetch or cache retrieval in state | N/A | Serialized prompt huge |
| **Eval / regression** | Golden set + chunk IDs | Harder — no cites | Prompt versioning |

For **agents that call tools repeatedly**, RAG as a tool usually beats stuffing everything in context.

## 13. Decision Scorecard

Score each strategy 1–5 for your use case (higher = better fit).

````python
SCORECARD: dict[str, dict[str, int]] = {
    "Fresh changing policies": {"rag": 5, "fine_tuning": 2, "long_context": 4, "hybrid": 5},
    "Strict citations / audit": {"rag": 5, "fine_tuning": 1, "long_context": 3, "hybrid": 5},
    "Brand tone / format": {"rag": 3, "fine_tuning": 5, "long_context": 3, "hybrid": 5},
    "Corpus > 100k chunks": {"rag": 5, "fine_tuning": 2, "long_context": 1, "hybrid": 5},
    "Corpus < 10 pages stable": {"rag": 4, "fine_tuning": 4, "long_context": 5, "hybrid": 4},
    "Lowest inference latency": {"rag": 4, "fine_tuning": 5, "long_context": 2, "hybrid": 4},
    "Lowest ops complexity": {"rag": 3, "fine_tuning": 2, "long_context": 5, "hybrid": 3},
}

strategies = ["rag", "fine_tuning", "long_context", "hybrid"]
totals = {s: 0 for s in strategies}

print(f"{'Criterion':<28}" + "".join(f"{s:>14}" for s in strategies))
print("-" * (28 + 14 * len(strategies)))
for criterion, scores in SCORECARD.items():
    for s in strategies:
        totals[s] += scores[s]
    row = "".join(f"{scores[s]:>14}" for s in strategies)
    print(f"{criterion:<28}{row}")
print("-" * (28 + 14 * len(strategies)))
print(f"{'TOTAL':<28}" + "".join(f"{totals[s]:>14}" for s in strategies))
````

## 14. Decision Tree (Plain Language)

```
Do facts change weekly/monthly?
  YES → prefer RAG (or hybrid)
  NO  → continue

Is the full corpus < ~50 pages AND stable?
  YES → long context is viable
  NO  → RAG

Do you need a specific brand voice / output JSON format?
  YES → add fine-tuning (or strong system prompt) ON TOP of RAG
  NO  → RAG or long context alone may suffice

Do auditors require source citations?
  YES → RAG (not fine-tuning alone)
```

## 15. Golden-Set Evaluation Across Strategies

````python
GOLDEN = [
    {"q": "What is the return window?", "expect": "30", "expect_cite": "pol-returns"},
    {"q": "When is shipping free?", "expect": "50", "expect_cite": "pol-shipping"},
    {"q": "billing statement name", "expect": "SHOPCO", "expect_cite": "pol-billing"},
]


def eval_strategy(fn, updated: bool = False) -> dict[str, Any]:
    corpus = UPDATED_CORPUS if updated else POLICY_DOCS
    hits = 0
    for row in GOLDEN:
        if fn == fine_tuning_strategy:
            r = fn(row["q"])
        elif fn in (rag_strategy, long_context_strategy):
            r = fn(row["q"], corpus=corpus)
        else:
            r = fn(row["q"], corpus=corpus)
        fact_ok = row["expect"].lower() in r.answer.lower()
        cite_ok = any(row["expect_cite"] in c for c in r.citations) if r.citations else fn == fine_tuning_strategy
        if fact_ok and cite_ok:
            hits += 1
    return {"score": hits / len(GOLDEN), "hits": hits}


print("Golden set (current corpus):")
for name, fn in STRATEGIES.items():
    e = eval_strategy(fn)
    print(f"  {name:<14} {e['score']:.0%} ({e['hits']}/{len(GOLDEN)})")

print("\nAfter 30→14 day update (RAG + long context use UPDATED_CORPUS):")
for name, fn in [("rag", rag_strategy), ("long_context", long_context_strategy)]:
    r = fn("What is the return window?", corpus=UPDATED_CORPUS)
    print(f"  {name}: {'14' in r.answer}")
````

## 16. What Fine-Tuning Should and Should Not Do

**Good fine-tuning targets:**

- Consistent JSON output schema for downstream parsers
- Brand voice (formal vs casual)
- Domain abbreviation disambiguation ("SKU" vs "order ID")
- Refusal style for out-of-scope requests

**Poor fine-tuning targets:**

- Weekly price lists or inventory counts
- Per-customer order status
- Regulatory text that legal must approve verbatim
- Anything requiring **provable** citations to source documents

---

## 17. Common Mistakes

| Mistake | Why it fails | Better approach |
|---------|--------------|------------------|
| Fine-tune for FAQ facts | Stale on first policy edit | RAG + optional FT for tone |
| Long context for 500-page wiki | Cost + context rot | RAG with chunking |
| RAG without eval set | Silent retrieval regressions | Golden questions + expected chunk IDs |
| Ignore hybrid | False trichotomy | RAG facts + FT/prompt style |
| No truncation strategy in long ctx | Missing policies at tail | Retrieval or hierarchical summarize |

## 20. Summary

| Strategy | Use when |
|----------|----------|
| **RAG** | Large or changing knowledge; citations required; agent tool loops |
| **Fine-tuning** | Stable behavior, tone, output shape — not live facts |
| **Long context** | Small stable corpus; simplest ops; one-shot Q&A |
| **Hybrid** | Production support bots: RAG facts + tuned tone |

On ShopCo Support, the **30 → 14 day** policy change showed the freshness gap: RAG and long context read the updated corpus; fine-tuned weights stayed stale until retrain. Choose based on **freshness, scale, auditability, and cost** — not ideology.
