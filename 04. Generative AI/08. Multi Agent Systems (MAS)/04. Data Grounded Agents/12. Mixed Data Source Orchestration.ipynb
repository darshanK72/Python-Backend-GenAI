# Mixed Data Source Orchestration

Real agents ground answers in **many systems at once**:

- **Vector indexes** — policy PDFs, FAQs, runbooks
- **Databases** — orders, inventory, billing
- **REST APIs** — carriers, CRM, payment gateways
- **Internal microservices** — loyalty, fraud, recommendations

Notebook 11 combined RAG + SQL in one workflow. This notebook generalizes that pattern into a **source-agnostic orchestrator** that fans out to any adapter, merges heterogeneous evidence, and handles failures gracefully.

```
                     ShopCo Multi-Source Orchestrator
                                  │
                         extract entities + plan
                                  │
       ┌──────────────┬───────────┼───────────┬──────────────┐
       ▼              ▼           ▼           ▼              ▼
   Policy RAG     Orders SQL   Tracking API   CRM API    (future)
   (FAISS)        (SQLite)     (REST mock)    (REST mock)
       │              │           │           │
       └──────────────┴───────────┴───────────┘
                                  ▼
              merge + rank + conflict check + citations
                                  ▼
                         unified grounded answer
```

You will build:

1. A `DataSource` adapter interface
2. Four ShopCo sources (RAG, SQL, tracking API, CRM)
3. A planner that selects sources per question
4. Parallel fan-out execution with timeouts and fallbacks
5. Evidence merge with freshness metadata and conflict detection
6. Plain-Python and **LangGraph** orchestration variants

Self-contained — no prior notebooks required.

## 1. Why Orchestrate Multiple Sources?

Each data system answers a different slice of truth:

| Source | Best for | Staleness risk |
|--------|----------|----------------|
| Policy RAG | Return rules, billing FAQs | Index may lag doc updates |
| Orders SQL | Current order status, counts | Live (authoritative for status) |
| Tracking API | Carrier scan events, ETA | Live (authoritative for location) |
| CRM API | Customer tier, account notes | Minutes behind writes |

A question like *"I'm gold tier — when will ORD-1001 arrive and can I return it?"* needs **four sources**. Orchestration decides which to call, runs them in parallel when possible, and merges results with provenance.

## 2. The DataSource Adapter Pattern

Every backend exposes the same contract: `can_handle(context)` and `fetch(context)` returning standardized `SourceEvidence`.

````python
@dataclass
class QueryContext:
    question: str
    order_id: str | None = None
    tracking: str | None = None
    email: str | None = None


@dataclass
class SourceEvidence:
    source_id: str
    source_type: Literal["rag", "sql", "api"]
    ref: str
    snippet: str
    freshness: Literal["live", "indexed", "cached"] = "live"
    latency_ms: float = 0.0
    error: str = ""


class DataSource(ABC):
    source_id: str
    source_type: Literal["rag", "sql", "api"]

    @abstractmethod
    def can_handle(self, ctx: QueryContext) -> bool:
        ...

    @abstractmethod
    def fetch(self, ctx: QueryContext) -> list[SourceEvidence]:
        ...
````

---

## 3. Policy RAG Adapter

````python
class PolicyRAGSource(DataSource):
    source_id = "policy_rag"
    source_type = "rag"

    def __init__(self, docs: list[dict[str, str]]) -> None:
        self.docs = docs
        texts = [f"{d['title']}. {d['text']}" for d in docs]
        self.vectorizer = TfidfVectorizer(stop_words="english")
        matrix = self.vectorizer.fit_transform(texts).astype("float32").toarray()
        faiss.normalize_L2(matrix)
        self.index = faiss.IndexFlatIP(matrix.shape[1])
        self.index.add(matrix)

    def can_handle(self, ctx: QueryContext) -> bool:
        q = ctx.question.lower()
        return any(t in q for t in ("return", "refund", "shipping", "billing", "policy", "vip", "money back"))

    def fetch(self, ctx: QueryContext) -> list[SourceEvidence]:
        t0 = time.perf_counter()
        vec = self.vectorizer.transform([ctx.question]).astype("float32").toarray()
        faiss.normalize_L2(vec)
        scores, idx = self.index.search(vec, 2)
        hits: list[SourceEvidence] = []
        for s, i in zip(scores[0], idx[0]):
            if i < 0 or s < 0.05:
                continue
            d = self.docs[int(i)]
            hits.append(SourceEvidence(
                source_id=self.source_id, source_type="rag", ref=d["id"],
                snippet=d["text"], freshness="indexed",
                latency_ms=(time.perf_counter() - t0) * 1000,
            ))
        return hits


rag_source = PolicyRAGSource(POLICY_DOCS)
ctx = QueryContext("What is the return window?")
print(rag_source.fetch(ctx)[0].ref)
````

## 4. Orders SQL Adapter

````python
FORBIDDEN = re.compile(r"\b(INSERT|UPDATE|DELETE|DROP|ALTER|CREATE)\b", re.I)


class OrdersSQLSource(DataSource):
    source_id = "orders_sql"
    source_type = "sql"

    def __init__(self, conn: sqlite3.Connection) -> None:
        self.conn = conn

    def can_handle(self, ctx: QueryContext) -> bool:
        q = ctx.question.lower()
        return ctx.order_id is not None or "order" in q or "how many" in q

    def _plan_sql(self, ctx: QueryContext) -> str | None:
        if ctx.order_id:
            return f"SELECT order_id, customer_email, status, total_usd, tracking FROM orders WHERE order_id = '{ctx.order_id}'"
        q = ctx.question.lower()
        if "how many" in q and "shipped" in q:
            return "SELECT COUNT(*) AS cnt FROM orders WHERE status = 'shipped'"
        if "how many" in q:
            return "SELECT COUNT(*) AS cnt FROM orders"
        return None

    def fetch(self, ctx: QueryContext) -> list[SourceEvidence]:
        t0 = time.perf_counter()
        sql = self._plan_sql(ctx)
        if not sql:
            return []
        cleaned = sql.strip().rstrip(";")
        if FORBIDDEN.search(cleaned) or not re.match(r"^SELECT", cleaned, re.I):
            return [SourceEvidence(self.source_id, "sql", "blocked", "", error="only SELECT allowed")]
        if "limit" not in cleaned.lower():
            cleaned += " LIMIT 50"
        try:
            rows = [dict(r) for r in self.conn.execute(cleaned).fetchall()]
            if not rows:
                return []
            return [SourceEvidence(
                source_id=self.source_id, source_type="sql", ref=cleaned[:50],
                snippet=pretty(rows[0]), freshness="live",
                latency_ms=(time.perf_counter() - t0) * 1000,
            )]
        except sqlite3.Error as exc:
            return [SourceEvidence(self.source_id, "sql", "error", "", error=str(exc))]


sql_source = OrdersSQLSource(DB)
print(sql_source.fetch(QueryContext("ORD-1001 status", order_id="ORD-1001"))[0].snippet[:80])
````

## 5. Tracking API Adapter

````python
class TrackingAPISource(DataSource):
    source_id = "tracking_api"
    source_type = "api"

    def __init__(self, api_data: dict[str, dict[str, Any]], *, simulate_down: bool = False) -> None:
        self.api_data = api_data
        self.simulate_down = simulate_down

    def can_handle(self, ctx: QueryContext) -> bool:
        q = ctx.question.lower()
        return ctx.tracking is not None or "tracking" in q or "arrive" in q or "eta" in q or "where is" in q

    def fetch(self, ctx: QueryContext) -> list[SourceEvidence]:
        t0 = time.perf_counter()
        if self.simulate_down:
            return [SourceEvidence(self.source_id, "api", "tracking_api", "", error="503 Service Unavailable")]
        tracking = ctx.tracking
        if not tracking and ctx.order_id:
            row = DB.execute("SELECT tracking FROM orders WHERE order_id = ?", (ctx.order_id,)).fetchone()
            tracking = row["tracking"] if row else None
        if not tracking:
            return []
        payload = self.api_data.get(tracking)
        if not payload:
            return [SourceEvidence(self.source_id, "api", tracking, "No carrier data", freshness="live")]
        return [SourceEvidence(
            source_id=self.source_id, source_type="api", ref=tracking,
            snippet=pretty(payload), freshness="live",
            latency_ms=(time.perf_counter() - t0) * 1000,
        )]


tracking_source = TrackingAPISource(TRACKING_API)
print(tracking_source.fetch(QueryContext("Where is my package?", order_id="ORD-1001"))[0].snippet)
````

## 6. CRM API Adapter

````python
class CRMAPISource(DataSource):
    source_id = "crm_api"
    source_type = "api"

    def __init__(self, profiles: dict[str, dict[str, Any]]) -> None:
        self.profiles = profiles

    def can_handle(self, ctx: QueryContext) -> bool:
        q = ctx.question.lower()
        return ctx.email is not None or "tier" in q or "vip" in q or "gold" in q or "account" in q

    def fetch(self, ctx: QueryContext) -> list[SourceEvidence]:
        t0 = time.perf_counter()
        email = ctx.email
        if not email and ctx.order_id:
            row = DB.execute("SELECT customer_email FROM orders WHERE order_id = ?", (ctx.order_id,)).fetchone()
            email = row["customer_email"] if row else None
        if not email:
            return []
        profile = self.profiles.get(email)
        if not profile:
            return []
        return [SourceEvidence(
            source_id=self.source_id, source_type="api", ref=email,
            snippet=pretty(profile), freshness="cached",
            latency_ms=(time.perf_counter() - t0) * 1000,
        )]


crm_source = CRMAPISource(CRM_PROFILES)
print(crm_source.fetch(QueryContext("Am I gold tier?", order_id="ORD-1001"))[0].snippet)
````

## 7. Entity Extraction and Source Planner

Before fan-out, extract entities (order ID, tracking, email) and select eligible sources.

````python
def extract_context(question: str) -> QueryContext:
    order_m = re.search(r"ORD-[0-9]{4}", question.upper())
    email_m = re.search(r"[\w.+-]+@shopco\.com", question.lower())
    tracking_m = re.search(r"1Z[0-9A-Z]{16}", question.upper())
    return QueryContext(
        question=question,
        order_id=order_m.group(0) if order_m else None,
        email=email_m.group(0) if email_m else None,
        tracking=tracking_m.group(0) if tracking_m else None,
    )


SOURCE_REGISTRY: list[DataSource] = [
    rag_source, sql_source, tracking_source, crm_source,
]


def plan_sources(ctx: QueryContext, registry: list[DataSource]) -> list[str]:
    return [s.source_id for s in registry if s.can_handle(ctx)]


ctx = extract_context("Gold tier refund for ORD-1001 — when will it arrive?")
print("Entities:", ctx)
print("Planned sources:", plan_sources(ctx, SOURCE_REGISTRY))
````

## 8. Parallel Fan-Out Executor

````python
def fan_out_fetch(
    ctx: QueryContext,
    source_ids: list[str],
    registry: list[DataSource],
    timeout_s: float = 2.0,
) -> tuple[list[SourceEvidence], list[str]]:
    by_id = {s.source_id: s for s in registry}
    evidence: list[SourceEvidence] = []
    trace: list[str] = []

    with ThreadPoolExecutor(max_workers=4) as pool:
        futures = {pool.submit(by_id[sid].fetch, ctx): sid for sid in source_ids if sid in by_id}
        for fut in as_completed(futures, timeout=timeout_s):
            sid = futures[fut]
            try:
                hits = fut.result()
                evidence.extend(hits)
                trace.append(f"{sid}:ok:{len(hits)}")
            except Exception as exc:
                trace.append(f"{sid}:error:{exc}")
    return evidence, trace


ctx = extract_context("Return policy and ORD-1001 tracking ETA")
ids = plan_sources(ctx, SOURCE_REGISTRY)
ev, tr = fan_out_fetch(ctx, ids, SOURCE_REGISTRY)
print("Trace:", tr)
print("Sources hit:", [e.source_id for e in ev])
````

## 9. Evidence Merge, Ranking, and Conflict Detection

When SQL says `shipped` but tracking API says `delivered`, flag the conflict and prefer the **live API** for carrier state.

````python
@dataclass
class OrchestrationResult:
    question: str
    context: QueryContext
    planned_sources: list[str]
    evidence: list[SourceEvidence] = field(default_factory=list)
    conflicts: list[str] = field(default_factory=list)
    answer: str = ""
    trace: list[str] = field(default_factory=list)
    grounded: bool = False


FRESHNESS_RANK = {"live": 3, "cached": 2, "indexed": 1}


def rank_evidence(items: list[SourceEvidence]) -> list[SourceEvidence]:
    return sorted(
        [e for e in items if not e.error],
        key=lambda e: FRESHNESS_RANK.get(e.freshness, 0),
        reverse=True,
    )


def detect_conflicts(evidence: list[SourceEvidence]) -> list[str]:
    conflicts: list[str] = []
    sql_status = None
    api_status = None
    for e in evidence:
        if e.source_id == "orders_sql" and "status" in e.snippet:
            m = re.search(r'"status":\s*"(\w+)"', e.snippet)
            if m:
                sql_status = m.group(1)
        if e.source_id == "tracking_api":
            m = re.search(r'"status":\s*"(\w+)"', e.snippet)
            if m:
                api_status = m.group(1)
    if sql_status and api_status and sql_status != api_status:
        if sql_status == "shipped" and api_status == "delivered":
            conflicts.append(f"SQL status '{sql_status}' vs tracking '{api_status}' — prefer tracking API for carrier state")
        else:
            conflicts.append(f"SQL status '{sql_status}' vs tracking '{api_status}'")
    return conflicts


def compose_mixed_answer(evidence: list[SourceEvidence], conflicts: list[str]) -> str:
    if not evidence:
        return "I don't have verified ShopCo information for that question."
    ranked = rank_evidence(evidence)
    lines = [f"[{e.source_id}:{e.ref}] ({e.freshness}) {e.snippet[:100]}" for e in ranked]
    cites = ", ".join(f"{e.source_id}:{e.ref}" for e in ranked)
    body = "\n".join(lines)
    if conflicts:
        body += "\n\n⚠ Conflicts:\n" + "\n".join(f"  - {c}" for c in conflicts)
    return body + f"\n\nSources: {cites}"
````

## 10. MixedDataOrchestrator (Plain Python)

````python
class MixedDataOrchestrator:
    def __init__(self, registry: list[DataSource]) -> None:
        self.registry = registry

    def run(self, question: str) -> OrchestrationResult:
        ctx = extract_context(question)
        planned = plan_sources(ctx, self.registry)
        result = OrchestrationResult(question=question, context=ctx, planned_sources=planned)
        result.trace.append(f"plan:{planned}")

        if not planned:
            result.answer = "Hello! Ask about ShopCo orders, tracking, policies, or your account tier."
            return result

        evidence, fetch_trace = fan_out_fetch(ctx, planned, self.registry)
        result.evidence = evidence
        result.trace.extend(fetch_trace)
        result.conflicts = detect_conflicts(evidence)
        result.grounded = any(not e.error for e in evidence)
        result.answer = compose_mixed_answer(evidence, result.conflicts)
        result.trace.append("compose:done")
        return result


orchestrator = MixedDataOrchestrator(SOURCE_REGISTRY)

demo = orchestrator.run("I'm gold — return policy and when will ORD-1001 arrive?")
print("Planned:", demo.planned_sources)
print("Evidence sources:", [e.source_id for e in demo.evidence])
print(demo.answer[:300], "...")
````

## 11. LangGraph Multi-Source Workflow

Explicit graph: extract → plan → parallel fetch → merge → compose.

````python
class GraphState(TypedDict):
    question: str
    context: dict[str, str | None]
    planned: list[str]
    evidence: list[dict[str, Any]]
    conflicts: list[str]
    answer: str
    trace: list[str]


def extract_node(state: GraphState) -> dict[str, Any]:
    ctx = extract_context(state["question"])
    return {
        "context": {"order_id": ctx.order_id, "email": ctx.email, "tracking": ctx.tracking},
        "trace": state.get("trace", []) + ["extract:done"],
    }


def plan_node(state: GraphState) -> dict[str, Any]:
    ctx = QueryContext(state["question"], **{k: v for k, v in state["context"].items()})
    planned = plan_sources(ctx, SOURCE_REGISTRY)
    return {"planned": planned, "trace": state["trace"] + [f"plan:{planned}"]}


def fetch_node(state: GraphState) -> dict[str, Any]:
    ctx = QueryContext(state["question"], **{k: v for k, v in state["context"].items()})
    if not state["planned"]:
        return {"evidence": [], "trace": state["trace"] + ["fetch:skip"]}
    ev, tr = fan_out_fetch(ctx, state["planned"], SOURCE_REGISTRY)
    return {
        "evidence": [e.__dict__ for e in ev],
        "trace": state["trace"] + tr,
    }


def merge_node(state: GraphState) -> dict[str, Any]:
    items = [SourceEvidence(**e) for e in state.get("evidence", [])]
    conflicts = detect_conflicts(items)
    return {"conflicts": conflicts, "trace": state["trace"] + [f"merge:conflicts={len(conflicts)}"]}


def compose_node(state: GraphState) -> dict[str, Any]:
    if not state["planned"]:
        return {"answer": "Hello! Ask about ShopCo orders, tracking, policies, or your account tier.", "trace": state["trace"] + ["compose:greeting"]}
    items = [SourceEvidence(**e) for e in state.get("evidence", [])]
    return {"answer": compose_mixed_answer(items, state.get("conflicts", [])), "trace": state["trace"] + ["compose:done"]}


builder = StateGraph(GraphState)
for name, fn in [("extract", extract_node), ("plan", plan_node), ("fetch", fetch_node),
                 ("merge", merge_node), ("compose", compose_node)]:
    builder.add_node(name, fn)
builder.add_edge(START, "extract")
builder.add_edge("extract", "plan")
builder.add_edge("plan", "fetch")
builder.add_edge("fetch", "merge")
builder.add_edge("merge", "compose")
builder.add_edge("compose", END)

mixed_graph = builder.compile()

gout = mixed_graph.invoke({
    "question": "ORD-1003 status and tracking location",
    "context": {}, "planned": [], "evidence": [], "conflicts": [], "answer": "", "trace": [],
})
print("Trace:", " → ".join(gout["trace"]))
print(gout["answer"][:200], "...")
````

## 15. Orchestration Patterns Compared

| Pattern | When to use |
|---------|-------------|
| **Sequential chain** | Source B needs output from source A (e.g. SQL → tracking number → API) |
| **Parallel fan-out** | Independent sources; minimize latency |
| **Conditional plan** | Only call sources `can_handle` accepts |
| **Fallback** | Primary API down → degrade to SQL-only answer |
| **Conflict resolution** | Prefer live API over indexed RAG for carrier state |

---

## 16. Common Mistakes

| Mistake | Symptom | Fix |
|---------|---------|-----|
| Hard-coded source order | Slow sequential calls | Parallel fan-out with timeout |
| No adapter interface | Tight coupling per backend | `DataSource` contract |
| Ignoring freshness | Stale RAG overrides live API | Rank by `freshness` |
| Silent API failures | Missing evidence, no trace | Capture `error` on `SourceEvidence` |
| No entity linking | CRM can't find customer | SQL lookup enriches `ctx.email` |
| One mega-tool | LLM picks wrong backend | Planner + `can_handle` gates |

## 18. Optional Live LLM Compose Step

Fetch deterministically from all sources; optional LLM synthesizes the final answer from merged evidence only.

````python
def compose_with_optional_llm(result: OrchestrationResult) -> str:
    if not USE_LIVE_LLM or not result.evidence:
        return result.answer
    from langchain_openai import ChatOpenAI

    llm = ChatOpenAI(model="gpt-4o-mini", api_key=OPENAI_API_KEY, temperature=0)
    ranked = rank_evidence(result.evidence)
    ctx = "\n".join(f"[{e.source_id}:{e.ref}] {e.snippet}" for e in ranked)
    conflicts = "\n".join(result.conflicts) if result.conflicts else "none"
    prompt = (
        f"Answer using ONLY this evidence. Cite source_id:ref. Note conflicts if any.\n"
        f"Evidence:\n{ctx}\n\nConflicts: {conflicts}\n\nQuestion: {result.question}"
    )
    return llm.invoke(prompt).content


base = orchestrator.run("Gold tier return and ORD-1001 tracking")
print(compose_with_optional_llm(base)[:180], "...")
````

## 20. Summary

**Mixed data source orchestration** extends RAG + SQL into a general pattern:

1. **Adapters** — one `DataSource` per backend (RAG, SQL, REST APIs)
2. **Plan** — extract entities, select sources via `can_handle`
3. **Fan-out** — parallel fetch with timeouts and error capture
4. **Merge** — rank by freshness, detect conflicts, compose with citations

ShopCo's orchestrator handles policy docs, order database, carrier tracking, and CRM profiles in one workflow — the capstone of **data-grounded agents**. The next module covers **agentic memory**: how agents retain context across turns and sessions.
