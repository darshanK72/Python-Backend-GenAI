# RAG with FAISS for Agent Workflows

**FAISS** (Facebook AI Similarity Search) is an in-memory vector index optimized for fast nearest-neighbor search. In agent workflows, it backs the **retrieval tool** — `search_policy`, `search_runbook`, `find_similar_ticket` — that grounds the model in your private documents.

```
  Agent loop                    FAISS stack
  ─────────                     ───────────
  user question                 chunk + embed corpus
       │                              │
       ▼                              ▼
  planner decides              IndexFlatIP (cosine)
  to call search_policy  ───►  top-k chunks + scores
       │                              │
       ▼                              ▼
  compose with evidence ◄──── doc_id metadata map
```

This notebook builds a complete **ShopCo Policy RAG** stack:

1. Chunk and embed policy documents
2. Index vectors in **FAISS**
3. Expose retrieval as an **agent tool**
4. Run a multi-step **support agent workflow** grounded on FAISS hits
5. Persist and reload the index for production deploys

Embeddings use **TF-IDF** (offline, no API). An optional cell swaps in OpenAI embeddings. Everything else is plain Python + FAISS + scikit-learn.

````python
"""
ShopCo Policy RAG — FAISS + agent workflow lab.
"""

import json
import os
import re
import uuid
from dataclasses import dataclass, field
from datetime import datetime, timezone
from pathlib import Path
from typing import Any, Callable

import faiss
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer

OPENAI_API_KEY = os.environ.get("OPENAI_API_KEY", "sk-proj-placeholder-replace-with-your-real-key")
os.environ.setdefault("OPENAI_API_KEY", OPENAI_API_KEY)

USE_LIVE_LLM = False
USE_OPENAI_EMBEDDINGS = False

INDEX_DIR = Path("shopco_faiss_index")


def pretty(obj: Any) -> str:
    return json.dumps(obj, indent=2, default=str)


POLICY_DOCS: list[dict[str, str]] = [
    {
        "id": "pol-returns-01",
        "title": "Returns Policy",
        "text": (
            "Customers may return items within 30 days of delivery. Items must be unused "
            "and in original packaging. Refunds are processed within 5-7 business days "
            "after warehouse receipt. Final-sale items cannot be returned."
        ),
    },
    {
        "id": "pol-shipping-02",
        "title": "Shipping Policy",
        "text": (
            "Standard shipping is free on orders over $50. Express shipping is available "
            "at checkout. Tracking numbers are emailed when the carrier scans the package."
        ),
    },
    {
        "id": "pol-billing-03",
        "title": "Billing Policy",
        "text": (
            "Charges appear as SHOPCO* on credit card statements. Partial refunds may take "
            "up to two billing cycles to display. Contact support with order ID for disputes."
        ),
    },
    {
        "id": "faq-tracking-04",
        "title": "Tracking FAQ",
        "text": (
            "If order status is shipped, use the tracking number from your confirmation email. "
            "Processing orders do not yet have tracking numbers assigned."
        ),
    },
    {
        "id": "faq-warranty-05",
        "title": "Warranty FAQ",
        "text": (
            "Electronics include a 1-year manufacturer warranty. Warranty claims require "
            "proof of purchase and serial number. Warranty does not cover accidental damage."
        ),
    },
]

ORDERS_DB: dict[str, dict[str, Any]] = {
    "ORD-1001": {"order_id": "ORD-1001", "status": "shipped", "tracking": "1Z999AA10123456784"},
    "ORD-1002": {"order_id": "ORD-1002", "status": "processing", "tracking": None},
}

print(f"Corpus: {len(POLICY_DOCS)} policy documents")
````

## 1. Why FAISS in Agent Workflows?

| Requirement | How FAISS helps |
|-------------|-----------------|
| **Low latency** | Millisecond search over millions of vectors |
| **Agent tool** | `search()` returns top-k in one call |
| **Metadata join** | Store `doc_id` per vector row |
| **Deployable** | Serialize index to disk between restarts |
| **Composable** | One retriever service, many agents |

Agents should call retrieval as a **tool** with structured results (chunk id, score, text) — not embed search logic inside prompt strings.

## 2. Chunking — Index Units

FAISS indexes **chunks**, not whole documents. Small corpora can use one chunk per doc; larger wikis need sliding windows with overlap.

````python
@dataclass
class Chunk:
    chunk_id: str
    doc_id: str
    title: str
    text: str


def chunk_documents(docs: list[dict[str, str]], max_chars: int = 400) -> list[Chunk]:
    """One chunk per doc for small policies; split long docs by sentence groups."""
    chunks: list[Chunk] = []
    for doc in docs:
        body = doc["text"].strip()
        if len(body) <= max_chars:
            chunks.append(Chunk(
                chunk_id=f"{doc['id']}#0",
                doc_id=doc["id"],
                title=doc["title"],
                text=body,
            ))
            continue
        sentences = re.split(r"(?<=[.!?])\s+", body)
        buf, idx = "", 0
        for sent in sentences:
            if len(buf) + len(sent) > max_chars and buf:
                chunks.append(Chunk(f"{doc['id']}#{idx}", doc["id"], doc["title"], buf.strip()))
                idx += 1
                buf = sent
            else:
                buf = f"{buf} {sent}".strip()
        if buf:
            chunks.append(Chunk(f"{doc['id']}#{idx}", doc["id"], doc["title"], buf.strip()))
    return chunks


CHUNKS = chunk_documents(POLICY_DOCS)
print(f"Chunks: {len(CHUNKS)}")
for c in CHUNKS[:3]:
    print(f"  {c.chunk_id} ({len(c.text)} chars)")
````

---

## 3. Embeddings — TF-IDF for Offline Search

Production systems use embedding APIs or `sentence-transformers`. **TF-IDF** gives semantic-ish overlap without network calls — enough to demonstrate FAISS integration and agent tooling.

Vectors are **L2-normalized** so inner product in FAISS equals cosine similarity.

````python
def build_tfidf_embeddings(texts: list[str]) -> tuple[TfidfVectorizer, np.ndarray]:
    vectorizer = TfidfVectorizer(stop_words="english")
    matrix = vectorizer.fit_transform(texts).astype("float32")
    dense = matrix.toarray().astype("float32")
    faiss.normalize_L2(dense)
    return vectorizer, dense


corpus_texts = [f"{c.title}. {c.text}" for c in CHUNKS]
VECTORIZER, CORPUS_VECTORS = build_tfidf_embeddings(corpus_texts)
print(f"Embedding matrix: {CORPUS_VECTORS.shape} (chunks × dims)")
````

---

## 4. Build the FAISS Index

`IndexFlatIP` = exact inner-product search. For normalized vectors, scores are cosine similarities in `[0, 1]`.

````python
def create_faiss_index(vectors: np.ndarray) -> faiss.IndexFlatIP:
    dim = vectors.shape[1]
    index = faiss.IndexFlatIP(dim)
    index.add(vectors)
    return index


FAISS_INDEX = create_faiss_index(CORPUS_VECTORS)
print(f"FAISS index: {FAISS_INDEX.ntotal} vectors, dim={FAISS_INDEX.d}")
````

---

## 5. FaissRetriever — Search + Metadata

````python
@dataclass
class SearchHit:
    chunk_id: str
    doc_id: str
    title: str
    text: str
    score: float


class FaissRetriever:
    """FAISS-backed retriever with chunk metadata and score threshold."""

    def __init__(
        self,
        index: faiss.Index,
        vectorizer: TfidfVectorizer,
        chunks: list[Chunk],
        min_score: float = 0.05,
    ) -> None:
        self.index = index
        self.vectorizer = vectorizer
        self.chunks = chunks
        self.min_score = min_score
        self.search_log: list[dict[str, Any]] = []

    def _embed_query(self, query: str) -> np.ndarray:
        vec = self.vectorizer.transform([query]).astype("float32").toarray()
        faiss.normalize_L2(vec)
        return vec

    def search(self, query: str, top_k: int = 3) -> list[SearchHit]:
        qvec = self._embed_query(query)
        scores, indices = self.index.search(qvec, top_k)
        hits: list[SearchHit] = []
        for score, idx in zip(scores[0], indices[0]):
            if idx < 0 or score < self.min_score:
                continue
            chunk = self.chunks[int(idx)]
            hits.append(SearchHit(
                chunk_id=chunk.chunk_id,
                doc_id=chunk.doc_id,
                title=chunk.title,
                text=chunk.text,
                score=float(score),
            ))
        self.search_log.append({
            "query": query,
            "top_k": top_k,
            "hits": [h.chunk_id for h in hits],
            "scores": [round(h.score, 4) for h in hits],
            "ts": datetime.now(timezone.utc).isoformat(),
        })
        return hits


retriever = FaissRetriever(FAISS_INDEX, VECTORIZER, CHUNKS)

demo_hits = retriever.search("money back guarantee window")
print("Semantic query 'money back guarantee window':")
for h in demo_hits:
    print(f"  [{h.doc_id}] score={h.score:.3f} — {h.title}")
````

## 6. Keyword vs FAISS — Why Vectors Matter

Users rarely use exact policy vocabulary. Vector search maps **"money back"** → returns policy even without the word "return" in the query.

````python
def keyword_search(query: str, top_k: int = 2) -> list[str]:
    terms = [t for t in re.split(r"\W+", query.lower()) if len(t) > 2]
    scored = []
    for c in CHUNKS:
        hay = f"{c.title} {c.text}".lower()
        score = sum(1 for t in terms if t in hay)
        if score:
            scored.append((score, c.doc_id))
    scored.sort(reverse=True)
    return [d for _, d in scored[:top_k]]


SEMANTIC_QUERIES = [
    "money back guarantee window",
    "when will I see tracking info",
    "charge on my credit card statement",
]

print(f"{'Query':<40} {'Keyword':<20} {'FAISS'}")
print("-" * 75)
for q in SEMANTIC_QUERIES:
    kw = keyword_search(q) or ["(none)"]
    fv = [h.doc_id for h in retriever.search(q, top_k=1)] or ["(none)"]
    print(f"{q[:38]:<40} {kw[0]:<20} {fv[0]}")
````

## 7. Agent Tool — `search_policy_docs`

Wrap the retriever as a tool the agent loop can call with structured JSON results.

````python
TOOL_SCHEMA = {
    "name": "search_policy_docs",
    "description": (
        "Search ShopCo policy and FAQ documents. Use for returns, shipping, billing, "
        "tracking, and warranty questions. Do not use for live order status."
    ),
    "parameters": {
        "type": "object",
        "properties": {
            "query": {"type": "string", "description": "Natural language search query"},
            "top_k": {"type": "integer", "default": 3},
        },
        "required": ["query"],
    },
}


def search_policy_docs(query: str, top_k: int = 3) -> dict[str, Any]:
    hits = retriever.search(query, top_k=top_k)
    if not hits:
        return {"isError": True, "content": "No policy documents matched the query."}
    return {
        "content": [
            {
                "chunk_id": h.chunk_id,
                "doc_id": h.doc_id,
                "title": h.title,
                "score": round(h.score, 4),
                "text": h.text,
            }
            for h in hits
        ]
    }


TOOL_REGISTRY: dict[str, Callable[..., dict[str, Any]]] = {
    "search_policy_docs": lambda query, top_k=3: search_policy_docs(query, top_k),
}

print(pretty(search_policy_docs("refund timeline")))
````

---

## 8. Agent Workflow State

````python
@dataclass
class AgentState:
    run_id: str
    question: str
    intent: str = ""
    order_id: str | None = None
    retrieval_hits: list[SearchHit] = field(default_factory=list)
    order_facts: dict[str, Any] | None = None
    answer: str = ""
    tool_calls: list[dict[str, Any]] = field(default_factory=list)
    trace: list[str] = field(default_factory=list)

    def log(self, msg: str) -> None:
        self.trace.append(msg)


def classify_intent(question: str) -> tuple[str, str | None]:
    q = question.lower()
    m = re.search(r"ORD-[0-9]{4}", question.upper())
    oid = m.group(0) if m else None
    policy = any(t in q for t in ("return", "refund", "shipping", "billing", "track", "warranty", "policy"))
    if oid and policy:
        return "mixed", oid
    if oid:
        return "order_status", oid
    if policy:
        return "policy", oid
    return "general", oid
````

---

## 9. Offline Planner — When to Call FAISS Tool

The planner mimics an LLM: policy intents trigger `search_policy_docs`; order intents call structured lookup.

````python
def lookup_order(order_id: str) -> dict[str, Any] | None:
    return ORDERS_DB.get(order_id)


def offline_planner(state: AgentState) -> AgentState:
    intent, oid = classify_intent(state.question)
    state.intent = intent
    state.order_id = oid
    state.log(f"plan:intent={intent}")

    if intent in ("policy", "mixed", "general"):
        call = {"tool": "search_policy_docs", "args": {"query": state.question, "top_k": 2}}
        state.tool_calls.append(call)
        result = TOOL_REGISTRY["search_policy_docs"](**call["args"])
        if not result.get("isError"):
            for item in result["content"]:
                state.retrieval_hits.append(SearchHit(
                    chunk_id=item["chunk_id"], doc_id=item["doc_id"],
                    title=item["title"], text=item["text"], score=item["score"],
                ))
        state.log(f"tool:search_policy_docs:hits={len(state.retrieval_hits)}")

    if intent in ("order_status", "mixed") and oid:
        state.order_facts = lookup_order(oid)
        state.log(f"tool:lookup_order:found={state.order_facts is not None}")

    return state


def compose_answer(state: AgentState) -> AgentState:
    parts: list[str] = []
    if state.retrieval_hits:
        top = state.retrieval_hits[0]
        parts.append(f"[{top.doc_id}] {top.text[:120]}...")
    if state.order_facts:
        parts.append(
            f"Order {state.order_facts['order_id']}: status {state.order_facts['status']}."
        )
        if state.order_facts.get("tracking"):
            parts.append(f"Tracking: {state.order_facts['tracking']}.")
    state.answer = " ".join(parts) if parts else "I could not find verified ShopCo information."
    state.log("compose:done")
    return state


def run_agent(question: str) -> AgentState:
    state = AgentState(run_id=str(uuid.uuid4()), question=question)
    state = offline_planner(state)
    state = compose_answer(state)
    return state
````

## 10. Run the Agent Workflow

````python
TICKETS = [
    "How long until I get my money back?",
    "Where is order ORD-1001?",
    "Return policy and status of ORD-1002",
]

for q in TICKETS:
    out = run_agent(q)
    print(f"Q: {q}")
    print(f"  intent={out.intent} | FAISS hits={[h.doc_id for h in out.retrieval_hits]}")
    print(f"  {out.answer[:90]}...\n")
````

## 11. Score Threshold and Abstention

Low FAISS scores mean weak matches — the agent should abstain instead of citing irrelevant chunks.

````python
STRICT_RETRIEVER = FaissRetriever(FAISS_INDEX, VECTORIZER, CHUNKS, min_score=0.25)

OUT_OF_SCOPE = [
    "What is the weather in Paris?",
    "Explain quantum computing",
]

for q in OUT_OF_SCOPE:
    hits = STRICT_RETRIEVER.search(q)
    print(f"Q: {q}")
    print(f"  hits={len(hits)} (min_score=0.25) → {'abstain' if not hits else hits[0].doc_id}")
````

## 12. Persist and Reload the Index

Production agents rebuild or load FAISS indexes at startup. Save **index**, **vectorizer vocabulary**, and **chunk metadata** together.

````python
import pickle


def save_index_bundle(path: Path, index: faiss.Index, vectorizer: TfidfVectorizer, chunks: list[Chunk]) -> None:
    path.mkdir(parents=True, exist_ok=True)
    faiss.write_index(index, str(path / "index.faiss"))
    with open(path / "vectorizer.pkl", "wb") as f:
        pickle.dump(vectorizer, f)
    with open(path / "chunks.json", "w", encoding="utf-8") as f:
        json.dump([c.__dict__ for c in chunks], f, indent=2)
    print(f"Saved bundle to {path}")


def load_index_bundle(path: Path) -> FaissRetriever:
    index = faiss.read_index(str(path / "index.faiss"))
    with open(path / "vectorizer.pkl", "rb") as f:
        vectorizer = pickle.load(f)
    raw = json.loads((path / "chunks.json").read_text(encoding="utf-8"))
    chunks = [Chunk(**row) for row in raw]
    return FaissRetriever(index, vectorizer, chunks)


save_index_bundle(INDEX_DIR, FAISS_INDEX, VECTORIZER, CHUNKS)
loaded = load_index_bundle(INDEX_DIR)
print("Reloaded search:", [h.doc_id for h in loaded.search("free delivery threshold")])
````

## 13. Incremental Index Update

When a new policy arrives, embed the chunk and `index.add()`. For large corpora, schedule periodic full rebuilds.

````python
def add_document_to_index(
    doc: dict[str, str],
    index: faiss.Index,
    vectorizer: TfidfVectorizer,
    chunks: list[Chunk],
) -> Chunk:
    new_chunks = chunk_documents([doc])
    if not new_chunks:
        raise ValueError("no chunks produced")
    chunk = new_chunks[0]
    text = f"{chunk.title}. {chunk.text}"
    vec = vectorizer.transform([text]).astype("float32").toarray()
    faiss.normalize_L2(vec)
    index.add(vec)
    chunks.append(chunk)
    return chunk


new_doc = {
    "id": "pol-gift-06",
    "title": "Gift Card Policy",
    "text": "Gift cards are non-refundable and do not expire. Balance inquiries available in account settings.",
}
added = add_document_to_index(new_doc, FAISS_INDEX, VECTORIZER, CHUNKS)
print(f"Added {added.chunk_id} | index size: {FAISS_INDEX.ntotal}")
print("Search gift card:", [h.doc_id for h in retriever.search("gift card refund")])
````

## 14. FAISS Index Types (When to Upgrade)

| Index | Use case |
|-------|----------|
| `IndexFlatIP` | Exact search, <100k vectors (this notebook) |
| `IndexIVFFlat` | Approximate, faster at scale — needs training |
| `IndexHNSW` | Low-latency approximate, memory-heavy |
| `IndexIDMap` | Attach custom IDs to vectors |

Start with **Flat** for agent prototypes; move to IVF/HNSW when latency SLOs tighten at millions of chunks.

## 16. Golden-Set Evaluation

````python
GOLDEN = [
    {"query": "money back timeline", "expect_doc": "pol-returns"},
    {"query": "free shipping minimum", "expect_doc": "pol-shipping"},
    {"query": "credit card statement charge", "expect_doc": "pol-billing"},
    {"query": "tracking number not showing", "expect_doc": "faq-tracking"},
]


def eval_retriever(r: FaissRetriever) -> float:
    hits = 0
    for row in GOLDEN:
        result = r.search(row["query"], top_k=1)
        if result and row["expect_doc"] in result[0].doc_id:
            hits += 1
    return hits / len(GOLDEN)


print(f"FAISS golden score: {eval_retriever(retriever):.0%}")
print(f"Keyword golden score: {sum(1 for row in GOLDEN if row['expect_doc'] in (keyword_search(row['query']) or [''])) / len(GOLDEN):.0%}")
````

## 18. Common Mistakes

| Mistake | Symptom | Fix |
|---------|---------|-----|
| No normalization before `IndexFlatIP` | Random ranking | `faiss.normalize_L2` |
| Index without metadata map | Can't cite doc IDs | Parallel `chunks` list |
| FAISS for live order rows | Stale/wrong status | SQL/API tool separate |
| No score threshold | Irrelevant chunks cited | `min_score` + abstain |
| Save index only | Reload fails | Bundle vectorizer + chunks |
| Giant flat index forever | Slow at 10M+ vectors | IVF/HNSW + profiling |

## 19. Optional OpenAI Embeddings + Live LLM

Set `USE_OPENAI_EMBEDDINGS = True` to rebuild vectors with `text-embedding-3-small`. Set `USE_LIVE_LLM = True` for live answer synthesis.

````python
def build_openai_embeddings(texts: list[str]) -> np.ndarray:
    from openai import OpenAI

    client = OpenAI(api_key=OPENAI_API_KEY)
    resp = client.embeddings.create(model="text-embedding-3-small", input=texts)
    vectors = np.array([d.embedding for d in resp.data], dtype="float32")
    faiss.normalize_L2(vectors)
    return vectors


if USE_OPENAI_EMBEDDINGS:
    oa_vectors = build_openai_embeddings(corpus_texts)
    oa_index = create_faiss_index(oa_vectors)
    print("OpenAI index:", oa_index.ntotal, "vectors")
elif USE_LIVE_LLM:
    from langchain_core.messages import HumanMessage, SystemMessage
    from langchain_openai import ChatOpenAI

    s = run_agent("What is the return window?")
    evidence = "\n".join(f"[{h.doc_id}] {h.text}" for h in s.retrieval_hits)
    llm = ChatOpenAI(model="gpt-4o-mini", api_key=OPENAI_API_KEY, temperature=0)
    resp = llm.invoke([
        SystemMessage(content="Answer from evidence only. Cite doc ids."),
        HumanMessage(content=f"Q: {s.question}\nEvidence:\n{evidence}"),
    ])
    print(resp.content)
else:
    print("Offline mode. FAISS eval:", f"{eval_retriever(retriever):.0%}")
    print("Flags: USE_OPENAI_EMBEDDINGS, USE_LIVE_LLM")
````

## 21. Summary

**FAISS** powers fast vector retrieval inside agent workflows. The pattern:

1. Chunk documents → embed → `IndexFlatIP`
2. Wrap `search()` as an agent tool returning chunk IDs and scores
3. Planner calls tool → composer cites evidence → validator checks grounding
4. Persist index + vectorizer + chunk metadata for production restarts

ShopCo Policy RAG showed semantic queries beating keyword search, score-threshold abstention, and a complete offline agent loop — the foundation you extend with better embeddings and approximate indexes at scale.
