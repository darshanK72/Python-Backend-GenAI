# Anti-patterns — Premature Multi-Agent and Hidden State

Two failures show up repeatedly in agent projects:

1. **Premature multi-agent** — many agents and handoffs where **routing** or a **single ReAct loop** would suffice.
2. **Hidden state** — facts stored in globals, closure mutations, or chat text that no orchestrator can audit.

```
  ANTI-PATTERN                         FIX
  ─────────────────────────────────    ─────────────────────────
  5 agents for FAQ hours                 lookup or routing
  order_id in module global            explicit CaseBoard field
  "the model remembers" in prompt      checkpointer + named state
  handoffs only in free text           HandoffEnvelope + trace
```

This notebook **demonstrates** both anti-patterns on **ShopCo Support Hub** and **ReleaseFlow**, measures the damage, and refactors to explicit state and simpler patterns.

````python
"""
Anti-patterns lab — ShopCo + ReleaseFlow.
"""

import json
import os
import re
import uuid
from dataclasses import dataclass, field
from datetime import datetime, timezone
from enum import Enum
from typing import Any, Callable

OPENAI_API_KEY = os.environ.get("OPENAI_API_KEY", "sk-proj-placeholder-replace-with-your-real-key")
os.environ.setdefault("OPENAI_API_KEY", OPENAI_API_KEY)

USE_LIVE_LLM = False


def pretty(obj: Any) -> str:
    return json.dumps(obj, indent=2, default=str)


def utc_now() -> str:
    return datetime.now(timezone.utc).isoformat()


POLICY_SNIPPETS = {
    "returns": "Returns within 30 days. [pol-returns-01]",
    "shipping": "Free shipping over $50. [pol-shipping-02]",
    "hours": "Support hours Mon–Fri 9am–6pm ET. [faq-hours-05]",
}

ORDERS_DB = {
    "ORD-1001": {"order_id": "ORD-1001", "status": "shipped"},
    "ORD-1002": {"order_id": "ORD-1002", "status": "processing"},
}

print("Anti-patterns lab ready.")
````

## 1. Anti-Pattern Catalog

| Anti-pattern | Symptom | Preferred fix |
|--------------|---------|---------------|
| **Premature multi-agent** | High latency, same answer as routing | Routing / ReAct |
| **Hidden global state** | Wrong order_id on concurrent tickets | Explicit `CaseBoard` |
| **Chat-only memory** | Lost facts after trim | Named state fields |
| **Ghost agents** | Agent runs but output unchanged | Remove or merge role |
| **Pattern soup** | 4 patterns chained for FAQ | Simplest tier |
| **Unbounded handoffs** | Cost spike, no termination | `MAX_ROUNDS` + done predicate |

## 2. Premature Multi-Agent — ShopCo FAQ Hours

Five agents for a static FAQ is **pattern prestige**, not engineering.

````python
@dataclass
class RunMetrics:
    pattern: str
    agent_calls: int
    llm_calls_est: int
    answer: str
    trace_visible: bool


def premature_multi_agent_faq() -> RunMetrics:
    agents_run = ["triage", "researcher", "policy", "writer", "qa"]
    answer = POLICY_SNIPPETS["hours"]  # same answer a lookup returns
    return RunMetrics(
        pattern="premature_multi_agent",
        agent_calls=len(agents_run),
        llm_calls_est=len(agents_run) * 2,
        answer=answer,
        trace_visible=False,
    )


def lookup_faq() -> RunMetrics:
    return RunMetrics(
        pattern="lookup",
        agent_calls=0,
        llm_calls_est=0,
        answer=POLICY_SNIPPETS["hours"],
        trace_visible=True,
    )


bad = premature_multi_agent_faq()
good = lookup_faq()
print(f"Bad: {bad.agent_calls} agents, ~{bad.llm_calls_est} LLM calls")
print(f"Good: {good.agent_calls} agents, ~{good.llm_calls_est} LLM calls")
print(f"Same answer: {bad.answer == good.answer}")
````

## 3. Premature Multi-Agent — Order Status (Routing Wins)

````python
def multi_agent_order_status(order_id: str) -> RunMetrics:
    agents = ["triage", "order", "formatter", "qa"]
    status = ORDERS_DB.get(order_id, {}).get("status", "unknown")
    return RunMetrics("multi_agent", len(agents), 8, f"Order {order_id}: {status}", False)


def routing_order_status(order_id: str) -> RunMetrics:
    status = ORDERS_DB.get(order_id, {}).get("status", "unknown")
    return RunMetrics("routing", 1, 2, f"Order {order_id}: {status}", True)


m = multi_agent_order_status("ORD-1002")
r = routing_order_status("ORD-1002")
print(pretty({"multi_agent": m, "routing": r}))
````

## 4. Hidden State — Module Global `CURRENT_ORDER`

Globals break under **concurrent tickets** — classic hidden state bug.

````python
# ANTI-PATTERN: module-level mutable state
CURRENT_ORDER: str | None = None


def triage_hidden(message: str) -> None:
    global CURRENT_ORDER
    m = re.search(r"ORD-[0-9]{4}", message.upper())
    CURRENT_ORDER = m.group(0) if m else None


def order_agent_hidden() -> str:
    global CURRENT_ORDER
    status = ORDERS_DB.get(CURRENT_ORDER or "", {}).get("status", "unknown")
    return f"Order {CURRENT_ORDER}: {status}"


def simulate_concurrent_hidden() -> dict[str, str]:
    triage_hidden("Where is ORD-1001?")
    ticket_a = order_agent_hidden()
    triage_hidden("Where is ORD-1002?")  # overwrites global before A is "done"
    ticket_b = order_agent_hidden()
    triage_hidden("Where is ORD-1001?")
    ticket_a_late = order_agent_hidden()  # now wrong — reads 1001 but race possible
    return {"ticket_a_first": ticket_a, "ticket_b": ticket_b, "ticket_a_late": ticket_a_late}


print(pretty(simulate_concurrent_hidden()))
````

## 5. Fix — Explicit `CaseBoard` (No Globals)

````python
@dataclass
class CaseBoard:
    case_id: str
    customer_message: str = ""
    order_id: str | None = None
    order_status: str | None = None
    answer: str | None = None


def triage_explicit(board: CaseBoard) -> CaseBoard:
    board.customer_message = board.customer_message
    m = re.search(r"ORD-[0-9]{4}", board.customer_message.upper())
    board.order_id = m.group(0) if m else None
    return board


def order_agent_explicit(board: CaseBoard) -> CaseBoard:
    rec = ORDERS_DB.get(board.order_id or "", {})
    board.order_status = rec.get("status", "unknown")
    board.answer = f"Order {board.order_id}: {board.order_status}"
    return board


def handle_ticket_explicit(case_id: str, message: str) -> CaseBoard:
    board = CaseBoard(case_id=case_id, customer_message=message)
    triage_explicit(board)
    order_agent_explicit(board)
    return board


a = handle_ticket_explicit("CASE-A", "ORD-1001 status?")
b = handle_ticket_explicit("CASE-B", "ORD-1002 status?")
print(a.case_id, a.answer)
print(b.case_id, b.answer)
````

## 6. Hidden State — Facts Only in Chat History

Storing `order_id` only in messages means **trimming** or **wrong agent** loses the fact.

````python
def chat_only_flow() -> dict[str, Any]:
    messages = [
        {"role": "user", "content": "My order is ORD-1001"},
        {"role": "assistant", "content": "Noted."},
    ]
    # Simulate context trim — keep last message only
    trimmed = messages[-1:]
    order_in_trimmed = re.search(r"ORD-[0-9]{4}", " ".join(m["content"] for m in trimmed))
    return {
        "anti_pattern": "chat_only_state",
        "order_id_recoverable_after_trim": bool(order_in_trimmed),
        "trimmed_messages": trimmed,
    }


def named_state_flow() -> dict[str, Any]:
    state = {"messages": [{"role": "user", "content": "My order is ORD-1001"}], "order_id": "ORD-1001"}
    state["messages"] = state["messages"][-1:]  # trim chat
    return {
        "pattern": "named_state",
        "order_id_after_trim": state["order_id"],
    }


print(pretty(chat_only_flow()))
print(pretty(named_state_flow()))
````

## 7. ReleaseFlow — Hidden Draft in Closure

````python
def release_anti_pattern() -> dict[str, Any]:
    draft_box: list[str] = []  # hidden mutable closure

    def writer(text: str) -> None:
        draft_box.append(text)

    def gatekeeper() -> str:
        return draft_box[-1] if draft_box else "(empty)"

    writer("v2.4.0 — stuff fixed")
    writer("v2.4.0 — better draft")  # overwrites intent unclear
    return {
        "anti_pattern": "hidden_closure_list",
        "draft_visible_to_ops": False,
        "final": gatekeeper(),
        "all_drafts_lost": len(draft_box),
    }


@dataclass
class ReleaseBoard:
    release_id: str
    version: str = "2.4.0"
    draft: str | None = None
    revision: int = 0
    approved: bool = False


def release_explicit(board: ReleaseBoard, text: str) -> ReleaseBoard:
    board.revision += 1
    board.draft = text
    return board


def release_gate_explicit(board: ReleaseBoard) -> ReleaseBoard:
    board.approved = "breaking" in (board.draft or "").lower()
    return board


rb = ReleaseBoard(release_id="REL-1")
release_explicit(rb, "v2.4.0 — stuff fixed")
release_explicit(rb, "v2.4.0 — breaking: webhook deprecated")
release_gate_explicit(rb)
print("Anti-pattern:", pretty(release_anti_pattern()))
print("Explicit board:", pretty(rb))
````

## 8. Ghost Agents — No Effect on Output

````python
def run_with_ghost_agents(message: str) -> dict[str, Any]:
    agents_run = ["triage", "sentiment", "emoji_decorator", "order"]
    m = re.search(r"ORD-[0-9]{4}", message.upper())
    oid = m.group(0) if m else "N/A"
    answer = f"Status: {ORDERS_DB.get(oid, {}).get('status', 'unknown')}"
    # sentiment and emoji_decorator change nothing
    return {"agents_run": agents_run, "answer": answer, "ghost_agents": ["sentiment", "emoji_decorator"]}


def run_minimal(message: str) -> dict[str, Any]:
    m = re.search(r"ORD-[0-9]{4}", message.upper())
    oid = m.group(0) if m else "N/A"
    return {
        "agents_run": ["order"],
        "answer": f"Status: {ORDERS_DB.get(oid, {}).get('status', 'unknown')}",
    }


msg = "ORD-1001 status"
ghost = run_with_ghost_agents(msg)
minimal = run_minimal(msg)
print("Same answer:", ghost["answer"] == minimal["answer"])
print("Ghost overhead:", len(ghost["agents_run"]) - len(minimal["agents_run"]), "extra agents")
````

## 9. Anti-Pattern Detector

````python
class Smell(str, Enum):
    PREMATURE_MULTI_AGENT = "premature_multi_agent"
    HIDDEN_GLOBAL = "hidden_global"
    CHAT_ONLY_STATE = "chat_only_state"
    GHOST_AGENT = "ghost_agent"
    NO_TRACE = "no_trace"


@dataclass
class ArchitectureReview:
    system_name: str
    agent_count: int
    uses_explicit_state: bool
    has_trace: bool
    task_is_static_faq: bool
    ghost_agent_names: list[str] = field(default_factory=list)

    def detect(self) -> list[dict[str, str]]:
        findings: list[dict[str, str]] = []
        if self.task_is_static_faq and self.agent_count >= 3:
            findings.append({"smell": Smell.PREMATURE_MULTI_AGENT.value, "fix": "Use lookup or routing"})
        if not self.uses_explicit_state:
            findings.append({"smell": Smell.HIDDEN_GLOBAL.value, "fix": "CaseBoard / ReleaseBoard"})
        if not self.has_trace:
            findings.append({"smell": Smell.NO_TRACE.value, "fix": "CollaborationTrace per case"})
        for g in self.ghost_agent_names:
            findings.append({"smell": Smell.GHOST_AGENT.value, "fix": f"Remove agent '{g}'"})
        return findings


review_bad = ArchitectureReview(
    system_name="ShopCo-overbuilt",
    agent_count=5,
    uses_explicit_state=False,
    has_trace=False,
    task_is_static_faq=True,
    ghost_agent_names=["sentiment"],
)
review_good = ArchitectureReview(
    system_name="ShopCo-routing",
    agent_count=1,
    uses_explicit_state=True,
    has_trace=True,
    task_is_static_faq=True,
)
print("Bad:", pretty(review_bad.detect()))
print("Good:", pretty(review_good.detect()))
````

## 11. Before / After — ShopCo Support

````python
@dataclass
class RefactorReport:
    scenario: str
    before: RunMetrics
    after: RunMetrics

    def savings(self) -> dict[str, int]:
        return {
            "agents_removed": self.before.agent_calls - self.after.agent_calls,
            "llm_calls_saved_est": self.before.llm_calls_est - self.after.llm_calls_est,
        }


report_faq = RefactorReport(
    "FAQ hours",
    premature_multi_agent_faq(),
    lookup_faq(),
)
report_order = RefactorReport(
    "Order status",
    multi_agent_order_status("ORD-1001"),
    routing_order_status("ORD-1001"),
)

for r in (report_faq, report_order):
    print(f"{r.scenario}: saved {r.savings()}")
````

````python
def auditability_score(hidden: bool, has_revision_field: bool, has_trace: bool) -> int:
    score = 0
    if not hidden:
        score += 40
    if has_revision_field:
        score += 30
    if has_trace:
        score += 30
    return score


print("Hidden closure:", auditability_score(hidden=True, has_revision_field=False, has_trace=False))
print("ReleaseBoard:", auditability_score(hidden=False, has_revision_field=True, has_trace=True))
````

## 13. When Multi-Agent Is Justified

Multi-agent is appropriate when **all** apply:

1. **Distinct roles** with separate contracts (policy vs order vs QA)
2. **Measurable value** — QA catches real errors, not 0% reject rate
3. **Explicit shared board** — not globals or chat-only facts
4. **Trace** — handoffs logged for compliance

If only one intent and no quality gate → **routing**.

## 17. Summary

**Premature multi-agent** wastes agents on tasks routing or lookup handles. **Hidden state** — globals, closures, chat-only facts — causes wrong answers and impossible debugging.

Refactor to **explicit boards**, **visible traces**, and the **simplest pattern** that passes eval. Remove ghost agents. Document why each remaining role exists.

Multi-agent is a **scalpel**, not a default.
