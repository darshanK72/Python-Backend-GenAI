# Introduction to Agent Design Patterns

An **agent design pattern** is a reusable **control-flow template** for LLM-powered systems: how you route work, call tools, loop until done, split tasks across workers, or critique outputs before shipping.

Patterns are not frameworks — they are **architectural choices** you implement in plain Python, LangGraph, or any orchestrator. Picking the wrong pattern (e.g. five agents for a FAQ bot) wastes latency, cost, and debuggability.

```
  User request
       │
       ▼
  ┌─────────────┐     Which pattern?
  │  Pattern    │ ──► Routing / Chaining / ReAct / Plan-Execute / ...
  │  selection  │
  └──────┬──────┘
         ▼
  ┌─────────────┐
  │  Agent run  │  state + tools + memory + termination
  └─────────────┘
```

This notebook introduces the **pattern catalog** used throughout this module, maps two recurring scenarios — **ShopCo Support Hub** and **ReleaseFlow** — to candidate patterns, and runs **minimal plain-Python demos** of each pattern family so you can recognize them in the wild.

````python
"""
Agent design patterns — introduction lab.
"""

import json
import os
import re
import uuid
from dataclasses import dataclass, field
from datetime import datetime, timezone
from enum import Enum
from typing import Any, Callable

OPENAI_API_KEY = os.environ.get("OPENAI_API_KEY", "sk-proj-placeholder-replace-with-your-real-key")
os.environ.setdefault("OPENAI_API_KEY", OPENAI_API_KEY)

USE_LIVE_LLM = False
MAX_REACT_STEPS = 4


def pretty(obj: Any) -> str:
    return json.dumps(obj, indent=2, default=str)


def utc_now() -> str:
    return datetime.now(timezone.utc).isoformat()


POLICY_SNIPPETS = {
    "returns": "Returns within 30 days. [pol-returns-01]",
    "shipping": "Free shipping over $50. [pol-shipping-02]",
}

ORDERS_DB = {
    "ORD-1001": {"order_id": "ORD-1001", "status": "shipped"},
    "ORD-1002": {"order_id": "ORD-1002", "status": "processing"},
}

RELEASE_CHECKS = {
    "unit_tests": {"status": "pass", "detail": "412 tests passed"},
    "lint": {"status": "pass", "detail": "no issues"},
    "security_scan": {"status": "warn", "detail": "1 medium CVE in dev dependency"},
}

print("Pattern lab environment ready.")
````

## 1. What Makes Something an "Agent"?

| System | Control flow | Tools | Stateful |
|--------|--------------|-------|----------|
| **Single-shot prompt** | One call | No | No |
| **Prompt chain** | Fixed sequence | Maybe | Light |
| **Router** | Branch on intent | Per branch | Light |
| **ReAct agent** | Model decides loop | Yes | Yes |
| **Multi-agent** | Orchestrator + workers | Yes | Yes |

A pattern names the **shape** of that control flow. The LLM is one component inside the shape.

---

## 2. Building Blocks Every Pattern Composes

1. **State** — messages, scratch fields, retrieved docs
2. **Tools** — order lookup, policy search, CI status
3. **Termination** — max steps, success predicate, human gate
4. **Memory** — thread history, user profile (optional)

Patterns differ in **how** these blocks connect, not whether they exist.

````python
@dataclass
class AgentState:
    run_id: str = field(default_factory=lambda: f"run-{uuid.uuid4().hex[:8]}")
    messages: list[dict[str, str]] = field(default_factory=list)
    scratch: dict[str, Any] = field(default_factory=dict)
    steps: int = 0
    done: bool = False

    def add(self, role: str, content: str) -> None:
        self.messages.append({"role": role, "content": content, "ts": utc_now()})


state = AgentState()
state.add("user", "Where is ORD-1001?")
print("State keys:", list(state.__dataclass_fields__.keys()))
````

## 3. The Pattern Catalog

| Pattern | One-line description | Best when |
|---------|---------------------|-----------|
| **Prompt chaining** | Fixed A → B → C pipeline | Steps always run in order |
| **Routing** | Classify intent, branch | Disjoint specialists |
| **Parallelization** | Run independent tasks together | Latency-sensitive fan-out |
| **Orchestrator-workers** | Planner delegates subtasks | Complex multi-step cases |
| **ReAct** | Reason → act (tool) → observe loop | Dynamic tool selection |
| **Evaluator-optimizer** | Generate → score → revise | Quality bar matters |
| **Plan-and-execute** | Plan once, execute steps | Long horizons |
| **Reflection** | Self-critique then improve | Draft quality issues |

Later notebooks implement each pattern in depth. Here we **recognize** and **sketch** them.

````python
class PatternId(str, Enum):
    PROMPT_CHAINING = "prompt_chaining"
    ROUTING = "routing"
    PARALLELIZATION = "parallelization"
    ORCHESTRATOR_WORKERS = "orchestrator_workers"
    REACT = "react"
    EVALUATOR_OPTIMIZER = "evaluator_optimizer"
    PLAN_AND_EXECUTE = "plan_and_execute"
    REFLECTION = "reflection"


@dataclass
class PatternCard:
    pattern_id: PatternId
    name: str
    summary: str
    signals: list[str]
    risks: list[str]


PATTERN_CATALOG: list[PatternCard] = [
    PatternCard(PatternId.PROMPT_CHAINING, "Prompt Chaining",
                "Linear pipeline of LLM or function steps.",
                ["fixed stages", "ETL on text", "summarize then classify"],
                ["rigid if inputs vary wildly"]),
    PatternCard(PatternId.ROUTING, "Routing",
                "Classifier picks a specialist path.",
                ["disjoint intents", "different tools per path"],
                ["misroutes hurt UX"]),
    PatternCard(PatternId.PARALLELIZATION, "Parallelization",
                "Independent subtasks run concurrently.",
                ["fan-out lookups", "merge results"],
                ["merge logic complexity"]),
    PatternCard(PatternId.ORCHESTRATOR_WORKERS, "Orchestrator-Workers",
                "Lead agent delegates to workers.",
                ["multi-domain ticket", "dynamic subtasks"],
                ["coordination overhead"]),
    PatternCard(PatternId.REACT, "ReAct",
                "Loop: thought → tool → observation.",
                ["unknown tool sequence", "exploratory lookup"],
                ["runaway loops", "higher cost"]),
    PatternCard(PatternId.EVALUATOR_OPTIMIZER, "Evaluator-Optimizer",
                "Generate, score, revise until threshold.",
                ["policy compliance", "release notes quality"],
                ["extra latency per iteration"]),
    PatternCard(PatternId.PLAN_AND_EXECUTE, "Plan-and-Execute",
                "Explicit plan then step execution.",
                ["multi-step SOP", "release pipeline"],
                ["stale plan if world changes"]),
    PatternCard(PatternId.REFLECTION, "Reflection",
                "Agent critiques its own draft.",
                ["tone/quality polish", "code review style"],
                ["may over-edit"]),
]

print("Catalog:", [p.name for p in PATTERN_CATALOG])
````

## 4. Scenario A — ShopCo Support Hub

Customer tickets mix **policy questions**, **order status**, and **general FAQ**. Different patterns fit different ticket shapes.

````python
SHOPCO_TICKETS = [
    "Can I return ORD-1001?",
    "Where is my order ORD-1002?",
    "Summarize return policy then tell me if final-sale items qualify.",
    "I need help — not sure if this is billing or shipping.",
]


def recommend_pattern_shopco(message: str) -> PatternId:
    low = message.lower()
    if "summarize" in low and "then" in low:
        return PatternId.PROMPT_CHAINING
    if "not sure" in low or "billing or shipping" in low:
        return PatternId.ORCHESTRATOR_WORKERS
    if "return" in low and "ord-" in low:
        return PatternId.REACT  # may need policy + order tools in unknown order
    if "where is" in low or "status" in low:
        return PatternId.ROUTING
    return PatternId.ROUTING


for t in SHOPCO_TICKETS:
    print(f"{t[:50]:50} → {recommend_pattern_shopco(t).value}")
````

## 5. Scenario B — ReleaseFlow

**ReleaseFlow** automates release readiness: run checks, draft notes, gate on security, publish. Patterns map to pipeline stages.

````python
RELEASE_TASKS = [
    "Run unit tests and lint in parallel before merge",
    "Draft release notes then improve until tone matches style guide",
    "Plan v2.4 rollout: staging → canary → production",
    "Investigate which check failed and fix iteratively",
]


def recommend_pattern_release(task: str) -> PatternId:
    low = task.lower()
    if "parallel" in low:
        return PatternId.PARALLELIZATION
    if "improve until" in low or "style guide" in low:
        return PatternId.EVALUATOR_OPTIMIZER
    if "plan" in low and "rollout" in low:
        return PatternId.PLAN_AND_EXECUTE
    if "investigate" in low or "iteratively" in low:
        return PatternId.REACT
    return PatternId.PROMPT_CHAINING


for t in RELEASE_TASKS:
    print(f"{t[:45]:45} → {recommend_pattern_release(t).value}")
````

## 6. Mini-Demo — Routing Pattern

Classify → dispatch to a specialist function.

````python
def classify_intent(msg: str) -> str:
    low = msg.lower()
    if "return" in low:
        return "policy"
    if "order" in low or re.search(r"ord-[0-9]{4}", msg, re.I):
        return "order"
    return "general"


SPECIALISTS: dict[str, Callable[[str], str]] = {
    "policy": lambda m: POLICY_SNIPPETS["returns"] if "return" in m.lower() else POLICY_SNIPPETS["shipping"],
    "order": lambda m: (
        f"Status: {ORDERS_DB.get(re.search(r'ORD-[0-9]{4}', m.upper()).group(0), {}).get('status', 'unknown')}"
        if re.search(r"ORD-[0-9]{4}", m.upper()) else "Please provide order ID."
    ),
    "general": lambda m: "ShopCo Support — how can we help?",
}


def routing_agent(message: str) -> dict[str, Any]:
    intent = classify_intent(message)
    answer = SPECIALISTS[intent](message)
    return {"pattern": PatternId.ROUTING.value, "intent": intent, "answer": answer}


print(routing_agent("Can I return ORD-1001?"))
print(routing_agent("Where is ORD-1002?"))
````

## 7. Mini-Demo — Prompt Chaining

Step 1 output becomes step 2 input — order is fixed.

````python
def step_extract_topic(msg: str) -> str:
    return "returns" if "return" in msg.lower() else "shipping"


def step_fetch_policy(topic: str) -> str:
    return POLICY_SNIPPETS.get(topic, "No policy found.")


def step_format_answer(policy: str, msg: str) -> str:
    return f"Re your question ('{msg[:40]}...'): {policy}"


def prompt_chain(message: str) -> dict[str, Any]:
    topic = step_extract_topic(message)
    policy = step_fetch_policy(topic)
    answer = step_format_answer(policy, message)
    return {"pattern": PatternId.PROMPT_CHAINING.value, "topic": topic, "answer": answer}


print(prompt_chain("What is the return window?"))
````

## 8. Mini-Demo — ReAct Loop

The agent chooses tools until it can answer or hits `MAX_REACT_STEPS`.

````python
TOOLS: dict[str, Callable[[str], str]] = {
    "lookup_order": lambda oid: f"order:{ORDERS_DB.get(oid, {'status': 'not_found'})}",
    "search_policy": lambda q: POLICY_SNIPPETS["returns"] if "return" in q.lower() else POLICY_SNIPPETS["shipping"],
}


def plan_tool_call(message: str, observations: list[str]) -> tuple[str | None, str | None]:
    if not any(o.startswith("order:") for o in observations):
        m = re.search(r"ORD-[0-9]{4}", message.upper())
        if m:
            return "lookup_order", m.group(0)
    if "return" in message.lower() and not any("pol-" in o for o in observations):
        return "search_policy", "returns"
    return None, None


def react_agent(message: str) -> dict[str, Any]:
    st = AgentState()
    st.add("user", message)
    observations: list[str] = []
    trace: list[str] = []

    while st.steps < MAX_REACT_STEPS:
        tool, arg = plan_tool_call(message, observations)
        if tool is None:
            break
        obs = TOOLS[tool](arg or "")
        observations.append(obs)
        trace.append(f"act:{tool}({arg}) → {obs[:60]}")
        st.steps += 1

    answer = " | ".join(observations) if observations else "Could not resolve."
    return {"pattern": PatternId.REACT.value, "steps": st.steps, "trace": trace, "answer": answer}


print(pretty(react_agent("Can I return ORD-1001?")))
````

## 9. Mini-Demo — Parallelization

ReleaseFlow runs independent checks, then merges.

````python
def run_check(name: str) -> dict[str, str]:
    rec = RELEASE_CHECKS[name]
    return {"check": name, **rec}


def parallel_release_gate(check_names: list[str]) -> dict[str, Any]:
    # Simulated parallel — independent results merged in Python
    results = [run_check(n) for n in check_names]
    blocking = [r for r in results if r["status"] == "fail"]
    warnings = [r for r in results if r["status"] == "warn"]
    gate = "block" if blocking else ("warn" if warnings else "pass")
    return {"pattern": PatternId.PARALLELIZATION.value, "gate": gate, "results": results}


print(pretty(parallel_release_gate(["unit_tests", "lint", "security_scan"])))
````

## 10. Mini-Demo — Orchestrator-Workers

Lead agent decomposes an ambiguous ticket into worker tasks.

````python
def orchestrator_plan(message: str) -> list[str]:
    tasks = []
    if "billing" in message.lower():
        tasks.append("policy:billing")
    if "shipping" in message.lower():
        tasks.append("policy:shipping")
    if not tasks:
        tasks.append("general:triage")
    return tasks


WORKERS: dict[str, Callable[[], str]] = {
    "policy:billing": lambda: "Charges appear as SHOPCO* on statements. [pol-billing-03]",
    "policy:shipping": lambda: POLICY_SNIPPETS["shipping"],
    "general:triage": lambda: "Please clarify billing or shipping.",
}


def orchestrator_workers(message: str) -> dict[str, Any]:
    tasks = orchestrator_plan(message)
    outputs = {t: WORKERS[t]() for t in tasks}
    merged = " ".join(outputs.values())
    return {"pattern": PatternId.ORCHESTRATOR_WORKERS.value, "tasks": tasks, "answer": merged}


print(orchestrator_workers("Not sure if this is billing or shipping issue."))
````

## 11. Mini-Demo — Evaluator-Optimizer

Generate release notes, score, revise until score threshold.

````python
def draft_release_notes(version: str) -> str:
    return f"v{version} - stuff fixed and things added."


def score_notes(text: str) -> int:
    score = 0
    if re.search(r"v\d+\.\d+", text):
        score += 30
    if "security" in text.lower() or "CVE" in text:
        score += 25
    if len(text.split()) >= 8:
        score += 25
    if not re.search(r"stuff|things", text.lower()):
        score += 20
    return score


def optimize_notes(draft: str) -> str:
    return draft.replace("stuff fixed and things added", "bug fixes, performance improvements, and security patches including CVE remediation")


def evaluator_optimizer(version: str, threshold: int = 80) -> dict[str, Any]:
    draft = draft_release_notes(version)
    iterations: list[dict[str, Any]] = []
    for i in range(3):
        s = score_notes(draft)
        iterations.append({"round": i + 1, "score": s, "draft": draft})
        if s >= threshold:
            break
        draft = optimize_notes(draft)
    return {"pattern": PatternId.EVALUATOR_OPTIMIZER.value, "final": draft, "iterations": iterations}


print(pretty(evaluator_optimizer("2.4.0")))
````

## 12. Mini-Demo — Plan-and-Execute

Explicit plan steps, then execute each.

````python
ROLLOUT_PLAN = [
    {"step": 1, "action": "deploy_staging", "detail": "Deploy v2.4.0 to staging"},
    {"step": 2, "action": "smoke_tests", "detail": "Run smoke suite on staging"},
    {"step": 3, "action": "canary", "detail": "Route 5% traffic to v2.4.0"},
    {"step": 4, "action": "full_rollout", "detail": "Promote to 100% production"},
]


def execute_step(action: str) -> str:
    outcomes = {
        "deploy_staging": "staging OK",
        "smoke_tests": "smoke pass",
        "canary": "canary healthy",
        "full_rollout": "production updated",
    }
    return outcomes.get(action, "unknown")


def plan_and_execute(plan: list[dict[str, Any]]) -> dict[str, Any]:
    log = []
    for item in plan:
        result = execute_step(item["action"])
        log.append({**item, "result": result})
    return {"pattern": PatternId.PLAN_AND_EXECUTE.value, "execution_log": log}


print(pretty(plan_and_execute(ROLLOUT_PLAN)))
````

## 13. Mini-Demo — Reflection

Draft → critique → revised draft.

````python
def draft_support_reply(order_status: str) -> str:
    return f"Your order is {order_status}. Deal with it."


def reflect_on_draft(text: str) -> list[str]:
    issues = []
    if "deal with it" in text.lower():
        issues.append("tone too harsh")
    if "[" not in text:
        issues.append("missing policy citation")
    return issues


def revise_draft(text: str, issues: list[str]) -> str:
    revised = text.replace("Deal with it.", "Let us know if you need further help.")
    if "missing policy citation" in issues:
        revised += f" {POLICY_SNIPPETS['shipping']}"
    return revised


def reflection_agent(order_status: str) -> dict[str, Any]:
    draft = draft_support_reply(order_status)
    issues = reflect_on_draft(draft)
    final = revise_draft(draft, issues) if issues else draft
    return {"pattern": PatternId.REFLECTION.value, "draft": draft, "issues": issues, "final": final}


print(pretty(reflection_agent("shipped")))
````

## 14. Unified Runner — Pattern Dispatch

A thin facade picks a pattern implementation from the catalog recommendation.

````python
PATTERN_RUNNERS: dict[PatternId, Callable[[str], dict[str, Any]]] = {
    PatternId.ROUTING: routing_agent,
    PatternId.PROMPT_CHAINING: prompt_chain,
    PatternId.REACT: react_agent,
    PatternId.ORCHESTRATOR_WORKERS: orchestrator_workers,
    PatternId.EVALUATOR_OPTIMIZER: lambda _: evaluator_optimizer("2.4.0"),
    PatternId.PLAN_AND_EXECUTE: lambda _: plan_and_execute(ROLLOUT_PLAN),
    PatternId.REFLECTION: lambda _: reflection_agent("shipped"),
    PatternId.PARALLELIZATION: lambda _: parallel_release_gate(["unit_tests", "lint"]),
}


def run_with_recommended_pattern(message: str, domain: str = "shopco") -> dict[str, Any]:
    pid = recommend_pattern_shopco(message) if domain == "shopco" else recommend_pattern_release(message)
    result = PATTERN_RUNNERS[pid](message)
    result["recommended_pattern"] = pid.value
    return result


print(pretty(run_with_recommended_pattern("Can I return ORD-1001?")))
````

## 15. Patterns vs Anti-Patterns (Preview)

| Anti-pattern | Why it fails | Prefer |
|--------------|--------------|--------|
| Premature multi-agent | 3 agents for FAQ | Routing or chain |
| Hidden shared state | Untraceable bugs | Explicit `AgentState` |
| Unbounded ReAct | Cost + loops | `MAX_REACT_STEPS` + success check |
| Pattern soup | Every pattern at once | Simplest that works |

## 16. Optional Live LLM Router

Set `USE_LIVE_LLM = True` to classify ShopCo tickets with `gpt-4o-mini`.

````python
def classify_intent_live(message: str) -> str:
    from langchain_core.messages import HumanMessage, SystemMessage
    from langchain_openai import ChatOpenAI

    llm = ChatOpenAI(model="gpt-4o-mini", api_key=OPENAI_API_KEY, temperature=0)
    resp = llm.invoke([
        SystemMessage(content="Classify as: policy, order, or general. Reply one word only."),
        HumanMessage(content=message),
    ])
    return str(resp.content).strip().lower()


msg = "Is ORD-1001 eligible for return?"
if USE_LIVE_LLM:
    print("Live intent:", classify_intent_live(msg))
else:
    print("Rule intent:", classify_intent(msg))
````

## 17. Common Mistakes

| Mistake | Symptom | Fix |
|---------|---------|-----|
| Pattern before problem | Over-engineered graph | Map user journey first |
| No termination | Infinite tool loop | Step cap + done predicate |
| Opaque state | Can't debug misroutes | Log pattern + trace |
| Same pattern everywhere | Slow FAQ path | Routing for disjoint intents |
| Skip evaluator on compliance | Bad policy answers | Evaluator-optimizer on citations |

## 19. Summary

**Agent design patterns** name reusable control-flow shapes: chain, route, parallelize, orchestrate, react, evaluate, plan, reflect. They help you **communicate architecture** and **avoid reinventing** fragile one-off loops.

This notebook built a **pattern catalog**, mapped **ShopCo** and **ReleaseFlow** tasks to recommendations, and ran minimal implementations of each family in plain Python.

Start with the **simplest pattern that solves the problem** — then adopt richer patterns when routing errors, quality bars, or multi-step SOPs demand them.
