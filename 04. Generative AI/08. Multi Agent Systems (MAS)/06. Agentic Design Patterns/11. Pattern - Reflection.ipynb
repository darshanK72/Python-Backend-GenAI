# Pattern — Reflection

**Reflection** is a generate → **critique** → **revise** loop. The agent (or a paired critic) inspects its own draft, names concrete issues, and produces an improved version — often for **tone**, **structure**, or **completeness** rather than numeric scores alone.

```
  Task + context
       │
       ▼
  ┌──────────────┐
  │   GENERATE   │  first draft
  └──────┬───────┘
         ▼
  ┌──────────────┐
  │   REFLECT    │  "missing citation", "tone too harsh"
  └──────┬───────┘
         ▼
  ┌──────────────┐
  │   REVISE     │  patched draft
  └──────┬───────┘
         │ repeat until no issues or max_rounds
         ▼
     Final output
```

| Pattern | Feedback style |
|---------|----------------|
| **Single-shot** | None |
| **Reflection** | Qualitative critique list |
| **Evaluator-optimizer** | Numeric score + threshold |

This notebook implements reflection for **ShopCo Support Hub** replies and **ReleaseFlow** announcements in plain Python — self-reflection, separate critic role, bounded loops, and traces.

````python
"""
Reflection pattern lab — ShopCo + ReleaseFlow.
"""

import json
import os
import re
import uuid
from dataclasses import dataclass, field
from datetime import datetime, timezone
from enum import Enum
from typing import Any, Callable

OPENAI_API_KEY = os.environ.get("OPENAI_API_KEY", "sk-proj-placeholder-replace-with-your-real-key")
os.environ.setdefault("OPENAI_API_KEY", OPENAI_API_KEY)

USE_LIVE_LLM = False
MAX_REFLECTION_ROUNDS = 3


def pretty(obj: Any) -> str:
    return json.dumps(obj, indent=2, default=str)


def utc_now() -> str:
    return datetime.now(timezone.utc).isoformat()


POLICY_SNIPPETS = {
    "returns": "Returns within 30 days with receipt. [pol-returns-01]",
    "shipping": "Free shipping on orders over $50. [pol-shipping-02]",
}

ORDERS_DB = {
    "ORD-1001": {"order_id": "ORD-1001", "status": "shipped"},
    "ORD-1002": {"order_id": "ORD-1002", "status": "processing"},
}

CHANGELOG = [
    {"version": "2.4.0", "item": "Added bulk export endpoint"},
    {"version": "2.4.0", "item": "Fixed session timeout on mobile"},
    {"version": "2.4.0", "item": "Deprecated legacy v1 webhook format"},
]

STYLE_GUIDE = {
    "must_include": ["version", "breaking", "support"],
    "support_channel": "#releases",
    "forbidden_phrases": ["stuff", "things", "deal with it"],
}

print("Reflection lab ready.")
````

---

## 1. When Reflection Fits

| Good fit | Poor fit |
|----------|----------|
| Tone and empathy matter | Exact numeric SLA scoring → evaluator-optimizer |
| Checklist-style critique (citations, sections) | Single factual lookup |
| Draft quality varies run to run | Fixed template always correct |
| One or two revision passes help | Needs formal pass/fail metrics only |

Reflection teaches the model (or pipeline) to **articulate what's wrong** before fixing it — useful for debugging and audit logs.

---

## 2. Core Types — Critique and Reflection Round

````python
@dataclass
class CritiqueItem:
    issue_id: str
    category: str
    message: str
    severity: str  # low | medium | high


@dataclass
class ReflectionRound:
    round_num: int
    draft: str
    critiques: list[CritiqueItem]
    revised: str | None = None
    accepted: bool = False


@dataclass
class ReflectionResult:
    task_id: str
    final_text: str
    rounds: list[ReflectionRound]
    total_revisions: int

    def trace_summary(self) -> list[str]:
        lines = []
        for r in self.rounds:
            issues = [c.message for c in r.critiques]
            lines.append(f"Round {r.round_num}: {len(issues)} issues → accepted={r.accepted}")
        return lines


print("Types ready.")
````

---

## 3. Generator — First Draft (ShopCo)

````python
def draft_shopco_reply(customer_message: str, order_id: str | None = None) -> str:
    """Intentionally imperfect first draft for reflection demos."""
    if "return" in customer_message.lower():
        if order_id:
            status = ORDERS_DB.get(order_id, {}).get("status", "unknown")
            return f"Your order is {status}. Maybe you can return it. Deal with it if not."
        return "Returns are allowed sometimes."
    if order_id:
        status = ORDERS_DB.get(order_id, {}).get("status", "unknown")
        return f"Order {order_id} is {status}."
    return "Thanks for contacting ShopCo."


sample_draft = draft_shopco_reply("Can I return ORD-1001?", "ORD-1001")
print("Raw draft:", sample_draft)
````

---

## 4. Reflector — Rule-Based Critique Rules

````python
def reflect_shopco(draft: str, context: dict[str, Any]) -> list[CritiqueItem]:
    critiques: list[CritiqueItem] = []
    low = draft.lower()

    if context.get("needs_policy") and "[pol-" not in draft:
        critiques.append(CritiqueItem(
            "c1", "citation", "Missing policy citation [pol-*]", "high",
        ))
    if any(p in low for p in ["deal with it", "maybe", "sometimes"]):
        critiques.append(CritiqueItem(
            "c2", "tone", "Tone is vague or harsh — use professional empathy", "high",
        ))
    if len(draft.split()) < 12 and context.get("needs_policy"):
        critiques.append(CritiqueItem(
            "c3", "completeness", "Reply too brief for policy question", "medium",
        ))
    if "return" in context.get("topic", "") and "30" not in draft and "[pol-" not in draft:
        critiques.append(CritiqueItem(
            "c4", "accuracy", "Return window not stated", "medium",
        ))
    return critiques


ctx = {"needs_policy": True, "topic": "return"}
print(pretty(reflect_shopco(sample_draft, ctx)))
````

---

## 5. Reviser — Apply Critiques to Draft

````python
def revise_shopco(draft: str, critiques: list[CritiqueItem], context: dict[str, Any]) -> str:
    text = draft
    cats = {c.category for c in critiques}

    if "tone" in cats:
        text = text.replace("Deal with it if not.", "Contact us if you need further assistance.")
        text = text.replace("Maybe you can return it.", "You may be eligible for a return.")
        text = text.replace("Returns are allowed sometimes.", "Our returns policy may apply to your order.")
    if "citation" in cats or "accuracy" in cats:
        text += f" {POLICY_SNIPPETS['returns']}"
    if "completeness" in cats and "[pol-" in text:
        oid = context.get("order_id", "")
        if oid:
            text = f"Regarding order {oid}: {text}"
    return text.strip()


crits = reflect_shopco(sample_draft, ctx)
revised = revise_shopco(sample_draft, crits, {**ctx, "order_id": "ORD-1001"})
print("Revised:", revised)
````

---

## 6. Reflection Engine — Bounded Loop

````python
ReflectFn = Callable[[str, dict[str, Any]], list[CritiqueItem]]
ReviseFn = Callable[[str, list[CritiqueItem], dict[str, Any]], str]
GenerateFn = Callable[[dict[str, Any]], str]


class ReflectionEngine:
    def __init__(
        self,
        generate: GenerateFn,
        reflect: ReflectFn,
        revise: ReviseFn,
        max_rounds: int = MAX_REFLECTION_ROUNDS,
        accept_if_no_high: bool = True,
    ) -> None:
        self.generate = generate
        self.reflect = reflect
        self.revise = revise
        self.max_rounds = max_rounds
        self.accept_if_no_high = accept_if_no_high

    def run(self, task_id: str, context: dict[str, Any]) -> ReflectionResult:
        rounds: list[ReflectionRound] = []
        draft = self.generate(context)

        for n in range(1, self.max_rounds + 1):
            critiques = self.reflect(draft, context)
            high = [c for c in critiques if c.severity == "high"]
            accepted = len(critiques) == 0 or (self.accept_if_no_high and len(high) == 0)
            revised = None if accepted else self.revise(draft, critiques, context)
            rounds.append(ReflectionRound(n, draft, critiques, revised, accepted))
            if accepted:
                return ReflectionResult(task_id, draft, rounds, n - 1)
            draft = revised or draft

        return ReflectionResult(task_id, draft, rounds, len(rounds))


def gen_shopco(ctx: dict[str, Any]) -> str:
    return draft_shopco_reply(ctx["message"], ctx.get("order_id"))


shopco_engine = ReflectionEngine(gen_shopco, reflect_shopco, revise_shopco)
print("Engine ready.")
````

---

## 7. Demo — ShopCo Return Ticket

````python
shopco_result = shopco_engine.run(
    "CASE-RET-01",
    {"message": "Can I return ORD-1001?", "order_id": "ORD-1001", "needs_policy": True, "topic": "return"},
)
print("Final:", shopco_result.final_text)
print("Revisions:", shopco_result.total_revisions)
for line in shopco_result.trace_summary():
    print(" ", line)
````

## 8. ReleaseFlow — Announcement Reflection

````python
def draft_release_announcement(version: str) -> str:
    items = ", ".join(c["item"] for c in CHANGELOG if c["version"] == version)
    return f"v{version} shipped. We fixed stuff and added things. Changes: {items}."


def reflect_release(draft: str, context: dict[str, Any]) -> list[CritiqueItem]:
    critiques: list[CritiqueItem] = []
    low = draft.lower()
    version = context.get("version", "")

    if version and version not in draft and f"v{version}" not in draft:
        critiques.append(CritiqueItem("r1", "version", "Version number not prominent", "high"))
    if "breaking" not in low and "deprecated" not in low:
        critiques.append(CritiqueItem("r2", "structure", "Missing breaking changes section", "high"))
    if STYLE_GUIDE["support_channel"] not in draft and "support" not in low:
        critiques.append(CritiqueItem("r3", "structure", "Missing support channel", "medium"))
    for phrase in STYLE_GUIDE["forbidden_phrases"]:
        if phrase in low:
            critiques.append(CritiqueItem("r4", "tone", f"Forbidden phrase: '{phrase}'", "high"))
    return critiques


def revise_release(draft: str, critiques: list[CritiqueItem], context: dict[str, Any]) -> str:
    version = context.get("version", "2.4.0")
    text = draft
    cats = {c.category for c in critiques}
    if "tone" in cats:
        text = text.replace("fixed stuff and added things", "includes bug fixes and new features")
    if "version" in cats:
        text = f"Release {version} is now available. " + text
    if "structure" in cats:
        if "breaking" not in text.lower():
            text += "\n\nBreaking changes: legacy v1 webhook format is deprecated."
        if STYLE_GUIDE["support_channel"] not in text:
            text += f"\nQuestions? Join {STYLE_GUIDE['support_channel']}."
    return text.strip()


release_engine = ReflectionEngine(
    lambda ctx: draft_release_announcement(ctx["version"]),
    reflect_release,
    revise_release,
)

release_result = release_engine.run("REL-2.4.0", {"version": "2.4.0"})
print(release_result.final_text)
````

## 9. Self-Reflection vs Separate Critic Agent

| Mode | Pros | Cons |
|------|------|------|
| **Self-reflection** | One model, simpler deploy | May repeat same blind spots |
| **Separate critic** | Fresh perspective, role clarity | Extra latency + orchestration |

Same engine — swap the `reflect` function for a critic persona.

````python
def critic_agent_reflect(draft: str, context: dict[str, Any]) -> list[CritiqueItem]:
    """Separate 'critic' — stricter on empathy keywords."""
    base = reflect_shopco(draft, context)
    if "thank" not in draft.lower() and "contact" not in draft.lower():
        base.append(CritiqueItem("cx", "tone", "Critic: add courteous closing", "medium"))
    return base


def revise_with_closing(draft: str, critiques: list[CritiqueItem], context: dict[str, Any]) -> str:
    text = revise_shopco(draft, critiques, context)
    if any(c.message.startswith("Critic:") for c in critiques):
        text += " Thank you for choosing ShopCo."
    return text


critic_engine = ReflectionEngine(gen_shopco, critic_agent_reflect, revise_with_closing)
critic_result = critic_engine.run(
    "CASE-CRITIC",
    {"message": "Return question", "order_id": "ORD-1001", "needs_policy": True, "topic": "return"},
)
print(critic_result.final_text)
````

## 10. Reflection vs Evaluator-Optimizer

```
  Reflection:     draft → critique items → revise (qualitative)
  Eval-optimizer:   draft → score 72/100 → revise until score ≥ 80 (quantitative)
```

Use **reflection** when issues are **named in prose** (tone, missing section). Use **evaluator-optimizer** when you have **measurable rubrics** and pass thresholds.

````python
def score_release(draft: str) -> int:
    s = 0
    if re.search(r"2\.4\.0", draft):
        s += 30
    if "breaking" in draft.lower():
        s += 25
    if "#releases" in draft:
        s += 25
    if not re.search(r"stuff|things", draft.lower()):
        s += 20
    return s


draft = draft_release_announcement("2.4.0")
print("Draft score:", score_release(draft))
refined = release_engine.run("cmp", {"version": "2.4.0"}).final_text
print("After reflection score:", score_release(refined))
print("Reflection issues were qualitative; score improved as side effect.")
````

## 11. Termination and Safety Caps

- `MAX_REFLECTION_ROUNDS` prevents infinite polish loops
- Accept when **no high-severity** critiques remain (configurable)
- Log every round for operators

````python
def render_reflection_timeline(result: ReflectionResult) -> str:
    lines = [f"Task {result.task_id}"]
    for r in result.rounds:
        lines.append(f"  Round {r.round_num}: draft={r.draft[:50]}...")
        for c in r.critiques:
            lines.append(f"    - [{c.severity}] {c.category}: {c.message}")
        if r.revised:
            lines.append(f"    → revised: {r.revised[:50]}...")
    lines.append(f"  FINAL: {result.final_text[:80]}...")
    return "\n".join(lines)


print(render_reflection_timeline(shopco_result))
````

## 12. Optional Live LLM Reflection

Set `USE_LIVE_LLM = True` for critique via `gpt-4o-mini` (JSON issues list).

````python
def reflect_live(draft: str, context: dict[str, Any]) -> list[CritiqueItem]:
    from langchain_core.messages import HumanMessage, SystemMessage
    from langchain_openai import ChatOpenAI

    llm = ChatOpenAI(model="gpt-4o-mini", api_key=OPENAI_API_KEY, temperature=0)
    resp = llm.invoke([
        SystemMessage(content=(
            "Critique this support draft. Return JSON list of {issue_id, category, message, severity}. "
            "Check tone, policy citations [pol-*], completeness."
        )),
        HumanMessage(content=draft),
    ])
    try:
        raw = json.loads(str(resp.content))
        return [CritiqueItem(**item) for item in raw]
    except (json.JSONDecodeError, TypeError):
        return reflect_shopco(draft, context)


if USE_LIVE_LLM:
    live_engine = ReflectionEngine(gen_shopco, reflect_live, revise_shopco)
    print(live_engine.run("LIVE", {"message": "return ORD-1001", "order_id": "ORD-1001", "needs_policy": True, "topic": "return"}).final_text)
else:
    print("Offline reflector:", [c.message for c in reflect_shopco(sample_draft, ctx)])
````

## 13. Common Mistakes

| Mistake | Symptom | Fix |
|---------|---------|-----|
| Unbounded reflection | Cost + over-edited text | `MAX_REFLECTION_ROUNDS` |
| Vague critiques | Revision doesn't help | Structured `CritiqueItem` categories |
| Reflection for facts | Still wrong policy | Ground with retrieval first |
| No trace | Can't audit tone fixes | Log `ReflectionRound` |
| Confuse with eval-opt | Arbitrary loops | Pick qualitative vs score threshold |

## 15. Summary

**Reflection** improves drafts through explicit **critique → revise** cycles. ShopCo support replies gain citations and professional tone; ReleaseFlow announcements gain version headers, breaking-change sections, and support channels.

The `ReflectionEngine` bounds rounds, logs traces, and supports self-reflection or a separate critic. Pair reflection with **grounded retrieval** for facts — reflection polishes presentation, not unknown policy text.

When you need numeric pass thresholds, use **evaluator-optimizer**; when you need named, explainable fixes, use **reflection**.
