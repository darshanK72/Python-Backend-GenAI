# Pattern — Orchestrator-Workers

The **orchestrator-workers** pattern places a central **orchestrator** in charge of decomposing a request, assigning **specialist workers**, and **synthesizing** their outputs into a final result. Workers are narrow — each owns a domain, tool set, or subtask.

```
  Complex customer ticket
           │
           ▼
    ┌──────────────┐
    │ ORCHESTRATOR │  plan + delegate + merge
    └──────┬───────┘
           │
     ┌─────┼─────┬─────────┐
     ▼     ▼     ▼         ▼
  Order  Policy Billing  Escalation
  worker worker worker   worker
     │     │     │         │
     └─────┴─────┴─────────┘
           ▼
    Synthesized reply
```

Compared to **routing** (pick one handler) and **parallelization** (fixed fan-out), the orchestrator **dynamically plans** which workers run and in what order.

This notebook implements orchestrator-workers for **ShopCo Support Hub** and **ReleaseFlow** in plain Python — task plans, worker registry, parallel dispatch, synthesis, and failure handling.

````python
"""
Orchestrator-Workers Pattern Lab — ShopCo + ReleaseFlow.
"""

import json
import re
import time
import uuid
from concurrent.futures import ThreadPoolExecutor, as_completed
from dataclasses import dataclass, field
from datetime import datetime, timezone
from enum import Enum
from typing import Any, Callable

USE_LIVE_LLM = False
MAX_WORKERS_PER_RUN = 4


def pretty(obj: Any) -> str:
    return json.dumps(obj, indent=2, default=str)


def utcnow() -> str:
    return datetime.now(timezone.utc).isoformat()


ORDERS: dict[str, dict[str, Any]] = {
    "ORD-1001": {"order_id": "ORD-1001", "customer": "alice@shop.com", "status": "shipped", "total_usd": 1299.0},
    "ORD-1002": {"order_id": "ORD-1002", "customer": "bob@shop.com", "status": "processing", "total_usd": 449.0},
}

POLICIES: dict[str, str] = {
    "returns": "30-day returns. [pol-returns-01]",
    "refunds": "Escalate over $1000. [pol-refunds-03]",
}

BILLING: dict[str, dict[str, Any]] = {
    "ORD-1001": {"refund_eligible": True},
    "ORD-1002": {"refund_eligible": True},
}

RELEASE_CHECKS: dict[str, str] = {
    "unit_tests": "pass",
    "security_scan": "warn",
    "lint": "pass",
}

ORCH_TRACE: list[dict[str, Any]] = []

print("Orchestrator-workers lab ready")
````

---

## 1. Orchestrator vs Router vs Parallel Fan-Out

| Pattern | Who plans work? | Workers invoked |
|---------|-----------------|-----------------|
| **Routing** | Classifier picks 1 route | 1 handler |
| **Parallelization** | Developer fixed fan-out | All branches always |
| **Orchestrator-workers** | Orchestrator per request | Subset, dynamic plan |

The orchestrator adds a **planning** step: decompose → assign → execute → synthesize.

---

## 2. Worker Model — Specialist Agents

````python
@dataclass
class WorkerResult:
    worker_id: str
    success: bool
    output: str
    data: dict[str, Any] = field(default_factory=dict)
    error: str | None = None
    duration_ms: float = 0.0


@dataclass
class WorkerSpec:
    worker_id: str
    description: str
    handler: Callable[[dict[str, Any]], WorkerResult]
    capabilities: list[str] = field(default_factory=list)


print("Worker model ready")
````

## 3. ShopCo Workers

````python
def simulate_work(ms: float = 30) -> None:
    time.sleep(ms / 1000.0)


def order_worker(ctx: dict[str, Any]) -> WorkerResult:
    simulate_work(40)
    oid = ctx.get("order_id")
    if not oid:
        return WorkerResult("order", False, "No order ID provided.", error="missing order_id")
    order = ORDERS.get(oid)
    if not order:
        return WorkerResult("order", False, f"Order {oid} not found.", error="not found")
    return WorkerResult(
        "order", True,
        f"{oid} is {order['status']} (${order['total_usd']}).",
        {"order": order},
    )


def policy_worker(ctx: dict[str, Any]) -> WorkerResult:
    simulate_work(25)
    intent = ctx.get("intent", "general")
    key = "refunds" if intent == "refund" else "returns"
    return WorkerResult("policy", True, POLICIES[key], {"policy_key": key})


def billing_worker(ctx: dict[str, Any]) -> WorkerResult:
    simulate_work(35)
    oid = ctx.get("order_id")
    if not oid:
        return WorkerResult("billing", False, "Billing check needs order_id.", error="missing order_id")
    bill = BILLING.get(oid)
    if not bill:
        return WorkerResult("billing", False, "No billing record.", error="not found")
    return WorkerResult("billing", True, f"Refund eligible: {bill['refund_eligible']}.", {"billing": bill})


def escalation_worker(ctx: dict[str, Any]) -> WorkerResult:
    simulate_work(20)
    ticket = f"ESC-{uuid.uuid4().hex[:6].upper()}"
    return WorkerResult("escalation", True, f"Escalation ticket {ticket} created.", {"ticket": ticket})


WORKER_REGISTRY: dict[str, WorkerSpec] = {
    "order": WorkerSpec("order", "Order lookup", order_worker, ["order_status", "tracking"]),
    "policy": WorkerSpec("policy", "Policy citation", policy_worker, ["returns", "refunds"]),
    "billing": WorkerSpec("billing", "Billing/refund eligibility", billing_worker, ["refund"]),
    "escalation": WorkerSpec("escalation", "Human handoff", escalation_worker, ["escalate"]),
}

print(f"Workers registered: {list(WORKER_REGISTRY)}")
````

## 4. Task Plan — Orchestrator Decomposition

The orchestrator converts raw input into a **plan**: which workers to invoke and shared context for each.

````python
@dataclass
class WorkerAssignment:
    worker_id: str
    context: dict[str, Any]
    priority: int = 0


@dataclass
class TaskPlan:
    plan_id: str
    assignments: list[WorkerAssignment]
    synthesis_mode: str = "merge"  # merge | first_success | escalate_on_fail


class ShopCoOrchestratorPlanner:
    """Rule-based planner — production may use LLM for decomposition."""

    def plan(self, message: str) -> TaskPlan:
        msg = message.lower()
        oid_match = re.search(r"ORD-\d+", message.upper())
        order_id = oid_match.group() if oid_match else None
        intent = "general"
        if "refund" in msg:
            intent = "refund"
        elif "return" in msg:
            intent = "return"
        elif order_id:
            intent = "order_status"

        base_ctx = {"order_id": order_id, "intent": intent, "raw_message": message}
        assignments: list[WorkerAssignment] = []

        if order_id:
            assignments.append(WorkerAssignment("order", base_ctx.copy(), priority=1))
        if intent in ("refund", "return"):
            assignments.append(WorkerAssignment("policy", base_ctx.copy(), priority=2))
            if order_id:
                assignments.append(WorkerAssignment("billing", base_ctx.copy(), priority=2))
        if "manager" in msg or "human" in msg:
            assignments.append(WorkerAssignment("escalation", base_ctx.copy(), priority=3))
        if order_id and intent == "refund" and order_id in ORDERS:
            if ORDERS[order_id]["total_usd"] > 1000:
                assignments.append(WorkerAssignment("escalation", base_ctx.copy(), priority=3))
        if not assignments:
            assignments.append(WorkerAssignment("policy", {"intent": "general", "raw_message": message}))

        return TaskPlan(str(uuid.uuid4())[:8], assignments)


planner = ShopCoOrchestratorPlanner()
sample_plan = planner.plan("I want a refund for ORD-1001, get me a manager!")
print(f"Plan {sample_plan.plan_id}: {[a.worker_id for a in sample_plan.assignments]}")
````

## 5. Orchestrator Runtime — Dispatch and Synthesize

````python
@dataclass
class OrchestratorRun:
    run_id: str
    plan: TaskPlan
    worker_results: list[WorkerResult] = field(default_factory=list)
    final_reply: str = ""
    elapsed_ms: float = 0.0


class Orchestrator:
    def __init__(self, registry: dict[str, WorkerSpec], max_workers: int = MAX_WORKERS_PER_RUN):
        self.registry = registry
        self.max_workers = max_workers

    def _run_worker(self, assignment: WorkerAssignment) -> WorkerResult:
        spec = self.registry.get(assignment.worker_id)
        if not spec:
            return WorkerResult(assignment.worker_id, False, "Unknown worker.", error="unknown worker")
        t0 = time.perf_counter()
        try:
            result = spec.handler(assignment.context)
            result.duration_ms = (time.perf_counter() - t0) * 1000
            return result
        except Exception as e:
            return WorkerResult(assignment.worker_id, False, "", error=str(e), duration_ms=(time.perf_counter() - t0) * 1000)

    def execute_plan(self, plan: TaskPlan, *, parallel: bool = True) -> OrchestratorRun:
        run = OrchestratorRun(str(uuid.uuid4())[:8], plan)
        t0 = time.perf_counter()
        assignments = plan.assignments[: self.max_workers]

        if parallel and len(assignments) > 1:
            with ThreadPoolExecutor(max_workers=self.max_workers) as pool:
                futures = {pool.submit(self._run_worker, a): a for a in assignments}
                for future in as_completed(futures):
                    run.worker_results.append(future.result())
        else:
            for a in assignments:
                run.worker_results.append(self._run_worker(a))

        run.final_reply = self.synthesize(run.worker_results, plan.synthesis_mode)
        run.elapsed_ms = (time.perf_counter() - t0) * 1000
        ORCH_TRACE.append({
            "run_id": run.run_id, "plan_id": plan.plan_id,
            "workers": [r.worker_id for r in run.worker_results],
            "elapsed_ms": round(run.elapsed_ms, 1), "ts": utcnow(),
        })
        return run

    def synthesize(self, results: list[WorkerResult], mode: str) -> str:
        ok = [r for r in results if r.success]
        failed = [r for r in results if not r.success]
        if mode == "first_success" and ok:
            return ok[0].output
        parts = [r.output for r in ok]
        if failed:
            parts.append(f"(Warnings: {', '.join(r.worker_id for r in failed)})")
        return " ".join(parts) if parts else "Unable to complete request."

    def handle(self, message: str, planner: ShopCoOrchestratorPlanner) -> OrchestratorRun:
        plan = planner.plan(message)
        return self.execute_plan(plan)


orchestrator = Orchestrator(WORKER_REGISTRY)
print("Orchestrator runtime ready")
````

## 6. ShopCo End-to-End Runs

````python
test_tickets = [
    "Where is ORD-1002?",
    "Refund for ORD-1001 please — this is unacceptable!",
    "I need to speak to a human manager about my order.",
]

for ticket in test_tickets:
    run = orchestrator.handle(ticket, planner)
    print(f"\nTicket: {ticket}")
    print(f"Workers: {[r.worker_id for r in run.worker_results]}")
    print(f"Reply ({run.elapsed_ms:.0f}ms): {run.final_reply}")
````

## 7. Worker Failure — Orchestrator Synthesis with Partial Results

````python
def billing_fail_worker(ctx: dict[str, Any]) -> WorkerResult:
    simulate_work(10)
    return WorkerResult("billing", False, "", error="billing API down")


fail_registry = {**WORKER_REGISTRY, "billing": WorkerSpec("billing", "fail", billing_fail_worker, ["refund"])}
fail_orch = Orchestrator(fail_registry)
run = fail_orch.handle("Refund ORD-1001", planner)
print("Partial failure reply:", run.final_reply)
print("Successful workers:", [r.worker_id for r in run.worker_results if r.success])
````

## 8. Sequential vs Parallel Worker Dispatch

Orchestrators may run workers **in parallel** when independent, or **sequentially** when outputs chain.

````python
plan = planner.plan("Refund ORD-1001")
par_run = orchestrator.execute_plan(plan, parallel=True)
seq_run = orchestrator.execute_plan(plan, parallel=False)
print(f"Parallel dispatch: {par_run.elapsed_ms:.0f}ms")
print(f"Sequential dispatch: {seq_run.elapsed_ms:.0f}ms")
````

## 9. ReleaseFlow Orchestrator

Release prep: orchestrator assigns test, security, and docs workers; synthesizes GO/NO-GO.

````python
def rf_test_worker(ctx: dict[str, Any]) -> WorkerResult:
    simulate_work(50)
    return WorkerResult("tests", True, "412 unit tests passed.", {"status": "pass"})


def rf_security_worker(ctx: dict[str, Any]) -> WorkerResult:
    simulate_work(60)
    status = RELEASE_CHECKS["security_scan"]
    return WorkerResult(
        "security", status == "pass",
        f"Security scan: {status}.",
        {"status": status},
        error=None if status == "pass" else "warn",
    )


def rf_changelog_worker(ctx: dict[str, Any]) -> WorkerResult:
    simulate_work(30)
    ver = ctx.get("version", "unknown")
    return WorkerResult("changelog", True, f"Changelog drafted for {ver}.", {})


RF_REGISTRY = {
    "tests": WorkerSpec("tests", "CI tests", rf_test_worker, []),
    "security": WorkerSpec("security", "Security scan", rf_security_worker, []),
    "changelog": WorkerSpec("changelog", "Release notes", rf_changelog_worker, []),
}


class ReleaseOrchestratorPlanner:
    def plan(self, message: str) -> TaskPlan:
        ver = re.search(r"v\d+\.\d+\.\d+", message)
        ctx = {"version": ver.group() if ver else "unknown", "raw": message}
        return TaskPlan(
            str(uuid.uuid4())[:8],
            [WorkerAssignment(w, ctx.copy()) for w in ("tests", "security", "changelog")],
        )


class ReleaseOrchestrator(Orchestrator):
    def synthesize(self, results: list[WorkerResult], mode: str) -> str:
        blockers = [r for r in results if not r.success or r.data.get("status") == "warn"]
        if blockers:
            return f"NO-GO: {'; '.join(r.output for r in blockers)}"
        return "GO: " + " ".join(r.output for r in results if r.success)


rf_orch = ReleaseOrchestrator(RF_REGISTRY)
rf_run = rf_orch.handle("Ship v2.4.1 to prod?", ReleaseOrchestratorPlanner())  # type: ignore[arg-type]
print(f"Release: {rf_run.final_reply}")
````

## 10. Orchestrator Budget and Guardrails

````python
@dataclass
class OrchestratorBudget:
    max_workers: int = 4
    max_elapsed_ms: float = 5000
    max_cost_units: int = 10


class BudgetedOrchestrator(Orchestrator):
    def __init__(self, registry: dict[str, WorkerSpec], budget: OrchestratorBudget):
        super().__init__(registry, max_workers=budget.max_workers)
        self.budget = budget

    def execute_plan(self, plan: TaskPlan, *, parallel: bool = True) -> OrchestratorRun:
        if len(plan.assignments) > self.budget.max_workers:
            plan.assignments = plan.assignments[: self.budget.max_workers]
        run = super().execute_plan(plan, parallel=parallel)
        if run.elapsed_ms > self.budget.max_elapsed_ms:
            run.final_reply += " [Budget: slow run flagged]"
        return run


budget_orch = BudgetedOrchestrator(WORKER_REGISTRY, OrchestratorBudget(max_workers=3))
big_plan = planner.plan("Refund ORD-1001 and get manager")
print(f"Plan wants {len(big_plan.assignments)} workers; budget caps at 3")
br = budget_orch.execute_plan(big_plan)
print(f"Executed: {[r.worker_id for r in br.worker_results]}")
````

## 12. When to Use Orchestrator-Workers

| Good fit | Poor fit |
|----------|----------|
| Multi-domain tickets needing several specialists | Single FAQ with one answer |
| Dynamic decomposition (refund + escalate + policy) | Fixed 1:1 routing suffices |
| Release pipelines with varied checks | Simple prompt chain with 2 steps |
| Synthesis across worker outputs | One worker can do everything |

---

## 13. Anti-Patterns

| Anti-pattern | Problem | Fix |
|--------------|---------|-----|
| Orchestrator does all work | Workers are theater | Delegate real domain logic |
| Too many micro-workers | Planning overhead | Merge related workers |
| No synthesis step | User gets raw worker dumps | Orchestrator merges + formats |
| Unbounded worker spawn | Cost explosion | Budget + max_workers |
| Workers share mutable state | Race conditions | Pass immutable context dicts |
| Skip plan logging | Un-debuggable runs | Trace plan + worker results |

## 16. Summary

Orchestrator-workers centralizes **planning** and **synthesis**:

- **Plan** — decompose request into worker assignments with shared context
- **Dispatch** — run workers in parallel or sequence
- **Synthesize** — merge outputs; handle partial failures gracefully
- **ShopCo** — order + policy + billing + escalation workers for complex tickets
- **ReleaseFlow** — test + security + changelog workers → GO/NO-GO

Use when one request needs multiple specialists; keep workers narrow, log plans, and cap concurrency.
