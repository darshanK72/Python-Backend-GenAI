# Pattern — Parallelization

**Parallelization** runs **independent subtasks at the same time** and merges their results in a **fan-in** step. Use it when branches do not depend on each other's output — only on shared input.

```
                    Customer ticket (ORD-1001 refund)
                              │
              ┌───────────────┼───────────────┐
              ▼               ▼               ▼
        Order lookup    Policy search   Billing check
         (120ms sim)     (80ms sim)      (100ms sim)
              │               │               │
              └───────────────┼───────────────┘
                              ▼
                         AGGREGATE → draft reply
```

Sequential time ≈ sum of tasks. Parallel time ≈ max of tasks (plus merge overhead).

This notebook implements parallelization for **ShopCo Support Hub** and **ReleaseFlow** using `concurrent.futures` in plain Python — simulated I/O delays, fan-in aggregation, partial failure handling, and timing comparisons.

**When to use:** independent lookups, parallel CI checks, multi-source enrichment.

**When not to use:** step B needs step A's output (use prompt chaining instead).

````python
"""
Parallelization Pattern Lab — ShopCo + ReleaseFlow.
"""

import json
import time
import uuid
from concurrent.futures import ThreadPoolExecutor, as_completed
from dataclasses import dataclass, field
from datetime import datetime, timezone
from enum import Enum
from typing import Any, Callable

USE_LIVE_LLM = False


def pretty(obj: Any) -> str:
    return json.dumps(obj, indent=2, default=str)


def utcnow() -> str:
    return datetime.now(timezone.utc).isoformat()


ORDERS: dict[str, dict[str, Any]] = {
    "ORD-1001": {"order_id": "ORD-1001", "customer": "alice@shop.com", "status": "shipped", "total_usd": 1299.0},
    "ORD-1002": {"order_id": "ORD-1002", "customer": "bob@shop.com", "status": "processing", "total_usd": 449.0},
}

POLICIES: dict[str, str] = {
    "returns": "30-day returns on unused items. [pol-returns-01]",
    "refunds": "Escalate refunds over $1000. [pol-refunds-03]",
}

BILLING: dict[str, dict[str, Any]] = {
    "ORD-1001": {"refund_eligible": True, "payment_method": "visa-4242"},
    "ORD-1002": {"refund_eligible": True, "payment_method": "mc-8888"},
}

RELEASE_CHECKS: dict[str, dict[str, Any]] = {
    "unit_tests": {"status": "pass", "detail": "412 passed", "duration_ms": 120},
    "lint": {"status": "pass", "detail": "clean", "duration_ms": 45},
    "security_scan": {"status": "warn", "detail": "1 medium CVE", "duration_ms": 200},
    "integration": {"status": "pass", "detail": "89 passed", "duration_ms": 180},
}

PARALLEL_TRACE: list[dict[str, Any]] = []

print("Parallelization lab ready")
````

---

## 1. Parallel vs Sequential vs Routing

| Pattern | Shape | When |
|---------|-------|------|
| **Sequential chain** | A → B → C | B needs A's output |
| **Routing** | Pick one branch | Mutually exclusive intents |
| **Parallelization** | A \|\| B \|\| C → merge | Independent subtasks, same input |

Parallelization is **fan-out / fan-in**, not multi-agent debate. Each branch is usually the same kind of work on different data sources.

---

## 2. Parallel Task Model

````python
class TaskStatus(str, Enum):
    OK = "ok"
    ERROR = "error"
    TIMEOUT = "timeout"


@dataclass
class ParallelTask:
    task_id: str
    name: str
    fn: Callable[[], Any]
    timeout_s: float = 5.0


@dataclass
class TaskResult:
    task_id: str
    name: str
    status: TaskStatus
    data: Any = None
    error: str | None = None
    duration_ms: float = 0.0


print("Task model ready")
````

---

## 3. ParallelExecutor — Fan-Out / Fan-In

````python
class ParallelExecutor:
    def __init__(self, max_workers: int = 4):
        self.max_workers = max_workers

    def run(self, tasks: list[ParallelTask]) -> list[TaskResult]:
        run_id = str(uuid.uuid4())[:8]
        t0 = time.perf_counter()
        results: list[TaskResult] = []
        with ThreadPoolExecutor(max_workers=self.max_workers) as pool:
            future_map = {pool.submit(self._run_one, task): task for task in tasks}
            for future in as_completed(future_map):
                task = future_map[future]
                try:
                    results.append(future.result())
                except Exception as e:
                    results.append(TaskResult(task.task_id, task.name, TaskStatus.ERROR, error=str(e)))
        elapsed = (time.perf_counter() - t0) * 1000
        PARALLEL_TRACE.append({"run_id": run_id, "tasks": len(tasks), "elapsed_ms": round(elapsed, 1), "ts": utcnow()})
        return results

    def _run_one(self, task: ParallelTask) -> TaskResult:
        t0 = time.perf_counter()
        try:
            data = task.fn()
            ms = (time.perf_counter() - t0) * 1000
            return TaskResult(task.task_id, task.name, TaskStatus.OK, data=data, duration_ms=ms)
        except Exception as e:
            ms = (time.perf_counter() - t0) * 1000
            return TaskResult(task.task_id, task.name, TaskStatus.ERROR, error=str(e), duration_ms=ms)


executor = ParallelExecutor(max_workers=4)
print("ParallelExecutor ready")
````

---

## 4. ShopCo — Independent Lookup Workers

Three data sources with **no cross-dependencies** — ideal for parallel fetch.

````python
def simulate_io(delay_ms: float) -> None:
    time.sleep(delay_ms / 1000.0)


def fetch_order(order_id: str) -> dict[str, Any]:
    simulate_io(120)
    order = ORDERS.get(order_id)
    if not order:
        raise ValueError(f"Order {order_id} not found")
    return order


def fetch_policy(topic: str) -> str:
    simulate_io(80)
    return POLICIES.get(topic, "No policy found.")


def fetch_billing(order_id: str) -> dict[str, Any]:
    simulate_io(100)
    billing = BILLING.get(order_id)
    if not billing:
        raise ValueError(f"Billing record missing for {order_id}")
    return billing


def build_shopco_tasks(order_id: str) -> list[ParallelTask]:
    return [
        ParallelTask("t1", "order_lookup", lambda: fetch_order(order_id)),
        ParallelTask("t2", "policy_lookup", lambda: fetch_policy("returns")),
        ParallelTask("t3", "billing_check", lambda: fetch_billing(order_id)),
    ]


print("ShopCo lookup workers defined (120ms + 80ms + 100ms simulated I/O)")
````

---

## 5. Fan-In Aggregator

Merge parallel results into one context object for the draft step.

````python
@dataclass
class ShopCoContext:
    order: dict[str, Any] | None = None
    policy: str = ""
    billing: dict[str, Any] | None = None
    errors: list[str] = field(default_factory=list)
    task_timings: dict[str, float] = field(default_factory=dict)

    def to_reply(self) -> str:
        if self.errors and not self.order:
            return f"Could not complete lookup: {'; '.join(self.errors)}"
        parts = []
        if self.order:
            parts.append(f"Order {self.order['order_id']}: {self.order['status']} (${self.order['total_usd']}).")
        if self.policy:
            parts.append(f"Policy: {self.policy}")
        if self.billing:
            parts.append(f"Refund eligible: {self.billing['refund_eligible']}.")
            if self.order and self.order["total_usd"] > 1000:
                parts.append("Escalation required — amount over $1000.")
        if self.errors:
            parts.append(f"Partial data — {', '.join(self.errors)}")
        return " ".join(parts) if parts else "No data available."


def aggregate_shopco(results: list[TaskResult]) -> ShopCoContext:
    ctx = ShopCoContext()
    for r in results:
        ctx.task_timings[r.name] = r.duration_ms
        if r.status != TaskStatus.OK:
            ctx.errors.append(f"{r.name}: {r.error}")
            continue
        if r.name == "order_lookup":
            ctx.order = r.data
        elif r.name == "policy_lookup":
            ctx.policy = r.data
        elif r.name == "billing_check":
            ctx.billing = r.data
    return ctx


print("Aggregator ready")
````

---

## 6. ShopCo Parallel Run

````python
tasks = build_shopco_tasks("ORD-1001")
results = executor.run(tasks)
ctx = aggregate_shopco(results)
print("Reply:", ctx.to_reply())
print(f"\nWall time ~{PARALLEL_TRACE[-1]['elapsed_ms']}ms (sequential would be ~{sum(r.duration_ms for r in results):.0f}ms)")
print("Per-task:", ctx.task_timings)
````

---

## 7. Sequential Baseline — Timing Comparison

````python
def run_sequential(order_id: str) -> tuple[ShopCoContext, float]:
    t0 = time.perf_counter()
    results = []
    for task in build_shopco_tasks(order_id):
        results.append(executor._run_one(task))
    elapsed = (time.perf_counter() - t0) * 1000
    return aggregate_shopco(results), elapsed


def run_parallel(order_id: str) -> tuple[ShopCoContext, float]:
    t0 = time.perf_counter()
    results = executor.run(build_shopco_tasks(order_id))
    elapsed = (time.perf_counter() - t0) * 1000
    return aggregate_shopco(results), elapsed


_, seq_ms = run_sequential("ORD-1001")
_, par_ms = run_parallel("ORD-1001")
print(f"Sequential: {seq_ms:.0f}ms")
print(f"Parallel:   {par_ms:.0f}ms")
print(f"Speedup:    {seq_ms / par_ms:.1f}× (ideal ≈ 3× for 3 equal I/O tasks)")
````

## 8. ReleaseFlow — Parallel CI Checks

CI pipelines naturally parallelize: unit tests, lint, security, integration run on independent workers.

````python
def run_ci_check(name: str, config: dict[str, Any]) -> dict[str, Any]:
    simulate_io(config["duration_ms"])
    if config.get("fail"):
        raise RuntimeError(f"{name} failed")
    return {"check": name, "status": config["status"], "detail": config["detail"]}


def build_release_tasks() -> list[ParallelTask]:
    return [
        ParallelTask(f"ci-{k}", k, lambda c=v, n=k: run_ci_check(n, c))
        for k, v in RELEASE_CHECKS.items()
    ]


def aggregate_release(results: list[TaskResult]) -> dict[str, Any]:
    checks = []
    blockers = []
    errors = []
    for r in results:
        if r.status == TaskStatus.OK:
            checks.append(r.data)
            if r.data["status"] not in ("pass",):
                blockers.append(f"{r.name}: {r.data['detail']}")
        else:
            errors.append(f"{r.name}: {r.error}")
    decision = "NO-GO" if blockers or errors else "GO"
    return {"decision": decision, "checks": checks, "blockers": blockers, "errors": errors}


rf_results = executor.run(build_release_tasks())
rf_out = aggregate_release(rf_results)
print(f"Release decision: {rf_out['decision']}")
if rf_out["blockers"]:
    print("Blockers:", rf_out["blockers"])
````

## 9. Partial Failure — One Branch Fails

Parallel systems must **degrade gracefully** when one lookup fails.

````python
def fetch_billing_fail(order_id: str) -> dict[str, Any]:
    simulate_io(50)
    raise ConnectionError("Billing API timeout")


fail_tasks = [
    ParallelTask("t1", "order_lookup", lambda: fetch_order("ORD-1002")),
    ParallelTask("t2", "policy_lookup", lambda: fetch_policy("returns")),
    ParallelTask("t3", "billing_check", lambda: fetch_billing_fail("ORD-1002")),
]
fail_results = executor.run(fail_tasks)
fail_ctx = aggregate_shopco(fail_results)
print("Partial failure reply:", fail_ctx.to_reply())
````

## 10. Dependency Check — When NOT to Parallelize

If policy lookup must use **order status** to pick the right policy, tasks are **dependent** — run sequentially or pass order result into a second parallel batch.

````python
def dependent_pipeline(order_id: str) -> str:
    """Correct: order first, then policy based on status."""
    order = fetch_order(order_id)
    topic = "refunds" if order["status"] == "shipped" else "returns"
    policy = fetch_policy(topic)
    return f"{order_id} ({order['status']}): {policy}"


def wrong_parallel(order_id: str) -> str:
    """Wrong: policy fetched before order status known — may pick wrong policy."""
    results = executor.run([
        ParallelTask("o", "order_lookup", lambda: fetch_order(order_id)),
        ParallelTask("p", "policy_lookup", lambda: fetch_policy("returns")),  # always returns
    ])
    ctx = aggregate_shopco(results)
    return ctx.to_reply()


print("Dependent (correct):", dependent_pipeline("ORD-1001")[:80], "...")
print("\nParallel without deps may miss status-specific policy — use two-phase: parallel batch 1, then batch 2.")
````

---

## 11. Two-Phase Parallel — Wave Pattern

When some tasks depend on others, use **waves**: wave 1 parallel, wave 2 parallel with wave 1 results.

````python
def two_wave_shopco(order_id: str) -> ShopCoContext:
    wave1 = executor.run([ParallelTask("o", "order_lookup", lambda: fetch_order(order_id))])
    ctx = aggregate_shopco(wave1)
    if not ctx.order:
        return ctx
    topic = "refunds" if ctx.order["status"] == "shipped" else "returns"
    wave2 = executor.run([
        ParallelTask("p", "policy_lookup", lambda: fetch_policy(topic)),
        ParallelTask("b", "billing_check", lambda: fetch_billing(order_id)),
    ])
    ctx2 = aggregate_shopco(wave2)
    ctx.policy = ctx2.policy
    ctx.billing = ctx2.billing
    ctx.errors.extend(ctx2.errors)
    ctx.task_timings.update(ctx2.task_timings)
    return ctx


tw = two_wave_shopco("ORD-1001")
print(tw.to_reply())
print(f"Policy topic chosen from order status: refunds (shipped)")
````

## 12. Parallel LLM Calls (Map Pattern)

Another use: **map** the same prompt over many chunks (reviews, tickets), then **reduce** summaries. Simulated offline.

````python
TICKET_CHUNKS = [
    "Customer angry about late ORD-1001 delivery.",
    "Asked about return policy twice.",
    "Mentioned switching to competitor.",
]


def summarize_chunk(chunk: str) -> dict[str, str]:
    simulate_io(60)
    sentiment = "negative" if any(w in chunk.lower() for w in ("angry", "competitor")) else "neutral"
    return {"chunk": chunk[:40], "sentiment": sentiment, "summary": chunk[:50]}


map_tasks = [ParallelTask(f"c{i}", f"summarize_{i}", lambda c=ch: summarize_chunk(c)) for i, ch in enumerate(TICKET_CHUNKS)]
map_results = executor.run(map_tasks)
summaries = [r.data for r in map_results if r.status == TaskStatus.OK]
negative_count = sum(1 for s in summaries if s["sentiment"] == "negative")
print(f"Mapped {len(summaries)} chunks in parallel; {negative_count} negative signals.")
for s in summaries:
    print(f"  • [{s['sentiment']}] {s['summary']}")
````

---

## 13. When to Use Parallelization

| Good fit | Poor fit |
|----------|----------|
| Multi-source API lookups | Step N needs output of step N−1 |
| Parallel CI / validation | Single LLM call suffices |
| Map over document chunks | Branches need shared mutable state |
| Independent LLM extractions on same input | Strict ordering required for compliance |

---

## 14. Anti-Patterns

| Anti-pattern | Problem | Fix |
|--------------|---------|-----|
| Parallel dependent steps | Wrong results | Sequential or two-wave |
| No fan-in merge | Scattered outputs | Single aggregator schema |
| Fail entire request on one branch | Poor UX | Partial results + errors |
| Unbounded parallelism | Rate limits, OOM | `max_workers` cap |
| Parallel for 5ms tasks | Overhead > savings | Stay sequential |
| Shared mutable state across threads | Race conditions | Immutable results per task |

## 17. Summary

Parallelization **fans out** independent work and **fans in** through an aggregator:

- **ShopCo** — order, policy, and billing lookups in ~max(120,80,100)ms instead of ~300ms sequential
- **ReleaseFlow** — four CI checks run concurrently; merge into GO/NO-GO
- **Partial failure** — return best-effort reply with error list
- **Two-wave** — when later tasks depend on earlier results

Parallelize only independent branches; merge with a typed aggregator; cap workers and handle failures per branch.
