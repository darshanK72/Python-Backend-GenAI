# Simplest Pattern That Solves the Problem

The best agent architecture is the **least powerful pattern that still meets requirements**. Extra loops, agents, and memory tiers add latency, cost, and failure modes — often without improving outcomes.

```
  Complexity ladder (use the lowest rung that works)

  1. Lookup / single-shot ──► FAQ with static answer
  2. Prompt chain ──────────► Fixed summarize → classify
  3. Routing ───────────────► Disjoint intents (policy vs order)
  4. Tool call (one-shot) ──► Single DB lookup + template reply
  5. ReAct loop ────────────► Unknown tool sequence
  6. Multi-agent ───────────► Multi-domain + quality gates
```

This notebook teaches a **pattern escalation framework**: score the problem, pick the minimal pattern, and **escalate only when evaluation proves the simpler approach failed. Scenarios: **ShopCo Support Hub** and **ReleaseFlow**.

````python
"""
Simplest-pattern selection lab — ShopCo + ReleaseFlow.
"""

import json
import os
import re
import uuid
from dataclasses import dataclass, field
from datetime import datetime, timezone
from enum import Enum
from typing import Any, Callable

OPENAI_API_KEY = os.environ.get("OPENAI_API_KEY", "sk-proj-placeholder-replace-with-your-real-key")
os.environ.setdefault("OPENAI_API_KEY", OPENAI_API_KEY)

USE_LIVE_LLM = False


def pretty(obj: Any) -> str:
    return json.dumps(obj, indent=2, default=str)


def utc_now() -> str:
    return datetime.now(timezone.utc).isoformat()


POLICY_SNIPPETS = {
    "returns": "Returns within 30 days with receipt. [pol-returns-01]",
    "shipping": "Free shipping on orders over $50. [pol-shipping-02]",
    "hours": "Support hours Mon–Fri 9am–6pm ET. [faq-hours-05]",
}

ORDERS_DB = {
    "ORD-1001": {"order_id": "ORD-1001", "status": "shipped"},
    "ORD-1002": {"order_id": "ORD-1002", "status": "processing"},
}

CHANGELOG = [
    {"version": "2.4.0", "item": "Added bulk export endpoint"},
    {"version": "2.4.0", "item": "Fixed session timeout on mobile"},
]

print("Simplicity lab ready.")
````

## 1. The Simplicity Principle

| Question | If **yes** → simpler pattern may suffice |
|----------|------------------------------------------|
| Is the answer in a static FAQ? | Lookup / single-shot |
| Are steps always the same order? | Prompt chain |
| Are intents disjoint with different tools? | Routing |
| Is exactly one tool call enough? | One-shot tool + template |
| Is tool order unknown upfront? | ReAct |
| Do multiple roles need separate contracts + QA? | Multi-agent |

**Default assumption:** you need less than you think. Prove otherwise with metrics.

## 2. Problem Signals — What the Request Tells You

````python
class PatternTier(str, Enum):
    LOOKUP = "lookup"
    PROMPT_CHAIN = "prompt_chain"
    ROUTING = "routing"
    ONE_SHOT_TOOL = "one_shot_tool"
    REACT = "react"
    MULTI_AGENT = "multi_agent"


@dataclass
class ProblemProfile:
    task_id: str
    description: str
    static_faq: bool = False
    fixed_steps: bool = False
    disjoint_intents: bool = False
    single_tool_sufficient: bool = False
    dynamic_tool_sequence: bool = False
    multi_domain: bool = False
    quality_gate_required: bool = False
    eval_failed_simpler: bool = False

    def minimal_tier(self) -> PatternTier:
        if self.static_faq and not self.eval_failed_simpler:
            return PatternTier.LOOKUP
        if self.fixed_steps and not self.disjoint_intents:
            return PatternTier.PROMPT_CHAIN
        if self.disjoint_intents and self.single_tool_sufficient:
            return PatternTier.ROUTING
        if self.single_tool_sufficient and not self.dynamic_tool_sequence:
            return PatternTier.ONE_SHOT_TOOL
        if self.dynamic_tool_sequence and not (self.multi_domain or self.quality_gate_required):
            return PatternTier.REACT
        if self.multi_domain or self.quality_gate_required or self.eval_failed_simpler:
            return PatternTier.MULTI_AGENT
        return PatternTier.ROUTING


faq_profile = ProblemProfile(
    task_id="shopco-hours",
    description="What are support hours?",
    static_faq=True,
)
print(faq_profile.minimal_tier().value)
````

---

## 3. Complexity Budget — Score Patterns You Implement

````python
@dataclass
class PatternCost:
    tier: PatternTier
    llm_calls_est: int
    agent_count: int
    state_fields: int
    debug_difficulty: int  # 1 easy .. 5 hard


PATTERN_COSTS: dict[PatternTier, PatternCost] = {
    PatternTier.LOOKUP: PatternCost(PatternTier.LOOKUP, 0, 0, 1, 1),
    PatternTier.PROMPT_CHAIN: PatternCost(PatternTier.PROMPT_CHAIN, 2, 0, 2, 2),
    PatternTier.ROUTING: PatternCost(PatternTier.ROUTING, 2, 1, 3, 2),
    PatternTier.ONE_SHOT_TOOL: PatternCost(PatternTier.ONE_SHOT_TOOL, 1, 0, 3, 2),
    PatternTier.REACT: PatternCost(PatternTier.REACT, 4, 1, 5, 4),
    PatternTier.MULTI_AGENT: PatternCost(PatternTier.MULTI_AGENT, 6, 4, 8, 5),
}


def compare_tiers(chosen: PatternTier, minimal: PatternTier) -> dict[str, Any]:
    c = PATTERN_COSTS[chosen]
    m = PATTERN_COSTS[minimal]
    return {
        "chosen": chosen.value,
        "minimal": minimal.value,
        "over_engineered": chosen != minimal and PATTERN_COSTS[chosen].llm_calls_est > m.llm_calls_est,
        "extra_llm_calls": c.llm_calls_est - m.llm_calls_est,
        "extra_agents": c.agent_count - m.agent_count,
    }


print(pretty(compare_tiers(PatternTier.MULTI_AGENT, PatternTier.LOOKUP)))
````

## 4. Tier 1 — Lookup (ShopCo FAQ)

No LLM required when the answer is deterministic.

````python
FAQ_INDEX = {
    "hours": POLICY_SNIPPETS["hours"],
    "support hours": POLICY_SNIPPETS["hours"],
    "when are you open": POLICY_SNIPPETS["hours"],
}


def lookup_answer(message: str) -> dict[str, Any] | None:
    low = message.lower().strip().rstrip("?")
    for key, answer in FAQ_INDEX.items():
        if key in low:
            return {"pattern": PatternTier.LOOKUP.value, "answer": answer, "llm_calls": 0}
    return None


print(lookup_answer("What are your support hours?"))
````

---

## 5. Tier 2 — Prompt Chain (ReleaseFlow Summarize → Format)

Fixed pipeline when step order never changes.

````python
def step_collect_changes(version: str) -> list[str]:
    return [c["item"] for c in CHANGELOG if c["version"] == version]


def step_format_bullets(items: list[str]) -> str:
    return "\n".join(f"- {i}" for i in items)


def step_wrap_release(version: str, body: str) -> str:
    return f"Release {version} notes:\n{body}"


def prompt_chain_release(version: str) -> dict[str, Any]:
    items = step_collect_changes(version)
    bullets = step_format_bullets(items)
    notes = step_wrap_release(version, bullets)
    return {"pattern": PatternTier.PROMPT_CHAIN.value, "notes": notes, "llm_calls": 0}


print(prompt_chain_release("2.4.0")["notes"])
````

---

## 6. Tier 3 — Routing (ShopCo Policy vs Order)

One classifier branch — not three collaborating agents.

````python
def classify_shopco(msg: str) -> str:
    low = msg.lower()
    if "return" in low or "policy" in low:
        return "policy"
    if "order" in low or re.search(r"ord-[0-9]{4}", msg, re.I):
        return "order"
    return "general"


def routing_shopco(message: str) -> dict[str, Any]:
    intent = classify_shopco(message)
    if intent == "policy":
        ans = POLICY_SNIPPETS["returns"] if "return" in message.lower() else POLICY_SNIPPETS["shipping"]
    elif intent == "order":
        m = re.search(r"ORD-[0-9]{4}", message.upper())
        oid = m.group(0) if m else ""
        ans = f"Status: {ORDERS_DB.get(oid, {}).get('status', 'unknown')}"
    else:
        ans = "How can ShopCo help you today?"
    return {"pattern": PatternTier.ROUTING.value, "intent": intent, "answer": ans, "agents": 1}


print(routing_shopco("Where is ORD-1002?"))
````

---

## 7. Tier 4 — One-Shot Tool + Template

Exactly one tool call, then a deterministic reply template.

````python
def lookup_order(order_id: str) -> dict[str, str]:
    return ORDERS_DB.get(order_id, {"order_id": order_id, "status": "not_found"})


def one_shot_order_status(message: str) -> dict[str, Any]:
    m = re.search(r"ORD-[0-9]{4}", message.upper())
    if not m:
        return {"pattern": PatternTier.ONE_SHOT_TOOL.value, "error": "missing order id"}
    rec = lookup_order(m.group(0))
    answer = f"Your order {rec['order_id']} is currently {rec['status']}."
    return {"pattern": PatternTier.ONE_SHOT_TOOL.value, "tool_calls": 1, "answer": answer}


print(one_shot_order_status("Please check ORD-1001"))
````

## 8. When Simpler Patterns Fail — Escalation Triggers

Escalate **only** when measurable criteria fail:

````python
@dataclass
class EvalResult:
    pattern: str
    accuracy: float
    citation_rate: float
    latency_ms: int

    def passes(self, min_accuracy: float = 0.85, min_citation: float = 0.8) -> bool:
        return self.accuracy >= min_accuracy and self.citation_rate >= min_citation


def recommend_escalation(results: list[EvalResult]) -> PatternTier:
    for r in results:
        if r.passes():
            return PatternTier(r.pattern)
    return PatternTier.MULTI_AGENT


eval_run = [
    EvalResult("lookup", 0.95, 1.0, 5),
    EvalResult("routing", 0.88, 0.9, 120),
    EvalResult("react", 0.91, 0.95, 800),
]
print("Pick:", recommend_escalation(eval_run).value)
````

## 9. Unified Resolver — Try Simplest First

````python
def react_fallback(message: str) -> dict[str, Any]:
    steps = []
    answer_parts = []
    if "return" in message.lower():
        steps.append("search_policy")
        answer_parts.append(POLICY_SNIPPETS["returns"])
    m = re.search(r"ORD-[0-9]{4}", message.upper())
    if m:
        steps.append("lookup_order")
        answer_parts.append(f"Order status: {ORDERS_DB[m.group(0)]['status']}")
    return {
        "pattern": PatternTier.REACT.value,
        "steps": steps,
        "answer": " | ".join(answer_parts) or "Unable to resolve.",
    }


def multi_agent_fallback(message: str) -> dict[str, Any]:
    draft = react_fallback(message)
    draft["pattern"] = PatternTier.MULTI_AGENT.value
    draft["agents"] = ["triage", "policy", "order", "qa"]
    draft["answer"] += " [QA approved]"
    return draft


def resolve_shopco(message: str, profile: ProblemProfile | None = None) -> dict[str, Any]:
    hit = lookup_answer(message)
    if hit:
        return hit

    prof = profile or ProblemProfile(
        task_id="auto", description=message,
        disjoint_intents=True,
        single_tool_sufficient="order" in message.lower() or "return" in message.lower(),
        dynamic_tool_sequence="return" in message.lower() and bool(re.search(r"ORD-[0-9]{4}", message, re.I)),
    )
    tier = prof.minimal_tier()

    if tier == PatternTier.ROUTING:
        return routing_shopco(message)
    if tier == PatternTier.ONE_SHOT_TOOL:
        return one_shot_order_status(message)
    if tier == PatternTier.REACT:
        return react_fallback(message)
    return multi_agent_fallback(message)


for q in [
    "What are your support hours?",
    "Where is ORD-1002?",
    "Can I return ORD-1001?",
]:
    r = resolve_shopco(q)
    print(f"{q[:40]:40} → {r['pattern']}")
````

## 10. Side-by-Side — Over-Engineered vs Minimal

````python
@dataclass
class RunComparison:
    question: str
    minimal_pattern: str
    overkill_pattern: str
    savings: dict[str, Any]


def compare_approaches(question: str) -> RunComparison:
    minimal = resolve_shopco(question)
    overkill = multi_agent_fallback(question)
    min_tier = PatternTier(minimal["pattern"])
    over_tier = PatternTier.MULTI_AGENT
    return RunComparison(
        question=question,
        minimal_pattern=minimal["pattern"],
        overkill_pattern=overkill["pattern"],
        savings=compare_tiers(over_tier, min_tier),
    )


print(pretty(compare_approaches("What are your support hours?")))
````

## 11. ReleaseFlow — Chain Beats Plan-and-Execute Here

Rollout with **fixed SOP steps** does not need a dynamic planner.

````python
FIXED_RELEASE_STEPS = ["run_tests", "build_artifact", "deploy_staging", "smoke_test"]


def run_fixed_chain(steps: list[str]) -> dict[str, Any]:
    log = [{"step": s, "result": "ok"} for s in steps]
    return {"pattern": PatternTier.PROMPT_CHAIN.value, "steps_run": len(log), "log": log}


def run_plan_and_execute(steps: list[str]) -> dict[str, Any]:
    plan = [{"n": i + 1, "action": s} for i, s in enumerate(steps)]
    log = [{"step": s, "result": "ok"} for s in steps]
    return {"pattern": "plan_and_execute", "plan_llm_call": 1, "log": log, "plan": plan}


chain_run = run_fixed_chain(FIXED_RELEASE_STEPS)
plan_run = run_plan_and_execute(FIXED_RELEASE_STEPS)
print("Chain LLM calls: 0 | Plan-and-execute extra planner call: 1")
print("Chain sufficient:", chain_run["steps_run"] == len(FIXED_RELEASE_STEPS))
````

## 12. Refactoring Down — Multi-Agent → Routing

When QA never rejects and intents never overlap, delete agents.

````python
@dataclass
class ArchitectureReview:
    current_agents: int
    qa_rejection_rate: float
    multi_domain_rate: float
    recommended_tier: PatternTier
    rationale: str


def review_architecture(
    current_agents: int,
    qa_rejection_rate: float,
    multi_domain_rate: float,
) -> ArchitectureReview:
    if multi_domain_rate < 0.05 and qa_rejection_rate < 0.02:
        return ArchitectureReview(
            current_agents, qa_rejection_rate, multi_domain_rate,
            PatternTier.ROUTING,
            "Low QA reject + rare multi-domain → routing is enough",
        )
    if qa_rejection_rate >= 0.1:
        return ArchitectureReview(
            current_agents, qa_rejection_rate, multi_domain_rate,
            PatternTier.MULTI_AGENT,
            "QA catches real issues → keep critique agent",
        )
    return ArchitectureReview(
        current_agents, qa_rejection_rate, multi_domain_rate,
        PatternTier.REACT,
        "Some dynamic tool need → ReAct before multi-agent",
    )


print(pretty(review_architecture(current_agents=4, qa_rejection_rate=0.01, multi_domain_rate=0.03)))
````

## 13. Decision Worksheet

Before adding an agent or loop, fill this in:

1. **Requirement** — what must be true in the answer?
2. **Simplest tier tried** — lookup / chain / route / …
3. **Eval metric** — accuracy, citation rate, latency
4. **Failure mode** — what did the simple pattern miss?
5. **Escalation** — next tier only if (4) is documented

````python
@dataclass
class DecisionRecord:
    requirement: str
    tiers_tried: list[str]
    metric: str
    failure_mode: str | None
    chosen_tier: str


record = DecisionRecord(
    requirement="Return answers must cite pol-returns-01",
    tiers_tried=["lookup", "routing"],
    metric="citation_rate >= 0.8",
    failure_mode="routing missed combined order+policy questions",
    chosen_tier="react",
)
print(pretty(record))
````

## 14. Optional Live LLM — Only When Lookup Fails

Set `USE_LIVE_LLM = True` to fall back to `gpt-4o-mini` after lookup misses.

````python
def resolve_with_live_fallback(message: str) -> dict[str, Any]:
    hit = lookup_answer(message)
    if hit:
        return hit
    if not USE_LIVE_LLM:
        return routing_shopco(message)
    from langchain_core.messages import HumanMessage, SystemMessage
    from langchain_openai import ChatOpenAI

    llm = ChatOpenAI(model="gpt-4o-mini", api_key=OPENAI_API_KEY, temperature=0)
    resp = llm.invoke([
        SystemMessage(content="Answer briefly using ShopCo policies if relevant."),
        HumanMessage(content=message),
    ])
    return {"pattern": "single_shot_llm", "answer": str(resp.content), "llm_calls": 1}


print(resolve_with_live_fallback("Do you ship internationally?"))
````

| Mistake | Symptom | Fix |
|---------|---------|-----|
| Pattern prestige | 5 agents for FAQ | Start at lookup |
| No eval before escalate | Cost spike | Measure simpler tier first |
| ReAct for fixed pipeline | Slow release notes | Prompt chain |
| Multi-agent without contracts | Overwritten state | Routing or board |
| Never refactor down | Stale complexity | Review QA + domain rates |

## 17. Summary

Choose the **simplest pattern that solves the problem**: lookup → chain → routing → one-shot tool → ReAct → multi-agent. Use `ProblemProfile` and eval results to justify each escalation step.

ShopCo hours need **lookup**, not ReAct. Order status needs **routing** or **one-shot tool**, not four agents. ReleaseFlow fixed steps need a **chain**, not a planner.

**Add complexity when metrics prove you must — remove it when production data says you do not.**
