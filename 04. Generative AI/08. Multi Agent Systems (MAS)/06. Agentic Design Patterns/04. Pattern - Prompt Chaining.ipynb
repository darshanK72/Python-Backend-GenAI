# Pattern — Prompt Chaining

**Prompt chaining** runs a **fixed sequence** of LLM steps where each step receives the previous step's output as context. There is no branching and no model-driven "what next?" — the control flow is predetermined.

```
  Customer message
       │
       ▼
  Step 1: EXTRACT  →  {order_id, intent, sentiment}
       │
       ▼
  Step 2: ENRICH   →  order facts + policy snippets
       │
       ▼
  Step 3: DRAFT    →  support reply draft
       │
       ▼
  Step 4: POLISH   →  final reply with citations
```

This notebook implements prompt chaining for **ShopCo Support Hub** and a shorter **ReleaseFlow** chain in plain Python. Offline step functions simulate LLM behavior; set `USE_LIVE_LLM = True` to run the same chain with a real model.

**When to use:** decomposable tasks with clear stages, quality gains from specialization, structured handoffs between steps.

**When not to use:** unpredictable tool loops, dynamic routing, or problems solvable in one well-crafted prompt.

````python
"""
Prompt Chaining Pattern Lab — ShopCo + ReleaseFlow.
"""

import json
import re
import uuid
from dataclasses import dataclass, field
from datetime import datetime, timezone
from enum import Enum
from typing import Any, Callable

USE_LIVE_LLM = False


def pretty(obj: Any) -> str:
    return json.dumps(obj, indent=2, default=str)


def utcnow() -> str:
    return datetime.now(timezone.utc).isoformat()


ORDERS: dict[str, dict[str, Any]] = {
    "ORD-1001": {"order_id": "ORD-1001", "customer": "alice@shop.com", "status": "shipped", "total_usd": 1299.0, "carrier": "FedEx"},
    "ORD-1002": {"order_id": "ORD-1002", "customer": "bob@shop.com", "status": "processing", "total_usd": 449.0, "carrier": None},
}

POLICIES: dict[str, str] = {
    "returns": "30-day returns on unused items. Refunds in 5–7 business days. [pol-returns-01]",
    "shipping": "Free shipping over $50. Tracking emailed at carrier scan. [pol-shipping-02]",
    "refunds": "Cite policy before promising refunds. Escalate over $1000. [pol-refunds-03]",
}

RELEASE_CHECKS: dict[str, dict[str, str]] = {
    "unit_tests": {"status": "pass", "detail": "412 tests passed"},
    "lint": {"status": "pass", "detail": "no issues"},
    "security_scan": {"status": "warn", "detail": "1 medium CVE in dev dependency"},
}

CHAIN_TRACE: list[dict[str, Any]] = []

print("Prompt chaining lab ready")
````

---

## 1. Prompt Chain vs Single Shot vs ReAct

| Pattern | Control flow | Steps decided by |
|---------|--------------|------------------|
| **Single shot** | One call | N/A |
| **Prompt chain** | Fixed sequence | Developer at build time |
| **Router** | Branch | Classifier / rules |
| **ReAct** | Loop until done | Model each iteration |

Prompt chaining trades **flexibility** for **predictability** and **per-step prompt specialization**. Each step can use a smaller, focused system prompt.

---

## 2. Chain Infrastructure

A chain is a list of named steps. Each step is a function `(state) → state`. The runner logs every handoff for debugging.

````python
@dataclass
class ChainState:
    """Shared state passed through every step."""
    run_id: str
    domain: str
    raw_input: str
    extracted: dict[str, Any] = field(default_factory=dict)
    enriched: dict[str, Any] = field(default_factory=dict)
    draft: str = ""
    final: str = ""
    errors: list[str] = field(default_factory=list)
    completed_steps: list[str] = field(default_factory=list)


@dataclass
class ChainStep:
    name: str
    handler: Callable[[ChainState], ChainState]
    required_fields: list[str] = field(default_factory=list)


class PromptChain:
    def __init__(self, name: str, steps: list[ChainStep]):
        self.name = name
        self.steps = steps

    def run(self, raw_input: str, *, domain: str = "shopco") -> ChainState:
        state = ChainState(run_id=str(uuid.uuid4())[:8], domain=domain, raw_input=raw_input)
        for step in self.steps:
            for req in step.required_fields:
                if not getattr(state, req, None) and req not in state.extracted and req not in state.enriched:
                    if req == "extracted" and not state.extracted:
                        state.errors.append(f"{step.name}: missing extracted")
                        CHAIN_TRACE.append({"chain": self.name, "step": step.name, "status": "skipped", "run_id": state.run_id})
                        return state
            try:
                state = step.handler(state)
                state.completed_steps.append(step.name)
                CHAIN_TRACE.append({"chain": self.name, "step": step.name, "status": "ok", "run_id": state.run_id, "ts": utcnow()})
            except Exception as e:
                state.errors.append(f"{step.name}: {e}")
                CHAIN_TRACE.append({"chain": self.name, "step": step.name, "status": "error", "run_id": state.run_id})
                break
        return state


print("PromptChain runner ready")
````

---

## 3. ShopCo Step 1 — Extract

First step: parse the customer message into **structured fields**. Downstream steps consume JSON, not raw prose.

````python
class Intent(str, Enum):
    ORDER_STATUS = "order_status"
    RETURN = "return"
    REFUND = "refund"
    GENERAL = "general"


def step_extract(state: ChainState) -> ChainState:
    """Simulated LLM extraction step."""
    msg = state.raw_input.lower()
    oid = re.search(r"ORD-\d+", state.raw_input.upper())
    intent = Intent.GENERAL
    if "refund" in msg:
        intent = Intent.REFUND
    elif "return" in msg:
        intent = Intent.RETURN
    elif oid or "status" in msg or "track" in msg or "ship" in msg:
        intent = Intent.ORDER_STATUS
    sentiment = "frustrated" if any(w in msg for w in ("angry", "late", "terrible", "awful")) else "neutral"
    state.extracted = {
        "order_id": oid.group() if oid else None,
        "intent": intent.value,
        "sentiment": sentiment,
        "customer_mentions": "alice" if "alice" in msg else "unknown",
    }
    return state


demo = ChainState("demo", "shopco", "I'm frustrated — ORD-1001 is late. Can I get a refund?")
demo = step_extract(demo)
print(pretty(demo.extracted))
````

---

## 4. ShopCo Step 2 — Enrich

Second step: use extracted fields to **fetch facts** from tools (orders, policies). No LLM required — but in production this step often mixes retrieval with a small "summarize facts" prompt.

````python
def step_enrich(state: ChainState) -> ChainState:
    ext = state.extracted
    enriched: dict[str, Any] = {"policies": [], "order": None, "flags": []}
    oid = ext.get("order_id")
    if oid:
        order = ORDERS.get(oid)
        if order:
            enriched["order"] = order
        else:
            state.errors.append(f"Order {oid} not found")
    intent = ext.get("intent", "general")
    if intent in ("return", "refund"):
        enriched["policies"].append(POLICIES["returns"])
        enriched["policies"].append(POLICIES["refunds"])
    if intent == "order_status":
        enriched["policies"].append(POLICIES["shipping"])
    if enriched.get("order") and enriched["order"].get("total_usd", 0) > 1000:
        enriched["flags"].append("escalation_required_over_1000")
    if ext.get("sentiment") == "frustrated":
        enriched["flags"].append("empathy_tone")
    state.enriched = enriched
    return state


demo = step_enrich(demo)
print(pretty(demo.enriched))
````

---

## 5. ShopCo Step 3 — Draft

Third step: compose a **draft reply** from structured extraction + enrichment. Focused prompt: "write support text given these facts."

````python
def step_draft(state: ChainState) -> ChainState:
    ext = state.extracted
    enr = state.enriched
    parts: list[str] = []
    if enr.get("flags") and "empathy_tone" in enr["flags"]:
        parts.append("I'm sorry for the frustration — let me help.")
    order = enr.get("order")
    if order:
        parts.append(f"Order {order['order_id']} is {order['status']} (${order['total_usd']}).")
        if order.get("carrier"):
            parts.append(f"Carrier: {order['carrier']}.")
    intent = ext.get("intent")
    if intent == "refund" and enr.get("policies"):
        parts.append(f"Per policy: {enr['policies'][0]}")
    if "escalation_required_over_1000" in enr.get("flags", []):
        parts.append("This refund requires supervisor escalation due to amount.")
    if not parts:
        parts.append("Please share your order ID so I can assist.")
    state.draft = " ".join(parts)
    return state


demo = step_draft(demo)
print("Draft:", demo.draft)
````

---

## 6. ShopCo Step 4 — Polish

Final step: enforce **citations**, tone, and compliance checks on the draft.

````python
def step_polish(state: ChainState) -> ChainState:
    draft = state.draft
    citations = re.findall(r"\[pol-[\w-]+\]", " ".join(state.enriched.get("policies", [])))
    if citations and not any(c in draft for c in citations):
        draft += f" {citations[0]}"
    if state.extracted.get("intent") == "refund" and "[pol-" not in draft:
        draft += " [pol-refunds-03]"
    draft = draft.strip()
    if not draft.endswith("."):
        draft += "."
    state.final = draft
    return state


demo = step_polish(demo)
print("Final:", demo.final)
````

---

## 7. ShopCo Support Chain — End to End

````python
shopco_chain = PromptChain("shopco-support", [
    ChainStep("extract", step_extract),
    ChainStep("enrich", step_enrich, required_fields=["extracted"]),
    ChainStep("draft", step_draft),
    ChainStep("polish", step_polish),
])

test_messages = [
    "What's the status of ORD-1002?",
    "I'm angry — ORD-1001 arrived late. I want a refund!",
    "Hello, I have a question.",
]

for msg in test_messages:
    result = shopco_chain.run(msg)
    print(f"\nInput: {msg}")
    print(f"Steps: {' → '.join(result.completed_steps)}")
    print(f"Reply: {result.final}")
    if result.errors:
        print(f"Errors: {result.errors}")
````

---

## 8. Structured Handoffs Between Steps

Chains fail when steps pass **unstructured prose** instead of **schemas**. Require JSON-shaped outputs at each boundary.

````python
EXTRACT_SCHEMA = {
    "type": "object",
    "required": ["intent", "sentiment"],
    "properties": {
        "order_id": {"type": ["string", "null"]},
        "intent": {"type": "string"},
        "sentiment": {"type": "string"},
    },
}


def validate_extracted(data: dict[str, Any]) -> tuple[bool, list[str]]:
    missing = [k for k in EXTRACT_SCHEMA["required"] if k not in data]
    return len(missing) == 0, missing


def step_extract_validated(state: ChainState) -> ChainState:
    state = step_extract(state)
    ok, missing = validate_extracted(state.extracted)
    if not ok:
        raise ValueError(f"Extract validation failed: missing {missing}")
    return state


validated_chain = PromptChain("shopco-validated", [
    ChainStep("extract", step_extract_validated),
    ChainStep("enrich", step_enrich),
    ChainStep("draft", step_draft),
    ChainStep("polish", step_polish),
])
r = validated_chain.run("Refund for ORD-1001 please")
print(f"Validated chain OK: {r.final[:80]}...")
````

---

## 9. Early Exit on Missing Data

If extraction finds no order and intent requires one, stop the chain and return a clarifying question instead of hallucinating through later steps.

````python
def step_gate_order_required(state: ChainState) -> ChainState:
    intent = state.extracted.get("intent")
    oid = state.extracted.get("order_id")
    if intent in ("refund", "return", "order_status") and not oid:
        state.final = "I'd be happy to help — please provide your order ID (e.g. ORD-1001)."
        state.completed_steps.append("gate_early_exit")
        raise StopIteration("early_exit")
    return state


class GatedPromptChain(PromptChain):
    def run(self, raw_input: str, *, domain: str = "shopco") -> ChainState:
        state = ChainState(run_id=str(uuid.uuid4())[:8], domain=domain, raw_input=raw_input)
        for step in self.steps:
            try:
                state = step.handler(state)
                if "gate_early_exit" in state.completed_steps:
                    return state
                state.completed_steps.append(step.name)
            except StopIteration:
                return state
            except Exception as e:
                state.errors.append(str(e))
                break
        return state


gated = GatedPromptChain("shopco-gated", [
    ChainStep("extract", step_extract),
    ChainStep("gate", step_gate_order_required),
    ChainStep("enrich", step_enrich),
    ChainStep("draft", step_draft),
    ChainStep("polish", step_polish),
])

early = gated.run("I want a refund, this is ridiculous!")
print(f"Early exit reply: {early.final}")
print(f"Steps run: {early.completed_steps}")
````

## 10. ReleaseFlow Chain — Three-Step Pipeline

Same pattern, different domain: parse release request → gather CI signals → produce go/no-go summary.

````python
def rf_extract(state: ChainState) -> ChainState:
    msg = state.raw_input.lower()
    version = re.search(r"v\d+\.\d+\.\d+", state.raw_input)
    state.extracted = {
        "version": version.group() if version else "unknown",
        "urgent": "hotfix" in msg or "urgent" in msg,
        "env": "prod" if "prod" in msg else "staging",
    }
    return state


def rf_gather(state: ChainState) -> ChainState:
    state.enriched = {"checks": RELEASE_CHECKS.copy(), "blockers": []}
    for name, chk in RELEASE_CHECKS.items():
        if chk["status"] in ("fail", "warn"):
            state.enriched["blockers"].append(f"{name}: {chk['detail']}")
    return state


def rf_summarize(state: ChainState) -> ChainState:
    blockers = state.enriched.get("blockers", [])
    ver = state.extracted.get("version", "unknown")
    if blockers:
        state.final = f"NO-GO for {ver}: {'; '.join(blockers)}"
    else:
        state.final = f"GO for {ver}: all checks passed."
    return state


release_chain = PromptChain("releaseflow", [
    ChainStep("extract", rf_extract),
    ChainStep("gather", rf_gather),
    ChainStep("summarize", rf_summarize),
])

rf = release_chain.run("Can we ship v2.4.1 to prod?", domain="releaseflow")
print(f"Release decision: {rf.final}")
````

## 11. Single Shot vs Chain — Quality Tradeoff

A single prompt must extract, lookup, draft, and cite in one pass. The chain specializes each step.

````python
def single_shot_support(message: str) -> str:
    """One-step baseline — often misses policy citation under complexity."""
    oid = re.search(r"ORD-\d+", message.upper())
    if oid and "refund" in message.lower():
        order = ORDERS.get(oid.group())
        if order:
            return f"Refund for {oid.group()} is {order['status']}. We'll process it."  # no citation
    return "How can I help?"


msg = "Refund for ORD-1001 please"
single = single_shot_support(msg)
chained = shopco_chain.run(msg).final
print(f"Single shot: {single}")
print(f"Chained:     {chained}")
print(f"\nChain includes policy citation: {'[pol-' in chained}")
````

## 12. Cost and Latency

Each chain step is typically **one LLM call**. A 4-step chain ≈ 4× single-step latency and cost — justified when quality or reliability gains exceed the overhead.

````python
STEP_COST_MS = 800  # illustrative LLM latency per step


def estimate_chain_cost(n_steps: int) -> dict[str, Any]:
    return {
        "steps": n_steps,
        "est_latency_ms": n_steps * STEP_COST_MS,
        "est_llm_calls": n_steps,
        "single_shot_latency_ms": STEP_COST_MS,
    }


print("ShopCo chain:", pretty(estimate_chain_cost(4)))
print("ReleaseFlow chain:", pretty(estimate_chain_cost(3)))
print("\nUse chaining when 4× cost buys citation accuracy, escalation flags, or stage-specific prompts.")
````

## 14. When to Use Prompt Chaining

| Good fit | Poor fit |
|----------|----------|
| Fixed pipeline: extract → retrieve → draft → review | Dynamic tool loops |
| Quality gates between stages | Intent requires unpredictable branches |
| Different prompt/system per stage | Single FAQ with one answer |
| Compliance citation as final step | Real-time chat needing minimal latency |
| Batch document processing | Exploration / research agents |

---

## 15. Anti-Patterns

| Anti-pattern | Problem | Fix |
|--------------|---------|-----|
| 10-step chain for simple FAQ | Latency, cost, fragility | Single shot or 2-step max |
| Unstructured prose handoffs | Parse errors downstream | JSON schema per step |
| Chain when ReAct needed | Can't adapt to tool results | Switch to ReAct / router |
| Skip validation gates | Hallucination on missing data | Early exit + clarify |
| Same prompt every step | No benefit over single shot | Specialized system prompt per step |
| Hidden state between steps | Untraceable bugs | `ChainState` + trace log |

## 17. Optional — Live LLM Chain

Set `USE_LIVE_LLM = True` to run extract + draft steps with a real model while keeping enrich as deterministic tool lookup.

````python
if USE_LIVE_LLM:
    import os
    from openai import OpenAI

    client = OpenAI(api_key=os.environ.get("OPENAI_API_KEY"))
    msg = "ORD-1001 is late, I want a refund"
    ext_resp = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": "Extract JSON: order_id, intent, sentiment."},
            {"role": "user", "content": msg},
        ],
        response_format={"type": "json_object"},
    )
    extracted = json.loads(ext_resp.choices[0].message.content or "{}")
    state = ChainState("live", "shopco", msg, extracted=extracted)
    state = step_enrich(state)
    draft_resp = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": f"Draft ShopCo support reply. Facts: {pretty(state.enriched)}"},
            {"role": "user", "content": msg},
        ],
    )
    print(draft_resp.choices[0].message.content)
else:
    print("Offline mode — prompt chaining demonstrated with simulated LLM steps.")
````

## 18. Summary

Prompt chaining executes a **fixed sequence** of specialized steps:

- **Extract** → structured fields from raw input
- **Enrich** → tool lookups and flags
- **Draft** → compose response from facts
- **Polish** → citations, tone, compliance

Use structured handoffs, validation gates, and trace logging. ShopCo support gained policy citations and escalation flags; ReleaseFlow gained a deterministic go/no-go pipeline. Reach for routing or ReAct when the path cannot be fixed at design time.
