# Collaboration Anti-patterns

Multi-agent workflows fail in predictable ways: agents **ping-pong** work forever, **duplicate** each other's research, **dump** entire chat logs into handoffs, or **contradict** each other with no merge step. These are **collaboration anti-patterns** — recurring designs that look sophisticated but produce loops, cost blowups, and unreliable answers.

This notebook names the most common anti-patterns, shows each one **breaking** a ShopCo support workflow in runnable code, then demonstrates the **fix**. You will learn to detect bad collaboration early using step counts, handoff size, and progress metrics.

Everything is self-contained plain Python — no frameworks required.

```python
"""
ShopCo Support Team — collaboration anti-pattern lab.
"""

import copy
import json
import os
import re
import uuid
from dataclasses import dataclass, field
from datetime import datetime, timezone
from enum import Enum
from typing import Any, Callable

OPENAI_API_KEY = os.environ.get("OPENAI_API_KEY", "sk-proj-placeholder-replace-with-your-real-key")


def pretty(obj: Any) -> str:
    return json.dumps(obj, indent=2, default=str)


ORDERS: dict[str, dict[str, Any]] = {
    "ORD-1001": {"customer": "alice@shop.com", "status": "shipped", "amount": 1299.00},
    "ORD-1002": {"customer": "bob@shop.com", "status": "processing", "amount": 449.00},
}

POLICIES = [
    {"id": "ret-02", "text": "Returns within 30 days if unused."},
    {"id": "cancel-03", "text": "Cancellations allowed only before shipment."},
]


def lookup_order(order_id: str) -> dict[str, Any]:
    oid = order_id.upper()
    order = ORDERS.get(oid)
    if not order:
        return {"found": False, "order_id": oid}
    return {"found": True, "order_id": oid, **order}


def search_policy(query: str) -> list[dict[str, str]]:
    terms = [t for t in re.split(r"\W+", query.lower()) if len(t) > 2]
    return [p for p in POLICIES if any(t in p["text"].lower() for t in terms)] or POLICIES[:1]


print(f"ShopCo backend — {len(ORDERS)} orders, {len(POLICIES)} policies")
```

## 1. Anti-Pattern Catalog

| # | Anti-pattern | Symptom | Fix |
|---|--------------|---------|-----|
| 1 | **Context dump** | Huge handoffs, slow + costly | Structured payload only |
| 2 | **Ping-pong loop** | Agents pass work back and forth | Max hops + single owner |
| 3 | **Duplicate work** | Same tool called N times | Shared artifacts + dedup |
| 4 | **Supervisor does everything** | No real delegation | Workers own tools |
| 5 | **Chatty group, no stop** | 20 messages, no answer | Termination + turn limits |
| 6 | **Contradictions, no merge** | Conflicting agent outputs | Validator / composer agent |
| 7 | **Over-agentification** | 8 agents for 2 steps | Simplest pattern that works |
| 8 | **Unguarded shared state** | Lost updates, races | Contract writes + merge |

Each section below runs the **broken** version, measures it, then shows the **fixed** version.

## 2. Metrics — Detect Bad Collaboration Early

Instrument every multi-agent run with collaboration health metrics.

```python
@dataclass
class CollaborationMetrics:
    agent_hops: int = 0
    tool_calls: int = 0
    handoff_bytes: int = 0
    duplicate_tool_calls: int = 0
    ping_pong_count: int = 0
    messages: int = 0
    terminated: bool = False
    termination_reason: str = ""

    def score(self) -> str:
        """Rough health: good / warn / bad"""
        if self.ping_pong_count >= 3 or self.agent_hops > 8:
            return "bad"
        if self.duplicate_tool_calls > 0 or self.handoff_bytes > 2000:
            return "warn"
        return "good"


@dataclass
class RunResult:
    answer: str
    metrics: CollaborationMetrics
    pattern: str
    fixed: bool


def handoff_size(payload: Any) -> int:
    return len(json.dumps(payload, default=str).encode("utf-8"))
```

---

## 3. Anti-Pattern #1 — Context Dump Handoff

**Bad:** Pass the entire conversation history to every agent.

**Good:** Pass a distilled `task_summary` + structured `payload` (order_id, intent).

```python
def bad_context_dump_handoff(chat_history: list[str]) -> dict[str, Any]:
    return {"messages": chat_history}  # anti-pattern


def good_structured_handoff(order_id: str, intent: str) -> dict[str, Any]:
    return {
        "task_summary": f"Research order {order_id} for {intent}",
        "payload": {"order_id": order_id, "intent": intent},
    }


fake_history = [f"User/Agent message turn {i}: " + "x" * 200 for i in range(25)]

bad = bad_context_dump_handoff(fake_history)
good = good_structured_handoff("ORD-1001", "return_eligibility")

print(f"Bad handoff size:  {handoff_size(bad):,} bytes")
print(f"Good handoff size: {handoff_size(good):,} bytes")
print(f"Reduction: {100 - 100 * handoff_size(good) / handoff_size(bad):.1f}%")
```

## 4. Anti-Pattern #2 — Ping-Pong Loop

**Bad:** Researcher says "you write it", writer says "need more research" — forever.

**Good:** Max hops, clear ownership, validator decides when done.

```python
MAX_HOPS = 4


def bad_ping_pong_workflow(query: str) -> RunResult:
    metrics = CollaborationMetrics()
    owner = "researcher"
    research_done = False

    for hop in range(12):
        metrics.agent_hops += 1
        metrics.messages += 1
        if owner == "researcher":
            if not research_done:
                lookup_order("ORD-1001")
                metrics.tool_calls += 1
                research_done = True
            owner = "writer"  # "I found data, you write"
        else:
            owner = "researcher"  # "Need more detail" — ping-pong!
            metrics.ping_pong_count += 1

    metrics.termination_reason = "never_finished"
    return RunResult("(no answer)", metrics, "ping_pong", fixed=False)


def good_bounded_workflow(query: str) -> RunResult:
    metrics = CollaborationMetrics()
    order = lookup_order("ORD-1001")
    metrics.tool_calls += 1
    policies = search_policy("return")
    metrics.tool_calls += 1
    metrics.agent_hops = 3  # researcher → composer → done

    answer = (
        f"Order {order['order_id']} is {order['status']}. "
        f"Policy [{policies[0]['id']}]: {policies[0]['text']}"
    )
    metrics.terminated = True
    metrics.termination_reason = "final_answer"
    return RunResult(answer, metrics, "ping_pong", fixed=True)


bad_r = bad_ping_pong_workflow("ORD-1001 return?")
good_r = good_bounded_workflow("ORD-1001 return?")

print(f"BAD:  hops={bad_r.metrics.agent_hops} ping_pong={bad_r.metrics.ping_pong_count} score={bad_r.metrics.score()}")
print(f"GOOD: hops={good_r.metrics.agent_hops} answer={good_r.answer[:60]}... score={good_r.metrics.score()}")
```

## 5. Anti-Pattern #3 — Duplicate Work

**Bad:** Three agents each call `lookup_order` independently.

**Good:** First agent writes `artifacts.order_info`; others read it.

```python
def bad_duplicate_research(order_id: str) -> RunResult:
    metrics = CollaborationMetrics()
    agents = ["triage_agent", "policy_agent", "compose_agent"]
    for agent in agents:
        lookup_order(order_id)
        metrics.tool_calls += 1
        metrics.agent_hops += 1
    metrics.duplicate_tool_calls = 2
    return RunResult("Order looked up 3 times", metrics, "duplicate_work", fixed=False)


def good_shared_artifacts(order_id: str) -> RunResult:
    metrics = CollaborationMetrics()
    artifacts: dict[str, Any] = {}

    if "order_info" not in artifacts:
        artifacts["order_info"] = lookup_order(order_id)
        metrics.tool_calls += 1
    metrics.agent_hops += 1

    if "policy_hits" not in artifacts:
        artifacts["policy_hits"] = search_policy("return")
        metrics.tool_calls += 1
    metrics.agent_hops += 1

    order = artifacts["order_info"]
    answer = f"{order['order_id']} is {order['status']} — {artifacts['policy_hits'][0]['text']}"
    metrics.agent_hops += 1
    return RunResult(answer, metrics, "duplicate_work", fixed=True)


bad_dup = bad_duplicate_research("ORD-1001")
good_dup = good_shared_artifacts("ORD-1001")

print(f"BAD:  tool_calls={bad_dup.metrics.tool_calls} duplicates={bad_dup.metrics.duplicate_tool_calls}")
print(f"GOOD: tool_calls={good_dup.metrics.tool_calls} answer={good_dup.answer[:50]}...")
```

## 6. Anti-Pattern #4 — Supervisor With All Tools

**Bad:** "Supervisor" agent has every tool — workers are decorative.

**Good:** Supervisor routes only; workers own tools.

```python
SUPERVISOR_TOOLS_BAD = ["lookup_order", "search_policy", "create_ticket", "send_email"]
WORKER_TOOLS_UNUSED = {"order_worker": [], "policy_worker": []}  # empty — supervisor did it all

HIERARCHY_GOOD = {
    "supervisor": [],
    "order_worker": ["lookup_order"],
    "policy_worker": ["search_policy"],
}


def bad_supervisor_does_all(query: str) -> RunResult:
    metrics = CollaborationMetrics(agent_hops=1)
    order = lookup_order("ORD-1001")
    policies = search_policy("return")
    metrics.tool_calls = 2
    answer = f"Supervisor did everything: {order['status']}, {policies[0]['id']}"
    return RunResult(answer, metrics, "supervisor_bloat", fixed=False)


def good_supervisor_delegates(query: str) -> RunResult:
    metrics = CollaborationMetrics()
    metrics.agent_hops += 1  # supervisor routes
    metrics.agent_hops += 1
    order = lookup_order("ORD-1001")
    metrics.tool_calls += 1
    metrics.agent_hops += 1
    policies = search_policy("return")
    metrics.tool_calls += 1
    answer = f"order_worker→{order['status']} | policy_worker→[{policies[0]['id']}]"
    return RunResult(answer, metrics, "supervisor_bloat", fixed=True)


print("BAD supervisor tools:", SUPERVISOR_TOOLS_BAD)
print("GOOD hierarchy:", HIERARCHY_GOOD)
print(good_supervisor_delegates("test").answer)
```

## 7. Anti-Pattern #5 — Chatty Group With No Termination

**Bad:** Group chat where every agent adds a message but nobody owns the final answer.

**Good:** Turn limit + designated **composer** agent produces final output.

```python
GROUP_AGENTS = ["agent_a", "agent_b", "agent_c", "agent_d"]
MAX_GROUP_TURNS = 3


def bad_group_chat(query: str) -> RunResult:
    metrics = CollaborationMetrics()
    chat: list[str] = []
    for turn in range(15):
        speaker = GROUP_AGENTS[turn % len(GROUP_AGENTS)]
        chat.append(f"{speaker}: I think we should also consider...")
        metrics.messages += 1
        metrics.agent_hops += 1
    metrics.termination_reason = "no_final_owner"
    return RunResult("(discussion ongoing)", metrics, "chatty_group", fixed=False)


def good_group_with_composer(query: str) -> RunResult:
    metrics = CollaborationMetrics()
    notes: list[str] = []

    for turn in range(min(MAX_GROUP_TURNS, len(GROUP_AGENTS))):
        agent = GROUP_AGENTS[turn]
        if agent == "agent_a":
            notes.append(f"order={lookup_order('ORD-1001')['status']}")
            metrics.tool_calls += 1
        elif agent == "agent_b":
            notes.append(f"policy={search_policy('return')[0]['id']}")
            metrics.tool_calls += 1
        else:
            notes.append(f"{agent}: ack")
        metrics.messages += 1
        metrics.agent_hops += 1

    metrics.agent_hops += 1
    answer = f"Composer final: {' | '.join(notes)}"
    metrics.terminated = True
    metrics.termination_reason = "composer_final_answer"
    return RunResult(answer, metrics, "chatty_group", fixed=True)


bad_chat = bad_group_chat("help")
good_chat = good_group_with_composer("help")

print(f"BAD:  messages={bad_chat.metrics.messages} terminated={bad_chat.metrics.terminated}")
print(f"GOOD: messages={good_chat.metrics.messages} answer={good_chat.answer[:55]}...")
```

## 8. Anti-Pattern #6 — Contradictions Without a Merge Step

**Bad:** Optimistic agent says "yes you can return"; cautious agent says "no" — user gets both.

**Good:** Validator or composer resolves conflicts against policy artifacts.

```python
def bad_contradictory_agents() -> RunResult:
    metrics = CollaborationMetrics(agent_hops=2)
    optimistic = "Yes, you can return shipped items anytime!"
    cautious = "No, shipped orders cannot be returned per policy."
    answer = f"{optimistic} ... however ... {cautious}"
    return RunResult(answer, metrics, "contradiction", fixed=False)


def good_validator_merge() -> RunResult:
    metrics = CollaborationMetrics()
    order = lookup_order("ORD-1001")
    policies = search_policy("return shipped")
    metrics.tool_calls = 2
    metrics.agent_hops = 3

    drafts = [
        "Yes, return anytime!",
        "Shipped items may not qualify — check [ret-02].",
    ]
    ground_truth = policies[0]["text"]
    if order["status"] == "shipped" and "30 days" in ground_truth:
        answer = f"Order is shipped. Policy [ret-02]: {ground_truth}"
    else:
        answer = drafts[0]

    metrics.terminated = True
    metrics.termination_reason = "validator_merged"
    return RunResult(answer, metrics, "contradiction", fixed=True)


bad_con = bad_contradictory_agents()
good_con = good_validator_merge()

print(f"BAD answer:  {bad_con.answer}")
print(f"GOOD answer: {good_con.answer}")
```

## 9. Anti-Pattern #7 — Over-Agentification

**Bad:** Eight agents for a two-tool lookup task.

**Good:** Single pipeline or three agents max (research → compose → validate).

```python
OVER_AGENT_PIPELINE = [
    "greeting_agent", "intent_agent", "entity_agent", "router_agent",
    "order_agent", "policy_agent", "tone_agent", "formatter_agent",
]

LEAN_PIPELINE = ["intake_agent", "research_agent", "compose_agent"]


def run_pipeline(agents: list[str], order_id: str) -> RunResult:
    metrics = CollaborationMetrics()
    order = None
    for agent in agents:
        metrics.agent_hops += 1
        if "order" in agent and order is None:
            order = lookup_order(order_id)
            metrics.tool_calls += 1
        if "policy" in agent:
            search_policy("return")
            metrics.tool_calls += 1
    answer = f"Done after {len(agents)} agents"
    return RunResult(answer, metrics, "over_agentification", fixed=len(agents) <= 4)


bloated = run_pipeline(OVER_AGENT_PIPELINE, "ORD-1001")
lean = run_pipeline(LEAN_PIPELINE, "ORD-1001")

print(f"Bloated: {bloated.metrics.agent_hops} hops, {bloated.metrics.tool_calls} tools")
print(f"Lean:    {lean.metrics.agent_hops} hops, {lean.metrics.tool_calls} tools")
```

## 10. Anti-Pattern #8 — Unguarded Shared State Writes

**Bad:** Two agents overwrite the same `artifacts.draft` key — last write wins, first work lost.

**Good:** Namespaced keys per agent or merge function with contract enforcement.

```python
def bad_unguarded_state() -> RunResult:
    metrics = CollaborationMetrics()
    shared: dict[str, str] = {}

    shared["draft"] = "Agent A: Order is processing, customer can cancel."
    metrics.agent_hops += 1
    shared["draft"] = "Agent B: Order is shipped, cannot cancel."  # overwrites A
    metrics.agent_hops += 1

    return RunResult(shared["draft"], metrics, "unguarded_state", fixed=False)


def good_namespaced_artifacts() -> RunResult:
    metrics = CollaborationMetrics()
    artifacts: dict[str, Any] = {}

    order = lookup_order("ORD-1001")
    metrics.tool_calls += 1
    artifacts["order_agent_notes"] = f"Status: {order['status']}"
    metrics.agent_hops += 1

    policies = search_policy("cancel")
    metrics.tool_calls += 1
    artifacts["policy_agent_notes"] = f"[{policies[0]['id']}] {policies[0]['text']}"
    metrics.agent_hops += 1

    answer = f"{artifacts['order_agent_notes']}. {artifacts['policy_agent_notes']}"
    metrics.agent_hops += 1
    return RunResult(answer, metrics, "unguarded_state", fixed=True)


bad_state = bad_unguarded_state()
good_state = good_namespaced_artifacts()

print(f"BAD (lost Agent A work):  {bad_state.answer}")
print(f"GOOD (merged): {good_state.answer}")
```

## 12. Detection Heuristics — Automate in CI

Flag runs in eval pipelines when:

```python
COLLABORATION_ALERTS = {
    "handoff_bytes > 2000": "likely context dump",
    "ping_pong_count >= 3": "agent ping-pong loop",
    "duplicate_tool_calls > 0": "redundant research",
    "agent_hops > 8": "over-agentification or missing termination",
    "messages > 10 and not terminated": "chatty group without composer",
    "tool_calls > agent_hops * 2": "supervisor doing worker tools",
}


def detect_alerts(metrics: CollaborationMetrics) -> list[str]:
    alerts: list[str] = []
    if metrics.handoff_bytes > 2000:
        alerts.append(COLLABORATION_ALERTS["handoff_bytes > 2000"])
    if metrics.ping_pong_count >= 3:
        alerts.append(COLLABORATION_ALERTS["ping_pong_count >= 3"])
    if metrics.duplicate_tool_calls > 0:
        alerts.append(COLLABORATION_ALERTS["duplicate_tool_calls > 0"])
    if metrics.agent_hops > 8:
        alerts.append(COLLABORATION_ALERTS["agent_hops > 8"])
    if metrics.messages > 10 and not metrics.terminated:
        alerts.append(COLLABORATION_ALERTS["messages > 10 and not terminated"])
    return alerts


for r in ALL_RUNS:
    if not r.fixed:
        alerts = detect_alerts(r.metrics)
        if alerts:
            print(f"{r.pattern}: {alerts}")
```

## 13. Refactored Pipeline — All Fixes Combined

A lean ShopCo support pipeline applying every fix: structured handoffs, shared artifacts, bounded hops, namespaced writes, validator merge.

```python
@dataclass
class WorkflowState:
    thread_id: str
    user_query: str
    entities: dict[str, Any] = field(default_factory=dict)
    artifacts: dict[str, Any] = field(default_factory=dict)
    handoff_log: list[dict[str, Any]] = field(default_factory=list)


def structured_handoff(from_a: str, to_a: str, task: str, payload: dict) -> dict:
    return {
        "from": from_a, "to": to_a,
        "task_summary": task,
        "payload": payload,
    }


class HealthyShopCoPipeline:
    MAX_HOPS = 6

    def run(self, query: str) -> RunResult:
        metrics = CollaborationMetrics()
        state = WorkflowState(thread_id=str(uuid.uuid4())[:8], user_query=query)

        # Intake
        oid = re.search(r"ORD-\d{4}", query.upper())
        state.entities = {
            "order_id": oid.group(0) if oid else None,
            "intent": "return" if "return" in query.lower() else "status",
        }
        metrics.agent_hops += 1

        # Research (deduped)
        h = structured_handoff("intake", "research", "Lookup order and policy", state.entities)
        metrics.handoff_bytes += handoff_size(h)
        state.handoff_log.append(h)
        if state.entities.get("order_id"):
            state.artifacts["order_info"] = lookup_order(state.entities["order_id"])
            metrics.tool_calls += 1
        state.artifacts["policy_hits"] = search_policy(state.entities.get("intent", "general"))
        metrics.tool_calls += 1
        metrics.agent_hops += 1

        # Compose
        h2 = structured_handoff("research", "compose", "Draft answer", {"keys": list(state.artifacts)})
        metrics.handoff_bytes += handoff_size(h2)
        state.handoff_log.append(h2)
        order = state.artifacts.get("order_info", {})
        pol = state.artifacts["policy_hits"][0]
        draft = (
            f"Order {order.get('order_id', 'N/A')} status: {order.get('status', 'unknown')}. "
            f"[{pol['id']}] {pol['text']}"
            if order.get("found") else f"[{pol['id']}] {pol['text']}"
        )
        state.artifacts["compose_draft"] = draft
        metrics.agent_hops += 1

        # Validate
        if "[" not in draft:
            draft += " [ret-02]"
        metrics.agent_hops += 1
        metrics.terminated = True
        metrics.termination_reason = "validator_approved"

        if metrics.agent_hops > self.MAX_HOPS:
            metrics.termination_reason = "max_hops"

        return RunResult(draft, metrics, "healthy_pipeline", fixed=True)


healthy = HealthyShopCoPipeline().run("Can I return order ORD-1001?")
print(f"Answer: {healthy.answer}")
print(f"Metrics: hops={healthy.metrics.agent_hops} tools={healthy.metrics.tool_calls} "
      f"handoff_bytes={healthy.metrics.handoff_bytes} score={healthy.metrics.score()}")
```

## 14. Prevention Checklist

| # | Before shipping a multi-agent workflow |
|---|----------------------------------------|
| 1 | Handoffs use structured payload, not chat dumps |
| 2 | Every run has a **single final answer owner** (composer or validator) |
| 3 | `max_hops` and termination enforced by orchestrator |
| 4 | Shared artifacts prevent duplicate tool calls |
| 5 | Supervisor routes; workers own tools |
| 6 | Group chats have turn limits + composer |
| 7 | Validator resolves contradictions against ground truth |
| 8 | Agent count matches task complexity (≤4 for simple tasks) |
| 9 | State writes are namespaced or contract-guarded |
| 10 | CI alerts on collaboration metrics thresholds |

## 16. Summary

Multi-agent collaboration fails repeatedly in the same ways:

- **Context dumps** → structured handoffs
- **Ping-pong loops** → hop limits + single owner
- **Duplicate work** → shared artifacts
- **Fake supervisors** → workers own tools
- **Chatty groups** → turn caps + composer
- **Contradictions** → validator against policy ground truth
- **Too many agents** → simplest pipeline that works
- **Unguarded state** → namespaced artifact keys

Measure every workflow with **CollaborationMetrics** and alert in CI when hops, handoff size, or ping-pong counts cross thresholds. The `HealthyShopCoPipeline` combined all fixes in one lean run — that is the target shape for production.
