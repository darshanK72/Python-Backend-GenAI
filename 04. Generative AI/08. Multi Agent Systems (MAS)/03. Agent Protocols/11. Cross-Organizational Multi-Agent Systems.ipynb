# Cross-Organizational Multi-Agent Systems

Most real agent deployments do not live in a single company's VPC. A retail support agent at **ShopCo** may need shipment status from **ShipFast Logistics**, refund authorization from **PaySecure**, and warranty terms from **GearGuard** — each a separate organization with its own data, policies, and agents.

Cross-organizational multi-agent systems (MAS) add a layer of problems beyond in-house orchestration:

- **Trust boundaries** — who may call whom, and with what scope?
- **Discovery** — how does ShopCo find ShipFast's agent capabilities without a shared database?
- **Data minimization** — share tracking status, not the full customer PII file
- **Auditability** — every cross-border call logged for compliance
- **Failure isolation** — partner outage must not crash ShopCo's entire support flow

This notebook implements a **federated agent mesh** in plain Python. Each organization publishes an **Agent Card** (A2A-style manifest), exposes local tools via an MCP-like server, and communicates through a **Cross-Org Broker** that enforces trust policies and scoped credentials. The scenario: resolve a customer refund that requires data from three organizations.

Everything runs offline. An optional cell at the end shows how a live LLM would drive the same cross-org orchestration.

````python
"""
Cross-Organizational MAS Lab — ShopCo federated agent mesh.
"""

import json
import re
import uuid
from dataclasses import dataclass, field
from datetime import datetime, timezone
from enum import Enum
from typing import Any, Callable


def pretty(obj: Any) -> str:
    return json.dumps(obj, indent=2, default=str)


def utcnow() -> str:
    return datetime.now(timezone.utc).isoformat()


# --- ShopCo (retailer) ---
SHOPCO_ORDERS: dict[str, dict[str, Any]] = {
    "ORD-1001": {
        "order_id": "ORD-1001",
        "customer_ref": "CUST-A1",  # opaque ID — not email
        "status": "shipped",
        "total_usd": 1299.00,
        "items": ["Laptop Stand"],
        "shipfast_ref": "SF-77821",
        "paysecure_ref": "PS-TXN-9912",
    },
    "ORD-1002": {
        "order_id": "ORD-1002",
        "customer_ref": "CUST-B2",
        "status": "processing",
        "total_usd": 449.00,
        "items": ["Wireless Keyboard"],
        "shipfast_ref": None,
        "paysecure_ref": "PS-TXN-9913",
    },
}

# --- ShipFast Logistics (shipping partner) ---
SHIPFAST_SHIPMENTS: dict[str, dict[str, Any]] = {
    "SF-77821": {
        "shipment_id": "SF-77821",
        "status": "delivered_late",
        "promised_date": "2026-07-01",
        "actual_date": "2026-07-08",
        "days_late": 7,
        "carrier": "FedEx",
    },
}

# --- PaySecure (payment processor) ---
PAYSECURE_TXNS: dict[str, dict[str, Any]] = {
    "PS-TXN-9912": {
        "txn_id": "PS-TXN-9912",
        "amount_usd": 1299.00,
        "refund_eligible": True,
        "max_refund_usd": 1299.00,
        "chargeback_window_days": 120,
    },
    "PS-TXN-9913": {
        "txn_id": "PS-TXN-9913",
        "amount_usd": 449.00,
        "refund_eligible": True,
        "max_refund_usd": 449.00,
        "chargeback_window_days": 120,
    },
}

# --- GearGuard (warranty partner) ---
GEARGUARD_WARRANTIES: dict[str, dict[str, Any]] = {
    "ORD-1001": {
        "order_id": "ORD-1001",
        "warranty_months": 12,
        "covers_shipping_delay": False,
        "notes": "Manufacturer warranty only; shipping issues handled by retailer.",
    },
}

CROSS_ORG_AUDIT: list[dict[str, Any]] = []

print("Federated mesh ready: ShopCo, ShipFast, PaySecure, GearGuard")
````

## 1. The Cross-Org Problem

Inside one company, an orchestrator can call any internal API. Across companies, every hop crosses a **legal and technical boundary**:

```
  Customer ──▶ ShopCo Support Agent
                    │
         ┌──────────┼──────────┐
         │          │          │
         ▼          ▼          ▼
     ShopCo DB   ShipFast    PaySecure
     (orders)    (tracking)  (refunds)
         │          │          │
         └─ separate orgs, separate agents, separate trust ─┘
```

| In-house MAS | Cross-organizational MAS |
|--------------|--------------------------|
| Shared identity (SSO) | Federated credentials with scoped claims |
| Implicit trust | Explicit allowlists and contracts |
| Full data access | Data minimization per request |
| Single audit log | Per-org logs + cross-org correlation IDs |
| One protocol stack | MCP inside org; A2A (or similar) between orgs |

---

## 2. Protocol Layers — MCP Inside, A2A Across

A practical pattern:

- **Within** each organization: **MCP** servers expose tools/resources to that org's agents (orders DB, shipment API, payment API).
- **Between** organizations: **Agent-to-Agent (A2A)** messages carry tasks, agent cards, and structured responses across trust boundaries.

```
  ShopCo Host                    ShipFast Org
  ┌─────────────┐               ┌─────────────┐
  │ ShopCo Agent│──A2A task────▶│ Logistics   │
  │             │◀─A2A result───│ Agent       │
  │     │       │               │     │       │
  │  MCP client │               │  MCP server │
  └─────┼───────┘               └─────┼───────┘
        ▼                             ▼
   local orders                  shipment API
```

The ShopCo agent never calls ShipFast's database directly. It sends an A2A task to ShipFast's published agent, which uses its own MCP tools internally.

## 3. Agent Cards — Cross-Org Discovery

An **Agent Card** is a JSON manifest describing who an agent is, what skills it offers, and how to reach it. Partners publish cards to a registry (or exchange them during onboarding). ShopCo discovers ShipFast's `get_shipment_status` skill without reading ShipFast's source code.

````python
@dataclass
class AgentSkill:
    id: str
    name: str
    description: str
    input_schema: dict[str, Any]
    output_schema: dict[str, Any] | None = None


@dataclass
class AgentCard:
    org_id: str
    agent_id: str
    name: str
    description: str
    version: str
    skills: list[AgentSkill]
    auth_scheme: str = "scoped_bearer"

    def to_dict(self) -> dict[str, Any]:
        return {
            "org_id": self.org_id,
            "agent_id": self.agent_id,
            "name": self.name,
            "description": self.description,
            "version": self.version,
            "auth_scheme": self.auth_scheme,
            "skills": [
                {"id": s.id, "name": s.name, "description": s.description, "input_schema": s.input_schema}
                for s in self.skills
            ],
        }


AGENT_REGISTRY: dict[str, AgentCard] = {}


def register_agent(card: AgentCard) -> None:
    key = f"{card.org_id}/{card.agent_id}"
    AGENT_REGISTRY[key] = card


register_agent(AgentCard(
    org_id="shopco",
    agent_id="support-orchestrator",
    name="ShopCo Support Orchestrator",
    description="Coordinates cross-partner resolution for customer tickets.",
    version="2.1.0",
    skills=[
        AgentSkill("get_order", "get_order", "Look up order by ID",
                   {"type": "object", "properties": {"order_id": {"type": "string"}}, "required": ["order_id"]}),
    ],
))

register_agent(AgentCard(
    org_id="shipfast",
    agent_id="logistics-agent",
    name="ShipFast Logistics Agent",
    description="Shipment tracking and delivery attestation for retail partners.",
    version="1.4.0",
    skills=[
        AgentSkill("shipment_status", "get_shipment_status", "Return delivery status for a shipment ref",
                   {"type": "object", "properties": {"shipment_id": {"type": "string"}}, "required": ["shipment_id"]}),
    ],
))

register_agent(AgentCard(
    org_id="paysecure",
    agent_id="refund-agent",
    name="PaySecure Refund Agent",
    description="Refund eligibility and authorization for partner retailers.",
    version="3.0.1",
    skills=[
        AgentSkill("refund_check", "check_refund_eligibility", "Check if txn is eligible for refund",
                   {"type": "object", "properties": {"txn_id": {"type": "string"}}, "required": ["txn_id"]}),
        AgentSkill("refund_auth", "authorize_refund", "Authorize a refund up to max amount",
                   {"type": "object", "properties": {"txn_id": {"type": "string"}, "amount_usd": {"type": "number"}}, "required": ["txn_id", "amount_usd"]}),
    ],
))

register_agent(AgentCard(
    org_id="gearguard",
    agent_id="warranty-agent",
    name="GearGuard Warranty Agent",
    description="Warranty coverage lookup for registered products.",
    version="1.0.0",
    skills=[
        AgentSkill("warranty_lookup", "get_warranty", "Warranty terms for an order",
                   {"type": "object", "properties": {"order_id": {"type": "string"}}, "required": ["order_id"]}),
    ],
))

print(f"Agent registry: {len(AGENT_REGISTRY)} agents")
for key, card in AGENT_REGISTRY.items():
    skills = ", ".join(s.name for s in card.skills)
    print(f"  • {key}: {skills}")
````

## 4. Local MCP Servers — One Per Organization

Each org wraps its internal data behind an MCP-style tool server. External callers never see the raw database — only tool results filtered by policy.

````python
@dataclass
class ToolResult:
    data: Any
    is_error: bool = False


class OrgMCPServer:
    def __init__(self, org_id: str):
        self.org_id = org_id
        self._tools: dict[str, Callable[..., Any]] = {}

    def tool(self, name: str):
        def decorator(fn: Callable):
            self._tools[name] = fn
            return fn
        return decorator

    def call(self, name: str, args: dict[str, Any]) -> ToolResult:
        fn = self._tools.get(name)
        if not fn:
            return ToolResult(f"Unknown tool: {name}", is_error=True)
        try:
            return ToolResult(fn(**args))
        except (ValueError, KeyError) as e:
            return ToolResult(str(e), is_error=True)


# ShopCo local server
shopco_mcp = OrgMCPServer("shopco")

@shopco_mcp.tool("get_order")
def shopco_get_order(order_id: str) -> dict[str, Any]:
    order = SHOPCO_ORDERS.get(order_id.strip().upper())
    if not order:
        raise ValueError(f"Order {order_id} not found")
    return order


# ShipFast local server
shipfast_mcp = OrgMCPServer("shipfast")

@shipfast_mcp.tool("get_shipment_status")
def shipfast_status(shipment_id: str) -> dict[str, Any]:
    rec = SHIPFAST_SHIPMENTS.get(shipment_id.strip().upper())
    if not rec:
        raise ValueError(f"Shipment {shipment_id} not found")
    # Data minimization: partner sees status, not internal routing codes
    return {
        "shipment_id": rec["shipment_id"],
        "status": rec["status"],
        "days_late": rec.get("days_late", 0),
        "carrier": rec["carrier"],
    }


# PaySecure local server
paysecure_mcp = OrgMCPServer("paysecure")

@paysecure_mcp.tool("check_refund_eligibility")
def paysecure_check(txn_id: str) -> dict[str, Any]:
    txn = PAYSECURE_TXNS.get(txn_id)
    if not txn:
        raise ValueError(f"Transaction {txn_id} not found")
    return {"txn_id": txn_id, "refund_eligible": txn["refund_eligible"], "max_refund_usd": txn["max_refund_usd"]}

@paysecure_mcp.tool("authorize_refund")
def paysecure_authorize(txn_id: str, amount_usd: float) -> dict[str, Any]:
    txn = PAYSECURE_TXNS.get(txn_id)
    if not txn:
        raise ValueError(f"Transaction {txn_id} not found")
    if not txn["refund_eligible"]:
        raise ValueError("Transaction not eligible for refund")
    if amount_usd > txn["max_refund_usd"]:
        raise ValueError(f"Amount {amount_usd} exceeds max {txn['max_refund_usd']}")
    auth_id = f"AUTH-{uuid.uuid4().hex[:8].upper()}"
    return {"authorization_id": auth_id, "txn_id": txn_id, "amount_usd": amount_usd, "status": "authorized"}


# GearGuard local server
gearguard_mcp = OrgMCPServer("gearguard")

@gearguard_mcp.tool("get_warranty")
def gearguard_warranty(order_id: str) -> dict[str, Any]:
    rec = GEARGUARD_WARRANTIES.get(order_id.strip().upper())
    if not rec:
        return {"order_id": order_id, "warranty_months": 0, "covers_shipping_delay": False, "notes": "No warranty on file."}
    return rec


ORG_SERVERS = {"shopco": shopco_mcp, "shipfast": shipfast_mcp, "paysecure": paysecure_mcp, "gearguard": gearguard_mcp}
print(f"MCP servers online: {list(ORG_SERVERS)}")
````

## 5. A2A Message Envelope

Cross-org messages carry metadata for routing, correlation, and audit. Every request gets a **correlation ID** that propagates across all partner hops so support can trace one customer ticket end to end.

````python
class MessageKind(str, Enum):
    TASK = "task"
    RESULT = "result"
    ERROR = "error"


@dataclass
class A2AMessage:
    message_id: str
    correlation_id: str
    kind: MessageKind
    from_org: str
    from_agent: str
    to_org: str
    to_agent: str
    skill_id: str
    payload: dict[str, Any]
    timestamp: str = field(default_factory=utcnow)
    credential_id: str | None = None

    def to_dict(self) -> dict[str, Any]:
        return {
            "message_id": self.message_id,
            "correlation_id": self.correlation_id,
            "kind": self.kind.value,
            "from": f"{self.from_org}/{self.from_agent}",
            "to": f"{self.to_org}/{self.to_agent}",
            "skill_id": self.skill_id,
            "payload": self.payload,
            "timestamp": self.timestamp,
            "credential_id": self.credential_id,
        }


def make_task(
    *, from_org: str, from_agent: str, to_org: str, to_agent: str,
    skill_id: str, payload: dict[str, Any], correlation_id: str,
    credential_id: str | None = None,
) -> A2AMessage:
    return A2AMessage(
        message_id=str(uuid.uuid4()),
        correlation_id=correlation_id,
        kind=MessageKind.TASK,
        from_org=from_org, from_agent=from_agent,
        to_org=to_org, to_agent=to_agent,
        skill_id=skill_id, payload=payload,
        credential_id=credential_id,
    )


sample = make_task(
    from_org="shopco", from_agent="support-orchestrator",
    to_org="shipfast", to_agent="logistics-agent",
    skill_id="shipment_status", payload={"shipment_id": "SF-77821"},
    correlation_id="TICKET-4421",
    credential_id="cred-shopco-shipfast-read",
)
print(pretty(sample.to_dict()))
````

## 6. Trust Policies and Scoped Credentials

Partners do not grant blanket API access. ShopCo receives a **scoped credential** that allows only specific skills on specific agents. The broker rejects out-of-scope calls before they reach the partner's MCP server.

````python
@dataclass
class ScopedCredential:
    credential_id: str
    holder_org: str
    target_org: str
    target_agent: str
    allowed_skills: list[str]
    expires_at: str
    max_calls_per_day: int = 1000
    calls_today: int = 0

    def allows(self, skill_id: str) -> bool:
        return skill_id in self.allowed_skills

    def is_expired(self) -> bool:
        return datetime.now(timezone.utc) > datetime.fromisoformat(self.expires_at)


CREDENTIAL_VAULT: dict[str, ScopedCredential] = {
    "cred-shopco-shipfast-read": ScopedCredential(
        credential_id="cred-shopco-shipfast-read",
        holder_org="shopco", target_org="shipfast", target_agent="logistics-agent",
        allowed_skills=["shipment_status"],
        expires_at="2027-01-01T00:00:00+00:00",
    ),
    "cred-shopco-paysecure-refund": ScopedCredential(
        credential_id="cred-shopco-paysecure-refund",
        holder_org="shopco", target_org="paysecure", target_agent="refund-agent",
        allowed_skills=["refund_check", "refund_auth"],
        expires_at="2027-01-01T00:00:00+00:00",
    ),
    "cred-shopco-gearguard-read": ScopedCredential(
        credential_id="cred-shopco-gearguard-read",
        holder_org="shopco", target_org="gearguard", target_agent="warranty-agent",
        allowed_skills=["warranty_lookup"],
        expires_at="2027-01-01T00:00:00+00:00",
    ),
}


class TrustPolicyEngine:
    def validate(self, msg: A2AMessage) -> tuple[bool, str]:
        if msg.kind != MessageKind.TASK:
            return True, "ok"
        if not msg.credential_id:
            return False, "Missing credential_id"
        cred = CREDENTIAL_VAULT.get(msg.credential_id)
        if not cred:
            return False, f"Unknown credential: {msg.credential_id}"
        if cred.is_expired():
            return False, "Credential expired"
        if cred.holder_org != msg.from_org:
            return False, f"Credential holder mismatch: {cred.holder_org} != {msg.from_org}"
        if cred.target_org != msg.to_org or cred.target_agent != msg.to_agent:
            return False, "Credential not valid for target agent"
        if not cred.allows(msg.skill_id):
            return False, f"Skill {msg.skill_id} not in allowed list"
        if cred.calls_today >= cred.max_calls_per_day:
            return False, "Daily call quota exceeded"
        cred.calls_today += 1
        return True, "ok"


trust = TrustPolicyEngine()
ok, reason = trust.validate(sample)
print(f"Sample task trust check: ok={ok}, reason={reason}")
````

## 7. Cross-Org Broker

The broker routes A2A tasks to the target org's agent runtime, which invokes local MCP tools and returns results. It logs every hop to the cross-org audit ledger.

````python
SKILL_TO_TOOL: dict[str, dict[str, str]] = {
    "shopco/support-orchestrator": {"get_order": "get_order"},
    "shipfast/logistics-agent": {"shipment_status": "get_shipment_status"},
    "paysecure/refund-agent": {"refund_check": "check_refund_eligibility", "refund_auth": "authorize_refund"},
    "gearguard/warranty-agent": {"warranty_lookup": "get_warranty"},
}


class CrossOrgBroker:
    def __init__(self, trust_engine: TrustPolicyEngine):
        self.trust = trust_engine

    def _audit(self, event: str, msg: A2AMessage, extra: dict[str, Any] | None = None) -> None:
        CROSS_ORG_AUDIT.append({
            "event": event,
            "correlation_id": msg.correlation_id,
            "message_id": msg.message_id,
            "from": f"{msg.from_org}/{msg.from_agent}",
            "to": f"{msg.to_org}/{msg.to_agent}",
            "skill_id": msg.skill_id,
            "timestamp": utcnow(),
            **(extra or {}),
        })

    def send(self, msg: A2AMessage) -> A2AMessage:
        ok, reason = self.trust.validate(msg)
        if not ok:
            self._audit("denied", msg, {"reason": reason})
            return A2AMessage(
                message_id=str(uuid.uuid4()), correlation_id=msg.correlation_id,
                kind=MessageKind.ERROR, from_org=msg.to_org, from_agent=msg.to_agent,
                to_org=msg.from_org, to_agent=msg.from_agent, skill_id=msg.skill_id,
                payload={"error": reason},
            )

        agent_key = f"{msg.to_org}/{msg.to_agent}"
        card = AGENT_REGISTRY.get(agent_key)
        if not card:
            self._audit("agent_not_found", msg)
            return A2AMessage(
                message_id=str(uuid.uuid4()), correlation_id=msg.correlation_id,
                kind=MessageKind.ERROR, from_org=msg.to_org, from_agent=msg.to_agent,
                to_org=msg.from_org, to_agent=msg.from_agent, skill_id=msg.skill_id,
                payload={"error": f"Agent not found: {agent_key}"},
            )

        tool_map = SKILL_TO_TOOL.get(agent_key, {})
        tool_name = tool_map.get(msg.skill_id)
        if not tool_name:
            self._audit("skill_unmapped", msg)
            return A2AMessage(
                message_id=str(uuid.uuid4()), correlation_id=msg.correlation_id,
                kind=MessageKind.ERROR, from_org=msg.to_org, from_agent=msg.to_agent,
                to_org=msg.from_org, to_agent=msg.from_agent, skill_id=msg.skill_id,
                payload={"error": f"No tool mapping for skill {msg.skill_id}"},
            )

        server = ORG_SERVERS.get(msg.to_org)
        if not server:
            return A2AMessage(
                message_id=str(uuid.uuid4()), correlation_id=msg.correlation_id,
                kind=MessageKind.ERROR, from_org=msg.to_org, from_agent=msg.to_agent,
                to_org=msg.from_org, to_agent=msg.from_agent, skill_id=msg.skill_id,
                payload={"error": f"No MCP server for org {msg.to_org}"},
            )

        self._audit("task_received", msg)
        result = server.call(tool_name, msg.payload)
        kind = MessageKind.ERROR if result.is_error else MessageKind.RESULT
        response = A2AMessage(
            message_id=str(uuid.uuid4()), correlation_id=msg.correlation_id,
            kind=kind, from_org=msg.to_org, from_agent=msg.to_agent,
            to_org=msg.from_org, to_agent=msg.from_agent, skill_id=msg.skill_id,
            payload={"error": result.data} if result.is_error else {"result": result.data},
        )
        self._audit("task_completed" if not result.is_error else "task_failed", response)
        return response


broker = CrossOrgBroker(trust)
resp = broker.send(sample)
print("Broker response:")
print(pretty(resp.to_dict()))
````

## 8. ShopCo Orchestrator — Cross-Org Workflow

A customer reports: *"My order ORD-1001 arrived a week late — I want a full refund."* The ShopCo orchestrator must:

1. Load the order locally
2. Ask ShipFast for delivery attestation
3. Ask PaySecure for refund eligibility and authorization
4. Optionally check GearGuard warranty (shipping delay not covered)
5. Synthesize a resolution with audit trail

````python
@dataclass
class CrossOrgResolution:
    correlation_id: str
    order_id: str
    shipment: dict[str, Any] | None
    refund: dict[str, Any] | None
    warranty: dict[str, Any] | None
    decision: str
    partner_errors: list[str] = field(default_factory=list)


class ShopCoOrchestrator:
    def __init__(self, broker: CrossOrgBroker):
        self.broker = broker
        self.org = "shopco"
        self.agent = "support-orchestrator"

    def _partner_call(
        self, *, correlation_id: str, to_org: str, to_agent: str,
        skill_id: str, payload: dict[str, Any], credential_id: str,
    ) -> dict[str, Any] | None:
        task = make_task(
            from_org=self.org, from_agent=self.agent,
            to_org=to_org, to_agent=to_agent,
            skill_id=skill_id, payload=payload,
            correlation_id=correlation_id, credential_id=credential_id,
        )
        reply = self.broker.send(task)
        if reply.kind == MessageKind.ERROR:
            return None
        if reply.kind == MessageKind.RESULT:
            return reply.payload.get("result")
        return None

    def resolve_late_delivery_refund(self, order_id: str) -> CrossOrgResolution:
        correlation_id = f"TICKET-{uuid.uuid4().hex[:6].upper()}"
        errors: list[str] = []

        order = shopco_mcp.call("get_order", {"order_id": order_id})
        if order.is_error:
            return CrossOrgResolution(correlation_id, order_id, None, None, None, "order_not_found", [str(order.data)])
        order_data = order.data

        shipment = None
        if order_data.get("shipfast_ref"):
            shipment = self._partner_call(
                correlation_id=correlation_id, to_org="shipfast", to_agent="logistics-agent",
                skill_id="shipment_status", payload={"shipment_id": order_data["shipfast_ref"]},
                credential_id="cred-shopco-shipfast-read",
            )
            if shipment is None:
                errors.append("ShipFast unavailable or denied")

        refund_elig = None
        refund_auth = None
        if order_data.get("paysecure_ref"):
            refund_elig = self._partner_call(
                correlation_id=correlation_id, to_org="paysecure", to_agent="refund-agent",
                skill_id="refund_check", payload={"txn_id": order_data["paysecure_ref"]},
                credential_id="cred-shopco-paysecure-refund",
            )
            if refund_elig and refund_elig.get("refund_eligible") and shipment and shipment.get("days_late", 0) >= 5:
                refund_auth = self._partner_call(
                    correlation_id=correlation_id, to_org="paysecure", to_agent="refund-agent",
                    skill_id="refund_auth",
                    payload={"txn_id": order_data["paysecure_ref"], "amount_usd": order_data["total_usd"]},
                    credential_id="cred-shopco-paysecure-refund",
                )
            elif refund_elig is None:
                errors.append("PaySecure unavailable or denied")

        warranty = self._partner_call(
            correlation_id=correlation_id, to_org="gearguard", to_agent="warranty-agent",
            skill_id="warranty_lookup", payload={"order_id": order_id},
            credential_id="cred-shopco-gearguard-read",
        )

        if refund_auth:
            decision = f"Full refund authorized: {refund_auth['authorization_id']} for ${refund_auth['amount_usd']}"
        elif shipment and shipment.get("days_late", 0) < 5:
            decision = "Delay below refund threshold (5 days). Offer shipping credit."
        elif errors:
            decision = "Partial resolution — partner errors prevented full refund. Escalate to human."
        else:
            decision = "Refund not eligible or insufficient evidence."

        return CrossOrgResolution(
            correlation_id=correlation_id, order_id=order_id,
            shipment=shipment, refund=refund_auth or refund_elig, warranty=warranty,
            decision=decision, partner_errors=errors,
        )


orchestrator = ShopCoOrchestrator(broker)
resolution = orchestrator.resolve_late_delivery_refund("ORD-1001")
print(f"Correlation: {resolution.correlation_id}")
print(f"Decision: {resolution.decision}")
print(f"\nShipment: {pretty(resolution.shipment)}")
print(f"Refund: {pretty(resolution.refund)}")
````

## 12. Topology — Hub Broker vs Peer-to-Peer

| Topology | How it works | Best for |
|----------|--------------|----------|
| **Hub broker** | All cross-org traffic routes through a neutral broker (or ShopCo's integration hub) | Many partners, centralized audit, policy enforcement |
| **Peer-to-peer** | ShopCo agent calls ShipFast agent directly after discovery | Few trusted partners, lower latency |
| **Federated registry** | Each org publishes Agent Cards; discovery is decentralized, routing may be direct | Industry consortia, open marketplaces |

```
Hub:     ShopCo ──▶ Broker ◀── ShipFast
                      ▲
                 PaySecure

P2P:     ShopCo ──────────▶ ShipFast
           └──────────────▶ PaySecure
```

Our lab uses a hub broker because it centralizes trust checks and audit — the pattern most enterprises adopt first.

## 13. Inter-Org Contracts

Beyond credentials, partners sign **capability contracts** specifying SLAs, data fields, and escalation paths. The orchestrator validates responses against the contract before acting.

````python
@dataclass
class PartnerContract:
    partner_org: str
    skill_id: str
    required_response_fields: list[str]
    sla_seconds: int
    escalation_contact: str


CONTRACTS: list[PartnerContract] = [
    PartnerContract("shipfast", "shipment_status", ["shipment_id", "status", "days_late"], sla_seconds=5, escalation_contact="shipfast-ops@partner.com"),
    PartnerContract("paysecure", "refund_check", ["txn_id", "refund_eligible", "max_refund_usd"], sla_seconds=10, escalation_contact="paysecure-b2b@partner.com"),
    PartnerContract("paysecure", "refund_auth", ["authorization_id", "amount_usd", "status"], sla_seconds=30, escalation_contact="paysecure-b2b@partner.com"),
]


def validate_contract(partner_org: str, skill_id: str, response: dict[str, Any]) -> tuple[bool, list[str]]:
    contract = next((c for c in CONTRACTS if c.partner_org == partner_org and c.skill_id == skill_id), None)
    if not contract:
        return True, []
    missing = [f for f in contract.required_response_fields if f not in response]
    return len(missing) == 0, missing


if resolution.shipment:
    ok, missing = validate_contract("shipfast", "shipment_status", resolution.shipment)
    print(f"ShipFast contract valid: {ok}" + (f" missing={missing}" if missing else ""))
if resolution.refund:
    skill = "refund_auth" if "authorization_id" in resolution.refund else "refund_check"
    ok, missing = validate_contract("paysecure", skill, resolution.refund)
    print(f"PaySecure contract valid: {ok}" + (f" missing={missing}" if missing else ""))
````

## 14. Agent Card Discovery Flow

When ShopCo onboards a new partner, the integration flow is:

1. Partner publishes Agent Card to registry
2. Legal/procurement signs data-sharing agreement
3. Ops issues scoped credential for allowed skills
4. ShopCo orchestrator maps `skill_id` → MCP tool name
5. Contract defines required response fields and SLA
6. Smoke test via broker before production traffic

````python
def discover_partner_skills(target_org: str) -> list[dict[str, Any]]:
    skills = []
    for key, card in AGENT_REGISTRY.items():
        if card.org_id == target_org:
            for s in card.skills:
                cred = next(
                    (c for c in CREDENTIAL_VAULT.values() if c.target_org == target_org and c.allows(s.id)),
                    None,
                )
                skills.append({
                    "agent": key,
                    "skill": s.name,
                    "description": s.description,
                    "credential_ready": cred is not None,
                })
    return skills


for org in ("shipfast", "paysecure", "gearguard"):
    print(f"\n{org.upper()} discoverable skills:")
    for s in discover_partner_skills(org):
        flag = "✓" if s["credential_ready"] else "✗"
        print(f"  [{flag}] {s['skill']}: {s['description']}")
````

## 15. Offline Multi-Turn Cross-Org Dialogue

Simulate a support conversation where each "turn" may trigger partner calls. The orchestrator decides which partners to invoke based on keywords and order context.

````python
def offline_cross_org_support(user_message: str) -> str:
    oid_match = re.search(r"ORD-\d+", user_message.upper())
    if not oid_match:
        return "Please provide an order ID (e.g. ORD-1001)."
    order_id = oid_match.group()

    if any(w in user_message.lower() for w in ("late", "delay", "refund")):
        res = orchestrator.resolve_late_delivery_refund(order_id)
        lines = [
            f"Ticket {res.correlation_id}",
            f"Decision: {res.decision}",
        ]
        if res.shipment:
            lines.append(f"Delivery: {res.shipment['status']} ({res.shipment.get('days_late', 0)} days late)")
        if res.warranty and not res.warranty.get("covers_shipping_delay"):
            lines.append(f"Warranty note: {res.warranty.get('notes', 'N/A')}")
        if res.partner_errors:
            lines.append(f"Partner issues: {', '.join(res.partner_errors)}")
        return "\n".join(lines)

    order = shopco_mcp.call("get_order", {"order_id": order_id})
    if order.is_error:
        return str(order.data)
    return f"Order {order_id}: status={order.data['status']}, total=${order.data['total_usd']}"


print(offline_cross_org_support("What's the status of ORD-1001?"))
print("\n" + "=" * 50 + "\n")
print(offline_cross_org_support("ORD-1001 arrived a week late, I want a full refund!"))
````

## 16. Production Checklist for Cross-Org MAS

### Trust and identity
- Issue scoped credentials per partner, per skill set, with expiry
- Rotate credentials on schedule; never embed long-lived secrets in agent prompts
- Validate caller org matches credential holder on every request

### Data governance
- Pass opaque reference IDs across boundaries, not PII
- Document allowed fields in partner contracts
- Log what was sent and received (hashed if sensitive)

### Reliability
- Timeouts and circuit breakers per partner
- Partial resolution paths when partners are unavailable
- Human escalation for denied or incomplete cross-org flows

### Observability
- Propagate `correlation_id` across all hops
- Per-partner success rate, latency, and denial dashboards
- Alert on credential expiry and quota exhaustion

## 17. Anti-Patterns

| Anti-pattern | Risk | Fix |
|--------------|------|-----|
| Shared database across orgs | Legal exposure, tight coupling | MCP/A2A with opaque refs |
| One API key for all partners | Blast radius on leak | Per-partner scoped credentials |
| Forwarding full customer records | GDPR/CCPA violations | Data minimization per request |
| No correlation IDs | Un-auditable cross-org flows | Propagate ID from ticket creation |
| Direct partner DB access from host | Bypasses partner policy | A2A task to partner's agent |
| Ignoring partner SLA breaches | Silent quality degradation | Contract validation + alerts |
| Failing entire ticket on one partner error | Poor customer experience | Partial resolution + escalation |

## 20. Summary

Cross-organizational multi-agent systems extend MAS across legal and technical trust boundaries:

- **Agent Cards** enable discovery of partner agent skills without shared infrastructure
- **MCP servers** wrap each org's internal data; external traffic never hits raw databases
- **A2A messages** carry tasks with correlation IDs, credentials, and structured payloads
- **Scoped credentials** and a **trust policy engine** enforce least-privilege access
- A **hub broker** routes, audits, and isolates failures across ShopCo, ShipFast, PaySecure, and GearGuard

The ShopCo orchestrator resolved a late-delivery refund by coordinating three partner agents — authorizing payment only after ShipFast attested the delay and PaySecure confirmed eligibility. Build for partial failure, minimize data at every boundary, and make every cross-org hop auditable.
