# Model Context Protocol (MCP) — Architecture

**Model Context Protocol (MCP)** is an open standard for connecting AI applications to external **tools**, **resources**, and **prompt templates** through a client–server model. MCP separates:

- **Host** — the AI application (IDE, chat app, agent runtime)
- **Client** — one connection from host to a server
- **Server** — exposes capabilities (order API, file system, database)

This notebook explains MCP architecture in depth and implements a **ShopCo MCP Stack** — a faithful in-memory simulation of initialization, discovery, tools, resources, prompts, and multi-server hosts. No external processes or CLI required.

````python
"""
ShopCo MCP Stack — in-memory MCP architecture simulation.
"""

import json
import re
import uuid
from dataclasses import dataclass, field
from datetime import datetime, timezone
from enum import Enum
from typing import Any, Callable


def pretty(obj: Any) -> str:
    return json.dumps(obj, indent=2, default=str)


ORDERS: dict[str, dict[str, Any]] = {
    "ORD-1001": {"customer": "alice@shop.com", "status": "shipped", "amount": 1299.00},
    "ORD-1002": {"customer": "bob@shop.com", "status": "processing", "amount": 449.00},
}

POLICIES = [
    {"id": "ret-02", "text": "Returns within 30 days if unused."},
    {"id": "cancel-03", "text": "Cancellations allowed only before shipment."},
]

MCP_TRACE: list[dict[str, Any]] = []

print("ShopCo MCP Stack ready")
````

## 1. MCP Actors — Host, Client, Server

```
┌─────────────────────────────────────────────────────────────┐
│  MCP Host (ShopCo Support App)                              │
│  ├── MCP Client ──► Orders Server    (lookup_order)         │
│  ├── MCP Client ──► Policies Server  (search_policy)        │
│  └── MCP Client ──► Prompts Server   (support templates)    │
│                                                             │
│  Agent loop uses aggregated tools/resources from all clients │
└─────────────────────────────────────────────────────────────┘
```

| Actor | Role | Count per host |
|-------|------|----------------|
| **Host** | Owns the agent loop; embeds clients | 1 |
| **Client** | One session to one server | 1 per server |
| **Server** | Exposes tools, resources, prompts | Many (external) |

The host **never** talks to servers directly — always through a client.

---

## 2. MCP Primitives — Tools, Resources, Prompts

| Primitive | Purpose | MCP methods | ShopCo example |
|-----------|---------|-------------|----------------|
| **Tools** | Model-invoked actions with side effects | `tools/list`, `tools/call` | `lookup_order` |
| **Resources** | Readable context (files, docs, records) | `resources/list`, `resources/read` | `shopco://policies/returns` |
| **Prompts** | Reusable prompt templates with arguments | `prompts/list`, `prompts/get` | `support_triage` template |

**Tools** are for actions. **Resources** are for context the model reads. **Prompts** are pre-built instruction templates the host can inject.

---

## 3. Transport Layer — How Bytes Move

MCP supports multiple transports; the **application protocol** stays the same.

| Transport | Use case | How it works |
|-----------|----------|--------------|
| **stdio** | Local subprocess servers | Host spawns server; JSON-RPC over stdin/stdout |
| **SSE** | Remote HTTP servers | Server-sent events stream + HTTP POST |
| **Streamable HTTP** | Modern remote deployments | Single HTTP endpoint for bidirectional messages |

This notebook uses **in-process transport** — `client.send()` calls `server.handle()` directly. The message shape is identical to production MCP.

## 4. JSON-RPC Message Layer

All MCP messages use JSON-RPC 2.0: requests, responses, errors, and notifications.

````python
class JsonRpcKind(str, Enum):
    REQUEST = "request"
    RESPONSE = "response"
    ERROR = "error"
    NOTIFICATION = "notification"


@dataclass
class JsonRpcMessage:
    id: str | None
    method: str
    kind: JsonRpcKind
    params: dict[str, Any] = field(default_factory=dict)
    result: Any = None
    error: dict[str, Any] | None = None

    def to_wire(self) -> dict[str, Any]:
        wire: dict[str, Any] = {"jsonrpc": "2.0", "method": self.method}
        if self.id is not None:
            wire["id"] = self.id
        if self.kind == JsonRpcKind.REQUEST:
            wire["params"] = self.params
        elif self.kind == JsonRpcKind.RESPONSE:
            wire["result"] = self.result
        elif self.kind == JsonRpcKind.ERROR:
            wire["error"] = self.error or {"code": -1, "message": "unknown"}
        elif self.kind == JsonRpcKind.NOTIFICATION:
            wire["params"] = self.params
        return wire


example_req = JsonRpcMessage(
    id="1", method="tools/call", kind=JsonRpcKind.REQUEST,
    params={"name": "lookup_order", "arguments": {"order_id": "ORD-1001"}},
)
print(pretty(example_req.to_wire()))
````

## 5. Server Capabilities — What a Server Advertises

During `initialize`, the server declares which primitives it supports.

````python
@dataclass
class ServerCapabilities:
    tools: bool = False
    resources: bool = False
    prompts: bool = False
    logging: bool = False

    def to_dict(self) -> dict[str, dict]:
        caps: dict[str, dict] = {}
        if self.tools:
            caps["tools"] = {"listChanged": True}
        if self.resources:
            caps["resources"] = {"subscribe": False, "listChanged": True}
        if self.prompts:
            caps["prompts"] = {"listChanged": True}
        if self.logging:
            caps["logging"] = {}
        return caps


@dataclass
class ServerInfo:
    name: str
    version: str


ORDERS_CAPS = ServerCapabilities(tools=True)
POLICIES_CAPS = ServerCapabilities(tools=True, resources=True)
PROMPTS_CAPS = ServerCapabilities(prompts=True)

print("Orders server caps:", ORDERS_CAPS.to_dict())
print("Policies server caps:", POLICIES_CAPS.to_dict())
````

## 6. MCP Server Implementation — ShopCo Orders

A server registers handlers for each MCP method it supports.

````python
@dataclass
class McpTool:
    name: str
    description: str
    input_schema: dict[str, Any]
    handler: Callable[..., Any] = field(repr=False)


@dataclass
class McpResource:
    uri: str
    name: str
    description: str
    mime_type: str = "application/json"
    reader: Callable[[], str] = field(repr=False, default=lambda: "")


@dataclass
class McpPrompt:
    name: str
    description: str
    arguments: list[dict[str, str]]
    renderer: Callable[[dict[str, str]], list[dict[str, str]]] = field(repr=False)


def _lookup_order(order_id: str) -> dict[str, Any]:
    oid = order_id.upper()
    row = ORDERS.get(oid)
    if not row:
        return {"found": False, "order_id": oid}
    return {"found": True, "order_id": oid, **row}


def _search_policy(query: str) -> list[dict[str, str]]:
    terms = [t for t in re.split(r"\W+", query.lower()) if len(t) > 2]
    return [p for p in POLICIES if any(t in p["text"].lower() for t in terms)] or POLICIES[:1]


class MCPServer:
    """In-memory MCP server with tools, resources, and prompts."""

    PROTOCOL_VERSION = "2024-11-05"

    def __init__(self, info: ServerInfo, capabilities: ServerCapabilities):
        self.info = info
        self.capabilities = capabilities
        self.tools: dict[str, McpTool] = {}
        self.resources: dict[str, McpResource] = {}
        self.prompts: dict[str, McpPrompt] = {}
        self._initialized = False

    def register_tool(self, tool: McpTool) -> None:
        self.tools[tool.name] = tool

    def register_resource(self, resource: McpResource) -> None:
        self.resources[resource.uri] = resource

    def register_prompt(self, prompt: McpPrompt) -> None:
        self.prompts[prompt.name] = prompt

    def handle(self, msg: JsonRpcMessage) -> JsonRpcMessage | None:
        MCP_TRACE.append({"server": self.info.name, "method": msg.method, "kind": msg.kind.value})

        if msg.method == "initialize":
            self._initialized = True
            return self._ok(msg.id, "initialize", {
                "protocolVersion": self.PROTOCOL_VERSION,
                "capabilities": self.capabilities.to_dict(),
                "serverInfo": {"name": self.info.name, "version": self.info.version},
            })

        if msg.method == "notifications/initialized":
            return None  # notification — no response

        if not self._initialized and msg.method != "initialize":
            return self._err(msg.id, msg.method, -32002, "Server not initialized")

        if msg.method == "tools/list":
            tools = [
                {"name": t.name, "description": t.description, "inputSchema": t.input_schema}
                for t in self.tools.values()
            ]
            return self._ok(msg.id, msg.method, {"tools": tools})

        if msg.method == "tools/call":
            name = msg.params.get("name", "")
            args = msg.params.get("arguments", {})
            tool = self.tools.get(name)
            if tool is None:
                return self._err(msg.id, msg.method, -32602, f"Unknown tool: {name}")
            try:
                result = tool.handler(**args)
                content = [{"type": "text", "text": json.dumps(result, default=str)}]
                return self._ok(msg.id, msg.method, {"content": content, "isError": False})
            except Exception as exc:
                content = [{"type": "text", "text": str(exc)}]
                return self._ok(msg.id, msg.method, {"content": content, "isError": True})

        if msg.method == "resources/list":
            resources = [
                {"uri": r.uri, "name": r.name, "description": r.description, "mimeType": r.mime_type}
                for r in self.resources.values()
            ]
            return self._ok(msg.id, msg.method, {"resources": resources})

        if msg.method == "resources/read":
            uri = msg.params.get("uri", "")
            res = self.resources.get(uri)
            if res is None:
                return self._err(msg.id, msg.method, -32602, f"Unknown resource: {uri}")
            return self._ok(msg.id, msg.method, {
                "contents": [{"uri": uri, "mimeType": res.mime_type, "text": res.reader()}]
            })

        if msg.method == "prompts/list":
            prompts = [
                {"name": p.name, "description": p.description, "arguments": p.arguments}
                for p in self.prompts.values()
            ]
            return self._ok(msg.id, msg.method, {"prompts": prompts})

        if msg.method == "prompts/get":
            name = msg.params.get("name", "")
            args = {a["name"]: a.get("value", "") for a in msg.params.get("arguments", [])}
            prompt = self.prompts.get(name)
            if prompt is None:
                return self._err(msg.id, msg.method, -32602, f"Unknown prompt: {name}")
            messages = prompt.renderer(args)
            return self._ok(msg.id, msg.method, {"description": prompt.description, "messages": messages})

        return self._err(msg.id, msg.method, -32601, f"Method not found: {msg.method}")

    def _ok(self, req_id: str | None, method: str, result: Any) -> JsonRpcMessage:
        return JsonRpcMessage(id=req_id, method=method, kind=JsonRpcKind.RESPONSE, result=result)

    def _err(self, req_id: str | None, method: str, code: int, message: str) -> JsonRpcMessage:
        return JsonRpcMessage(
            id=req_id, method=method, kind=JsonRpcKind.ERROR,
            error={"code": code, "message": message},
        )
````

## 7. Build ShopCo MCP Servers

Three specialized servers — orders (tools), policies (tools + resources), prompts (templates).

````python
def build_orders_server() -> MCPServer:
    srv = MCPServer(ServerInfo("shopco-orders", "1.0.0"), ServerCapabilities(tools=True))
    srv.register_tool(McpTool(
        name="lookup_order",
        description="Look up order by ORD-#### id",
        input_schema={
            "type": "object",
            "properties": {"order_id": {"type": "string"}},
            "required": ["order_id"],
        },
        handler=_lookup_order,
    ))
    return srv


def build_policies_server() -> MCPServer:
    srv = MCPServer(ServerInfo("shopco-policies", "1.0.0"), ServerCapabilities(tools=True, resources=True))
    srv.register_tool(McpTool(
        name="search_policy",
        description="Search return and cancellation policies",
        input_schema={
            "type": "object",
            "properties": {"query": {"type": "string"}},
            "required": ["query"],
        },
        handler=_search_policy,
    ))
    srv.register_resource(McpResource(
        uri="shopco://policies/all",
        name="all_policies",
        description="Full policy document list",
        reader=lambda: pretty(POLICIES),
    ))
    return srv


def build_prompts_server() -> MCPServer:
    srv = MCPServer(ServerInfo("shopco-prompts", "1.0.0"), ServerCapabilities(prompts=True))
    srv.register_prompt(McpPrompt(
        name="support_triage",
        description="Triage template for support agents",
        arguments=[
            {"name": "customer_name", "description": "Customer display name", "required": True},
            {"name": "issue_type", "description": "order|billing|policy", "required": True},
        ],
        renderer=lambda args: [
            {"role": "system", "content": {
                "type": "text",
                "text": (
                    f"You are ShopCo support for {args.get('customer_name', 'customer')}. "
                    f"Issue type: {args.get('issue_type', 'general')}. "
                    "Use MCP tools. Cite policy ids. Be concise."
                ),
            }},
        ],
    ))
    return srv


orders_server = build_orders_server()
policies_server = build_policies_server()
prompts_server = build_prompts_server()
print("Servers built: orders, policies, prompts")
````

## 8. MCP Client — Connection Lifecycle

Every client connection follows: **initialize → initialized notification → ready**.

````python
class MCPClient:
    """One client session to one MCP server."""

    def __init__(self, server: MCPServer, client_name: str = "shopco-host"):
        self._server = server
        self._client_name = client_name
        self._ready = False
        self._server_caps: dict[str, Any] = {}

    def connect(self) -> dict[str, Any]:
        init_resp = self._request("initialize", {
            "protocolVersion": MCPServer.PROTOCOL_VERSION,
            "capabilities": {"roots": {"listChanged": True}},
            "clientInfo": {"name": self._client_name, "version": "1.0.0"},
        })
        self._server_caps = init_resp.result
        self._notify("notifications/initialized", {})
        self._ready = True
        return self._server_caps

    def _request(self, method: str, params: dict) -> JsonRpcMessage:
        msg = JsonRpcMessage(id=str(uuid.uuid4())[:8], method=method, kind=JsonRpcKind.REQUEST, params=params)
        resp = self._server.handle(msg)
        if resp is None:
            raise RuntimeError(f"No response for {method}")
        if resp.kind == JsonRpcKind.ERROR:
            raise RuntimeError(resp.error.get("message", "unknown error"))
        return resp

    def _notify(self, method: str, params: dict) -> None:
        msg = JsonRpcMessage(id=None, method=method, kind=JsonRpcKind.NOTIFICATION, params=params)
        self._server.handle(msg)

    def list_tools(self) -> list[dict]:
        return self._request("tools/list", {}).result.get("tools", [])

    def call_tool(self, name: str, arguments: dict) -> Any:
        resp = self._request("tools/call", {"name": name, "arguments": arguments})
        content = resp.result.get("content", [{}])[0].get("text", "{}")
        return json.loads(content)

    def list_resources(self) -> list[dict]:
        return self._request("resources/list", {}).result.get("resources", [])

    def read_resource(self, uri: str) -> str:
        return self._request("resources/read", {"uri": uri}).result["contents"][0]["text"]

    def list_prompts(self) -> list[dict]:
        return self._request("prompts/list", {}).result.get("prompts", [])

    def get_prompt(self, name: str, arguments: dict[str, str]) -> list[dict]:
        args = [{"name": k, "value": v} for k, v in arguments.items()]
        return self._request("prompts/get", {"name": name, "arguments": args}).result["messages"]


MCP_TRACE.clear()
orders_client = MCPClient(orders_server)
init_result = orders_client.connect()
print("Initialize result:")
print(pretty(init_result))
print("\nDiscovered tools:", [t["name"] for t in orders_client.list_tools()])
````

## 9. MCP Host — Aggregating Multiple Servers

The host creates one client per server and **merges** capabilities for the agent loop.

````python
class MCPHost:
    """ShopCo support app — embeds multiple MCP clients."""

    def __init__(self, name: str = "shopco-support-host"):
        self.name = name
        self.clients: dict[str, MCPClient] = {}

    def connect_server(self, alias: str, server: MCPServer) -> dict[str, Any]:
        client = MCPClient(server, client_name=self.name)
        caps = client.connect()
        self.clients[alias] = client
        return caps

    def all_tools(self) -> list[dict[str, Any]]:
        """Merge tools from all connected servers (prefix with server alias)."""
        merged = []
        for alias, client in self.clients.items():
            for tool in client.list_tools():
                merged.append({
                    "server": alias,
                    "name": tool["name"],
                    "qualified_name": f"{alias}.{tool['name']}",
                    "description": tool["description"],
                    "inputSchema": tool["inputSchema"],
                })
        return merged

    def call_tool(self, server_alias: str, tool_name: str, arguments: dict) -> Any:
        client = self.clients[server_alias]
        return client.call_tool(tool_name, arguments)

    def tools_for_openai(self) -> list[dict]:
        return [
            {
                "type": "function",
                "function": {
                    "name": t["qualified_name"].replace(".", "__"),
                    "description": f"[{t['server']}] {t['description']}",
                    "parameters": t["inputSchema"],
                },
            }
            for t in self.all_tools()
        ]


host = MCPHost()
host.connect_server("orders", orders_server)
host.connect_server("policies", policies_server)
host.connect_server("prompts", prompts_server)

print(f"Host connected to {len(host.clients)} servers")
print("\nAggregated tools:")
for t in host.all_tools():
    print(f"  {t['qualified_name']}: {t['description'][:50]}")
````

## 11. Agent Loop Through MCP Host

The agent loop stays in the host; tool execution routes through MCP clients.

````python
class ShopCoMcpAgent:
    def __init__(self, host: MCPHost):
        self._host = host

    def run(self, user_query: str) -> dict[str, Any]:
        trace: list[str] = []
        parts: list[str] = []

        if "ord-" in user_query.lower():
            match = re.search(r"ORD-\d{4}", user_query.upper())
            oid = match.group(0) if match else "ORD-1001"
            data = self._host.call_tool("orders", "lookup_order", {"order_id": oid})
            trace.append(f"orders.lookup_order({oid})")
            if data.get("found"):
                parts.append(f"Order {data['order_id']} is {data['status']} (${data['amount']:.2f})")

        if any(w in user_query.lower() for w in ("return", "policy", "cancel")):
            hits = self._host.call_tool("policies", "search_policy", {"query": user_query[:40]})
            trace.append("policies.search_policy")
            parts.append(f"[{hits[0]['id']}] {hits[0]['text']}")

        return {
            "answer": " | ".join(parts) if parts else "How can I help?",
            "trace": trace,
            "tools_registered": len(self._host.all_tools()),
        }


agent = ShopCoMcpAgent(host)
result = agent.run("Where is ORD-1001 and can I return it?")
print(result["answer"])
print("MCP trace:", result["trace"])
print("Tools available to agent:", result["tools_registered"])
````

## 13. Security Architecture

| Concern | MCP pattern |
|---------|-------------|
| **Least privilege** | Host connects only to needed servers |
| **Per-server scope** | Each client session is isolated |
| **Untrusted observations** | Tool results are model input — validate |
| **Resource URI allowlist** | Host filters `shopco://` schemes |
| **User consent** | Host prompts before connecting new servers |

````python
ALLOWED_URI_SCHEMES = {"shopco://"}


def safe_read_resource(client: MCPClient, uri: str) -> str:
    if not any(uri.startswith(s) for s in ALLOWED_URI_SCHEMES):
        raise PermissionError(f"URI scheme not allowed: {uri}")
    return client.read_resource(uri)


print(safe_read_resource(host.clients["policies"], "shopco://policies/all")[:80], "...")
try:
    safe_read_resource(host.clients["policies"], "file:///etc/passwd")
except PermissionError as e:
    print(f"Blocked: {e}")
````

## 18. Summary

MCP architecture separates concerns cleanly:

- **Host** — your agent app; owns the loop and embeds clients.
- **Client** — one JSON-RPC session per server; handles initialize and calls.
- **Server** — exposes **tools**, **resources**, and **prompts**.
- **Transport** — stdio, SSE, or HTTP; same message format everywhere.
- **Lifecycle** — `initialize` → `notifications/initialized` → discover → invoke.

The ShopCo MCP Stack connected three servers through one host, aggregated tools for the agent, and demonstrated all three primitives. Production deployments use the same architecture — only the transport changes from in-process calls to stdio or HTTP.
