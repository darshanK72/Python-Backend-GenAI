# MCP vs A2A — Complementary Roles

Modern multi-agent systems rarely choose **one** protocol. They use **both**:

- **MCP (Model Context Protocol)** — connect your agent to **tools and resources** you control (order DB, policy docs, calculators).
- **A2A (Agent-to-Agent)** — delegate work to **other agents** that have their own reasoning, policies, and lifecycle (logistics partner, payment processor, legal review bot).

They are **complementary**, not competing standards:

```
                    ShopCo Coordinator
                           │
         ┌─────────────────┼─────────────────┐
         ▼                 ▼                 ▼
    MCP: orders      MCP: policies     A2A: ShipFast
    (tool call)      (resource read)   (agent task)
         │                 │                 │
    Your CRM API      Your doc store    Partner agent
```

This notebook is a **unified protocol lab** for ShopCo Support Hub. You will:

1. Run MCP tools/resources and A2A agents side by side
2. Apply a **decision framework** — MCP vs A2A per request type
3. Build a **hybrid coordinator** with unified tracing
4. Spot **anti-patterns** (wrong protocol for the job)
5. Score architecture readiness for production

Everything is self-contained plain Python — no external servers or CLI.

````python
"""
ShopCo Unified Protocol Hub — MCP + A2A complementary roles lab.
"""

import json
import os
import re
import uuid
from dataclasses import dataclass, field
from datetime import datetime, timezone
from enum import Enum
from typing import Any, Callable, Literal

OPENAI_API_KEY = os.environ.get("OPENAI_API_KEY", "sk-proj-placeholder-replace-with-your-real-key")
os.environ.setdefault("OPENAI_API_KEY", OPENAI_API_KEY)

USE_LIVE_LLM = False
ProtocolChoice = Literal["mcp_tool", "mcp_resource", "a2a_agent", "local_only"]


def pretty(obj: Any) -> str:
    return json.dumps(obj, indent=2, default=str)


def utc_now() -> str:
    return datetime.now(timezone.utc).isoformat()


ORDERS: dict[str, dict[str, Any]] = {
    "ORD-1001": {
        "order_id": "ORD-1001", "customer": "alice@shopco.com",
        "status": "shipped", "total_usd": 129.00,
        "tracking": "1Z999AA10123456784", "carrier": "UPS",
    },
    "ORD-1002": {
        "order_id": "ORD-1002", "customer": "bob@shopco.com",
        "status": "processing", "total_usd": 49.00,
        "tracking": None, "carrier": None,
    },
}

POLICIES: dict[str, str] = {
    "returns": "# Returns\n30-day window. Unused items only. Refunds in 5-7 business days.",
    "shipping": "# Shipping\nFree over $50. Tracking at carrier scan.",
    "billing": "# Billing\nDisputes within 60 days. Case ID within 24h.",
}

UNIFIED_TRACE: list[dict[str, Any]] = []

print(f"ShopCo unified hub — {len(ORDERS)} orders, {len(POLICIES)} policies")
````

---

## 1. The Core Distinction

| Question | If YES → | If NO → |
|----------|----------|--------|
| Is the remote party a **function/API** with a fixed schema? | **MCP tool** | Consider A2A |
| Is the data **read-only context** (policy, doc, config)? | **MCP resource** | Tool or A2A |
| Does the remote party **reason autonomously** with its own policies? | **A2A agent** | MCP tool |
| Is the remote party **outside your org** with separate SLA? | **A2A** (usually) | MCP if you host it |
| Do you need **task lifecycle** (working, input-required, streaming)? | **A2A** | MCP for quick calls |

**MCP = capability invocation. A2A = delegation.**

---

## 2. Side-by-Side Protocol Comparison

| Dimension | MCP | A2A |
|-----------|-----|-----|
| **Discovery** | `tools/list`, `resources/list` | Agent Card at `/.well-known/agent.json` |
| **Invoke** | `tools/call`, `resources/read` | `message/send`, `tasks/get` |
| **Unit of work** | Tool call / resource URI | Task with state machine |
| **Who owns logic** | Your server handler | Remote agent |
| **Streaming** | Tool result (single response) | Task status / artifact SSE streams |
| **Auth model** | Server-scoped API keys | Per-agent OAuth, mTLS, JWT |
| **Typical latency** | Low (ms–s) | Medium–high (agent loop) |
| **Best example** | `get_order(order_id)` | ShipFast tracking specialist agent |

## 3. MCP Layer — Tools and Resources

ShopCo-owned capabilities: order lookup (tool) and policy documents (resources).

````python
MCP_TOOLS: list[dict[str, Any]] = [
    {
        "name": "get_order",
        "description": "Fetch order from ShopCo CRM. Use for owned order data only.",
        "inputSchema": {
            "type": "object",
            "properties": {"order_id": {"type": "string"}},
            "required": ["order_id"],
        },
    },
]

MCP_RESOURCES: list[dict[str, Any]] = [
    {"uri": f"shopco://policies/{k}", "name": f"{k} policy", "mimeType": "text/markdown"}
    for k in POLICIES
]


class ShopCoMCPServer:
    def list_tools(self) -> list[dict[str, Any]]:
        return MCP_TOOLS

    def list_resources(self) -> list[dict[str, Any]]:
        return MCP_RESOURCES

    def call_tool(self, name: str, args: dict[str, Any]) -> dict[str, Any]:
        if name != "get_order":
            return {"isError": True, "content": [{"type": "text", "text": f"unknown tool {name}"}]}
        order = ORDERS.get(args.get("order_id", ""))
        if not order:
            return {"isError": True, "content": [{"type": "text", "text": "order not found"}]}
        return {"content": [{"type": "text", "text": pretty(order)}]}

    def read_resource(self, uri: str) -> str:
        key = uri.split("/")[-1]
        if key not in POLICIES:
            raise ValueError(f"unknown resource {uri}")
        return POLICIES[key]


class MCPClient:
    def __init__(self, server: ShopCoMCPServer) -> None:
        self.server = server
        self.tools = server.list_tools()
        self.resources = server.list_resources()

    def call_tool(self, name: str, args: dict[str, Any]) -> str:
        result = self.server.call_tool(name, args)
        text = result["content"][0]["text"]
        return f"ERROR: {text}" if result.get("isError") else text

    def read_resource(self, uri: str) -> str:
        return self.server.read_resource(uri)


mcp = MCPClient(ShopCoMCPServer())
print("MCP:", [t["name"] for t in mcp.tools], "| resources:", len(mcp.resources))
````

## 4. A2A Layer — Partner Agents

External specialists with autonomous handlers and task lifecycle.

````python
class TaskState(str, Enum):
    WORKING = "TASK_STATE_WORKING"
    COMPLETED = "TASK_STATE_COMPLETED"
    INPUT_REQUIRED = "TASK_STATE_INPUT_REQUIRED"


@dataclass
class AgentCard:
    agent_id: str
    name: str
    organization: str
    skills: list[str]
    url: str
    streaming: bool = False


@dataclass
class A2ATask:
    id: str
    state: TaskState
    artifacts: list[dict[str, Any]] = field(default_factory=list)

    def to_dict(self) -> dict[str, Any]:
        return {"id": self.id, "status": {"state": self.state.value}, "artifacts": self.artifacts}


A2A_AGENTS: dict[str, AgentCard] = {
    "shipfast-tracker": AgentCard(
        "shipfast-tracker", "ShipFast Tracker", "ShipFast Logistics",
        ["tracking", "logistics"], "https://a2a.shipfast.com/v1", streaming=True,
    ),
    "payright-disputes": AgentCard(
        "payright-disputes", "PayRight Disputes", "PayRight Inc",
        ["billing", "disputes"], "https://a2a.payright.io/v1", streaming=False,
    ),
}


def handle_shipfast(message: str) -> A2ATask:
    task = A2ATask(id=f"task-{uuid.uuid4().hex[:8]}", state=TaskState.WORKING)
    m = re.search(r"1Z[0-9A-Z]+", message.upper())
    if not m:
        oid = re.search(r"ORD-[0-9]{4}", message.upper())
        tracking = ORDERS.get(oid.group(0), {}).get("tracking") if oid else None
        if not tracking:
            task.state = TaskState.INPUT_REQUIRED
            task.artifacts = [{"type": "text", "text": "Need tracking or order ID"}]
            return task
        tracking_id = tracking
    else:
        tracking_id = m.group(0)
    task.state = TaskState.COMPLETED
    task.artifacts = [{"type": "data", "data": {"tracking": tracking_id, "eta": "2026-07-12", "scan": "In transit"}}]
    return task


def handle_payright(message: str, email: str = "") -> A2ATask:
    task = A2ATask(id=f"task-{uuid.uuid4().hex[:8]}", state=TaskState.WORKING)
    task.state = TaskState.COMPLETED
    task.artifacts = [{"type": "data", "data": {
        "case_id": f"DISP-{uuid.uuid4().hex[:6].upper()}",
        "customer": email or "unknown", "status": "opened",
    }}]
    return task


A2A_HANDLERS: dict[str, Callable[..., A2ATask]] = {
    "shipfast-tracker": handle_shipfast,
    "payright-disputes": handle_payright,
}


class A2AClient:
    def discover(self, agent_id: str) -> AgentCard:
        return A2A_AGENTS[agent_id]

    def send_message(self, agent_id: str, text: str, **ctx: Any) -> A2ATask:
        handler = A2A_HANDLERS[agent_id]
        if agent_id == "payright-disputes":
            return handler(text, ctx.get("customer_email", ""))
        return handler(text)


a2a = A2AClient()
print("A2A agents:", list(A2A_AGENTS.keys()))
````

## 5. Decision Framework — `choose_protocol()`

Encode routing rules explicitly so teams can test and audit them.

````python
@dataclass
class ProtocolDecision:
    protocol: ProtocolChoice
    target: str
    reason: str


def choose_protocol(user_text: str) -> list[ProtocolDecision]:
    """Return one or more protocol steps for a user request."""
    q = user_text.lower()
    decisions: list[ProtocolDecision] = []

    if any(k in q for k in ("return", "policy", "shipping policy")):
        decisions.append(ProtocolDecision(
            "mcp_resource", "shopco://policies/returns",
            "Static policy text — read-only, owned by ShopCo",
        ))

    if any(k in q for k in ("charged twice", "dispute", "billing", "invoice")):
        decisions.append(ProtocolDecision(
            "a2a_agent", "payright-disputes",
            "External payment processor owns dispute workflow",
        ))

    if any(k in q for k in ("tracking", "carrier", "shipment", "in transit", "delivered")):
        decisions.append(ProtocolDecision(
            "a2a_agent", "shipfast-tracker",
            "Logistics partner agent has live carrier integrations",
        ))

    if re.search(r"ORD-[0-9]{4}", user_text.upper()) or "order status" in q:
        decisions.append(ProtocolDecision(
            "mcp_tool", "get_order",
            "Owned CRM data — deterministic tool with schema",
        ))

    if not decisions:
        decisions.append(ProtocolDecision(
            "local_only", "coordinator",
            "No protocol match — clarify with user",
        ))
    return decisions


TEST_QUERIES = [
    "Where is order ORD-1001?",
    "Live tracking — carrier says in transit for 1Z999AA10123456784",
    "I was charged twice, open a dispute",
    "What is your return policy?",
    "ORD-1001 status and return policy",
]

print(f"{'Query':<45} Decisions")
print("-" * 90)
for q in TEST_QUERIES:
    decs = choose_protocol(q)
    labels = [f"{d.protocol}→{d.target}" for d in decs]
    print(f"{q[:44]:<45} {labels}")
````

## 6. Unified Trace — One Log for Both Protocols

Production observability should not split MCP and A2A into separate silos.

````python
def trace_event(protocol: str, operation: str, target: str, status: str, **extra: Any) -> None:
    UNIFIED_TRACE.append({
        "ts": utc_now(),
        "protocol": protocol,
        "operation": operation,
        "target": target,
        "status": status,
        **extra,
    })


def execute_mcp(decision: ProtocolDecision, user_text: str) -> str:
    if decision.protocol == "mcp_tool":
        m = re.search(r"ORD-[0-9]{4}", user_text.upper())
        oid = m.group(0) if m else "ORD-1001"
        result = mcp.call_tool("get_order", {"order_id": oid})
        trace_event("MCP", "tools/call", "get_order", "ok", order_id=oid)
        return result
    if decision.protocol == "mcp_resource":
        text = mcp.read_resource(decision.target)
        trace_event("MCP", "resources/read", decision.target, "ok")
        return text
    return ""


def execute_a2a(decision: ProtocolDecision, user_text: str, **ctx: Any) -> dict[str, Any]:
    agent_id = decision.target
    card = a2a.discover(agent_id)
    task = a2a.send_message(agent_id, user_text, **ctx)
    trace_event("A2A", "message/send", agent_id, task.state.value, org=card.organization)
    return task.to_dict()


print("Trace helpers ready")
````

## 7. Hybrid Coordinator

The coordinator plans protocol steps, executes them in order, and merges results.

````python
@dataclass
class CoordinatorResult:
    user_query: str
    decisions: list[ProtocolDecision]
    outputs: list[dict[str, Any]]
    answer: str


class HybridCoordinator:
    def __init__(self, mcp_client: MCPClient, a2a_client: A2AClient) -> None:
        self.mcp = mcp_client
        self.a2a = a2a_client

    def handle(self, user_text: str, **ctx: Any) -> CoordinatorResult:
        decisions = choose_protocol(user_text)
        outputs: list[dict[str, Any]] = []
        parts: list[str] = []

        for dec in decisions:
            if dec.protocol in ("mcp_tool", "mcp_resource"):
                out = execute_mcp(dec, user_text)
                outputs.append({"protocol": "MCP", "target": dec.target, "result": out})
                parts.append(f"[MCP {dec.target}] {out[:120]}")
            elif dec.protocol == "a2a_agent":
                out = execute_a2a(dec, user_text, **ctx)
                outputs.append({"protocol": "A2A", "target": dec.target, "result": out})
                parts.append(f"[A2A {dec.target}] {out['artifacts']}")
            else:
                parts.append("How can I help with your ShopCo order, tracking, or billing question?")

        trace_event("COORDINATOR", "merge", "hybrid", "ok", steps=len(decisions))
        return CoordinatorResult(
            user_query=user_text,
            decisions=decisions,
            outputs=outputs,
            answer="\n".join(parts),
        )


coordinator = HybridCoordinator(mcp, a2a)

demo = coordinator.handle(
    "ORD-1001 status and carrier tracking update",
    customer_email="alice@shopco.com",
)
print(demo.answer)
print("\nDecisions:", [(d.protocol, d.reason[:40]) for d in demo.decisions])
````

| User intent | Protocol | Why |
|-------------|----------|-----|
| Order status from ShopCo DB | MCP `get_order` | Owned, schema-stable API |
| Return policy text | MCP `resources/read` | Static doc, no agent reasoning |
| Carrier scan events + ETA | A2A ShipFast | Partner owns logistics agent |
| Payment dispute case | A2A PayRight | External org + task lifecycle |
| "Summarize everything" | MCP + A2A hybrid | Coordinator merges both |

## 9. Anti-Patterns — Wrong Protocol Choice

| Anti-pattern | Problem | Correct approach |
|--------------|---------|------------------|
| Wrap partner's **agent** as single MCP tool | Lose task state, streaming, org boundary | A2A delegation |
| Expose **CRUD API** as A2A agent | Unnecessary task overhead | MCP tool |
| Put **policy PDF** behind tool call | Wastes tool turns | MCP resource prefetch |
| Duplicate same data in MCP + A2A | Inconsistent answers | Single source of truth per domain |
| No unified trace | Can't debug cross-protocol flows | One trace with `protocol` tag |
| Coordinator embeds CRM logic | Bypasses MCP layer | Call MCP tool |

````python
def anti_pattern_demo() -> dict[str, str]:
    """Simulate wrong vs right choice for policy question."""
    wrong = mcp.call_tool("get_order", {"order_id": "ORD-1001"})  # wrong tool for policy Q
    right = mcp.read_resource("shopco://policies/returns")
    return {
        "wrong_tool_for_policy": wrong[:60] + "...",
        "right_resource_read": right[:60] + "...",
    }


print(pretty(anti_pattern_demo()))
````

## 11. Latency and Cost Profile

| Protocol | Typical cost drivers | When to prefer |

| MCP tool | DB/API latency | Fast factual lookups |
| MCP resource | Storage read, chunking | Context prefetch |

Use MCP for **hot path** facts; use A2A when **delegation** is the product requirement.

## 16. Optional Live LLM Router

Replace rule-based `choose_protocol` with an LLM that outputs structured routing decisions.

````python
def llm_choose_protocol(user_text: str) -> list[ProtocolDecision]:
    if not USE_LIVE_LLM:
        return choose_protocol(user_text)
    from langchain_openai import ChatOpenAI

    catalog = {
        "mcp_tools": [t["name"] for t in mcp.tools],
        "mcp_resources": [r["uri"] for r in mcp.resources],
        "a2a_agents": {k: v.skills for k, v in A2A_AGENTS.items()},
    }
    llm = ChatOpenAI(model="gpt-4o-mini", api_key=OPENAI_API_KEY, temperature=0)
    prompt = (
        f"Pick protocol steps for ShopCo support. Catalog: {json.dumps(catalog)}\n"
        f"Request: {user_text}\n"
        "Reply JSON list of {protocol, target, reason} where protocol is mcp_tool|mcp_resource|a2a_agent."
    )
    raw = llm.invoke(prompt).content
    try:
        items = json.loads(raw)
        return [ProtocolDecision(i["protocol"], i["target"], i["reason"]) for i in items]
    except (json.JSONDecodeError, KeyError):
        return choose_protocol(user_text)


llm_decs = llm_choose_protocol("Tracking for ORD-1001 and return policy")
print("LLM routing (or offline fallback):")
for d in llm_decs:
    print(f"  {d.protocol} → {d.target}: {d.reason[:50]}")
````

## 18. Summary

**MCP** and **A2A** solve different integration problems:

| MCP | A2A |
|-----|-----|
| Tools + resources | Agent delegation |
| Your infrastructure | Partner / specialist agents |
| Fast, schema-bound calls | Task lifecycle + streaming |

The **ShopCo Hybrid Coordinator** demonstrates the production pattern:

- `get_order` and policy resources → **MCP** (owned data)
- ShipFast tracking and PayRight disputes → **A2A** (partner agents)
- `choose_protocol()` + unified trace → auditable routing

Design systems that use **both** protocols deliberately — not one instead of the other.
