# Deploying A2A Agents

Understanding the **A2A protocol** is step one. **Deploying** an A2A agent is step two — turning a handler function into a reliable service that partners can discover, authenticate against, and call at scale.

A deployed A2A agent is a **network service** that exposes:

| Endpoint | Purpose |
|----------|--------|
| `GET /.well-known/agent.json` | Public Agent Card for discovery |
| `POST /a2a` (or binding-specific path) | JSON-RPC: `message/send`, `tasks/get`, … |
| `GET /health` | Liveness — process is up |
| `GET /ready` | Readiness — dependencies available |

This notebook deploys **ShopCo Support Agent** as a production-shaped service — entirely in-process (no separate server process or CLI). You will implement:

1. Environment-based configuration (dev / staging / prod)
2. Auth middleware (API keys, scoped clients)
3. Durable task store abstraction
4. Rate limiting and request validation
5. Metrics, audit logs, and health probes
6. Versioned rollout (v1 → v2 Agent Card)
7. A remote client calling the deployed agent through an API gateway

Everything is self-contained plain Python.

## 3. Core A2A Types (Task + Message)

Deployment code wraps the same task lifecycle from the protocol layer.

````python
class TaskState(str, Enum):
    SUBMITTED = "TASK_STATE_SUBMITTED"
    WORKING = "TASK_STATE_WORKING"
    COMPLETED = "TASK_STATE_COMPLETED"
    FAILED = "TASK_STATE_FAILED"
    INPUT_REQUIRED = "TASK_STATE_INPUT_REQUIRED"
    CANCELED = "TASK_STATE_CANCELED"

    @property
    def is_terminal(self) -> bool:
        return self in {TaskState.COMPLETED, TaskState.FAILED, TaskState.CANCELED}


@dataclass
class A2ATask:
    id: str
    state: TaskState = TaskState.SUBMITTED
    input_text: str = ""
    artifacts: list[dict[str, Any]] = field(default_factory=list)
    metadata: dict[str, Any] = field(default_factory=dict)
    created_at: str = field(default_factory=utc_now)
    updated_at: str = field(default_factory=utc_now)
    client_id: str = ""

    def to_dict(self) -> dict[str, Any]:
        return {
            "id": self.id,
            "status": {"state": self.state.value, "updatedAt": self.updated_at},
            "artifacts": self.artifacts,
            "metadata": self.metadata,
        }

    def set_state(self, state: TaskState) -> None:
        self.state = state
        self.updated_at = utc_now()
````

## 5. Task Store — Persistence Abstraction

In-memory stores lose tasks on restart. Production uses Redis, Postgres, or a managed task service. The interface stays the same.

````python
class TaskStore:
    def save(self, task: A2ATask) -> None:
        raise NotImplementedError

    def get(self, task_id: str) -> A2ATask | None:
        raise NotImplementedError


class InMemoryTaskStore(TaskStore):
    def __init__(self) -> None:
        self._tasks: dict[str, A2ATask] = {}

    def save(self, task: A2ATask) -> None:
        self._tasks[task.id] = task

    def get(self, task_id: str) -> A2ATask | None:
        return self._tasks.get(task_id)

    def count(self) -> int:
        return len(self._tasks)


class DurableTaskStore(TaskStore):
    """Simulates durable storage with a write-ahead log."""

    def __init__(self) -> None:
        self._tasks: dict[str, A2ATask] = {}
        self._wal: list[dict[str, Any]] = []

    def save(self, task: A2ATask) -> None:
        self._wal.append({"op": "upsert", "task_id": task.id, "state": task.state.value, "ts": utc_now()})
        self._tasks[task.id] = task

    def get(self, task_id: str) -> A2ATask | None:
        return self._tasks.get(task_id)

    @property
    def wal_entries(self) -> int:
        return len(self._wal)


TASK_STORE: TaskStore = DurableTaskStore()
print(f"Task store: {type(TASK_STORE).__name__}")
````

## 6. Auth Middleware

Validate `Authorization: Bearer <api_key>` before JSON-RPC dispatch. Map keys to `client_id` for audit scoping.

````python
@dataclass
class AuthContext:
    client_id: str
    authenticated: bool = True


class AuthMiddleware:
    def __init__(self, config: DeployConfig) -> None:
        self.config = config
        self._key_to_client = {v: k for k, v in config.api_keys.items()}

    def authenticate(self, headers: dict[str, str]) -> AuthContext:
        if not self.config.require_auth:
            return AuthContext(client_id="anonymous", authenticated=True)
        auth_header = headers.get("Authorization", "")
        if not auth_header.startswith("Bearer "):
            raise PermissionError("Missing Bearer token")
        token = auth_header.removeprefix("Bearer ").strip()
        client_id = self._key_to_client.get(token)
        if not client_id:
            raise PermissionError("Invalid API key")
        return AuthContext(client_id=client_id)


AUTH = AuthMiddleware(CONFIG)

try:
    ok = AUTH.authenticate({"Authorization": f"Bearer {CONFIG.api_keys['coordinator']}"})
    print(f"Auth OK: client={ok.client_id}")
    AUTH.authenticate({"Authorization": "Bearer bad-key"})
except PermissionError as e:
    print(f"Auth blocked (expected): {e}")
````

## 7. Rate Limiter

Protect deployed agents from abuse. Simple token-bucket per `client_id`.

````python
class RateLimiter:
    def __init__(self, limit_per_minute: int) -> None:
        self.limit = limit_per_minute
        self._windows: dict[str, list[float]] = {}

    def allow(self, client_id: str) -> bool:
        now = time.time()
        window = self._windows.setdefault(client_id, [])
        window[:] = [t for t in window if now - t < 60]
        if len(window) >= self.limit:
            return False
        window.append(now)
        return True


RATE_LIMITER = RateLimiter(CONFIG.rate_limit_per_minute)
print("Rate limiter ready:", CONFIG.rate_limit_per_minute, "req/min")
````

## 8. Agent Business Logic (Versioned Handlers)

Support **v1** and **v2** handlers to demonstrate blue-green rollout. The deployed service selects handler by `agent_version`.

````python
def support_handler_v1(task: A2ATask) -> A2ATask:
    task.set_state(TaskState.WORKING)
    text = task.input_text
    m = re.search(r"ORD-[0-9]{4}", text.upper())
    if m:
        order = ORDERS.get(m.group(0))
        if order:
            task.artifacts.append({"type": "text", "text": f"[v1] {m.group(0)} status={order['status']}"})
            task.set_state(TaskState.COMPLETED)
            return task
    if "return" in text.lower():
        task.artifacts.append({"type": "text", "text": f"[v1] {POLICIES['returns']}"})
        task.set_state(TaskState.COMPLETED)
        return task
    task.set_state(TaskState.INPUT_REQUIRED)
    task.artifacts.append({"type": "text", "text": "[v1] Provide order ID or policy question."})
    return task


def support_handler_v2(task: A2ATask) -> A2ATask:
    """v2 adds richer tracking details and structured artifacts."""
    task.set_state(TaskState.WORKING)
    text = task.input_text
    m = re.search(r"ORD-[0-9]{4}", text.upper())
    if m:
        order = ORDERS.get(m.group(0))
        if order:
            task.artifacts.append({
                "type": "data",
                "data": {
                    "version": "2.0",
                    "order_id": order["order_id"],
                    "status": order["status"],
                    "tracking": order.get("tracking"),
                    "customer": order.get("customer"),
                },
            })
            task.set_state(TaskState.COMPLETED)
            return task
    if "return" in text.lower():
        task.artifacts.append({"type": "text", "text": POLICIES["returns"]})
        task.set_state(TaskState.COMPLETED)
        return task
    task.set_state(TaskState.INPUT_REQUIRED)
    task.artifacts.append({"type": "text", "text": "Please share your ORD-XXXX order ID."})
    return task


HANDLERS: dict[str, Callable[[A2ATask], A2ATask]] = {
    "1.1.0": support_handler_v1,
    "1.2.0": support_handler_v2,
}
````

## 9. Deployed A2A Service

`DeployedA2AAgent` simulates an HTTP service with route dispatch, auth, rate limits, metrics, and JSON-RPC handling.

````python
@dataclass
class ServiceMetrics:
    requests_total: int = 0
    auth_failures: int = 0
    rate_limited: int = 0
    tasks_completed: int = 0
    tasks_failed: int = 0
    latency_ms: list[float] = field(default_factory=list)

    def record_latency(self, ms: float) -> None:
        self.latency_ms.append(ms)

    @property
    def p95_latency_ms(self) -> float:
        if not self.latency_ms:
            return 0.0
        sorted_lat = sorted(self.latency_ms)
        idx = int(len(sorted_lat) * 0.95) - 1
        return sorted_lat[max(0, idx)]


class DeployedA2AAgent:
    def __init__(
        self,
        config: DeployConfig,
        store: TaskStore,
        handler: Callable[[A2ATask], A2ATask],
    ) -> None:
        self.config = config
        self.store = store
        self.handler = handler
        self.auth = AuthMiddleware(config)
        self.rate_limiter = RateLimiter(config.rate_limit_per_minute)
        self.metrics = ServiceMetrics()
        self.audit_log: list[dict[str, Any]] = []
        self._dependencies_healthy = True

    def set_dependencies_healthy(self, healthy: bool) -> None:
        self._dependencies_healthy = healthy

    def handle_http(self, method: str, path: str, headers: dict[str, str], body: dict[str, Any] | None = None) -> tuple[int, dict[str, Any]]:
        started = time.perf_counter()
        self.metrics.requests_total += 1
        try:
            if path == "/health":
                return 200, {"status": "ok", "ts": utc_now()}
            if path == "/ready":
                if not self._dependencies_healthy:
                    return 503, {"status": "not_ready", "reason": "backing store unavailable"}
                return 200, {"status": "ready", "agent_version": self.config.agent_version}
            if path == "/.well-known/agent.json" and method == "GET":
                return 200, build_agent_card(self.config)
            if path == "/a2a/v1" and method == "POST":
                return self._handle_jsonrpc(headers, body or {})
            return 404, {"error": "not_found", "path": path}
        finally:
            self.metrics.record_latency((time.perf_counter() - started) * 1000)

    def _handle_jsonrpc(self, headers: dict[str, str], body: dict[str, Any]) -> tuple[int, dict[str, Any]]:
        rpc_id = body.get("id", str(uuid.uuid4()))
        try:
            auth_ctx = self.auth.authenticate(headers)
        except PermissionError:
            self.metrics.auth_failures += 1
            return 401, {"jsonrpc": "2.0", "id": rpc_id, "error": {"code": -32001, "message": "unauthorized"}}
        if not self.rate_limiter.allow(auth_ctx.client_id):
            self.metrics.rate_limited += 1
            return 429, {"jsonrpc": "2.0", "id": rpc_id, "error": {"code": -32005, "message": "rate limited"}}

        method = body.get("method", "")
        params = body.get("params", {})

        if method == "message/send":
            return self._message_send(rpc_id, params, auth_ctx)
        if method == "tasks/get":
            return self._tasks_get(rpc_id, params)
        if method == "tasks/cancel":
            return self._tasks_cancel(rpc_id, params)
        return 400, {"jsonrpc": "2.0", "id": rpc_id, "error": {"code": -32601, "message": f"unknown method {method}"}}

    def _message_send(self, rpc_id: str, params: dict[str, Any], auth_ctx: AuthContext) -> tuple[int, dict[str, Any]]:
        msg = params.get("message", {})
        parts = msg.get("parts", [])
        text = " ".join(p.get("text", "") for p in parts if "text" in p).strip()
        task = A2ATask(
            id=f"task-{uuid.uuid4().hex[:12]}",
            input_text=text,
            client_id=auth_ctx.client_id,
            metadata={"agent_version": self.config.agent_version},
        )
        processed = self.handler(task)
        self.store.save(processed)
        self.audit_log.append({
            "ts": utc_now(), "client": auth_ctx.client_id,
            "method": "message/send", "task_id": processed.id, "state": processed.state.value,
        })
        if processed.state == TaskState.COMPLETED:
            self.metrics.tasks_completed += 1
        elif processed.state == TaskState.FAILED:
            self.metrics.tasks_failed += 1
        return 200, {"jsonrpc": "2.0", "id": rpc_id, "result": {"task": processed.to_dict()}}

    def _tasks_get(self, rpc_id: str, params: dict[str, Any]) -> tuple[int, dict[str, Any]]:
        task_id = params.get("taskId") or params.get("id")
        task = self.store.get(task_id) if task_id else None
        if not task:
            return 404, {"jsonrpc": "2.0", "id": rpc_id, "error": {"code": -32001, "message": "task not found"}}
        return 200, {"jsonrpc": "2.0", "id": rpc_id, "result": {"task": task.to_dict()}}

    def _tasks_cancel(self, rpc_id: str, params: dict[str, Any]) -> tuple[int, dict[str, Any]]:
        task_id = params.get("taskId") or params.get("id")
        task = self.store.get(task_id) if task_id else None
        if not task:
            return 404, {"jsonrpc": "2.0", "id": rpc_id, "error": {"code": -32001, "message": "task not found"}}
        if task.state.is_terminal:
            return 409, {"jsonrpc": "2.0", "id": rpc_id, "error": {"code": -32002, "message": "not cancelable"}}
        task.set_state(TaskState.CANCELED)
        self.store.save(task)
        return 200, {"jsonrpc": "2.0", "id": rpc_id, "result": {"task": task.to_dict()}}


deployed_agent = DeployedA2AAgent(CONFIG, TASK_STORE, HANDLERS[CONFIG.agent_version])
print(f"Deployed agent v{CONFIG.agent_version} listening (simulated) at {CONFIG.base_url}")
````

## 11. Remote A2A Client (Through Gateway)

Clients discover the Agent Card, then POST JSON-RPC with auth headers — same flow whether the service runs locally or on Cloud Run.

````python
class RemoteA2AClient:
    """HTTP client that calls DeployedA2AAgent.handle_http in-process."""

    def __init__(self, service: DeployedA2AAgent, api_key: str) -> None:
        self.service = service
        self.headers = {"Authorization": f"Bearer {api_key}", "Content-Type": "application/json"}

    def fetch_agent_card(self) -> dict[str, Any]:
        code, body = self.service.handle_http("GET", "/.well-known/agent.json", {})
        if code != 200:
            raise RuntimeError(f"Agent Card fetch failed: {code}")
        return body

    def send_message(self, text: str) -> dict[str, Any]:
        payload = {
            "jsonrpc": "2.0",
            "method": "message/send",
            "params": {"message": {"role": "USER", "parts": [{"text": text}]}},
            "id": str(uuid.uuid4()),
        }
        code, body = self.service.handle_http("POST", "/a2a/v1", self.headers, payload)
        if code != 200:
            raise RuntimeError(f"message/send failed [{code}]: {body}")
        return body["result"]["task"]

    def get_task(self, task_id: str) -> dict[str, Any]:
        payload = {"jsonrpc": "2.0", "method": "tasks/get", "params": {"taskId": task_id}, "id": str(uuid.uuid4())}
        code, body = self.service.handle_http("POST", "/a2a/v1", self.headers, payload)
        if code != 200:
            raise RuntimeError(f"tasks/get failed: {body}")
        return body["result"]["task"]


client = RemoteA2AClient(deployed_agent, CONFIG.api_keys["coordinator"])
card = client.fetch_agent_card()
print(f"Remote discovered: {card['name']} v{card['version']}")

task = client.send_message("Where is order ORD-1001?")
print(f"Task {task['id']}: {task['status']['state']}")
print(f"Artifacts: {task['artifacts']}")
````

## 14. Deployment Readiness Scorecard

Automate pre-flight checks before exposing an agent to partners.

````python
READINESS_CHECKS: list[tuple[str, Callable[[DeployedA2AAgent], bool]]] = [
    ("agent_card_published", lambda s: s.handle_http("GET", "/.well-known/agent.json", {})[0] == 200),
    ("health_ok", lambda s: s.handle_http("GET", "/health", {})[0] == 200),
    ("ready_ok", lambda s: s.handle_http("GET", "/ready", {})[0] == 200),
    ("auth_enforced", lambda s: s.config.require_auth),
    ("handler_version_set", lambda s: bool(s.config.agent_version)),
    ("rate_limit_configured", lambda s: s.config.rate_limit_per_minute > 0),
]


def readiness_score(service: DeployedA2AAgent) -> dict[str, Any]:
    results = {name: check(service) for name, check in READINESS_CHECKS}
    passed = sum(results.values())
    return {"passed": passed, "total": len(results), "checks": results, "ready": passed == len(results)}


score = readiness_score(deployed_agent)
print(f"Readiness: {score['passed']}/{score['total']} — deploy={'YES' if score['ready'] else 'NO'}")
for name, ok in score["checks"].items():
    print(f"  {'✓' if ok else '✗'} {name}")
````

## 16. Gateway Pattern — Coordinator Calling Deployed Agents

A coordinator service does not embed business logic — it routes to deployed specialist agents using their Agent Cards.

````python
class CoordinatorGateway:
    def __init__(self, agents: dict[str, DeployedA2AAgent], api_key: str) -> None:
        self.agents = agents
        self.api_key = api_key

    def route(self, user_text: str) -> dict[str, Any]:
        target = "support" if "ord-" in user_text.lower() or "order" in user_text.lower() else "support"
        service = self.agents[target]
        client = RemoteA2AClient(service, self.api_key)
        card = client.fetch_agent_card()
        task = client.send_message(user_text)
        return {
            "routed_to": target,
            "agent_version": card["version"],
            "task_id": task["id"],
            "state": task["status"]["state"],
            "artifacts": task["artifacts"],
        }


gateway = CoordinatorGateway({"support": deployed_agent}, CONFIG.api_keys["coordinator"])
result = gateway.route("What is the status of ORD-1002?")
print(pretty(result))
````

## 17. Common Deployment Mistakes

| Mistake | Risk | Fix |
|---------|------|-----|
| No auth on `/a2a` | Anyone invokes your agent | Bearer tokens + gateway |
| Agent Card out of sync with code | Wrong capability ads | Generate card from config |
| In-memory tasks only | Lost work on restart | Durable task store |
| No `/ready` probe | Traffic to broken deps | Readiness checks backing systems |
| Skip rate limits | Cost / DoS exposure | Per-client token bucket |
| Hard-coded partner URLs | Brittle integrations | Discovery via Agent Card |
| No version in artifacts | Clients can't parse upgrades | Semantic versioning + structured output |

## 20. Summary

Deploying an A2A agent means wrapping protocol handlers in a **service shell**: Agent Card publication, authenticated JSON-RPC, durable tasks, health probes, rate limits, and observability.

The **ShopCo Support Agent** in this notebook demonstrates the full path:

- `DeployConfig` from environment variables
- `DeployedA2AAgent` with `/health`, `/ready`, `/.well-known/agent.json`, `/a2a/v1`
- `RemoteA2AClient` for partner-style integration
- Versioned handlers (v1 → v2) and a readiness scorecard

Swap the in-process `handle_http` dispatcher for FastAPI, Cloud Run, or Kubernetes — the boundaries and checks remain the same.
