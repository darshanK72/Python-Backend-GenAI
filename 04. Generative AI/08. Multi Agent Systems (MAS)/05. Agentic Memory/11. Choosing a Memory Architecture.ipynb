# Choosing a Memory Architecture

Agent memory is not one product decision — it is a **stack of layers** with different lifetimes, costs, and guarantees. Picking wrong leads to:

- **Context-only** → user repeats themselves every turn
- **Thread-only** → preferences lost when they open a new chat
- **LTM without RAG** → agent invents policy answers
- **RAG without memory** → agent forgets "my order is ORD-1001"

```
  L0  Context window     ─ messages in current LLM call
  L1  Working / scratch  ─ graph state fields (order_id, intent)
  L2  Thread memory      ─ checkpointer per thread_id
  L3  Long-term memory   ─ user-scoped store (LangMem, Mem0, BaseStore)
  L4  Retrieval (RAG)    ─ shared corpus, not per-user
```

This notebook gives you a **decision framework** and runnable **ShopCo Support** prototypes for each architecture tier, plus a scorer that recommends a stack from your requirements.

Self-contained — plain Python, offline by default.

````python
"""
ShopCo Support — memory architecture decision lab.
"""

import json
import os
import re
from dataclasses import dataclass, field
from enum import Enum
from typing import Any, Literal

OPENAI_API_KEY = os.environ.get("OPENAI_API_KEY", "sk-proj-placeholder-replace-with-your-real-key")
os.environ.setdefault("OPENAI_API_KEY", OPENAI_API_KEY)

USE_LIVE_LLM = False


def pretty(obj: Any) -> str:
    return json.dumps(obj, indent=2, default=str)


POLICY_DOCS: list[dict[str, str]] = [
    {"id": "pol-returns-01", "title": "Returns Policy",
     "text": "Returns within 30 days. Refunds in 5-7 business days."},
    {"id": "pol-shipping-02", "title": "Shipping Policy",
     "text": "Free shipping over $50. Tracking emailed at carrier scan."},
    {"id": "pol-billing-03", "title": "Billing Policy",
     "text": "Charges appear as SHOPCO* on statements. Disputes within 60 days."},
]

ORDERS_DB = {
    "ORD-1001": {"order_id": "ORD-1001", "status": "shipped"},
    "ORD-1002": {"order_id": "ORD-1002", "status": "processing"},
}

print("ShopCo architecture lab ready.")
````

## 1. The Five Memory Layers

| Layer | Lifetime | Key | ShopCo example |
|-------|----------|-----|----------------|
| **L0 Context** | Single LLM call | Prompt tokens | Last user message only |
| **L1 Working** | One graph step | State fields | `order_id` extracted this turn |
| **L2 Thread** | One conversation | `thread_id` | Full chat until tab closed |
| **L3 Long-term** | Weeks–months | `user_id` | "Prefers email contact" |
| **L4 Retrieval** | Corpus version | Doc chunk IDs | Official return policy text |

Production agents almost always need **L2 + at least one of L3 or L4**. L0 alone is never enough for support.

## 2. Architecture Profiles — What Each Stack Includes

````python
class MemoryArchitecture(str, Enum):
    CONTEXT_ONLY = "context_only"           # L0
    THREAD = "thread"                       # L0 + L1 + L2
    LONG_TERM = "long_term"                 # L0-L3
    RAG_ONLY = "rag_only"                   # L0 + L4
    THREAD_RAG = "thread_rag"               # L0-L2 + L4
    FULL_STACK = "full_stack"               # L0-L4


ARCHITECTURE_LAYERS: dict[MemoryArchitecture, set[str]] = {
    MemoryArchitecture.CONTEXT_ONLY: {"L0"},
    MemoryArchitecture.THREAD: {"L0", "L1", "L2"},
    MemoryArchitecture.LONG_TERM: {"L0", "L1", "L2", "L3"},
    MemoryArchitecture.RAG_ONLY: {"L0", "L4"},
    MemoryArchitecture.THREAD_RAG: {"L0", "L1", "L2", "L4"},
    MemoryArchitecture.FULL_STACK: {"L0", "L1", "L2", "L3", "L4"},
}

for arch, layers in ARCHITECTURE_LAYERS.items():
    print(f"{arch.value:14} → {sorted(layers)}")
````

## 3. Requirements Profile — Score Your Use Case

````python
@dataclass
class RequirementsProfile:
    """Score each dimension 0-5 (higher = more important)."""
    name: str
    multi_turn_context: int = 3      # needs prior turns in same chat
    cross_session_recall: int = 3    # new chat, same user
    policy_grounding: int = 3        # must cite official docs
    per_user_facts: int = 3          # names, prefs, order IDs
    audit_citations: int = 3         # prove answer source
    low_latency: int = 3             # minimize retrieval/write overhead
    low_complexity: int = 3          # small team, ship fast

    def as_dict(self) -> dict[str, int]:
        return {k: v for k, v in self.__dict__.items() if k != "name" and isinstance(v, int)}


SHOPCO_SUPPORT = RequirementsProfile(
    name="ShopCo Tier-1 Support",
    multi_turn_context=5,
    cross_session_recall=4,
    policy_grounding=5,
    per_user_facts=4,
    audit_citations=5,
    low_latency=3,
    low_complexity=2,
)

RELEASEFLOW_OPS = RequirementsProfile(
    name="ReleaseFlow Deploy Assistant",
    multi_turn_context=4,
    cross_session_recall=2,
    policy_grounding=5,   # runbooks, SLO docs
    per_user_facts=2,
    audit_citations=5,
    low_latency=4,
    low_complexity=3,
)

print(SHOPCO_SUPPORT.name, "→", pretty(SHOPCO_SUPPORT.as_dict()))
````

## 4. Architecture Scorer

````python
# How well each architecture satisfies each requirement (0-5)
ARCHITECTURE_FIT: dict[MemoryArchitecture, dict[str, int]] = {
    MemoryArchitecture.CONTEXT_ONLY: {
        "multi_turn_context": 1, "cross_session_recall": 0, "policy_grounding": 0,
        "per_user_facts": 1, "audit_citations": 0, "low_latency": 5, "low_complexity": 5,
    },
    MemoryArchitecture.THREAD: {
        "multi_turn_context": 5, "cross_session_recall": 1, "policy_grounding": 1,
        "per_user_facts": 4, "audit_citations": 1, "low_latency": 4, "low_complexity": 4,
    },
    MemoryArchitecture.LONG_TERM: {
        "multi_turn_context": 5, "cross_session_recall": 5, "policy_grounding": 1,
        "per_user_facts": 5, "audit_citations": 2, "low_latency": 3, "low_complexity": 2,
    },
    MemoryArchitecture.RAG_ONLY: {
        "multi_turn_context": 1, "cross_session_recall": 0, "policy_grounding": 5,
        "per_user_facts": 0, "audit_citations": 5, "low_latency": 3, "low_complexity": 3,
    },
    MemoryArchitecture.THREAD_RAG: {
        "multi_turn_context": 5, "cross_session_recall": 1, "policy_grounding": 5,
        "per_user_facts": 4, "audit_citations": 5, "low_latency": 2, "low_complexity": 3,
    },
    MemoryArchitecture.FULL_STACK: {
        "multi_turn_context": 5, "cross_session_recall": 5, "policy_grounding": 5,
        "per_user_facts": 5, "audit_citations": 5, "low_latency": 1, "low_complexity": 1,
    },
}


@dataclass
class ArchitectureScore:
    architecture: MemoryArchitecture
    weighted_score: float
    breakdown: dict[str, float] = field(default_factory=dict)


def score_architecture(profile: RequirementsProfile, arch: MemoryArchitecture) -> ArchitectureScore:
    weights = profile.as_dict()
    fit = ARCHITECTURE_FIT[arch]
    breakdown: dict[str, float] = {}
    total_w = sum(weights.values()) or 1
    for dim, importance in weights.items():
        breakdown[dim] = (importance / total_w) * fit.get(dim, 0)
    return ArchitectureScore(arch, sum(breakdown.values()), breakdown)


def recommend_architecture(profile: RequirementsProfile, top_k: int = 3) -> list[ArchitectureScore]:
    scores = [score_architecture(profile, arch) for arch in MemoryArchitecture]
    return sorted(scores, key=lambda s: s.weighted_score, reverse=True)[:top_k]


for profile in (SHOPCO_SUPPORT, RELEASEFLOW_OPS):
    print(f"\n=== {profile.name} ===")
    for s in recommend_architecture(profile):
        print(f"  {s.architecture.value:14} score={s.weighted_score:.2f}")
````

## 5. Runnable Prototypes — Six Architectures on ShopCo

````python
def rag_search(query: str, top_k: int = 1) -> list[dict[str, str]]:
    q = query.lower()
    scored = []
    for doc in POLICY_DOCS:
        text = (doc["title"] + " " + doc["text"]).lower()
        score = sum(1 for w in q.split() if w in text)
        if score:
            scored.append((score, doc))
    scored.sort(key=lambda x: x[0], reverse=True)
    return [{"id": d["id"], "text": d["text"]} for _, d in scored[:top_k]]


@dataclass
class TurnResult:
    answer: str
    grounded: bool = False
    cited: bool = False
    recalled_user_fact: bool = False
    architecture: str = ""


class ContextOnlyBot:
    architecture = MemoryArchitecture.CONTEXT_ONLY

    def reply(self, message: str, **_: Any) -> TurnResult:
        q = message.lower()
        if "return" in q:
            return TurnResult("Returns are allowed sometimes.", architecture=self.architecture.value)
        return TurnResult("How can I help?", architecture=self.architecture.value)


class ThreadBot:
    architecture = MemoryArchitecture.THREAD

    def __init__(self) -> None:
        self.threads: dict[str, dict[str, str]] = {}

    def reply(self, message: str, *, thread_id: str, **_: Any) -> TurnResult:
        state = self.threads.setdefault(thread_id, {})
        m = re.search(r"ORD-[0-9]{4}", message.upper())
        if m:
            state["order_id"] = m.group(0)
        if "my name is" in message.lower():
            state["name"] = message.split()[-1].rstrip(".")
        q = message.lower()
        if "status" in q or ("it" in q and state.get("order_id")):
            oid = state.get("order_id", "?")
            status = ORDERS_DB.get(oid, {}).get("status", "unknown")
            return TurnResult(f"Order {oid} is {status}.", recalled_user_fact=True, architecture=self.architecture.value)
        return TurnResult(f"Hi {state.get('name', 'there')}, how can I help?", architecture=self.architecture.value)


class LongTermBot:
    architecture = MemoryArchitecture.LONG_TERM

    def __init__(self) -> None:
        self.users: dict[str, dict[str, str]] = {}
        self.threads: dict[str, dict[str, str]] = {}

    def reply(self, message: str, *, user_id: str, thread_id: str, **_: Any) -> TurnResult:
        profile = self.users.setdefault(user_id, {})
        thread = self.threads.setdefault(thread_id, {})
        if "prefer" in message.lower() or "remember" in message.lower():
            profile["preference"] = message
        m = re.search(r"ORD-[0-9]{4}", message.upper())
        if m:
            thread["order_id"] = m.group(0)
        if "preference" in message.lower() or "remember" in message.lower():
            return TurnResult("Saved to long-term memory.", architecture=self.architecture.value)
        if profile.get("preference"):
            return TurnResult(f"Your saved preference: {profile['preference'][:60]}...", recalled_user_fact=True, architecture=self.architecture.value)
        return TurnResult("How can I help?", architecture=self.architecture.value)


class RAGOnlyBot:
    architecture = MemoryArchitecture.RAG_ONLY

    def reply(self, message: str, **_: Any) -> TurnResult:
        hits = rag_search(message)
        if hits:
            return TurnResult(f"{hits[0]['text']} [{hits[0]['id']}]", grounded=True, cited=True, architecture=self.architecture.value)
        return TurnResult("No policy found.", architecture=self.architecture.value)


class ThreadRAGBot:
    architecture = MemoryArchitecture.THREAD_RAG

    def __init__(self) -> None:
        self.threads: dict[str, dict[str, str]] = {}

    def reply(self, message: str, *, thread_id: str, **_: Any) -> TurnResult:
        state = self.threads.setdefault(thread_id, {})
        m = re.search(r"ORD-[0-9]{4}", message.upper())
        if m:
            state["order_id"] = m.group(0)
        q = message.lower()
        if "return" in q or "refund" in q:
            hits = rag_search(message)
            return TurnResult(f"{hits[0]['text']} [{hits[0]['id']}]", grounded=True, cited=True, architecture=self.architecture.value)
        if "status" in q and state.get("order_id"):
            oid = state["order_id"]
            status = ORDERS_DB.get(oid, {}).get("status", "unknown")
            return TurnResult(f"Order {oid}: {status}.", recalled_user_fact=True, architecture=self.architecture.value)
        return TurnResult("Ask about policy or your order.", architecture=self.architecture.value)


class FullStackBot:
    architecture = MemoryArchitecture.FULL_STACK

    def __init__(self) -> None:
        self.users: dict[str, dict[str, str]] = {}
        self.threads: dict[str, dict[str, str]] = {}

    def reply(self, message: str, *, user_id: str, thread_id: str, **_: Any) -> TurnResult:
        profile = self.users.setdefault(user_id, {})
        thread = self.threads.setdefault(thread_id, {})
        if "prefer" in message.lower():
            profile["preference"] = message
        m = re.search(r"ORD-[0-9]{4}", message.upper())
        if m:
            thread["order_id"] = m.group(0)
        q = message.lower()
        if "return" in q:
            hits = rag_search(message)
            pref = f" (contact: {profile['preference'][:30]}...)" if profile.get("preference") else ""
            return TurnResult(f"{hits[0]['text']} [{hits[0]['id']}]{pref}", grounded=True, cited=True, recalled_user_fact=bool(profile.get("preference")), architecture=self.architecture.value)
        if "preference" in q and profile.get("preference"):
            return TurnResult(profile["preference"], recalled_user_fact=True, architecture=self.architecture.value)
        return TurnResult("How can I help?", architecture=self.architecture.value)


BOTS: dict[MemoryArchitecture, Any] = {
    MemoryArchitecture.CONTEXT_ONLY: ContextOnlyBot(),
    MemoryArchitecture.THREAD: ThreadBot(),
    MemoryArchitecture.LONG_TERM: LongTermBot(),
    MemoryArchitecture.RAG_ONLY: RAGOnlyBot(),
    MemoryArchitecture.THREAD_RAG: ThreadRAGBot(),
    MemoryArchitecture.FULL_STACK: FullStackBot(),
}
````

## 6. Scenario Suite — Where Each Architecture Fails

````python
@dataclass
class Scenario:
    name: str
    turns: list[dict[str, str]]  # [{message, user_id, thread_id}]
    check: Literal["grounded", "cited", "recalled", "pronoun"]
    best_architectures: set[MemoryArchitecture]


SCENARIOS = [
    Scenario(
        "Policy grounding",
        [{"message": "What is the return window?", "user_id": "alice", "thread_id": "t1"}],
        "cited", {MemoryArchitecture.RAG_ONLY, MemoryArchitecture.THREAD_RAG, MemoryArchitecture.FULL_STACK},
    ),
    Scenario(
        "Pronoun resolution (thread)",
        [
            {"message": "My order is ORD-1001", "user_id": "alice", "thread_id": "t1"},
            {"message": "What's its status?", "user_id": "alice", "thread_id": "t1"},
        ],
        "recalled", {MemoryArchitecture.THREAD, MemoryArchitecture.THREAD_RAG, MemoryArchitecture.FULL_STACK},
    ),
    Scenario(
        "Cross-session preference",
        [
            {"message": "Remember I prefer email contact", "user_id": "alice", "thread_id": "mon"},
            {"message": "What are my contact preferences?", "user_id": "alice", "thread_id": "fri"},
        ],
        "recalled", {MemoryArchitecture.LONG_TERM, MemoryArchitecture.FULL_STACK},
    ),
    Scenario(
        "Policy + preference",
        [
            {"message": "I prefer email. What's the return policy?", "user_id": "alice", "thread_id": "t1"},
        ],
        "cited", {MemoryArchitecture.FULL_STACK},
    ),
]


def run_scenario(bot, scenario: Scenario) -> TurnResult:
    result = TurnResult("", architecture=getattr(bot, "architecture", MemoryArchitecture.CONTEXT_ONLY).value)
    for turn in scenario.turns:
        result = bot.reply(turn["message"], user_id=turn["user_id"], thread_id=turn["thread_id"])
    return result


def scenario_passed(result: TurnResult, scenario: Scenario) -> bool:
    if scenario.check == "cited":
        return result.cited and result.grounded
    if scenario.check == "recalled":
        return result.recalled_user_fact
    if scenario.check == "grounded":
        return result.grounded
    return False


for sc in SCENARIOS:
    print(f"\n=== {sc.name} ===")
    for arch, bot in BOTS.items():
        r = run_scenario(bot, sc)
        ok = scenario_passed(r, sc)
        best = "★" if arch in sc.best_architectures else " "
        print(f"  {best} {arch.value:14} {'✓' if ok else '✗'}  {r.answer[:55]}")
````

## 7. LTM Backend Choice — LangMem vs Mem0 vs BaseStore

````python
@dataclass
class LTMBackendOption:
    name: str
    langgraph_native: int   # 0-5
    managed_service: int
    extraction_builtin: int
    framework_agnostic: int
    notes: str


LTM_BACKENDS = [
    LTMBackendOption("BaseStore (manual)", 5, 2, 1, 3, "Full control; you write extract/load logic"),
    LTMBackendOption("LangMem", 5, 2, 4, 2, "Best when already on LangGraph + BaseStore"),
    LTMBackendOption("Mem0", 3, 5, 5, 5, "Best for managed LTM or framework-agnostic SDK"),
]


def pick_ltm_backend(needs_managed: bool, on_langgraph: bool) -> str:
    if needs_managed:
        return "Mem0 (platform)"
    if on_langgraph:
        return "LangMem + InMemoryStore/PostgresStore"
    return "Mem0 OSS or custom BaseStore"


print("ShopCo on LangGraph:", pick_ltm_backend(needs_managed=False, on_langgraph=True))
print("Startup wants managed:", pick_ltm_backend(needs_managed=True, on_langgraph=False))
print()
for b in LTM_BACKENDS:
    print(f"{b.name:22} native={b.langgraph_native} managed={b.managed_service} extract={b.extraction_builtin}")
````

## 8. Decision Tree

````python
def decide_stack(
    *,
    multi_turn: bool,
    cross_session: bool,
    needs_policy_docs: bool,
    needs_citations: bool,
) -> MemoryArchitecture:
    if not multi_turn and not cross_session and not needs_policy_docs:
        return MemoryArchitecture.CONTEXT_ONLY
    if needs_policy_docs and not multi_turn and not cross_session:
        return MemoryArchitecture.RAG_ONLY
    if multi_turn and needs_policy_docs and cross_session:
        return MemoryArchitecture.FULL_STACK
    if multi_turn and needs_policy_docs:
        return MemoryArchitecture.THREAD_RAG
    if cross_session:
        return MemoryArchitecture.LONG_TERM
    if multi_turn:
        return MemoryArchitecture.THREAD
    return MemoryArchitecture.THREAD_RAG


CASES = [
    ("ShopCo support", dict(multi_turn=True, cross_session=True, needs_policy_docs=True, needs_citations=True)),
    ("ReleaseFlow runbook Q&A", dict(multi_turn=True, cross_session=False, needs_policy_docs=True, needs_citations=True)),
    ("Simple FAQ bot", dict(multi_turn=False, cross_session=False, needs_policy_docs=True, needs_citations=True)),
    ("Personal prefs bot", dict(multi_turn=True, cross_session=True, needs_policy_docs=False, needs_citations=False)),
]

for label, kwargs in CASES:
    arch = decide_stack(**kwargs)
    print(f"{label:28} → {arch.value}")
````

## 11. Anti-Patterns — Architectures to Avoid

| Anti-pattern | Why it fails |
|--------------|--------------|
| **Giant system prompt** as "memory" | Stale, unscoped, no per-user facts |
| **RAG over chat logs** instead of LTM | Slow, noisy, privacy risk |
| **thread_id for long-term** | New chat = amnesia |
| **LTM without governance** | Contradictions accumulate |
| **Full stack for FAQ-only** | Over-engineering; use RAG_ONLY |
| **Context-only for support** | Fails pronoun and policy tests |

## 12. Eval Harness — Scorer vs Scenario Agreement

````python
def architecture_scenario_score(arch: MemoryArchitecture) -> int:
    passed = 0
    for sc in SCENARIOS:
        r = run_scenario(BOTS[arch], sc)
        if scenario_passed(r, sc):
            passed += 1
    return passed


ranking = sorted(MemoryArchitecture, key=lambda a: architecture_scenario_score(a), reverse=True)
scorer_top = recommend_architecture(SHOPCO_SUPPORT)[0].architecture
scenario_top = ranking[0]

print("Scenario pass counts:")
for arch in ranking:
    print(f"  {arch.value:14} {architecture_scenario_score(arch)}/{len(SCENARIOS)}")
print(f"\nScorer top pick:    {scorer_top.value}")
print(f"Scenario top pick:  {scenario_top.value}")
print(f"Agreement: {'✓' if scorer_top == scenario_top else '≈ (close — tune weights)'}")
````

## 14. Common Mistakes

| Mistake | Symptom | Fix |
|---------|---------|-----|
| Skipping requirements scoring | Wrong stack for use case | Run `RequirementsProfile` first |
| One-size-fits-all | Over-build simple FAQ | Match architecture to scenario |
| Ignoring L4 when policies matter | Hallucinated rules | Add RAG with citations |
| Ignoring L3 for returning users | Repeat preference questions | Add LTM backend |
| No migration plan | Big-bang complexity | Phase 1 → 2 → 3 |

## 16. Optional Live LLM — Natural Language Recommendation

````python
def explain_recommendation(profile: RequirementsProfile) -> str:
    top = recommend_architecture(profile, top_k=2)
    if not USE_LIVE_LLM:
        lines = [f"Recommended: {top[0].architecture.value} (score={top[0].weighted_score:.2f})"]
        lines.append(f"Runner-up: {top[1].architecture.value}")
        lines.append(f"Layers: {sorted(ARCHITECTURE_LAYERS[top[0].architecture])}")
        return "\n".join(lines)

    from langchain_openai import ChatOpenAI

    llm = ChatOpenAI(model="gpt-4o-mini", api_key=OPENAI_API_KEY, temperature=0)
    prompt = (
        f"Explain memory architecture choice for {profile.name}. "
        f"Requirements: {profile.as_dict()}. "
        f"Top scores: {[(t.architecture.value, round(t.weighted_score,2)) for t in top]}. "
        f"Layers L0=context, L1=working, L2=thread, L3=LTM, L4=RAG."
    )
    return llm.invoke(prompt).content


print(explain_recommendation(SHOPCO_SUPPORT))
````

## 18. Summary

Choosing memory architecture means matching **layers to requirements**:

1. **Score** your use case (`RequirementsProfile` + `recommend_architecture`)
2. **Validate** with scenario prototypes (policy, pronouns, cross-session)
3. **Pick LTM backend** — LangMem on LangGraph, Mem0 for managed/agnostic
4. **Migrate in phases** — THREAD_RAG → LTM → FULL_STACK + governance

ShopCo Support needs **thread + RAG + long-term** with citations and cross-session preferences. ReleaseFlow runbook Q&A can start with **thread + RAG** only. Use the scorer and scenario suite before committing infrastructure.
