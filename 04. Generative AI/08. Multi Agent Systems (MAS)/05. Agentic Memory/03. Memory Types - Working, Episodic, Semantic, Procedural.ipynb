# Memory Types — Working, Episodic, Semantic, Procedural

Production agent memory is not one blob of text. It is a **typed system** where each kind of information has different **lifetime**, **scope**, and **retrieval rules**.

| Type | Question it answers | Lifetime | ShopCo example |
|------|---------------------|----------|----------------|
| **Working** | What am I doing right now? | Minutes / single thread | "Checking refund eligibility for ORD-1001" |
| **Episodic** | What happened? | Days to years (archive) | "2026-07-01: Alice reported late delivery" |
| **Semantic** | What is true? | Long-term, revised | "Alice prefers email; VIP tier" |
| **Procedural** | How should I act? | Long-term, shared | "Cite return policy before promising refund" |

This notebook implements a **typed ShopCo Memory System** in plain Python. You will see when to write each type, how to recall them in the right order, how episodic events **consolidate** into semantic facts, and how retention policies differ per type.

Everything runs offline.

````python
"""
ShopCo Typed Memory Lab — working, episodic, semantic, procedural.
"""

import json
import re
import uuid
from dataclasses import dataclass, field
from datetime import datetime, timezone, timedelta
from enum import Enum
from typing import Any, Callable

USE_LIVE_LLM = False


def pretty(obj: Any) -> str:
    return json.dumps(obj, indent=2, default=str)


def utcnow() -> datetime:
    return datetime.now(timezone.utc)


ORDERS: dict[str, dict[str, Any]] = {
    "ORD-1001": {"order_id": "ORD-1001", "customer": "alice@shop.com", "status": "shipped", "total_usd": 1299.0},
    "ORD-1002": {"order_id": "ORD-1002", "customer": "bob@shop.com", "status": "processing", "total_usd": 449.0},
}

POLICIES = {"returns": "30-day returns on unused items.", "refunds": "Refunds require policy citation."}

print("ShopCo typed memory lab ready")
````

---

## 1. How the Four Types Compose

```
  User message
       │
       ▼
  ┌─────────────┐   always loaded first
  │ PROCEDURAL  │   "how to behave"
  └──────┬──────┘
         ▼
  ┌─────────────┐   if user identified
  │  SEMANTIC   │   "what we know about them"
  └──────┬──────┘
         ▼
  ┌─────────────┐   relevant past events
  │  EPISODIC   │   "what happened before"
  └──────┬──────┘
         ▼
  ┌─────────────┐   current thread scratchpad
  │   WORKING   │   "in-flight task state"
  └──────┬──────┘
         ▼
     Agent acts → writes working + episodic; may consolidate to semantic
```

Mixing all types into one undifferentiated store makes it impossible to apply the right retention, scope, and recall policy.

---

## 2. Memory Record and Type Enum

````python
class MemoryType(str, Enum):
    WORKING = "working"
    EPISODIC = "episodic"
    SEMANTIC = "semantic"
    PROCEDURAL = "procedural"


@dataclass
class MemoryRecord:
    memory_id: str
    memory_type: MemoryType
    content: str
    user_id: str | None = None
    thread_id: str | None = None
    tags: list[str] = field(default_factory=list)
    salience: float = 1.0
    created_at: datetime = field(default_factory=utcnow)
    expires_at: datetime | None = None

    def is_expired(self, now: datetime | None = None) -> bool:
        if self.expires_at is None:
            return False
        return (now or utcnow()) > self.expires_at

    def to_dict(self) -> dict[str, Any]:
        return {
            "memory_id": self.memory_id,
            "type": self.memory_type.value,
            "content": self.content,
            "user_id": self.user_id,
            "thread_id": self.thread_id,
            "tags": self.tags,
            "salience": self.salience,
            "created_at": self.created_at.isoformat(),
        }


print("MemoryRecord model ready")
````

---

## 3. Working Memory — The Scratchpad

**Working memory** holds in-flight task state: tool results being chained, intermediate conclusions, current order under investigation. It is **thread-scoped** and **short-lived**.

| Property | Value |
|----------|-------|
| Scope | `thread_id` |
| TTL | Minutes to hours |
| Cleared when | Task completes, thread closes |
| Read priority | High within active thread |

````python
class WorkingMemory:
    def __init__(self, store: list[MemoryRecord]):
        self._store = store

    def note(self, thread_id: str, content: str, tags: list[str] | None = None) -> MemoryRecord:
        rec = MemoryRecord(
            memory_id=f"work-{uuid.uuid4().hex[:8]}",
            memory_type=MemoryType.WORKING,
            content=content,
            thread_id=thread_id,
            tags=tags or [],
            salience=0.5,
            expires_at=utcnow() + timedelta(hours=4),
        )
        self._store.append(rec)
        return rec

    def recall(self, thread_id: str) -> list[MemoryRecord]:
        return [r for r in self._store if r.memory_type == MemoryType.WORKING and r.thread_id == thread_id and not r.is_expired()]

    def clear_thread(self, thread_id: str) -> int:
        before = len(self._store)
        self._store[:] = [r for r in self._store if not (r.memory_type == MemoryType.WORKING and r.thread_id == thread_id)]
        return before - len(self._store)


_backend: list[MemoryRecord] = []
working = WorkingMemory(_backend)

working.note("ticket-42", "Step 1: fetched ORD-1001 → shipped", tags=["ORD-1001"])
working.note("ticket-42", "Step 2: checking return policy for refund request")
print("Working notes:", [r.content for r in working.recall("ticket-42")])
````

## 4. Episodic Memory — The Event Log

**Episodic memory** records **events**: what the user said, what tools returned, what was decided. Episodes are timestamped and tagged for later retrieval.

| Property | Value |
|----------|-------|
| Scope | `user_id` (+ optional `thread_id`) |
| TTL | Long archive; may compress old episodes |
| Source | User messages, tool outputs, resolutions |
| Feeds | Semantic consolidation |

````python
class EpisodicMemory:
    def __init__(self, store: list[MemoryRecord]):
        self._store = store

    def record(self, user_id: str, content: str, *, thread_id: str | None = None, tags: list[str] | None = None) -> MemoryRecord:
        rec = MemoryRecord(
            memory_id=f"epi-{uuid.uuid4().hex[:8]}",
            memory_type=MemoryType.EPISODIC,
            content=content,
            user_id=user_id,
            thread_id=thread_id,
            tags=tags or [],
            salience=1.0,
        )
        self._store.append(rec)
        return rec

    def recall(self, user_id: str, query: str = "", limit: int = 5) -> list[MemoryRecord]:
        q = query.lower()
        eps = [r for r in self._store if r.memory_type == MemoryType.EPISODIC and r.user_id == user_id and not r.is_expired()]
        if q:
            eps = [r for r in eps if q in r.content.lower() or any(q in t for t in r.tags)]
        eps.sort(key=lambda r: r.created_at, reverse=True)
        return eps[:limit]


episodic = EpisodicMemory(_backend)
episodic.record("alice@shop.com", "2026-07-01: Reported ORD-1001 delivery 7 days late.", tags=["ORD-1001", "delay"])
episodic.record("alice@shop.com", "2026-07-05: Asked about refund eligibility.", tags=["ORD-1001", "refund"])
print("Episodes for Alice:")
for e in episodic.recall("alice@shop.com", query="refund"):
    print(f"  • {e.content}")
````

## 5. Semantic Memory — Stable Facts

**Semantic memory** stores **beliefs and facts** distilled from episodes and explicit user statements. Facts can be **updated** when new information contradicts old.

| Property | Value |
|----------|-------|
| Scope | `user_id` or global entity |
| TTL | Long-term until revised |
| Updates | Upsert / merge on conflict |
| Example | "VIP tier", "prefers email", "allergic to latex packaging" |

````python
class SemanticMemory:
    def __init__(self, store: list[MemoryRecord]):
        self._store = store

    def upsert(self, user_id: str, fact: str, *, tags: list[str] | None = None, salience: float = 1.5) -> MemoryRecord:
        tag_set = set(tags or [])
        for r in self._store:
            if r.memory_type == MemoryType.SEMANTIC and r.user_id == user_id and set(r.tags) & tag_set:
                r.content = fact
                r.salience = salience
                r.created_at = utcnow()
                return r
        rec = MemoryRecord(
            memory_id=f"sem-{uuid.uuid4().hex[:8]}",
            memory_type=MemoryType.SEMANTIC,
            content=fact,
            user_id=user_id,
            tags=tags or [],
            salience=salience,
        )
        self._store.append(rec)
        return rec

    def recall(self, user_id: str) -> list[MemoryRecord]:
        return [r for r in self._store if r.memory_type == MemoryType.SEMANTIC and r.user_id == user_id]


semantic = SemanticMemory(_backend)
semantic.upsert("alice@shop.com", "Prefers email notifications.", tags=["channel"])
semantic.upsert("alice@shop.com", "VIP tier customer since 2024.", tags=["tier"])
semantic.upsert("alice@shop.com", "Prefers email AND SMS for urgent issues.", tags=["channel"])  # updates channel fact
print("Semantic facts for Alice:")
for s in semantic.recall("alice@shop.com"):
    print(f"  • [{','.join(s.tags)}] {s.content}")
````

## 6. Procedural Memory — How to Behave

**Procedural memory** encodes **rules and playbooks**: org policies, agent SOPs, compliance constraints. Usually **global** (not per-user) and loaded on every turn.

| Property | Value |
|----------|-------|
| Scope | Global / team / agent role |
| TTL | Until policy version changes |
| Priority | Highest at recall — safety first |
| Example | "Never promise refund without policy citation" |

````python
class ProceduralMemory:
    def __init__(self, store: list[MemoryRecord]):
        self._store = store

    def add_rule(self, content: str, *, tags: list[str] | None = None, salience: float = 2.0) -> MemoryRecord:
        rec = MemoryRecord(
            memory_id=f"proc-{uuid.uuid4().hex[:8]}",
            memory_type=MemoryType.PROCEDURAL,
            content=content,
            tags=tags or [],
            salience=salience,
        )
        self._store.append(rec)
        return rec

    def recall(self, tags: list[str] | None = None) -> list[MemoryRecord]:
        rules = [r for r in self._store if r.memory_type == MemoryType.PROCEDURAL]
        if tags:
            rules = [r for r in rules if set(tags) & set(r.tags)]
        rules.sort(key=lambda r: r.salience, reverse=True)
        return rules


procedural = ProceduralMemory(_backend)
procedural.add_rule("Always cite return policy before promising a refund.", tags=["refund", "policy"])
procedural.add_rule("Escalate to human if refund exceeds $1000.", tags=["refund", "escalation"], salience=2.5)
procedural.add_rule("Use customer's preferred contact channel for follow-up.", tags=["channel"])
print("Refund procedures:")
for p in procedural.recall(tags=["refund"]):
    print(f"  • (salience={p.salience}) {p.content}")
````

## 7. TypedMemorySystem — Unified Facade

A facade exposes all four stores with a single `recall_for_turn()` that assembles context in the correct order.

````python
@dataclass
class TurnContext:
    procedural: list[MemoryRecord]
    semantic: list[MemoryRecord]
    episodic: list[MemoryRecord]
    working: list[MemoryRecord]

    def to_prompt_block(self) -> str:
        parts = []
        if self.procedural:
            parts.append("## Procedures\n" + "\n".join(f"- {r.content}" for r in self.procedural))
        if self.semantic:
            parts.append("## User facts\n" + "\n".join(f"- {r.content}" for r in self.semantic))
        if self.episodic:
            parts.append("## Past events\n" + "\n".join(f"- {r.content}" for r in self.episodic))
        if self.working:
            parts.append("## Working notes\n" + "\n".join(f"- {r.content}" for r in self.working))
        return "\n\n".join(parts)


class TypedMemorySystem:
    def __init__(self):
        self._backend: list[MemoryRecord] = []
        self.working = WorkingMemory(self._backend)
        self.episodic = EpisodicMemory(self._backend)
        self.semantic = SemanticMemory(self._backend)
        self.procedural = ProceduralMemory(self._backend)

    def recall_for_turn(self, *, user_id: str | None, thread_id: str, message: str) -> TurnContext:
        proc_tags = ["refund"] if "refund" in message.lower() else None
        oid = re.search(r"ORD-\d+", message.upper())
        epi_query = oid.group() if oid else message[:40]
        return TurnContext(
            procedural=self.procedural.recall(proc_tags),
            semantic=self.semantic.recall(user_id) if user_id else [],
            episodic=self.episodic.recall(user_id, epi_query) if user_id else [],
            working=self.working.recall(thread_id),
        )

    def count_by_type(self) -> dict[str, int]:
        counts: dict[str, int] = {}
        for r in self._backend:
            counts[r.memory_type.value] = counts.get(r.memory_type.value, 0) + 1
        return counts


memory = TypedMemorySystem()
memory.procedural.add_rule("Always cite return policy before promising a refund.", tags=["refund"])
memory.procedural.add_rule("Use preferred contact channel.", tags=["channel"])
memory.semantic.upsert("alice@shop.com", "Prefers email. VIP tier.", tags=["channel", "tier"])
memory.episodic.record("alice@shop.com", "Prior: ORD-1001 late delivery complaint.", tags=["ORD-1001"])
memory.working.note("t-1", "Investigating refund path for ORD-1001")

ctx = memory.recall_for_turn(user_id="alice@shop.com", thread_id="t-1", message="Refund for ORD-1001?")
print(ctx.to_prompt_block())
````

## 8. Consolidation — Episode to Semantic Fact

Humans sleep-consolidate episodes into long-term facts. Agents run an explicit **consolidation** step: extract stable facts from episodic logs and upsert semantic memory.

````python
CONSOLIDATION_RULES: list[tuple[re.Pattern[str], str, list[str]]] = [
    (re.compile(r"prefers? (\w+)", re.I), "Contact preference: {0}.", ["channel"]),
    (re.compile(r"VIP", re.I), "VIP tier customer.", ["tier"]),
    (re.compile(r"(ORD-\d+)", re.I), "Associated order: {0}.", ["order"]),
]


def consolidate_episodes_to_semantic(mem: TypedMemorySystem, user_id: str) -> list[str]:
    extracted: list[str] = []
    for ep in mem.episodic.recall(user_id, limit=20):
        for pattern, template, tags in CONSOLIDATION_RULES:
            m = pattern.search(ep.content)
            if m:
                fact = template.format(*m.groups()) if m.groups() else template
                mem.semantic.upsert(user_id, fact, tags=tags)
                extracted.append(fact)
    return list(dict.fromkeys(extracted))


mem2 = TypedMemorySystem()
mem2.episodic.record("bob@shop.com", "Bob said he prefers SMS for all updates.")
mem2.episodic.record("bob@shop.com", "Bob asked about ORD-1002 status.")
facts = consolidate_episodes_to_semantic(mem2, "bob@shop.com")
print("Consolidated facts:", facts)
print("Bob semantic store:", [r.content for r in mem2.semantic.recall("bob@shop.com")])
````

## 9. ShopCo Support Agent — All Four Types in One Turn

A complete turn: recall all types → act → write working + episodic → optionally consolidate.

````python
class ShopCoMemoryAgent:
    def __init__(self, mem: TypedMemorySystem):
        self.mem = mem

    def turn(self, message: str, *, user_id: str, thread_id: str) -> dict[str, Any]:
        ctx = self.mem.recall_for_turn(user_id=user_id, thread_id=thread_id, message=message)
        reply_parts: list[str] = []

        oid_match = re.search(r"ORD-\d+", message.upper())
        if oid_match:
            oid = oid_match.group()
            order = ORDERS.get(oid)
            if order:
                self.mem.working.note(thread_id, f"Looked up {oid}: status={order['status']}")
                reply_parts.append(f"{oid} is {order['status']} (${order['total_usd']}).")
            else:
                reply_parts.append(f"Order {oid} not found.")

        if "refund" in message.lower():
            for rule in ctx.procedural:
                if "policy" in rule.content.lower():
                    reply_parts.append(f"Policy reminder: {POLICIES['returns']}")
            order = ORDERS.get((oid_match or re.search(r"ORD-\d+", "ORD-1001")).group())
            if order and order["total_usd"] > 1000:
                reply_parts.append("Escalation required: refund exceeds $1000 threshold.")

        for fact in ctx.semantic:
            if "email" in fact.content.lower():
                reply_parts.append("I'll follow up via email.")
                break

        if ctx.episodic and "delay" in ctx.episodic[0].content.lower():
            reply_parts.append("I see you reported a prior delivery delay.")

        self.mem.episodic.record(user_id, f"{utcnow().date()}: User asked: {message[:60]}", thread_id=thread_id, tags=["interaction"])

        if not reply_parts:
            reply_parts.append("How can I help?")

        return {
            "reply": " ".join(reply_parts),
            "memories_recalled": sum(len(getattr(ctx, k)) for k in ("procedural", "semantic", "episodic", "working")),
            "context_block": ctx.to_prompt_block(),
        }


agent = ShopCoMemoryAgent(memory)
result = agent.turn("I need a refund for ORD-1001 please", user_id="alice@shop.com", thread_id="t-1")
print("Reply:", result["reply"])
print(f"Recalled {result['memories_recalled']} memories")
````

## 10. Retention Policies Per Type

| Type | Default retention | Eviction strategy |
|------|-------------------|-------------------|
| Working | 4 hours TTL | Clear on thread close |
| Episodic | Indefinite → summarize after 90 days | Archive + compress |
| Semantic | Until explicitly updated | Upsert on conflict |
| Procedural | Until policy version bump | Version tag on each rule |

````python
def purge_expired(store: list[MemoryRecord]) -> int:
    now = utcnow()
    before = len(store)
    store[:] = [r for r in store if not r.is_expired(now)]
    return before - len(store)


demo_store: list[MemoryRecord] = []
wm = WorkingMemory(demo_store)
wm.note("temp-thread", "Short-lived note")
demo_store[-1].expires_at = utcnow() - timedelta(seconds=1)  # force expire
removed = purge_expired(demo_store)
print(f"Purged {removed} expired working memory record(s)")
print("Episodic/semantic/procedural typically have no expires_at — governed by archive policy instead.")
````

## 11. Task Completion — Clear Working, Keep the Rest

When a ticket closes, flush working memory but keep episodic and semantic for future sessions.

````python
def close_ticket(mem: TypedMemorySystem, *, user_id: str, thread_id: str, resolution: str) -> dict[str, int]:
    mem.episodic.record(user_id, f"Ticket closed: {resolution}", thread_id=thread_id, tags=["resolution"])
    cleared = mem.working.clear_thread(thread_id)
    return {"working_cleared": cleared, "remaining": mem.count_by_type()}


stats = close_ticket(memory, user_id="alice@shop.com", thread_id="t-1", resolution="Refund escalated to supervisor")
print(pretty(stats))
````

## 12. Type Selection Guide — When to Write What

| Observation | Write to |
|-------------|----------|
| Tool returned order status | **Working** (in-flight) + **Episodic** (audit) |
| User said "I prefer email" | **Semantic** (durable fact) |
| User asked about refund today | **Episodic** (event) |
| Org policy loaded at startup | **Procedural** (rule) |
| Intermediate multi-step plan | **Working** only |
| Resolved ticket summary | **Episodic** (+ consolidate key facts to **Semantic**) |

````python
def recommend_write_type(observation: str) -> list[MemoryType]:
    obs = observation.lower()
    types: list[MemoryType] = []
    if any(w in obs for w in ("step", "checking", "looked up", "tool returned")):
        types.append(MemoryType.WORKING)
    if any(w in obs for w in ("said", "asked", "reported", "today", "resolved")):
        types.append(MemoryType.EPISODIC)
    if any(w in obs for w in ("prefers", "vip", "always", "never", "tier")):
        types.append(MemoryType.SEMANTIC)
    if any(w in obs for w in ("policy", "must", "escalate", "cite", "rule")):
        types.append(MemoryType.PROCEDURAL)
    return types or [MemoryType.EPISODIC]


samples = [
    "Tool returned ORD-1001 status=shipped",
    "User said she prefers email",
    "Org rule: cite policy before refund",
    "Step 2: checking eligibility",
]
for s in samples:
    print(f"{s!r} → {[t.value for t in recommend_write_type(s)]}")
````

## 13. Multi-User Isolation

Semantic and episodic memories are **user-scoped**. Procedural is shared. Working is **thread-scoped**.

````python
iso_mem = TypedMemorySystem()
iso_mem.semantic.upsert("alice@shop.com", "Prefers email.", tags=["channel"])
iso_mem.semantic.upsert("bob@shop.com", "Prefers SMS.", tags=["channel"])
iso_mem.procedural.add_rule("Be polite.", tags=["tone"])

for uid in ("alice@shop.com", "bob@shop.com"):
    ctx = iso_mem.recall_for_turn(user_id=uid, thread_id="shared-wrong", message="contact preference")
    print(f"{uid}: semantic={[r.content for r in ctx.semantic]}, procedural={len(ctx.procedural)}")
````

## 14. Anti-Patterns

| Anti-pattern | Problem | Fix |
|--------------|---------|-----|
| One store for everything | Wrong retention and recall | Typed stores with policies |
| Procedural facts in semantic | User-specific rules mixed with prefs | Separate procedural layer |
| Never clearing working | Stale scratchpad misleads agent | Clear on task/thread close |
| Episodes without consolidation | Repeated questions every session | Extract semantic facts periodically |
| Semantic without upsert | Duplicate conflicting facts | Tag-based merge |
| Skip procedural recall | Policy violations | Always load rules first |

## 17. Summary

Agentic memory works best when types are explicit:

- **Working** — scratchpad for the current thread; short TTL; clear on close
- **Episodic** — timestamped events; feeds consolidation and audit
- **Semantic** — stable user facts; upsert on change
- **Procedural** — global rules; highest recall priority

Recall in order: procedural → semantic → episodic → working. Write working and episodic during the turn; consolidate episodes into semantic facts over time. The ShopCo agent demonstrated all four types cooperating on a refund request while respecting escalation policy and Alice's contact preference.
