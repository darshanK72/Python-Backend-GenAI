# Hot-Path vs Background Memory Writes

When an agent learns something during a turn, **when** you persist it matters as much as **what** you store.

- **Hot-path writes** happen **synchronously** during the agent turn — before the user sees the reply. They block latency but guarantee the **next turn in the same thread** can read what was just written.
- **Background writes** are **queued** and processed after the reply ships — consolidation, embedding, summarization, analytics. They reduce user-facing latency but are **eventually consistent**.

This notebook implements a **ShopCo Support Write Pipeline** in plain Python:

1. A `MemoryWriteRouter` that classifies writes as hot or background
2. Hot-path store for working memory and critical episodes
3. Background queue with worker, retries, and failure handling
4. Latency comparison on the same support scenario
5. Read-your-writes guarantees within a thread vs cross-session eventual consistency

Everything runs offline.

````python
"""
ShopCo Write Pipeline Lab — hot-path vs background memory writes.
"""

import json
import time
import uuid
from collections import deque
from dataclasses import dataclass, field
from datetime import datetime, timezone
from enum import Enum
from typing import Any, Callable


def pretty(obj: Any) -> str:
    return json.dumps(obj, indent=2, default=str)


def utcnow() -> str:
    return datetime.now(timezone.utc).isoformat()


ORDERS: dict[str, dict[str, Any]] = {
    "ORD-1001": {"order_id": "ORD-1001", "customer": "alice@shop.com", "status": "shipped", "total_usd": 1299.0},
    "ORD-1002": {"order_id": "ORD-1002", "customer": "bob@shop.com", "status": "processing", "total_usd": 449.0},
}

WRITE_METRICS: list[dict[str, Any]] = []

print("ShopCo write pipeline lab ready")
````

---

## 1. Two Write Paths

```
  User message
       │
       ▼
  Agent turn ─────────────────────────────────────┐
       │                                          │
       ├── HOT PATH (sync, before reply)          │
       │     working memory, thread episodes      │
       │                                          │
       ├── Reply to user ◀────────────────────────┘
       │
       └── BACKGROUND (async queue, after reply)
             semantic consolidation, embeddings,
             episode archive, analytics
```

| Path | Latency impact | Consistency | Examples |
|------|----------------|-------------|----------|
| **Hot** | Adds to user wait time | Read-your-writes next turn | Working scratchpad, "user said X" this thread |
| **Background** | None on reply | Eventual (seconds later) | Extract VIP fact, index for search, nightly summary |

---

## 2. Write Job Model

````python
class MemoryKind(str, Enum):
    WORKING = "working"
    EPISODIC = "episodic"
    SEMANTIC = "semantic"
    PROCEDURAL = "procedural"


class WritePath(str, Enum):
    HOT = "hot"
    BACKGROUND = "background"


@dataclass
class MemoryWriteJob:
    job_id: str
    path: WritePath
    kind: MemoryKind
    content: str
    user_id: str | None = None
    thread_id: str | None = None
    tags: list[str] = field(default_factory=list)
    simulate_ms: float = 5.0  # simulated I/O latency
    attempts: int = 0
    max_attempts: int = 3

    def to_dict(self) -> dict[str, Any]:
        return {"job_id": self.job_id, "path": self.path.value, "kind": self.kind.value, "content": self.content[:60]}


print("Write job model ready")
````

---

## 3. MemoryWriteRouter — Classify Hot vs Background

Rules of thumb:

- **Hot**: working memory; episodic events needed **this thread** on the next turn; anything the reply depends on
- **Background**: semantic consolidation; embedding/index updates; cross-session analytics; bulk archive

````python
class MemoryWriteRouter:
    def route(self, kind: MemoryKind, *, same_thread_next_turn: bool = False) -> WritePath:
        if kind == MemoryKind.WORKING:
            return WritePath.HOT
        if kind == MemoryKind.EPISODIC and same_thread_next_turn:
            return WritePath.HOT
        if kind == MemoryKind.PROCEDURAL:
            return WritePath.HOT  # rare; usually loaded at startup
        if kind == MemoryKind.SEMANTIC:
            return WritePath.BACKGROUND
        if kind == MemoryKind.EPISODIC:
            return WritePath.BACKGROUND  # archive episode for long-term
        return WritePath.BACKGROUND


router = MemoryWriteRouter()
cases = [
    (MemoryKind.WORKING, False),
    (MemoryKind.EPISODIC, True),
    (MemoryKind.EPISODIC, False),
    (MemoryKind.SEMANTIC, False),
]
for kind, same_thread in cases:
    path = router.route(kind, same_thread_next_turn=same_thread)
    print(f"  {kind.value} (same_thread={same_thread}) → {path.value}")
````

---

## 4. In-Memory Stores

````python
@dataclass
class StoredMemory:
    memory_id: str
    kind: MemoryKind
    content: str
    user_id: str | None
    thread_id: str | None
    written_at: str


class MemoryStore:
    def __init__(self):
        self._records: list[StoredMemory] = []

    def write(self, job: MemoryWriteJob) -> StoredMemory:
        time.sleep(job.simulate_ms / 1000.0)
        rec = StoredMemory(
            memory_id=job.job_id,
            kind=job.kind,
            content=job.content,
            user_id=job.user_id,
            thread_id=job.thread_id,
            written_at=utcnow(),
        )
        self._records.append(rec)
        return rec

    def search(self, *, kind: MemoryKind | None = None, thread_id: str | None = None, user_id: str | None = None) -> list[StoredMemory]:
        out = self._records
        if kind:
            out = [r for r in out if r.kind == kind]
        if thread_id:
            out = [r for r in out if r.thread_id == thread_id]
        if user_id:
            out = [r for r in out if r.user_id == user_id]
        return out

    def count(self) -> int:
        return len(self._records)


hot_store = MemoryStore()
bg_store = MemoryStore()
print("Hot and background stores ready")
````

---

## 5. Hot-Path Writer — Synchronous

Hot writes run inline. User latency includes every hot write's simulated I/O.

````python
class HotPathWriter:
    def __init__(self, store: MemoryStore):
        self.store = store

    def execute(self, job: MemoryWriteJob) -> tuple[StoredMemory, float]:
        t0 = time.perf_counter()
        rec = self.store.write(job)
        elapsed_ms = (time.perf_counter() - t0) * 1000
        WRITE_METRICS.append({"path": "hot", "kind": job.kind.value, "latency_ms": round(elapsed_ms, 2)})
        return rec, elapsed_ms


hot_writer = HotPathWriter(hot_store)
job = MemoryWriteJob(str(uuid.uuid4())[:8], WritePath.HOT, MemoryKind.WORKING, "Fetched ORD-1001", thread_id="t-1", simulate_ms=8)
rec, ms = hot_writer.execute(job)
print(f"Hot write completed in {ms:.1f}ms: {rec.content}")
````

## 6. Background Queue and Worker

Background jobs enqueue instantly. A worker drains the queue **after** the reply — simulating Celery, SQS, or an async task.

````python
@dataclass
class QueueStats:
    enqueued: int = 0
    processed: int = 0
    failed: int = 0
    retried: int = 0


class BackgroundWriteQueue:
    def __init__(self, store: MemoryStore):
        self.store = store
        self._queue: deque[MemoryWriteJob] = deque()
        self.stats = QueueStats()
        self._fail_next: set[str] = set()

    def enqueue(self, job: MemoryWriteJob) -> str:
        self._queue.append(job)
        self.stats.enqueued += 1
        return job.job_id

    def drain(self) -> list[StoredMemory]:
        """Process all pending jobs — call after reply sent."""
        results: list[StoredMemory] = []
        while self._queue:
            job = self._queue.popleft()
            job.attempts += 1
            try:
                if job.job_id in self._fail_next:
                    self._fail_next.discard(job.job_id)
                    raise ConnectionError("simulated DB timeout")
                t0 = time.perf_counter()
                rec = self.store.write(job)
                elapsed_ms = (time.perf_counter() - t0) * 1000
                WRITE_METRICS.append({"path": "background", "kind": job.kind.value, "latency_ms": round(elapsed_ms, 2)})
                results.append(rec)
                self.stats.processed += 1
            except Exception:
                if job.attempts < job.max_attempts:
                    self._queue.append(job)
                    self.stats.retried += 1
                else:
                    self.stats.failed += 1
        return results

    def pending(self) -> int:
        return len(self._queue)


bg_queue = BackgroundWriteQueue(bg_store)
bg_queue.enqueue(MemoryWriteJob("j1", WritePath.BACKGROUND, MemoryKind.SEMANTIC, "Alice VIP tier", user_id="alice@shop.com", simulate_ms=15))
bg_queue.enqueue(MemoryWriteJob("j2", WritePath.BACKGROUND, MemoryKind.SEMANTIC, "Index ORD-1001 for search", tags=["index"], simulate_ms=20))
print(f"Queued {bg_queue.pending()} jobs (user reply not blocked yet)")
processed = bg_queue.drain()
print(f"Worker processed {len(processed)} jobs. Background store count: {bg_store.count()}")
````

---

## 7. Unified WritePipeline

The pipeline routes each job, executes hot writes immediately, and enqueues background writes.

````python
@dataclass
class TurnWriteResult:
    hot_writes: int
    hot_latency_ms: float
    background_enqueued: int
    reply: str


class WritePipeline:
    def __init__(self, router: MemoryWriteRouter, hot: HotPathWriter, bg: BackgroundWriteQueue):
        self.router = router
        self.hot = hot
        self.bg = bg

    def submit(self, job: MemoryWriteJob, *, same_thread_next_turn: bool = False) -> WritePath:
        path = self.router.route(job.kind, same_thread_next_turn=same_thread_next_turn)
        job.path = path
        if path == WritePath.HOT:
            self.hot.execute(job)
        else:
            self.bg.enqueue(job)
        return path

    def flush_background(self) -> int:
        return len(self.bg.drain())


pipeline = WritePipeline(router, hot_writer, bg_queue)
print("WritePipeline ready")
````

---

## 8. ShopCo Agent Turn — Split Writes

One support turn: hot writes for working state; background for semantic consolidation.

````python
class ShopCoWriteAgent:
    def __init__(self, pipeline: WritePipeline, hot_store: MemoryStore):
        self.pipeline = pipeline
        self.hot_store = hot_store

    def turn(self, message: str, *, user_id: str, thread_id: str) -> TurnWriteResult:
        t0 = time.perf_counter()
        hot_count = 0
        bg_count = 0

        # Hot: working note for multi-step flow in this thread
        self.pipeline.submit(MemoryWriteJob(
            str(uuid.uuid4())[:8], WritePath.HOT, MemoryKind.WORKING,
            f"Processing: {message[:50]}", user_id=user_id, thread_id=thread_id, simulate_ms=6,
        ))
        hot_count += 1

        order_id = None
        for oid in ("ORD-1001", "ORD-1002"):
            if oid in message.upper():
                order_id = oid
                break
        if order_id:
            order = ORDERS[order_id]
            self.pipeline.submit(MemoryWriteJob(
                str(uuid.uuid4())[:8], WritePath.HOT, MemoryKind.WORKING,
                f"{order_id} status={order['status']}", thread_id=thread_id, simulate_ms=5,
            ))
            hot_count += 1
            # Hot episodic — needed if user asks follow-up in same thread
            self.pipeline.submit(MemoryWriteJob(
                str(uuid.uuid4())[:8], WritePath.HOT, MemoryKind.EPISODIC,
                f"User asked about {order_id}", user_id=user_id, thread_id=thread_id, simulate_ms=7,
            ), same_thread_next_turn=True)
            hot_count += 1
            reply = f"{order_id} is {order['status']} (${order['total_usd']})."
        else:
            reply = "Please share your order ID."

        if "prefer" in message.lower() or "email" in message.lower():
            # Background: semantic fact — OK if available next session, not next second
            self.pipeline.submit(MemoryWriteJob(
                str(uuid.uuid4())[:8], WritePath.BACKGROUND, MemoryKind.SEMANTIC,
                "User prefers email contact.", user_id=user_id, simulate_ms=25,
            ))
            bg_count += 1
            reply += " Noted your contact preference."

        # Background: archive full interaction for analytics
        self.pipeline.submit(MemoryWriteJob(
            str(uuid.uuid4())[:8], WritePath.BACKGROUND, MemoryKind.EPISODIC,
            f"Interaction log: {message[:80]}", user_id=user_id, simulate_ms=12,
        ))
        bg_count += 1

        hot_latency = (time.perf_counter() - t0) * 1000
        return TurnWriteResult(hot_count, hot_latency, bg_count, reply)


agent = ShopCoWriteAgent(pipeline, hot_store)
turn1 = agent.turn("Check ORD-1001 — I prefer email updates", user_id="alice@shop.com", thread_id="thread-99")
print(f"Reply: {turn1.reply}")
print(f"Hot writes: {turn1.hot_writes} ({turn1.hot_latency_ms:.1f}ms user wait)")
print(f"Background queued: {turn1.background_enqueued} (not in user wait)")
print(f"Background still pending: {bg_queue.pending()}")
````

## 9. Read-Your-Writes vs Eventual Consistency

**Same thread, next turn:** hot working memory is visible immediately.

**Next session:** background semantic may not be flushed yet — demonstrate flush timing.

````python
# Turn 2 — same thread, before background flush
working_visible = hot_store.search(kind=MemoryKind.WORKING, thread_id="thread-99")
semantic_visible = bg_store.search(kind=MemoryKind.SEMANTIC, user_id="alice@shop.com")
print("Before background flush:")
print(f"  Working (hot) visible: {len(working_visible)} records")
print(f"  Semantic (bg) visible: {len(semantic_visible)} records")

n = pipeline.flush_background()
semantic_after = bg_store.search(kind=MemoryKind.SEMANTIC, user_id="alice@shop.com")
print(f"\nAfter flush ({n} jobs):")
print(f"  Semantic visible: {len(semantic_after)} — {[r.content for r in semantic_after]}")
````

## 10. Latency Comparison — All-Hot vs Split

If every write were hot-path, user latency sums all I/O. Splitting moves slow writes off the critical path.

````python
def simulate_all_hot(n_writes: int = 5, ms_each: float = 15) -> float:
    t0 = time.perf_counter()
    for _ in range(n_writes):
        time.sleep(ms_each / 1000.0)
    return (time.perf_counter() - t0) * 1000


def simulate_split(n_hot: int = 3, n_bg: int = 2, ms: float = 15) -> tuple[float, float]:
    t0 = time.perf_counter()
    for _ in range(n_hot):
        time.sleep(ms / 1000.0)
    user_wait = (time.perf_counter() - t0) * 1000
    t1 = time.perf_counter()
    for _ in range(n_bg):
        time.sleep(ms / 1000.0)
    bg_time = (time.perf_counter() - t1) * 1000
    return user_wait, bg_time


all_hot_ms = simulate_all_hot(5, 15)
user_ms, bg_ms = simulate_split(3, 2, 15)
print(f"All-hot user wait: {all_hot_ms:.0f}ms")
print(f"Split — user wait: {user_ms:.0f}ms, background: {bg_ms:.0f}ms (off critical path)")
print(f"User latency saved: {all_hot_ms - user_ms:.0f}ms ({100*(all_hot_ms-user_ms)/all_hot_ms:.0f}%)")
````

## 11. Background Failure and Retry

Background workers must retry transient failures. Hot-path failures are worse — they block the user and need immediate fallback.

````python
retry_queue = BackgroundWriteQueue(MemoryStore())
fail_job = MemoryWriteJob("fail-1", WritePath.BACKGROUND, MemoryKind.SEMANTIC, "VIP fact", simulate_ms=5)
retry_queue._fail_next.add("fail-1")
retry_queue.enqueue(fail_job)
retry_queue.drain()  # attempt 1 fails → requeued
print(f"After 1st drain: pending={retry_queue.pending()}, retried={retry_queue.stats.retried}")
retry_queue.drain()  # attempt 2 succeeds
print(f"After 2nd drain: processed={retry_queue.stats.processed}, failed={retry_queue.stats.failed}")
````

## 12. When to Choose Each Path

| Write | Path | Reason |
|-------|------|--------|
| Tool result for next step in thread | **Hot** | Next model call needs it |
| "User said X" this conversation | **Hot** episodic | Follow-up in same thread |
| Extract "prefers email" to profile | **Background** semantic | Next session is fine |
| Embedding / vector index update | **Background** | Search can lag seconds |
| Nightly episode summarization | **Background** | Batch job |
| Compliance audit log | **Hot** or **sync durable queue** | Must not lose before ACK |

## 13. Anti-Pattern: Background Everything

Queuing **all** writes breaks read-your-writes within the thread.

````python
class BackgroundOnlyAgent:
    def __init__(self, bg: BackgroundWriteQueue):
        self.bg = bg

    def turn(self, message: str, thread_id: str) -> str:
        self.bg.enqueue(MemoryWriteJob("w1", WritePath.BACKGROUND, MemoryKind.WORKING, f"Note: {message}", thread_id=thread_id))
        return "OK"

    def next_turn_recall(self, thread_id: str, store: MemoryStore) -> bool:
        return len(store.search(kind=MemoryKind.WORKING, thread_id=thread_id)) > 0


demo_store = MemoryStore()
demo_bg = BackgroundWriteQueue(demo_store)
bad_agent = BackgroundOnlyAgent(demo_bg)
bad_agent.turn("ORD-1001 status?", "t-bad")
recall_before_flush = bad_agent.next_turn_recall("t-bad", demo_store)
demo_bg.drain()
recall_after_flush = bad_agent.next_turn_recall("t-bad", demo_store)
print(f"Recall before flush: {recall_before_flush}  ← broken same-thread continuity")
print(f"Recall after flush:  {recall_after_flush}")
````

## 14. Write Metrics Dashboard

````python
def summarize_metrics(metrics: list[dict[str, Any]]) -> dict[str, Any]:
    hot = [m["latency_ms"] for m in metrics if m["path"] == "hot"]
    bg = [m["latency_ms"] for m in metrics if m["path"] == "background"]
    return {
        "hot_writes": len(hot),
        "hot_avg_ms": round(sum(hot) / len(hot), 2) if hot else 0,
        "background_writes": len(bg),
        "background_avg_ms": round(sum(bg) / len(bg), 2) if bg else 0,
    }


print(pretty(summarize_metrics(WRITE_METRICS)))
````

## 15. Anti-Patterns

| Anti-pattern | Problem | Fix |
|--------------|---------|-----|
| All writes hot | High p99 latency | Route consolidation to background |
| All writes background | Broken same-thread recall | Hot working + thread episodic |
| No retry on background | Silent memory loss | Exponential backoff + DLQ |
| Flush background before reply | Defeats the purpose | Flush after response sent |
| Huge hot writes | Blocks user on embedding | Index in background only |
| No metrics split | Can't tune path rules | Tag latency by hot vs bg |

## 17. Summary

Memory writes are not uniform:

- **Hot-path** — synchronous, on the user latency budget; working memory and thread-critical episodes
- **Background** — queued after reply; semantic consolidation, indexing, archive
- **Router** — classifies by memory kind and whether the next turn in the same thread needs the data
- **Consistency** — read-your-writes for hot store; eventual for background until worker drains

The ShopCo pipeline kept user wait to hot writes only (~18ms simulated) while semantic preference and interaction archive processed in the background — saving latency without breaking same-thread continuity.
