# Introduction to Agentic Memory

A stateless LLM call only knows what fits in the **context window** right now. An **agent with memory** can recall past conversations, learned user preferences, resolved tickets, and procedural rules — across sessions, tools, and time.

**Agentic memory** is the deliberate system that lets agents **read** what they need before acting and **write** what they learn after acting. It is not the same as dumping chat history into the prompt or hoping the model "remembers."

This notebook introduces memory concepts and implements a **ShopCo Support Memory Lab** in plain Python:

- Four memory types: **working**, **episodic**, **semantic**, and **procedural**
- Explicit **read** and **write** paths around each agent turn
- Thread-scoped and user-scoped recall
- A multi-turn support demo where the agent remembers preferences and past resolutions

Everything runs offline. Set `USE_LIVE_LLM = True` at the end only when you want a real model to synthesize replies from recalled memories.

````python
"""
ShopCo Support Memory Lab — introduction to agentic memory.
"""

import json
import re
import uuid
from dataclasses import dataclass, field
from datetime import datetime, timezone
from enum import Enum
from typing import Any

USE_LIVE_LLM = False


def pretty(obj: Any) -> str:
    return json.dumps(obj, indent=2, default=str)


def utcnow() -> str:
    return datetime.now(timezone.utc).isoformat()


ORDERS: dict[str, dict[str, Any]] = {
    "ORD-1001": {"order_id": "ORD-1001", "customer": "alice@shop.com", "status": "shipped", "total_usd": 1299.0},
    "ORD-1002": {"order_id": "ORD-1002", "customer": "bob@shop.com", "status": "processing", "total_usd": 449.0},
}

MEMORY_AUDIT: list[dict[str, Any]] = []

print("ShopCo memory lab ready")
````

---

## 1. The Memory Problem

Without memory, every support turn starts from zero:

```
Turn 1: "I'm Alice, I prefer email updates."
Turn 2: "What's my order status?"  → Agent: "What's your order ID?" (forgot Alice, forgot preference)
```

With agentic memory:

```
Turn 1: write semantic memory → user=alice, prefers=email
Turn 2: read memories → recall Alice + ORD-1001 from prior episodic turn
        → Agent: "ORD-1001 is shipped. I'll email you updates, per your preference."
```

| Stateless agent | Agent with memory |
|-----------------|-------------------|
| Context = current message only | Context = message + recalled memories |
| No learning across sessions | Persists facts, events, procedures |
| Repeats questions | Builds on prior interactions |
| Token cost grows if you paste full history | Retrieves only relevant slices |

---

## 2. Context Window ≠ Memory

The **context window** is a finite buffer the model sees **right now**. **Memory** is a managed store **outside** the model that you selectively inject into that buffer.

```
  ┌─────────────────────────────────────────┐
  │  Context window (this inference)        │
  │  system + recalled memories + messages  │
  └─────────────────────────────────────────┘
                    ▲
                    │ read (retrieve relevant items)
  ┌─────────────────┴───────────────────────┐
  │  Memory stores (persistent)             │
  │  working | episodic | semantic | proc.  │
  └─────────────────────────────────────────┘
                    ▲
                    │ write (after tool results / user facts)
              Agent turn completes
```

Pasting the entire chat log into context is a brittle substitute for memory: it burns tokens, dilutes attention, and still disappears when the session ends unless you persist it deliberately.

---

## 3. Four Memory Types (Overview)

| Type | What it holds | Lifetime | ShopCo example |
|------|---------------|----------|----------------|
| **Working** | Scratchpad for current task | Single turn / short thread | "Currently checking ORD-1001 refund eligibility" |
| **Episodic** | Events that happened | Session + long-term archive | "2026-07-01: Alice asked about late delivery" |
| **Semantic** | Facts and beliefs | Long-term per user | "Alice prefers email; VIP tier" |
| **Procedural** | How the agent should behave | Long-term, shared | "Always cite return policy before promising refund" |

Production systems often map these to different stores (Redis, vector DB, graph, config files). This lab keeps them in one `MemoryStore` with type tags for clarity.

---

## 4. Memory Record Model

Every memory item is a structured record with scope (who it belongs to), type, content, and metadata for retrieval.

````python
class MemoryType(str, Enum):
    WORKING = "working"
    EPISODIC = "episodic"
    SEMANTIC = "semantic"
    PROCEDURAL = "procedural"


@dataclass
class MemoryRecord:
    memory_id: str
    memory_type: MemoryType
    content: str
    user_id: str | None = None
    thread_id: str | None = None
    tags: list[str] = field(default_factory=list)
    created_at: str = field(default_factory=utcnow)
    salience: float = 1.0  # higher = more important at recall time

    def to_dict(self) -> dict[str, Any]:
        return {
            "memory_id": self.memory_id,
            "type": self.memory_type.value,
            "content": self.content,
            "user_id": self.user_id,
            "thread_id": self.thread_id,
            "tags": self.tags,
            "created_at": self.created_at,
            "salience": self.salience,
        }


sample = MemoryRecord(
    memory_id="mem-001",
    memory_type=MemoryType.SEMANTIC,
    content="Customer prefers email over SMS for notifications.",
    user_id="alice@shop.com",
    tags=["preference", "channel"],
)
print(pretty(sample.to_dict()))
````

---

## 5. MemoryStore — Write, Read, Search

The store supports **scoped queries**: by user, thread, type, and keyword overlap. Production systems add embeddings; here we use tag + keyword matching so the mechanics stay visible.

````python
class MemoryStore:
    def __init__(self):
        self._records: dict[str, MemoryRecord] = {}

    def write(self, record: MemoryRecord) -> str:
        self._records[record.memory_id] = record
        MEMORY_AUDIT.append({"op": "write", "memory_id": record.memory_id, "type": record.memory_type.value, "ts": utcnow()})
        return record.memory_id

    def get(self, memory_id: str) -> MemoryRecord | None:
        return self._records.get(memory_id)

    def search(
        self,
        *,
        user_id: str | None = None,
        thread_id: str | None = None,
        memory_type: MemoryType | None = None,
        query: str = "",
        limit: int = 5,
    ) -> list[MemoryRecord]:
        MEMORY_AUDIT.append({"op": "search", "user_id": user_id, "type": memory_type.value if memory_type else None, "query": query, "ts": utcnow()})
        q = query.lower()
        results: list[MemoryRecord] = []
        for rec in self._records.values():
            if user_id and rec.user_id and rec.user_id != user_id:
                continue
            if thread_id and rec.thread_id and rec.thread_id != thread_id:
                continue
            if memory_type and rec.memory_type != memory_type:
                continue
            if q and q not in rec.content.lower() and not any(q in t for t in rec.tags):
                continue
            results.append(rec)
        results.sort(key=lambda r: r.salience, reverse=True)
        return results[:limit]

    def clear_working(self, thread_id: str) -> int:
        to_remove = [mid for mid, r in self._records.items() if r.memory_type == MemoryType.WORKING and r.thread_id == thread_id]
        for mid in to_remove:
            del self._records[mid]
        return len(to_remove)

    def count_by_type(self) -> dict[str, int]:
        counts: dict[str, int] = {}
        for r in self._records.values():
            counts[r.memory_type.value] = counts.get(r.memory_type.value, 0) + 1
        return counts


store = MemoryStore()
print("MemoryStore initialized")
````

## 6. Seeding Memory Types

Bootstrap procedural rules (shared), semantic facts (per user), and an episodic event from a prior session.

````python
def seed_shopco_memory(store: MemoryStore) -> None:
    store.write(MemoryRecord(
        memory_id="proc-01", memory_type=MemoryType.PROCEDURAL,
        content="Always cite return policy before promising a refund.",
        tags=["policy", "refund"], salience=2.0,
    ))
    store.write(MemoryRecord(
        memory_id="proc-02", memory_type=MemoryType.PROCEDURAL,
        content="Use the customer's preferred contact channel when offering follow-up.",
        tags=["preference", "channel"], salience=1.5,
    ))
    store.write(MemoryRecord(
        memory_id="sem-alice-01", memory_type=MemoryType.SEMANTIC,
        content="Alice (alice@shop.com) prefers email notifications. VIP tier customer.",
        user_id="alice@shop.com", tags=["preference", "vip"], salience=1.8,
    ))
    store.write(MemoryRecord(
        memory_id="sem-bob-01", memory_type=MemoryType.SEMANTIC,
        content="Bob (bob@shop.com) prefers SMS. Standard tier.",
        user_id="bob@shop.com", tags=["preference"], salience=1.0,
    ))
    store.write(MemoryRecord(
        memory_id="epi-alice-01", memory_type=MemoryType.EPISODIC,
        content="Prior session: Alice asked about ORD-1001 delivery delay on 2026-07-01.",
        user_id="alice@shop.com", tags=["order", "ORD-1001", "delay"], salience=1.2,
    ))


seed_shopco_memory(store)
print("Seeded memory counts:", store.count_by_type())
````

## 7. The Read Path — Recall Before Respond

Before the model generates a reply, the agent **reads** memories: procedural rules (always), semantic facts (for identified user), episodic events (relevant history), and working scratchpad (current thread).

````python
@dataclass
class RecallBundle:
    procedural: list[MemoryRecord]
    semantic: list[MemoryRecord]
    episodic: list[MemoryRecord]
    working: list[MemoryRecord]

    def to_context_block(self) -> str:
        sections = []
        if self.procedural:
            sections.append("## Procedures\n" + "\n".join(f"- {r.content}" for r in self.procedural))
        if self.semantic:
            sections.append("## User facts\n" + "\n".join(f"- {r.content}" for r in self.semantic))
        if self.episodic:
            sections.append("## Past events\n" + "\n".join(f"- {r.content}" for r in self.episodic))
        if self.working:
            sections.append("## Working notes\n" + "\n".join(f"- {r.content}" for r in self.working))
        return "\n\n".join(sections)


class MemoryReader:
    def __init__(self, store: MemoryStore):
        self.store = store

    def recall(self, *, user_id: str | None, thread_id: str, user_message: str) -> RecallBundle:
        query = user_message[:80]
        oid = re.search(r"ORD-\d+", user_message.upper())
        extra_query = oid.group() if oid else ""
        return RecallBundle(
            procedural=self.store.search(memory_type=MemoryType.PROCEDURAL, limit=5),
            semantic=self.store.search(user_id=user_id, memory_type=MemoryType.SEMANTIC, limit=3) if user_id else [],
            episodic=self.store.search(user_id=user_id, memory_type=MemoryType.EPISODIC, query=extra_query or query, limit=3) if user_id else [],
            working=self.store.search(thread_id=thread_id, memory_type=MemoryType.WORKING, limit=5),
        )


reader = MemoryReader(store)
bundle = reader.recall(user_id="alice@shop.com", thread_id="thread-1", user_message="Any update on ORD-1001?")
print(bundle.to_context_block())
````

---

## 8. The Write Path — Persist After Acting

After tools run or the user reveals new facts, the agent **writes** memories. Working notes are ephemeral; semantic and episodic updates persist.

````python
class MemoryWriter:
    def __init__(self, store: MemoryStore):
        self.store = store

    def note_working(self, thread_id: str, content: str) -> str:
        return self.store.write(MemoryRecord(
            memory_id=f"work-{uuid.uuid4().hex[:8]}",
            memory_type=MemoryType.WORKING,
            content=content, thread_id=thread_id, salience=0.5,
        ))

    def record_episode(self, user_id: str, content: str, tags: list[str] | None = None) -> str:
        return self.store.write(MemoryRecord(
            memory_id=f"epi-{uuid.uuid4().hex[:8]}",
            memory_type=MemoryType.EPISODIC,
            content=content, user_id=user_id, tags=tags or [], salience=1.0,
        ))

    def upsert_semantic(self, user_id: str, content: str, tags: list[str] | None = None) -> str:
        return self.store.write(MemoryRecord(
            memory_id=f"sem-{uuid.uuid4().hex[:8]}",
            memory_type=MemoryType.SEMANTIC,
            content=content, user_id=user_id, tags=tags or [], salience=1.5,
        ))


writer = MemoryWriter(store)
writer.note_working("thread-1", "Looked up ORD-1001 → status=shipped")
writer.record_episode("alice@shop.com", "2026-07-10: Confirmed ORD-1001 still in transit inquiry.", tags=["ORD-1001"])
print("After writes:", store.count_by_type())
````

## 9. MemoryAwareAgent — Read → Act → Write Loop

The agent turn wraps memory around a simple support logic core: recall context, run tools, write new episodes, compose reply.

````python
@dataclass
class AgentTurnResult:
    reply: str
    memories_used: int
    memories_written: int
    context_preview: str


class MemoryAwareAgent:
    def __init__(self, store: MemoryStore):
        self.reader = MemoryReader(store)
        self.writer = MemoryWriter(store)
        self._writes_before = 0

    def _lookup_order(self, order_id: str) -> dict[str, Any] | None:
        return ORDERS.get(order_id.strip().upper())

    def _identify_user(self, message: str, explicit_user: str | None) -> str | None:
        if explicit_user:
            return explicit_user
        if "alice" in message.lower():
            return "alice@shop.com"
        if "bob" in message.lower():
            return "bob@shop.com"
        return None

    def turn(self, message: str, *, thread_id: str, user_id: str | None = None) -> AgentTurnResult:
        user_id = self._identify_user(message, user_id)
        audit_len = len(MEMORY_AUDIT)

        bundle = self.reader.recall(user_id=user_id, thread_id=thread_id, user_message=message)
        memories_used = sum(len(getattr(bundle, k)) for k in ("procedural", "semantic", "episodic", "working"))

        reply_parts: list[str] = []
        oid_match = re.search(r"ORD-\d+", message.upper())

        if oid_match:
            oid = oid_match.group()
            order = self._lookup_order(oid)
            if order:
                self.writer.note_working(thread_id, f"Resolved {oid} → status={order['status']}")
                if user_id:
                    self.writer.record_episode(user_id, f"Asked about {oid} status on {utcnow()[:10]}.", tags=[oid])
                reply_parts.append(f"{oid} is {order['status']} (${order['total_usd']}).")
            else:
                reply_parts.append(f"Order {oid} not found.")

        if "refund" in message.lower() and bundle.procedural:
            reply_parts.append(f"Reminder: {bundle.procedural[0].content}")

        if bundle.semantic:
            pref = bundle.semantic[0].content
            if "email" in pref.lower():
                reply_parts.append("I'll follow up via email per your preference.")
            elif "sms" in pref.lower():
                reply_parts.append("I'll follow up via SMS per your preference.")

        if bundle.episodic and not oid_match:
            reply_parts.append(f"I see from earlier: {bundle.episodic[0].content}")

        if not reply_parts:
            reply_parts.append("How can I help? Share an order ID or question.")

        writes = len(MEMORY_AUDIT) - audit_len
        return AgentTurnResult(
            reply=" ".join(reply_parts),
            memories_used=memories_used,
            memories_written=max(0, writes - 1),
            context_preview=bundle.to_context_block()[:200] + "...",
        )


agent = MemoryAwareAgent(store)
r1 = agent.turn("I'm Alice — what's the status of ORD-1001?", thread_id="thread-1", user_id="alice@shop.com")
print("Reply:", r1.reply)
print(f"Memories used: {r1.memories_used}, written: {r1.memories_written}")
````

## 10. Multi-Turn Conversation — Memory Across Turns

The same thread accumulates working notes and episodic events. Turn 2 benefits from Turn 1 without re-pasting the full transcript.

````python
def run_conversation(turns: list[str], thread_id: str, user_id: str) -> None:
    for i, msg in enumerate(turns, 1):
        result = agent.turn(msg, thread_id=thread_id, user_id=user_id)
        print(f"Turn {i} user: {msg}")
        print(f"Turn {i} agent: {result.reply}")
        print(f"  (recalled {result.memories_used} memories)\n")


run_conversation([
    "Hi, I'm Alice. Can you check ORD-1001?",
    "Is a refund possible for that order?",
    "Please use my preferred contact method for updates.",
], thread_id="thread-alice-42", user_id="alice@shop.com")
````

## 11. Scoping — User vs Thread vs Global

| Scope | Isolation | Use for |
|-------|-----------|--------|
| **Global** | Shared across all users | Procedural rules, org policies |
| **User** | Per `user_id` | Preferences, VIP status, semantic facts |
| **Thread** | Per `thread_id` | Working scratchpad, current ticket context |

Mixing scopes incorrectly leaks data — Bob must never see Alice's semantic memories.

````python
alice_recall = reader.recall(user_id="alice@shop.com", thread_id="t", user_message="preferences")
bob_recall = reader.recall(user_id="bob@shop.com", thread_id="t", user_message="preferences")

print("Alice semantic:", [r.content for r in alice_recall.semantic])
print("Bob semantic:", [r.content for r in bob_recall.semantic])
print("\nProcedural (shared):", alice_recall.procedural[0].content[:50], "...")
````

## 12. Memory vs Retrieval (RAG) Alone

**RAG** retrieves documents from a corpus (policies, manuals). **Agentic memory** also stores **interaction-derived** state: what this user said, what tools returned, what was decided.

| | RAG | Agentic memory |
|--|-----|----------------|
| Source | Static knowledge base | User interactions + tool results |
| Updates | Index rebuild / ingest pipeline | Per-turn writes |
| Scope | Usually global docs | User, thread, session scoped |
| Example | Return policy PDF | "Alice already received 10% credit on 2026-06-15" |

Production agents combine both: RAG for policies, memory for personalized context.

## 13. Working Memory Cleanup

Working memory should not grow forever. Clear it when a thread closes or a task completes.

````python
before = store.count_by_type().get("working", 0)
removed = store.clear_working("thread-alice-42")
after = store.count_by_type().get("working", 0)
print(f"Cleared {removed} working memories from thread-alice-42 ({before} → {after})")
print("Episodic and semantic memories persist for future sessions.")
````

## 15. Anti-Patterns

| Anti-pattern | Problem | Fix |
|--------------|---------|-----|
| Full chat log in every prompt | Token bloat, attention dilution | Retrieve relevant memories only |
| No write path | Agent never learns user facts | Persist semantic/episodic after turns |
| Single undifferentiated store | Can't tune retention or scope | Type tags + separate policies per type |
| User A sees User B memories | Privacy violation | Enforce `user_id` on recall |
| Working memory never cleared | Stale scratchpad pollutes recall | Clear on thread close |
| Memory = vector DB only | Loses structure and procedures | Hybrid: vectors + structured records |

## 18. Summary

Agentic memory gives agents persistence beyond the context window:

- **Working** — scratchpad for the current thread
- **Episodic** — what happened in past interactions
- **Semantic** — stable facts about users and world
- **Procedural** — how the agent should behave

Every turn follows **read → act → write**: recall scoped memories, run tools and reasoning, persist new facts and events. Scope by user and thread, audit reads and writes, and combine memory with RAG for policies. The ShopCo agent demonstrated multi-turn support where preferences and order history carry forward without stuffing the full chat log into every prompt.
