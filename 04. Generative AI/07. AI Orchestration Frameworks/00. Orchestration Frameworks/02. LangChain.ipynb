## 1. What LangChain Is

LangChain is a **Python and JavaScript library** for building LLM applications by composing reusable components. It does not host models — it wraps provider SDKs (OpenAI, Anthropic, Google, local models) and connects them to retrieval, tools, and memory.

Core value proposition:

- **Compose** multi-step flows without rewriting glue code
- **Swap** providers and vector stores behind stable interfaces
- **Integrate** quickly with 100+ data sources and services
- **Observe** runs via callbacks and LangSmith tracing

---

## 2. Mental Model — LCEL and Runnables

Modern LangChain centers on **LCEL** (LangChain Expression Language): pipe runnables together with the `|` operator.

```python
chain = prompt | llm | parser
result = chain.invoke({"question": "..."})
```

| Primitive | Role |
|-----------|------|
| **Runnable** | Anything with `invoke` / `stream` / `batch` |
| **Prompt template** | Variables → chat messages |
| **Chat model** | Messages → model response |
| **Output parser** | Text → structured data |
| **Retriever** | Query → relevant documents |
| **Tool** | Callable exposed to the model |

Chains are **directed acyclic graphs** of runnables. For cycles (agent loops), use **LangGraph** (notebook **03**), which is designed as LangChain's sibling for stateful graphs.

---

## 3. Architecture in a Typical App

```
┌─────────────────────────────────────────────┐
│  Your FastAPI / service layer               │
├─────────────────────────────────────────────┤
│  LangChain chain or agent                   │
│    prompt → retriever → llm → parser      │
├─────────────────────────────────────────────┤
│  Integrations: Chroma, OpenAI, Redis, ...   │
└─────────────────────────────────────────────┘
```

LangChain sits between your product API and infrastructure. Keep framework types behind your own service boundary so you can replace or simplify later.

## 4. Strengths

| Strength | Why it matters |

| **Integration breadth** | Vector DBs, loaders, embeddings, cloud auth |

| **LCEL composability** | Readable pipelines; streaming and async built in |

---

## 5. Trade-offs and Watch-outs

| Issue | Guidance |
|-------|----------|
| **API churn** | Pin versions; v0.1 → v0.2+ renamed imports — verify docs match your version |
| **Abstraction depth** | Debug by enabling callbacks; don't stack wrappers blindly |
| **Overlap with LlamaIndex** | Pick one primary data/indexing layer for RAG |
| **Agents can be opaque** | Prefer explicit LangGraph flows for production-critical paths |
| **Dependency weight** | Heavy for a single `chat.completions` call — use SDK directly when sufficient |

---

## 6. When to Choose LangChain

Choose LangChain when:

- You need **multiple integrations** (loaders, vector stores, providers) quickly
- Your core flow is **chains** (RAG Q&A, summarize → translate, classify → route)
- You want **LangSmith** or callback-based observability out of the box
- You plan to graduate agent loops to **LangGraph** while staying in one ecosystem

Skip or delay LangChain when:

- You have **one LLM call** and a fixed prompt
- Your team prefers **minimal dependencies** and full control
- **LlamaIndex** or **Haystack** already owns your retrieval stack end-to-end

## 7. Comparison Snapshot

| vs | LangChain position |
