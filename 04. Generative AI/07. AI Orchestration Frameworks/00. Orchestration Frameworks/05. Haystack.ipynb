# Haystack — Framework Overview

**Haystack** (by deepset) is a production-oriented framework for building **search and RAG pipelines**. It emphasizes typed components, pipeline composition, evaluation, and self-hosted deployments — with strong roots in NLP and information retrieval.

## 1. What Haystack Is

Haystack models applications as **pipelines** of **components** connected by explicit inputs and outputs. Each component has a typed contract — retriever, ranker, prompt builder, generator — making pipelines easier to test and version than ad-hoc scripts.

```python
# Conceptual Haystack 2.x shape
pipeline = Pipeline()
pipeline.add_component("retriever", retriever)
pipeline.add_component("prompt", prompt_builder)
pipeline.add_component("llm", generator)
pipeline.connect("retriever.documents", "prompt.documents")
```

---

## 2. Mental Model

```
Query ──► Retriever ──► (optional Ranker) ──► PromptBuilder ──► Generator ──► Answer
```

| Component | Role |
|-----------|------|
| **Retriever** | BM25, dense, hybrid search |
| **Ranker** | Cross-encoder reranking |
| **Prompt builder** | Assemble context + query |
| **Generator** | LLM or extractive reader |
| **Pipeline** | Wires components; supports branching (Pipeline 2.x) |

---

## 3. Strengths

| Strength | Detail |
|----------|--------|
| **Pipeline clarity** | Explicit component graph |
| **Eval-first culture** | deepset eval tools, benchmark integration |
| **On-prem friendly** | Self-hosted search + RAG stacks |
| **Hybrid retrieval** | Keyword + dense in one pipeline |
| **Enterprise adoption** | Used in regulated / air-gapped environments |

---

## 4. Trade-offs

| Issue | Guidance |
|-------|----------|
| **Smaller hobbyist community** than LangChain |
| **Agent features** less mature than LangGraph |
| **Haystack 1.x → 2.x** migration required for new projects |
| **Integration count** lower than LangChain for exotic connectors |

---

## 5. When to Choose Haystack

Choose Haystack when:

- **Enterprise search + RAG** is the primary workload
- You need **evaluable, typed pipelines** for compliance or QA teams
- **Self-hosted** deployment is mandatory
- **Hybrid retrieval** (BM25 + vectors) is a first-class requirement

Choose LangChain or LlamaIndex when integration breadth and rapid agent prototyping matter more than pipeline typing.

---

## 6. Haystack vs LlamaIndex vs LangChain

| | Haystack | LlamaIndex | LangChain |
|--|----------|------------|----------|
| **Center** | Pipeline components | Indices & query engines | Runnables & integrations |
| **RAG eval** | Strong | Good | Good (via LangSmith) |
| **Agents** | Growing | Supported | Strong + LangGraph |
| **Best fit** | Enterprise search | Data frameworks | General orchestration |

---

## 7. Summary and Next Steps

Haystack excels at **production RAG and search pipelines** with clear component boundaries and evaluation discipline. It is a strong alternative when LangChain's breadth is unnecessary and pipeline rigor is valued.
