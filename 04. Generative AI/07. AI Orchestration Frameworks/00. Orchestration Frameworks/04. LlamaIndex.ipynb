## 1. What LlamaIndex Is

LlamaIndex (formerly GPT Index) treats **data** as the primary abstraction. Instead of starting from "chain a prompt to an LLM," you start from "index these documents and query them intelligently."

Core responsibilities:

- **Load** documents (PDF, Notion, SQL, APIs)
- **Index** into vector, keyword, knowledge-graph, or composite structures
- **Query** via retrievers, query engines, and routers
- **Chat** over indexed data with memory

---

## 2. Mental Model

```
Documents ──► Nodes/Chunks ──► Index ──► Query Engine ──► Answer
                    │              │
                    └── Embeddings └── Retrievers
```

| Primitive | Role |
|-----------|------|
| **Document / Node** | Unit of stored content + metadata |
| **Index** | `VectorStoreIndex`, `SummaryIndex`, composable indices |
| **Retriever** | Fetch relevant nodes for a query |
| **Query engine** | Retrieve + synthesize response |
| **Chat engine** | Multi-turn Q&A over an index |
| **Agent** | Tool-using agent over data tools |

---

## 3. Strengths

| Strength | Detail |
|----------|--------|
| **RAG-first design** | Index types, routers, sub-question query engines |
| **Data connectors** | Broad loader ecosystem |
| **Structured data** | SQL, JSON, knowledge-graph indices |
| **Query transformations** | HyDE, multi-step decomposition built in |
| **Observability** | LlamaTrace and eval hooks |

---

## 4. Trade-offs

| Issue | Guidance |
|-------|----------|
| **Overlap with LangChain** | Both do RAG — avoid duplicating index + chain layers |
| **Abstraction layers** | Learn index → engine path before agents |
| **Version/API changes** | Check docs for v0.10+ module layout |
| **Not ideal for pure chitchat** | Optimized for data-grounded workloads |

---

## 5. When to Choose LlamaIndex

Choose LlamaIndex when:

- **Document Q&A** or enterprise search is the core product
- You need **multiple index types** or query routers
- Ingestion from **many sources** is a priority
- You want **structured + unstructured** retrieval in one framework

Prefer LangChain or custom code when orchestration is mostly **tools/APIs** with light retrieval, or when you already standardized on another data layer.

---

## 6. LlamaIndex vs LangChain for RAG

| Aspect | LlamaIndex | LangChain |
|--------|------------|----------|
| Starting point | Index & query engine | Runnable chain |
| Retrieval depth | Native, many patterns | Strong via retrievers + LCEL |
| General tools/agents | Supported | Broader integration catalog |
| Typical combo | LI index + LC/LG orchestration | LC end-to-end |

Many teams use **LlamaIndex for indexing** and **LangGraph for agentic flow** — but one primary owner for the data layer reduces confusion.
