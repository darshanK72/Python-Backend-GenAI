**DSPy** (Stanford NLP) treats LLM pipelines as **programs** composed of modules, not hand-crafted prompt strings. You define **signatures** (input/output fields), wire **modules** (predict, chain-of-thought, retrieve), and use **teleprompters** (compilers) to optimize prompts and few-shot examples against a **metric** on a dev set.

## 1. What DSPy Is

DSPy shifts prompt engineering from **manual string tuning** to **program optimization**:

| Traditional | DSPy |
|-------------|------|
| Write prompt in a text file | Declare `Signature` with typed fields |
| Manually add few-shot examples | Teleprompter searches for good demonstrations |
| Hope it generalizes | Optimize against labeled metric on dev set |
| Copy-paste between models | Recompile program for new LM |

DSPy is **not** a chain library like LangChain â€” it is a **programming model + optimizer** for LLM components.

| Primitive | Role |
|-----------|------|
| **Signature** | Declares I/O fields (contract) |

| **Metric** | Scores outputs for optimization |
| **LM** | Configured language model backend |

## 4. Trade-offs

| Issue | Guidance |
|-------|----------|
| **Requires dev set + metric** | Not for one-off prompts |
| **Compilation cost** | Optimization runs consume LLM calls |
| **Different paradigm** | Team must learn signatures/teleprompters |
| **Not a deployment server** | Compose with FastAPI; no built-in RAG stack |
| **Rapid API changes** | Pin versions; read migration notes |

---

## 5. When to Choose DSPy

Choose DSPy when:

- You have a **labeled evaluation set** and clear **metric**
- Prompt quality is the bottleneck and manual tuning plateaus
- You need to **swap models** without rewriting every prompt
- You are building **repeatable LLM programs** (classifiers, extractors, RAG answerers)

Skip DSPy when you have a **single static prompt** with no eval loop, or when integration breadth (vector DBs, tools) matters more than optimization.

---

## 6. DSPy vs LangChain

| | DSPy | LangChain |
|--|------|----------|
| **Focus** | Optimize programs | Wire integrations |
| **Prompts** | Compiled from data | Hand-authored templates |
| **RAG** | Via `Retrieve` modules | Native retriever ecosystem |
| **Typical combo** | DSPy module inside LC/LG service | LC for plumbing, DSPy for core LM logic |

They are **complementary**: LangChain moves data; DSPy improves the reasoning/extraction step.
