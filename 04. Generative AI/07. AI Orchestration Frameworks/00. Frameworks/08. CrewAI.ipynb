# CrewAI — Framework Overview

**CrewAI** is a high-level framework for defining **role-based agent teams** (crews) that collaborate on tasks. It emphasizes readable YAML/Python definitions of agents, goals, tools, and process flow — making multi-agent demos accessible without building conversation managers from scratch.

## 1. What CrewAI Is

CrewAI organizes work around three building blocks:

| Building block | Definition |
|----------------|------------|
| **Agent** | Role + goal + backstory + tools |
| **Task** | Specific assignment with expected output |
| **Crew** | Agents + tasks + **process** (sequential, hierarchical) |

```python
# Conceptual shape
crew = Crew(agents=[researcher, writer], tasks=[research_task, write_task], process=Process.sequential)
result = crew.kickoff(inputs={"topic": "RAG evaluation"})
```

---

## 2. Mental Model

```
Task 1 (Researcher) ──► Task 2 (Writer) ──► Task 3 (Editor)
         │                      │
         └── tools: search       └── tools: none
```

In **hierarchical** process, a manager agent delegates subtasks. In **sequential** process, tasks run in fixed order with outputs passed forward.

---

## 3. Strengths

| Strength | Detail |
|----------|--------|
| **Readable definitions** | Role/goal/task maps to business language |
| **Fast multi-agent demos** | Marketing, research, content pipelines |
| **Process types** | Sequential and hierarchical out of the box |
| **Tool integration** | Agents can use search, APIs, custom tools |
| **CrewAI+ / enterprise** | Hosted and enterprise offerings emerging |

---

## 4. Trade-offs

| Issue | Guidance |
|-------|----------|
| **Less low-level control** than LangGraph for complex branching |
| **Opaque internals** when debugging failed crews |
| **Cost** scales with agents × tasks × LLM calls |
| **Production hardening** requires external timeouts, eval, logging |

---

## 5. When to Choose CrewAI

Choose CrewAI when:

- You want **role-based teams** with minimal boilerplate
- Workflows are **sequential or manager-delegated** rather than arbitrary graphs
- Stakeholders need to **read agent definitions** (role, goal, task)
- You are prototyping **content, research, or ops automation** crews

Choose **LangGraph** or **AutoGen** when conversation topology or code-execution loops are more dynamic than fixed task lists.

---

## 6. CrewAI vs AutoGen

| | CrewAI | AutoGen |
|--|--------|--------|
| **Abstraction** | Crew / task / process | Conversable agents / group chat |
| **Definition style** | Declarative roles | Imperative conversation |
| **Best for** | Structured pipelines | Open-ended collaboration |
| **Debugging** | Task output chains | Message transcripts |

---

## 7. Summary and Next Steps

CrewAI lowers the barrier to **multi-agent automation** with clear roles and tasks. Use it for readable crew prototypes; graduate to LangGraph when you need explicit graphs, checkpointing, and production control flow.
