# Semantic Kernel — Framework Overview

**Semantic Kernel** (Microsoft) is an enterprise-oriented orchestration SDK for building **copilot-style applications** in **Python and .NET**. It centers on **plugins** (functions the model can invoke), **planners** that sequence steps, and tight integration with **Azure OpenAI** and Microsoft identity patterns.

## 1. What Semantic Kernel Is

Semantic Kernel (SK) wraps LLM calls in a **Kernel** that registers **plugins** — collections of functions described to the model as tools. A **planner** (automatic or manual) decides which functions to call to satisfy a user goal.

```
User goal ──► Planner ──► Kernel invokes plugins ──► LLM synthesizes result
```

SK targets teams already on **Microsoft Azure**, **.NET**, and **enterprise IAM** who want a structured plugin model similar to how Microsoft Copilot extensions work.

---

## 2. Core Abstractions

| Primitive | Role |
|-----------|------|
| **Kernel** | Runtime container for services and plugins |
| **Plugin** | Group of functions (native C#/Python or prompt-based) |
| **Function** | Callable unit with schema exposed to the model |
| **Planner** | Sequences function calls to achieve a goal |
| **Memory / connectors** | Store and retrieve contextual information |
| **Prompt template** | Semantic function defined as a prompt skill |

## 3. Strengths

| Strength | Detail |
|----------|--------|
| **Enterprise alignment** | Azure OpenAI, Entra ID, compliance narratives |
| **.NET + Python** | Same concepts across stacks |
| **Plugin model** | Clean separation of capabilities |
| **Planner abstraction** | Automatic step decomposition for demos |
| **Microsoft ecosystem** | Copilot Studio, Azure AI patterns |

---

## 4. Trade-offs

| Issue | Guidance |
|-------|----------|
| **Azure-centric docs** | Non-Azure teams may fight examples |
| **Planner unpredictability** | Production often prefers explicit graphs (LangGraph) |
| **Smaller open-source community** than LangChain |
| **Concept overlap** with LangChain tools/agents |

---

## 5. When to Choose Semantic Kernel

Choose Semantic Kernel when:

- Your org standardizes on **Microsoft Azure** and **.NET**
- You are building **internal copilots** with plugin boundaries
- **Enterprise security and IAM** patterns are non-negotiable
- You want alignment with **Microsoft Copilot extensibility** models

Choose LangChain/LangGraph for broader integration catalogs or polycloud deployments.

---

## 6. SK vs LangChain

| | Semantic Kernel | LangChain |
|--|-----------------|----------|
| **Unit** | Plugin / function | Runnable / tool |
| **Planning** | Built-in planners | Agents / LangGraph |
| **Cloud** | Azure-first | Provider-agnostic |
| **Languages** | Python, C#, Java | Python, JS/TS |

---

## 7. Summary and Next Steps

Semantic Kernel is the natural orchestration layer for **Microsoft-centric enterprise copilots** — plugins, planners, and Azure integration. For polyglot or integration-heavy stacks, LangChain remains the more common choice.
