# AutoGen — Framework Overview

**AutoGen** (Microsoft) is a framework for **multi-agent conversation**: multiple LLM-backed agents exchange messages, call tools, and run code until a task completes or a termination condition is met. It is popular for research workflows, coding assistants with critique loops, and exploratory automation.

## 1. What AutoGen Is

AutoGen models collaboration as **message passing between agents**. Each agent has a role, system prompt, and optional tools. A **group chat** or **two-agent** pattern orchestrates who speaks next until the user proxy or manager agent declares completion.

```
User ──► AssistantAgent ◄──► UserProxyAgent
              │
              ├── tools (web, code execution)
              └── other specialist agents
```

---

## 2. Core Abstractions

| Primitive | Role |
|-----------|------|
| **ConversableAgent** | Base agent that sends/receives messages |
| **AssistantAgent** | LLM-powered agent with instructions |
| **UserProxyAgent** | Represents human; can execute code on approval |
| **GroupChat** | Multiple agents in shared conversation |
| **GroupChatManager** | Selects next speaker |
| **Tools / functions** | Registered capabilities per agent |

---

## 3. Strengths

| Strength | Detail |
|----------|--------|
| **Flexible multi-agent** | Custom roles and conversation topologies |
| **Code execution** | User proxy can run generated code in sandbox |
| **Research-friendly** | Rapid experimentation with agent teams |
| **AutoGen Studio** | Low-code UI for prototyping |
| **Microsoft backing** | Active development (AutoGen 0.2+ / AgentChat API) |

---

## 4. Trade-offs

| Issue | Guidance |
|-------|----------|
| **Conversation unpredictability** | Harder to guarantee fixed step counts |
| **Cost** | Multiple agents = many LLM calls per task |
| **Production guardrails** | Need timeouts, tool sandboxes, human approval |
| **API evolution** | AutoGen 0.2 reworked APIs — verify version in docs |

---

## 5. When to Choose AutoGen

Choose AutoGen when:

- Tasks benefit from **multiple perspectives** (writer + critic, planner + coder)
- You need **code execution loops** with human oversight
- You are prototyping **multi-agent research** or internal automation
- Conversation-driven coordination fits better than a fixed graph

Prefer **LangGraph** when you need **explicit state machines**, checkpointing, and auditable transitions for production.

---

## 6. AutoGen vs CrewAI vs LangGraph

| | AutoGen | CrewAI | LangGraph |
|--|---------|--------|----------|
| **Model** | Message passing | Role/task crews | State graph |
| **Control** | Conversation manager | Process (sequential/hierarchical) | Declared edges |
| **Code execution** | Native patterns | Via tools | Via tool nodes |
| **Production fit** | Needs guardrails | Demo-friendly | Strongest structure |
