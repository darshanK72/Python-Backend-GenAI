# Introduction to LangGraph

**LangChain LCEL** excels at linear pipelines: `retrieve → prompt → llm → parse`. Real assistants, however, **loop** (call tools until done), **branch** (RAG vs calculator vs human approval), and **pause** (wait for a reviewer). Encoding those flows as nested `if` statements and manual `while` loops becomes unmaintainable.

**LangGraph** models LLM workflows as a **state graph**: **nodes** read and update shared **state**; **edges** define transitions, including **conditional routes** and **cycles**. It is the low-level runtime behind LangChain **`create_agent`** (**01. LangChain/13**) and the right tool when **control flow is the product**.

This notebook defines LangGraph, contrasts it with LCEL chains, previews the **StateGraph** API, builds minimal graphs without and with an LLM, maps concepts to the **02. LangGraph/** track, and verifies your environment. Framework landscape context lives in **00. Frameworks/03. LangGraph**.

Recommended background: **01. LangChain/** track (especially **02**, **03**, **12**, **13**) — LangGraph nodes often wrap LangChain runnables.

## 1. The Problem — Linear Chains Hit a Wall

A fixed LCEL chain runs **once**, top to bottom:

```
question ──► retriever ──► prompt ──► llm ──► parser ──► answer
```

Agentic behavior needs **cycles**:

```
question ──► llm ──► tool_calls? ──yes──► tools ──┐
                ▲                                  │
                └──────────────────────────────────┘
                └──no──► final answer
```

| Requirement | LCEL alone | LangGraph |
|-------------|------------|----------|
| Fixed step order | Natural (`a \| b \| c`) | Possible but verbose |
| **Tool loop** | Manual `while` + message list | First-class cycle |
| **Conditional routing** | `RunnableBranch` (limited) | `add_conditional_edges` |
| **Checkpoint / resume** | Manual session store | Built-in checkpointers |
| **Human approval gate** | Custom HTTP pause | `interrupt` API |
| **Visible control flow** | Hidden in executor code | Declared graph diagram |

LangGraph does not replace LangChain — it **orchestrates** LangChain (and plain Python) inside **nodes**.

---

## 2. What Is LangGraph?

**LangGraph** is a library for building **stateful, multi-step LLM applications** as graphs. Core ideas:

1. **Shared state** — a typed dict (or schema) every node reads and partially updates.
2. **Nodes** — functions (or runnables) that return **state patches**.
3. **Edges** — fixed transitions between nodes.
4. **Conditional edges** — router functions pick the next node at runtime.
5. **Compilation** — `graph.compile()` produces an executable app with `invoke`, `stream`, `batch`, async.
6. **Checkpointing** — optional persistence between steps and across HTTP requests.

LangGraph is maintained by the LangChain team but ships as a **separate package** (`langgraph`) with its own graph runtime.

---

## 3. LangChain vs LangGraph

| | **LangChain (LCEL)** | **LangGraph** |
|--|----------------------|---------------|
| **Shape** | Directed acyclic pipe | Graph with cycles |
| **State** | Passed along the chain | Central shared state object |
| **Best for** | RAG, extraction, single-shot Q&A | Agents, routers, HITL, multi-stage flows |
| **Debugging** | Callbacks, LangSmith traces | Node-level traces + graph visualization |
| **Entry API** | `prompt \| llm \| parser` | `StateGraph` → `compile()` → `invoke` |

**`create_agent`** (**01. LangChain/13**) returns a **pre-built LangGraph** — you get agent loops without drawing nodes. When you need **custom routing**, **extra nodes** (grading, guardrails), or **auditability**, author the graph explicitly in this track.

## 5. Mental Model — StateGraph

```
         ┌─────────────────────────────────────┐
         │           Shared State              │
         │  { question, messages, context, … } │
         └─────────────────────────────────────┘
              ▲           ▲           ▲
              │           │           │
         ┌────┴───┐  ┌────┴───┐  ┌────┴───┐
         │ node A │  │ node B │  │ node C │
         └────┬───┘  └────┬───┘  └────┬───┘
              │           │           │
         START ──► A ──► B ──► C ──► END
                      └──► (conditional) ──► C
```

| Concept | Meaning |
|---------|--------|
| **`StateGraph(schema)`** | Builder bound to a state type |
| **`add_node(name, fn)`** | Register a step |
| **`add_edge(a, b)`** | Always go from `a` to `b` |
| **`add_conditional_edges`** | Router picks next node (**04**) |
| **`compile(checkpointer=...)`** | Frozen executable graph |
| **`invoke(state)`** | Run from START to END |

## 6. Your First Graph — No LLM

A two-node linear graph to learn the lifecycle without API calls:

````python
from typing_extensions import TypedDict

from langgraph.graph import END, START, StateGraph


class PipelineState(TypedDict):
    question: str
    steps: list[str]
    answer: str


def receive(state: PipelineState) -> dict:
    return {"steps": state.get("steps", []) + ["received"]}


def respond(state: PipelineState) -> dict:
    return {
        "steps": state["steps"] + ["responded"],
        "answer": f"Notes API assistant processed: {state['question']}",
    }


builder = StateGraph(PipelineState)
builder.add_node("receive", receive)
builder.add_node("respond", respond)
builder.add_edge(START, "receive")
builder.add_edge("receive", "respond")
builder.add_edge("respond", END)

graph = builder.compile()

result = graph.invoke({
    "question": "What is Alembic?",
    "steps": [],
    "answer": "",
})

print(json.dumps(result, indent=2))
````

### 6.1 What Just Happened

1. **`PipelineState`** defines allowed state keys.
2. Each **node** returns a **partial update** — not the full state.
3. LangGraph **merges** patches into the running state.
4. **`compile()`** returns a **`CompiledStateGraph`** with Runnable methods.
5. **`invoke`** runs START → receive → respond → END.

## 8. Adding an LLM Node

Wrap **`ChatOpenAI`** inside a node — same model as **01. LangChain/03**:

````python
from langchain_core.messages import HumanMessage
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-4o-mini", api_key=OPENAI_API_KEY, temperature=0)


class QAState(TypedDict):
    question: str
    answer: str


def call_model(state: QAState) -> dict:
    response = llm.invoke([HumanMessage(content=state["question"])])
    return {"answer": response.content}


qa_builder = StateGraph(QAState)
qa_builder.add_node("model", call_model)
qa_builder.add_edge(START, "model")
qa_builder.add_edge("model", END)

qa_graph = qa_builder.compile()
print(qa_graph.invoke({"question": "What is FastAPI in one sentence?", "answer": ""}))
````

This single-node graph is equivalent to `llm.invoke` — the value appears when you **add more nodes** and **edges** around the model.

## 9. LangChain LCEL Inside a Node

Nodes can invoke full LCEL chains (**01. LangChain/02**):

````python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate

qa_chain = (
    ChatPromptTemplate.from_messages([
        ("system", "You answer backend questions concisely."),
        ("human", "{question}"),
    ])
    | llm
    | StrOutputParser()
)


def lcel_node(state: QAState) -> dict:
    return {"answer": qa_chain.invoke({"question": state["question"]})}


lcel_builder = StateGraph(QAState)
lcel_builder.add_node("qa_chain", lcel_node)
lcel_builder.add_edge(START, "qa_chain")
lcel_builder.add_edge("qa_chain", END)

lcel_graph = lcel_builder.compile()
print(lcel_graph.invoke({"question": "What header carries JWT?", "answer": ""}))
````

Production pattern: **LangGraph for control flow**, **LCEL for step internals** — prompt | llm | parser lives inside one node; routing lives in the graph.

## 10. Agent Loop — Why Graphs Exist

The ReAct tool loop (**06. Agent Loops and the ReAct Pattern**) is a **cycle** in the graph:

```
        ┌──────────┐
        │  START   │
        └────┬─────┘
             ▼
        ┌──────────┐
   ┌───►│   LLM    │◄───┐
   │    └────┬─────┘    │
   │         │          │
   │    tool calls?     │
   │    ┌────┴────┐     │
   │    ▼         ▼     │
   │  tools      END    │
   │    │               │
   └────┘               │
```

| Approach | Visibility | Flexibility |
|----------|------------|-------------|
| **`create_agent`** | Pre-built graph | Fastest start |
| **Custom StateGraph** | Every node explicit | Add graders, guards, logging nodes |
| **Manual while loop** | Code only | Hard to checkpoint and visualize |

## 11. Checkpointing Preview

Compiled graphs accept a **checkpointer** — the same mechanism **`create_agent`** uses for multi-turn threads (**01. LangChain/13**, **14**):

````python
from langgraph.checkpoint.memory import MemorySaver

checkpointed = builder.compile(checkpointer=MemorySaver())

cfg = {"configurable": {"thread_id": "intro-thread-1"}}
r1 = checkpointed.invoke({"question": "Tell me about Alembic.", "steps": [], "answer": ""}, config=cfg)
r2 = checkpointed.invoke({"question": "Summarize in three words.", "steps": [], "answer": ""}, config=cfg)

print("turn 1 steps:", r1["steps"])
print("turn 2 answer:", r2["answer"])
````

**`MemorySaver`** is for learning and tests. Production uses Postgres, Redis, or hosted checkpointers — **08. Checkpointing and Thread Persistence**.

## 14. When to Choose LangGraph

**Choose LangGraph** when:

- Control flow includes **loops**, **branches**, or **parallel paths**
- You need **checkpointing**, **time travel**, or **resume after crash**
- **Human approval** is required before some steps
- Compliance asks for a **diagram of every transition**
- You outgrow **`create_agent`** defaults (extra grading/guard nodes)

**Stay with LCEL** (**01. LangChain/**) when:

- Pipeline is **strictly linear** (RAG Q&A, summarization, extraction)
- No tool loop, no HITL, no dynamic routing
- Team prefers minimal abstraction

Many products use **both**: LCEL RAG chain as a **node** inside a larger LangGraph router.

## 17. Design Principles to Carry Forward

1. **State is the contract** — design the schema before writing nodes.
2. **Nodes are small** — one responsibility; compose via the graph.
3. **Keep LLM work in LCEL** — use LangGraph for routing, not prompt string soup.
4. **Declare cycles explicitly** — no hidden `while True` in route handlers.
5. **Checkpoint early** — if users can return tomorrow, you need a checkpointer.
6. **Graph = documentation** — reviewers and auditors read your edges.
7. **Start from `create_agent`** — refactor to custom graph only when requirements demand it.

## 19. Common Beginner Mistakes

| Mistake | Consequence | Fix |
|---------|-------------|-----|
| LangGraph for simple RAG | Boilerplate without benefit | Use LCEL (**01. LangChain/11**) |
| Returning full state from nodes | Overwrites other keys | Return **partial** updates only |
| No state schema | Key errors at runtime | `TypedDict` or Pydantic model (**02**) |
| Forgetting `compile()` | Builder is not runnable | Always `graph = builder.compile()` |
| Hidden loops in node code | Unbounded cost | Model loop as graph cycle + step limit |
| `MemorySaver` in production | Lost threads on restart | Durable checkpointer (**08**) |
| Skipping LangChain basics | Reinvent prompts/tools | Complete **01. LangChain/** core notebooks |

## 20. Summary

**LangGraph** builds **stateful LLM workflows** as graphs: **nodes** update shared **state**, **edges** define flow, **compilation** produces a runnable app with optional **checkpointing** and **interrupts**. It complements **LangChain LCEL** — chains inside nodes, graphs around them.

Key takeaways:

- Linear LCEL is ideal for fixed pipelines; **cycles and branches** need graphs.
- **`StateGraph` → `compile()` → `invoke`** is the basic lifecycle.
- Nodes return **partial state patches**; LangGraph merges them.
- **`create_agent`** is a pre-built LangGraph; custom graphs add control.
- **Checkpointers** and **interrupts** are production differentiators.
- Patterns map to notebooks **02–16** in this track.
