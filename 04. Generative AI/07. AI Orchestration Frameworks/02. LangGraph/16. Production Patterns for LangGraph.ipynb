## 3. `build_graph()` Factory — Singleton at Startup

Compile **once** when the app boots — not per request (**05**, **03**):

````python
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from langgraph.checkpoint.memory import MemorySaver
from langgraph.graph import END, START, StateGraph


class NotesGraphState(TypedDict):
    question: str
    intent: str
    blocked: bool
    context: str
    source_ids: list[str]
    grade: str
    answer: str
    attempts: int
    trace: Annotated[list[str], operator.add]


CORPUS = [
    {"id": "c1", "text": "The Notes API is built with FastAPI. Routes live under /notes with GET, POST, PUT, DELETE."},
    {"id": "c2", "text": "Alembic applies SQLAlchemy schema migrations. Run alembic upgrade head after pulling revisions."},
    {"id": "c3", "text": "JWT bearer tokens authenticate API requests. Send Authorization: Bearer token header."},
    {"id": "c4", "text": "Pytest fixtures share database setup. Use conftest.py for session-scoped engines."},
    {"id": "c5", "text": "Alembic autogenerate compares SQLAlchemy models to the live schema and drafts revision files."},
]

BLOCKED_TERMS = {"password", "secret", "api key", "private key"}
DOCS_KEYWORDS = {"jwt", "alembic", "pytest", "fastapi", "migration", "token", "route", "conftest"}
MAX_ATTEMPTS = 2
INITIAL = {
    "intent": "", "blocked": False, "context": "", "source_ids": [],
    "grade": "", "answer": "", "attempts": 0, "trace": [],
}


def keyword_retrieve(query: str, k: int = 2) -> tuple[str, list[str]]:
    tokens = set(query.lower().split())
    scored = [(len(tokens & set(d["text"].lower().split())), d) for d in CORPUS]
    scored.sort(key=lambda x: x[0], reverse=True)
    top = [d for s, d in scored if s > 0][:k] or [CORPUS[0]]
    context = "\n".join(f"[{d['id']}] {d['text']}" for d in top)
    return context, [d["id"] for d in top]


def make_llm(cfg: AppSettings, tier: str = "economy") -> ChatOpenAI:
    model = cfg.premium_model if tier == "premium" else cfg.default_model
    return ChatOpenAI(model=model, api_key=cfg.openai_api_key, temperature=0)


def build_graph(cfg: AppSettings, checkpointer=None):
    """Factory — call once at app startup. Returns compiled graph."""
    llm = make_llm(cfg)
    generate_chain = (
        ChatPromptTemplate.from_messages([
            ("system", "Answer using ONLY the context. Cite chunk ids like [c2]."),
            ("human", "Context:\n{context}\n\nQuestion: {question}"),
        ])
        | llm
    )

    def guard_node(state: NotesGraphState) -> dict:
        blocked = any(t in state["question"].lower() for t in BLOCKED_TERMS)
        patch: dict = {"blocked": blocked, "trace": ["guard"]}
        if blocked:
            patch["answer"] = "I cannot help with credentials or secrets."
        return patch

    def classify_node(state: NotesGraphState) -> dict:
        intent = "docs" if any(k in state["question"].lower() for k in DOCS_KEYWORDS) else "general"
        return {"intent": intent, "trace": ["classify"]}

    def retrieve_node(state: NotesGraphState) -> dict:
        context, ids = keyword_retrieve(state["question"])
        return {"context": context, "source_ids": ids, "attempts": state.get("attempts", 0) + 1, "trace": ["retrieve"]}

    def grade_node(state: NotesGraphState) -> dict:
        q, ctx = state["question"].lower(), state.get("context", "").lower()
        relevant = any(k in ctx for k in q.split() if len(k) > 3) or any(k in q and k in ctx for k in DOCS_KEYWORDS)
        return {"grade": "relevant" if relevant else "irrelevant", "trace": ["grade"]}

    def rewrite_node(state: NotesGraphState) -> dict:
        return {"question": state["question"] + " (FastAPI Notes API alembic jwt pytest)", "trace": ["rewrite"]}

    def chitchat_node(state: NotesGraphState) -> dict:
        return {"answer": "Ask me about the Notes API — migrations, JWT, pytest, or FastAPI routes.", "trace": ["chitchat"]}

    def generate_node(state: NotesGraphState) -> dict:
        msg = generate_chain.invoke({"context": state["context"], "question": state["question"]})
        return {"answer": msg.content, "trace": ["generate"]}

    def route_guard(state: NotesGraphState):
        return END if state["blocked"] else "classify"

    def route_intent(state: NotesGraphState) -> Literal["docs", "general"]:
        return "docs" if state["intent"] == "docs" else "general"

    def route_grade(state: NotesGraphState) -> Literal["relevant", "irrelevant"]:
        return "relevant" if state["grade"] == "relevant" else "irrelevant"

    def route_rewrite(state: NotesGraphState):
        return "retrieve" if state.get("attempts", 0) < MAX_ATTEMPTS else END

    builder = StateGraph(NotesGraphState)
    builder.add_node("guard", guard_node)
    builder.add_node("classify", classify_node)
    builder.add_node("retrieve", retrieve_node)
    builder.add_node("grade", grade_node)
    builder.add_node("rewrite", rewrite_node)
    builder.add_node("generate", generate_node)
    builder.add_node("chitchat", chitchat_node)
    builder.add_edge(START, "guard")
    builder.add_edge("retrieve", "grade")
    builder.add_edge("generate", END)
    builder.add_edge("chitchat", END)
    builder.add_conditional_edges("guard", route_guard)
    builder.add_conditional_edges("classify", route_intent, {"docs": "retrieve", "general": "chitchat"})
    builder.add_conditional_edges("grade", route_grade, {"relevant": "generate", "irrelevant": "rewrite"})
    builder.add_conditional_edges("rewrite", route_rewrite)

    cp = checkpointer if checkpointer is not None else MemorySaver()
    return builder.compile(checkpointer=cp, debug=cfg.graph_debug)


# App startup — singleton
NOTES_GRAPH = build_graph(settings)
print("singleton compiled:", type(NOTES_GRAPH).__name__)
print("nodes:", [n for n in NOTES_GRAPH.get_graph().nodes if not n.startswith("__")])
````

Store **`NOTES_GRAPH`** on application state (`app.state.graph`) — routes never call **`build_graph()`** per request.

## 4. Integrate Guard + Agent + Checkpoint

Production stacks combine patterns from **05** (guard + RAG), **06** (agent loop), **08** (checkpoint), and **14** (guardrails):

```
START → guard ──blocked──► END
          │
          ▼
       classify ──general──► chitchat ──► END
          │
          └──docs──► retrieve → grade → generate ──► END

Optional parallel endpoint: create_agent + checkpointer for tool-heavy queries (06)
```

The factory above already includes **guard** and **checkpoint**. For tool agents, compile a separate graph with **`MemorySaver`** or Postgres (**08**).

````python
from langchain.agents import create_agent
from langchain_core.messages import HumanMessage
from langchain_core.tools import tool


@tool
def search_docs(query: str) -> str:
    """Search Notes API documentation chunks."""
    ctx, _ = keyword_retrieve(query)
    return ctx


def build_agent_graph(cfg: AppSettings, checkpointer=None):
    llm = make_llm(cfg)
    cp = checkpointer if checkpointer is not None else MemorySaver()
    return create_agent(
        model=llm,
        tools=[search_docs],
        system_prompt="Use search_docs for Notes API documentation questions.",
        checkpointer=cp,
    )


AGENT_GRAPH = build_agent_graph(settings)
print("agent nodes:", list(AGENT_GRAPH.get_graph().nodes)[:6], "...")
````

Expose **two endpoints** or route by intent: RAG graph for grounded doc Q&A; agent for multi-tool exploration (**06**, **12**).

## 5. `run_notes_service()` — Service Wrapper

Stable response contract for HTTP routes — evolved from **`run_notes_assistant()`** in **05**:

````python
@dataclass
class NotesServiceResponse:
    question: str
    answer: str
    sources: list[str]
    path: list[str]
    blocked: bool
    thread_id: str


class NotesGraphService:
    def __init__(self, graph, settings: AppSettings):
        self._graph = graph
        self._settings = settings

    def _config(self, thread_id: str, user_tier: str = "free") -> dict:
        return {
            "configurable": {"thread_id": thread_id},
            "tags": ["notes-api", user_tier, "rag"],
            "metadata": {"user_tier": user_tier, "service": "notes_graph"},
            "run_name": "notes_rag",
        }

    def run_notes_service(
        self,
        question: str,
        *,
        thread_id: str | None = None,
        user_tier: str = "free",
    ) -> NotesServiceResponse:
        tid = thread_id or str(uuid.uuid4())
        state = self._graph.invoke(
            {**INITIAL, "question": question},
            config=self._config(tid, user_tier),
        )
        return NotesServiceResponse(
            question=question,
            answer=state.get("answer", ""),
            sources=state.get("source_ids", []),
            path=state.get("trace", []),
            blocked=state.get("blocked", False),
            thread_id=tid,
        )


notes_service = NotesGraphService(NOTES_GRAPH, settings)
resp = notes_service.run_notes_service("Alembic upgrade command?", thread_id="user-42-session-1")
print(json.dumps(resp.__dict__, indent=2))
````

Routes call **`run_notes_service()`** — never **`NOTES_GRAPH.invoke()`** directly. This is your swap point if you migrate off LangGraph.

## 6. Dependency Injection for Checkpointer

Inject the checkpointer at **factory** time — swap backends without changing graph nodes (**08**):

````python
def make_checkpointer(backend: str):
    if backend == "memory":
        return MemorySaver()
    if backend == "postgres":
        # from langgraph.checkpoint.postgres import PostgresSaver
        # return PostgresSaver.from_conn_string(os.environ["DATABASE_URL"])
        raise NotImplementedError("Wire PostgresSaver in production")
    if backend == "redis":
        raise NotImplementedError("Wire Redis checkpointer in production")
    raise ValueError(f"Unknown backend: {backend}")


def create_app_graph(cfg: AppSettings):
    cp = make_checkpointer(cfg.checkpointer_backend)
    return build_graph(cfg, checkpointer=cp)


# FastAPI lifespan sketch:
LIFESPAN = '''
@asynccontextmanager
async def lifespan(app: FastAPI):
    cfg = AppSettings()
    app.state.settings = cfg
    app.state.checkpointer = make_checkpointer(cfg.checkpointer_backend)
    app.state.graph = build_graph(cfg, checkpointer=app.state.checkpointer)
    app.state.service = NotesGraphService(app.state.graph, cfg)
    warmup_graph(app.state.graph)  # see section 9
    yield
    # teardown: close pools if using Postgres checkpointer
'''
print(LIFESPAN)
````

**Postgres** checkpointers survive restarts; **`MemorySaver`** is fine for notebooks and single-node dev.

## 8. Manual Graph vs `create_agent` — When to Use Which

| Scenario | Use | Why |
|----------|-----|-----|
| Guard + RAG + grade + rewrite | **Manual `StateGraph`** (**05**) | Explicit control flow |
| Standard tool ReAct loop | **`create_agent`** (**06**) | Less boilerplate |
| HITL before generate | **Manual graph** (**09**) | `interrupt_before` |
| Multi-agent supervisor | **Manual graph** (**11**) | Custom routing |
| Simple doc search agent | **`create_agent`** | Ships fast |

**Rule of thumb:** start with **`create_agent`** for tool agents; drop to **`StateGraph`** when you need **nodes between** model steps (guard, grade, formatter, HITL).

## 9. Health Check and Warmup Invoke

Cold-start latency hurts first requests. **Warm up** the graph at startup with a cheap invoke:

````python
def warmup_graph(graph, question: str = "health ping") -> dict:
    """Run a minimal invoke to load clients and validate compile."""
    return graph.invoke(
        {**INITIAL, "question": question},
        config={"configurable": {"thread_id": "__warmup__"}, "tags": ["warmup"]},
    )


def health_check(service: NotesGraphService) -> dict:
    try:
        r = service.run_notes_service("ping", thread_id="__health__")
        return {"status": "ok", "path": r.path, "blocked": r.blocked}
    except Exception as exc:
        return {"status": "error", "detail": type(exc).__name__}


warmup_out = warmup_graph(NOTES_GRAPH)
print("warmup trace:", warmup_out.get("trace"))
print("health:", health_check(notes_service))
````

Expose **`GET /health`** that returns graph status without calling the LLM provider — check checkpointer connectivity separately in prod.

## 11. FastAPI Endpoint Sketch — `POST /chat`

Thin route — validate input, delegate to service, map errors (**01. LangChain/17**):

````python
FASTAPI_SKETCH = '''
from contextlib import asynccontextmanager
from fastapi import FastAPI, HTTPException, Depends
from pydantic import BaseModel, Field


class ChatRequest(BaseModel):
    message: str = Field(min_length=1, max_length=4000)
    thread_id: str | None = None
    user_tier: str = "free"


class ChatResponse(BaseModel):
    answer: str
    sources: list[str]
    path: list[str]
    blocked: bool
    thread_id: str


def get_service(request):
    return request.app.state.service


@asynccontextmanager
async def lifespan(app: FastAPI):
    cfg = AppSettings()
    cp = make_checkpointer(cfg.checkpointer_backend)
    app.state.graph = build_graph(cfg, checkpointer=cp)
    app.state.service = NotesGraphService(app.state.graph, cfg)
    warmup_graph(app.state.graph)
    yield


app = FastAPI(title="Notes API Graph Assistant", lifespan=lifespan)


@app.get("/health")
def health(service: NotesGraphService = Depends(get_service)):
    return health_check(service)


@app.post("/chat", response_model=ChatResponse)
def chat(req: ChatRequest, service: NotesGraphService = Depends(get_service)):
    try:
        result = service.run_notes_service(
            req.message,
            thread_id=req.thread_id,
            user_tier=req.user_tier,
        )
        return ChatResponse(**result.__dict__)
    except Exception as exc:
        raise HTTPException(status_code=502, detail="Graph execution failed") from exc
'''
print(FASTAPI_SKETCH)
````

Add **`POST /chat/stream`** with **`graph.astream(..., stream_mode="updates")`** for SSE progress (**13**, **15**).

## 16. Common Mistakes

| Mistake | Consequence | Fix |
|---------|-------------|-----|
| Rebuild graph per request | Slow, leaky | Startup singleton |
| Routes call `graph.invoke` directly | Tangled HTTP + graph | `run_notes_service()` |
| Shared `thread_id` | Cross-user state leak | Derive from auth |
| `MemorySaver` in prod | Lost on restart | Postgres checkpointer |
| Agent for every query | Cost + latency | RAG graph for doc Q&A |
| No warmup | Cold first request | `warmup_graph()` in lifespan |
| Rename nodes after deploy | Broken checkpoints | Treat names as API |
