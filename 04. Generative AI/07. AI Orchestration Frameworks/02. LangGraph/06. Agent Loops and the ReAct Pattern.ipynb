# Agent Loops and the ReAct Pattern

An **agent** alternates between **reasoning** (LLM decides what to do) and **acting** (execute tools, observe results). This **ReAct** loop — **Reason + Act** — is the core pattern behind **`create_agent`** (**01. LangChain/13**). Notebook **04** previewed the cycle; here you build it **explicitly** in LangGraph.

This notebook covers **`MessagesState`**, model and tool nodes, the **`should_continue`** router, step limits, multi-tool traces, streaming the loop, comparison with **`create_agent`**, and a preview of **`ToolNode`** (**07**).

## 1. The ReAct Loop

**ReAct** (Yao et al.) interleaves natural-language **thought** with **tool actions** and **observations**:

```
User question
     │
     ▼
┌─────────┐     tool_calls?     ┌─────────┐
│  model  │ ──── yes ─────────► │  tools  │
│ (reason)│                       │  (act)  │
└────┬────┘                       └────┬────┘
     │                                 │
     │ no tool_calls                   │ ToolMessage
     ▼                                 │
    END ◄──────────────────────────────┘
              (loop back to model)
```

| Step | Message type | Who produces it |
|------|--------------|-----------------|
| User input | `HumanMessage` | Caller |
| Model decides | `AIMessage` (+ optional `tool_calls`) | LLM |
| Tool result | `ToolMessage` | Tool node |
| Final answer | `AIMessage` (text, no `tool_calls`) | LLM |

LangGraph models this as a **cycle**: `model → tools → model → … → END`.

---

## 2. Agent State — Messages + Counters

Agents use **`MessagesState`** or an equivalent **`messages`** key with **`add_messages`** reducer (**02**). Add **`step_count`** to cap runaway loops (**14**):

````python
from typing import Annotated
from typing_extensions import TypedDict

from langgraph.graph.message import add_messages


class ReActState(TypedDict):
    messages: Annotated[list, add_messages]
    step_count: int
````

**`add_messages`** appends new messages and **updates by message id** when the same id is resent — important for checkpointing and edits (**08**).

## 3. Fake Agent Loop (No API)

Understand the graph **before** adding a live LLM — same pattern as **04**, expanded:

````python
from langchain_core.messages import AIMessage, HumanMessage, ToolMessage
from langgraph.graph import END, START, StateGraph


def fake_model(state: ReActState) -> dict:
    last = state["messages"][-1]
    text = last.content.lower() if hasattr(last, "content") else ""
    # Second visit: we already have a ToolMessage — answer directly
    if any(isinstance(m, ToolMessage) for m in state["messages"]):
        ai = AIMessage(content="Run alembic upgrade head after pulling migrations [c2].")
    elif "alembic" in text:
        ai = AIMessage(
            content="",
            tool_calls=[{
                "name": "search_docs",
                "args": {"query": "alembic"},
                "id": "call_1",
                "type": "tool_call",
            }],
        )
    else:
        ai = AIMessage(content="Ask about Notes API docs — JWT, Alembic, pytest.")
    return {"messages": [ai], "step_count": state.get("step_count", 0) + 1}


def fake_tools(state: ReActState) -> dict:
    return {"messages": [ToolMessage(
        content="[c2] Run alembic upgrade head after pulling revisions.",
        tool_call_id="call_1",
    )]}


def should_continue(state: ReActState):
    last = state["messages"][-1]
    if isinstance(last, AIMessage) and last.tool_calls:
        return "tools"
    return END


fb = StateGraph(ReActState)
fb.add_node("model", fake_model)
fb.add_node("tools", fake_tools)
fb.add_edge(START, "model")
fb.add_conditional_edges("model", should_continue)
fb.add_edge("tools", "model")

fake_agent = fb.compile()
out = fake_agent.invoke({"messages": [HumanMessage(content="Alembic migration command?")], "step_count": 0})
print([m.type for m in out["messages"]])
print("answer:", out["messages"][-1].content)
````

Trace: **`human → ai (tool_calls) → tool → ai (final)`** — the canonical ReAct path.

## 4. Tools for the Notes API Agent

Reuse the teaching corpus from **01. LangChain/12**:

````python
from langchain_core.tools import tool
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-4o-mini", api_key=OPENAI_API_KEY, temperature=0)

NOTES = {
    "n1": "The Notes API is built with FastAPI. Routes live under /notes.",
    "n2": "Run alembic upgrade head after pulling database migrations.",
    "n3": "JWT bearer tokens use Authorization: Bearer header.",
    "n4": "Pytest fixtures for DB setup belong in conftest.py.",
}

DOC_CHUNKS = [
    {"id": "c1", "text": NOTES["n1"]},
    {"id": "c2", "text": NOTES["n2"]},
    {"id": "c3", "text": NOTES["n3"]},
    {"id": "c4", "text": NOTES["n4"]},
]


@tool
def get_note(note_id: str) -> str:
    """Fetch a note by id (n1, n2, n3, n4)."""
    return NOTES.get(note_id, f"Unknown note id: {note_id}")


@tool
def search_docs(query: str) -> str:
    """Keyword search over documentation chunks."""
    tokens = set(query.lower().split())
    hits = []
    for chunk in DOC_CHUNKS:
        if tokens & set(chunk["text"].lower().split()):
            hits.append(f"[{chunk['id']}] {chunk['text']}")
    return "\n".join(hits) if hits else "No matching chunks."


@tool
def add(a: int, b: int) -> int:
    """Add two integers."""
    return a + b


tools = [get_note, search_docs, add]
TOOLS_BY_NAME = {t.name: t for t in tools}
llm_with_tools = llm.bind_tools(tools)
````

---

## 5. Model Node — `bind_tools` + Invoke

The **model node** calls the LLM with the full message history:

````python
SYSTEM = """You are a Notes API assistant.
Use search_docs and get_note before answering factual questions.
Use add for arithmetic. Cite chunk ids like [c2] when possible."""


def call_model(state: ReActState) -> dict:
    messages = [{"role": "system", "content": SYSTEM}] + state["messages"]
    response = llm_with_tools.invoke(messages)
    return {"messages": [response], "step_count": state.get("step_count", 0) + 1}
````

Inject **`system_prompt`** as the first message — same effect as **`create_agent(system_prompt=...)`**.

---

## 6. Tool Node — Execute `tool_calls`

Mirror **`execute_tool_calls`** from **01. LangChain/12**:

````python
def execute_tool_calls(ai_message: AIMessage) -> list[ToolMessage]:
    results = []
    for tc in ai_message.tool_calls:
        tool = TOOLS_BY_NAME.get(tc["name"])
        try:
            if tool is None:
                content = f"Unknown tool: {tc['name']}"
            else:
                content = str(tool.invoke(tc["args"]))
        except Exception as exc:
            content = f"Tool error: {exc}"
        results.append(ToolMessage(content=content, tool_call_id=tc["id"]))
    return results


def run_tools(state: ReActState) -> dict:
    last = state["messages"][-1]
    if not isinstance(last, AIMessage) or not last.tool_calls:
        return {}
    return {"messages": execute_tool_calls(last)}
````

**Rules:** one **`ToolMessage`** per **`tool_call`**, matching **`tool_call_id`**. Parallel tool calls in one **`AIMessage`** are all executed here.

---

## 7. Router with Step Limit

Prevent infinite loops with **`MAX_STEPS`** (**14**):

````python
MAX_STEPS = 8


def should_continue_limited(state: ReActState):
    if state.get("step_count", 0) >= MAX_STEPS:
        return END
    last = state["messages"][-1]
    if isinstance(last, AIMessage) and last.tool_calls:
        return "tools"
    return END
````

---

## 8. Build the ReAct Graph

Assemble nodes, cycle, and compile:

````python
def build_react_agent():
    builder = StateGraph(ReActState)
    builder.add_node("model", call_model)
    builder.add_node("tools", run_tools)
    builder.add_edge(START, "model")
    builder.add_conditional_edges("model", should_continue_limited)
    builder.add_edge("tools", "model")
    return builder.compile()


react_agent = build_react_agent()
print("nodes:", list(react_agent.get_graph().nodes))
````

This is the **minimal agent graph** — the same topology **`create_agent`** compiles under the hood.

---

## 9. Invoke and Read the Trace

Input: **`{"messages": [HumanMessage(...)], "step_count": 0}`**

````python
def last_ai_text(result: dict) -> str:
    for msg in reversed(result["messages"]):
        if isinstance(msg, AIMessage) and msg.content:
            return msg.content if isinstance(msg.content, str) else str(msg.content)
    return ""


def print_trace(result: dict) -> None:
    for i, msg in enumerate(result["messages"]):
        if isinstance(msg, AIMessage) and msg.tool_calls:
            names = [tc["name"] for tc in msg.tool_calls]
            print(f"{i:02d} ai     tool_calls={names}")
        elif isinstance(msg, ToolMessage):
            print(f"{i:02d} tool   {str(msg.content)[:70]}...")
        else:
            print(f"{i:02d} {msg.type:6s} {str(msg.content)[:70]}...")


result = react_agent.invoke({
    "messages": [HumanMessage(content="What pytest file holds DB fixtures? Use tools.")],
    "step_count": 0,
})

print_trace(result)
print("\nanswer:", last_ai_text(result))
print("steps:", result["step_count"])
````

## 10. Multi-Tool and Multi-Step Runs

The model may call several tools across multiple loop iterations:

````python
multi = react_agent.invoke({
    "messages": [HumanMessage(
        content="Search docs for jwt, then get note n3. Also what is 12 + 30?"
    )],
    "step_count": 0,
})

tool_rounds = sum(
    1 for m in multi["messages"]
    if isinstance(m, AIMessage) and m.tool_calls
)
print(f"tool rounds: {tool_rounds}, total messages: {len(multi['messages'])}")
print_trace(multi)
````

## 11. Stream the Agent Loop

Watch each **`model`** / **`tools`** iteration (**13**):

````python
print("Streaming agent steps:")
for update in react_agent.stream(
    {"messages": [HumanMessage(content="Alembic upgrade command?")], "step_count": 0},
    stream_mode="updates",
):
    node = list(update.keys())[0]
    patch = update[node]
    if "messages" in patch:
        last = patch["messages"][-1]
        kind = last.type
        extra = ""
        if isinstance(last, AIMessage) and last.tool_calls:
            extra = f" calls={[tc['name'] for tc in last.tool_calls]}"
        print(f"  {node}: {kind}{extra}")
````

## 12. Manual Graph vs `create_agent`

**01. LangChain/13** hides the wiring — compare topology:

````python
from langchain.agents import create_agent

factory_agent = create_agent(
    model=llm,
    tools=tools,
    system_prompt=SYSTEM,
)

print("create_agent nodes:", list(factory_agent.get_graph().nodes))
print("manual react nodes:", list(react_agent.get_graph().nodes))

factory_result = factory_agent.invoke({
    "messages": [HumanMessage(content="JWT header format? Use tools.")],
})
print("\nfactory answer:", last_ai_text(factory_result))
````

| | **Manual ReAct graph** | **`create_agent`** |
|--|------------------------|-------------------|
| **Control** | Full — add nodes around loop | Factory defaults |
| **Tool node** | You write `run_tools` | Built-in |
| **Middleware** | Custom nodes / wrappers | `middleware=[...]` param |
| **When to use** | Custom RAG, HITL, graders (**05**, **09**, **12**) | Standard tool agents |

Start with **`create_agent`** for simple agents; drop to manual graphs when you need **extra nodes inside the loop**.

## 13. Prebuilt `ToolNode` (Preview)

**07. Tool Nodes and Tool Execution** covers this in depth. LangGraph ships a ready-made tool executor:

````python
from langgraph.prebuilt import ToolNode, tools_condition


def build_react_with_toolnode():
    builder = StateGraph(ReActState)
    builder.add_node("model", call_model)
    builder.add_node("tools", ToolNode(tools))
    builder.add_edge(START, "model")
    builder.add_conditional_edges("model", tools_condition)
    builder.add_edge("tools", "model")
    return builder.compile()


toolnode_agent = build_react_with_toolnode()
tn_result = toolnode_agent.invoke({
    "messages": [HumanMessage(content="What note id covers JWT?")],
    "step_count": 0,
})
print("ToolNode answer:", last_ai_text(tn_result))
````

**`tools_condition`** is the standard router: `tool_calls` → `"tools"`, else → **`END`**. Pair with your own step-limit wrapper in production.

## 15. Design Guidelines

| Guideline | Rationale |
|-----------|----------|
| **Cap `step_count`** | Models can loop forever |
| **System prompt in model node** | Consistent instructions each turn |
| **Tool errors → ToolMessage** | Model can recover; don't crash graph |
| **Log full message trace** | Debug routing and bad tool args |
| **Keep router thin** | Only inspect last message + counters |
| **One graph instance** | Compile once at startup (**16**) |
| **Prefer `create_agent`** until you need custom nodes | Less boilerplate |

---

## 16. Common Mistakes

| Mistake | Consequence | Fix |
|---------|-------------|-----|
| Missing `tools → model` edge | Graph stops after one tool call | Add cycle edge |
| Wrong `tool_call_id` | Model ignores tool result | Match `id` from `tool_calls` |
| No `bind_tools` on LLM | Empty `tool_calls` | `llm.bind_tools(tools)` |
| Router checks wrong message | Skips or repeats tools | Inspect **last** message only |
| No step limit | Runaway API cost | `MAX_STEPS` in router |
| Tool node runs on non-AI message | Crash | Guard `isinstance(last, AIMessage)` |
| Recompile graph per request | Slow | `build_react_agent()` once |

---

## 17. Summary

The **ReAct loop** is a **`model ↔ tools`** cycle terminated when the model returns an **`AIMessage`** without **`tool_calls`**. LangGraph implements it with **`add_conditional_edges`** on the model node and a fixed edge from **tools back to model**.

Key takeaways:

- **`MessagesState`** + **`add_messages`** accumulates the conversation trace.
- **Model node** = `bind_tools` + invoke; **tool node** = execute `tool_calls` → `ToolMessage`.
- **`should_continue`** routes to **`tools`** or **`END`**; add **`step_count`** limits.
- **`create_agent`** is the same topology with factory wiring (**13**).
- **`ToolNode`** + **`tools_condition`** replace hand-written tool execution (**07**).
