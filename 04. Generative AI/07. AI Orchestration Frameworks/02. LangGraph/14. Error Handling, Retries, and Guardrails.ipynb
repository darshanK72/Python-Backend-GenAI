# Error Handling, Retries, and Guardrails

Agents fail — tools throw, models rate-limit, users send blocked prompts, and loops run forever. Production graphs need **structured recovery**: errors as **`ToolMessage`** (**07**), **retry** wrappers, **step limits**, **guard** nodes, **fallback** routes, and **error traces** in state.

This notebook covers tool error handling, tenacity-style retry loops inside nodes, **`MAX_STEPS`** / **`tool_calls_count`** guards, blocked-term guards, fallback models, conditional routing on error flags, **`with_retry`** on LLMs, a circuit-breaker sketch, and logging errors to a **`trace`** list.

## 1. Failure Modes in Agent Graphs

```
User input
     │
     ▼
┌─────────┐   blocked term   ┌──────────┐
│  guard  │ ───────────────► │ fallback │
└────┬────┘                  └────┬─────┘
     │ ok                          │
     ▼                             ▼
┌─────────┐   tool error    ┌──────────┐
│  model  │ ◄── ToolMessage │  tools   │
└────┬────┘                  └──────────┘
     │ rate limit / timeout
     ▼
┌─────────┐   still failing  ┌──────────┐
│  retry  │ ───────────────► │ circuit  │
│  node   │                  │ breaker  │
└─────────┘                  └──────────┘
```

| Failure | Symptom | Graph-level response |
|---------|---------|---------------------|
| **Tool exception** | Crash or empty result | Catch → `ToolMessage` with error text |
| **LLM timeout** | Node raises | Retry node or `with_retry` |
| **Runaway loop** | Many tool rounds | `MAX_STEPS` / `tool_calls_count` |
| **Policy violation** | Blocked prompt | Guard node → static fallback |
| **Repeated outages** | All retries fail | Circuit breaker → fallback |

## 2. Shared State — Messages, Counters, Errors, Trace

Extend agent state with fields routers and observability nodes read:

````python
from typing_extensions import TypedDict

from langgraph.graph.message import add_messages


def append_trace(existing: list[str], new: list[str]) -> list[str]:
    return existing + new


class ResilientState(TypedDict):
    messages: Annotated[list, add_messages]
    step_count: int
    tool_calls_count: int
    error: str | None
    blocked: bool
    trace: Annotated[list[str], append_trace]
````

**`error`** and **`blocked`** drive conditional routes. **`trace`** accumulates node names and error strings for debugging (**15**).

## 3. Tool Errors as `ToolMessage` (from **07**)

Never let a tool exception crash the graph — return a **`ToolMessage`** so the model can recover:

````python
from langchain_core.messages import AIMessage, HumanMessage, ToolMessage
from langchain_core.tools import tool
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-4o-mini", api_key=OPENAI_API_KEY, temperature=0)

NOTES = {
    "n1": "The Notes API is built with FastAPI. Routes live under /notes.",
    "n2": "Run alembic upgrade head after pulling database migrations.",
    "n3": "JWT bearer tokens use Authorization: Bearer header.",
    "n4": "Pytest fixtures for DB setup belong in conftest.py.",
}


@tool
def get_note(note_id: str) -> str:
    """Fetch a note by id (n1–n4)."""
    if note_id == "n0":
        raise ValueError("Simulated database timeout")
    return NOTES.get(note_id, f"Unknown note id: {note_id}")


@tool
def search_docs(query: str) -> str:
    """Keyword search over Notes API documentation."""
    tokens = set(query.lower().split())
    hits = [f"[{k}] {v}" for k, v in NOTES.items() if tokens & set(v.lower().split())]
    return "\n".join(hits) if hits else "No matching chunks."


TOOLS = [get_note, search_docs]
TOOLS_BY_NAME = {t.name: t for t in TOOLS}


def execute_tool_calls(ai_message: AIMessage, trace: list[str]) -> tuple[list[ToolMessage], list[str]]:
    """Execute tool_calls; never raise — errors become ToolMessage content."""
    results = []
    new_trace = list(trace)
    for tc in ai_message.tool_calls:
        tool_fn = TOOLS_BY_NAME.get(tc["name"])
        try:
            if tool_fn is None:
                content = f"Unknown tool: {tc['name']}"
                new_trace.append(f"tool_error:unknown:{tc['name']}")
            else:
                content = str(tool_fn.invoke(tc["args"]))
        except Exception as exc:
            content = f"Tool error ({tc['name']}): {exc}"
            new_trace.append(f"tool_error:{tc['name']}:{exc}")
        results.append(ToolMessage(content=content, tool_call_id=tc["id"]))
    return results, new_trace


# Demo: fake AIMessage with a failing tool call
fake_ai = AIMessage(
    content="",
    tool_calls=[{"name": "get_note", "args": {"note_id": "n0"}, "id": "call_fail", "type": "tool_call"}],
)
msgs, tr = execute_tool_calls(fake_ai, ["tools"])
print("ToolMessage content:", msgs[0].content)
print("trace:", tr)
````

The model sees **`Tool error (get_note): Simulated database timeout`** and can retry with a different `note_id` or explain the failure to the user.

## 4. Retry Node — Tenacity-Style Loop Inside a Node

Wrap LLM calls with **exponential backoff** inside a single node — no extra graph edges needed:

````python
MAX_RETRIES = 3
BASE_DELAY = 0.5  # seconds; use 1–2 in production


def invoke_with_retry(invoke_fn, *args, **kwargs):
    """Tenacity-style retry loop without external dependency."""
    last_exc = None
    for attempt in range(1, MAX_RETRIES + 1):
        try:
            return invoke_fn(*args, **kwargs), attempt
        except Exception as exc:
            last_exc = exc
            if attempt < MAX_RETRIES:
                delay = BASE_DELAY * (2 ** (attempt - 1))
                time.sleep(delay)
    raise last_exc  # type: ignore[misc]


def retry_model_node(state: ResilientState) -> dict:
    llm_bound = llm.bind_tools(TOOLS)
    system = {"role": "system", "content": "You are a Notes API assistant. Use tools when needed."}
    messages = [system] + state["messages"]
    trace = state.get("trace", [])
    try:
        response, attempts = invoke_with_retry(llm_bound.invoke, messages)
        trace = trace + [f"model:ok:attempts={attempts}"]
        return {
            "messages": [response],
            "step_count": state.get("step_count", 0) + 1,
            "error": None,
            "trace": trace,
        }
    except Exception as exc:
        return {
            "error": str(exc),
            "trace": trace + [f"model:failed:{exc}"],
            "step_count": state.get("step_count", 0) + 1,
        }
````

Set **`error`** in state when all retries exhaust — a downstream router sends the run to a **fallback** node.

## 5. `MAX_STEPS` and `tool_calls_count` Guards

Cap total graph iterations and tool rounds separately — a model can call many tools in one step:

````python
MAX_STEPS = 8
MAX_TOOL_CALLS = 12


def run_tools_guarded(state: ResilientState) -> dict:
    last = state["messages"][-1]
    if not isinstance(last, AIMessage) or not last.tool_calls:
        return {}
    results, new_trace = execute_tool_calls(last, state.get("trace", []))
    return {
        "messages": results,
        "tool_calls_count": state.get("tool_calls_count", 0) + len(last.tool_calls),
        "trace": new_trace,
    }


def route_after_model(state: ResilientState) -> Literal["tools", "fallback", "end"]:
    if state.get("error"):
        return "fallback"
    if state.get("step_count", 0) >= MAX_STEPS:
        return "fallback"
    if state.get("tool_calls_count", 0) >= MAX_TOOL_CALLS:
        return "fallback"
    last = state["messages"][-1]
    if isinstance(last, AIMessage) and last.tool_calls:
        return "tools"
    return "end"
````

**`step_count`** increments each model visit; **`tool_calls_count`** increments per tool invocation batch.

## 6. Guard Node — Blocked Terms Before the Model

Run policy checks **before** any LLM call — same pattern as **05**'s `guard_node`:

````python
BLOCKED_TERMS = ("password", "api_key", "secret", "ssn", "credit card")


def guard_node(state: ResilientState) -> dict:
    last_human = ""
    for msg in reversed(state["messages"]):
        if isinstance(msg, HumanMessage):
            last_human = msg.content.lower() if isinstance(msg.content, str) else ""
            break
    blocked = any(term in last_human for term in BLOCKED_TERMS)
    trace = state.get("trace", []) + (["guard:blocked"] if blocked else ["guard:ok"])
    return {"blocked": blocked, "trace": trace}


def route_guard(state: ResilientState) -> Literal["model", "blocked_fallback"]:
    return "blocked_fallback" if state.get("blocked") else "model"
````

Extend with regex, LLM classifiers, or external policy APIs for production.

## 7. Fallback — Smaller Model or Static Message

When **`error`**, **`blocked`**, or limits trigger, route to a **cheap static response** or a **smaller model**:

````python
fallback_llm = ChatOpenAI(model="gpt-4o-mini", api_key=OPENAI_API_KEY, temperature=0)


def blocked_fallback_node(state: ResilientState) -> dict:
    msg = AIMessage(
        content="I cannot process requests involving passwords, API keys, or secrets. "
                "Ask about Notes API docs — FastAPI routes, Alembic, JWT, or pytest."
    )
    return {"messages": [msg], "trace": state.get("trace", []) + ["blocked_fallback"]}


def error_fallback_node(state: ResilientState) -> dict:
    reason = state.get("error") or "step/tool limit reached"
    # Option A: static message (fast, deterministic)
    static = AIMessage(
        content=f"Sorry, I could not complete your request ({reason}). "
                "Try rephrasing or ask about Notes API documentation."
    )
    # Option B: smaller/cheaper model summarizing failure (uncomment to use)
    # summary = fallback_llm.invoke([{"role": "user", "content": f"Apologize briefly: {reason}"}])
    # static = summary
    return {"messages": [static], "trace": state.get("trace", []) + [f"error_fallback:{reason}"]}
````

| Fallback type | When | Trade-off |
|---------------|------|----------|
| **Static `AIMessage`** | Policy block, hard limits | Zero extra LLM cost |
| **Smaller model** | Soft failures, partial context | Better UX, small cost |
| **Human handoff** | High-stakes errors (**09**) | Slowest, safest |

## 8. Conditional Route on `error` Flag in State

After the retry model node, the router inspects **`error`** before continuing the tool loop:

````python
from langgraph.graph import END, START, StateGraph


def route_after_model_to_end(state: ResilientState) -> str:
    route = route_after_model(state)
    if route == "end":
        return END
    return route  # "tools" or "fallback"


def build_resilient_agent():
    builder = StateGraph(ResilientState)
    builder.add_node("guard", guard_node)
    builder.add_node("model", retry_model_node)
    builder.add_node("tools", run_tools_guarded)
    builder.add_node("blocked_fallback", blocked_fallback_node)
    builder.add_node("error_fallback", error_fallback_node)

    builder.add_edge(START, "guard")
    builder.add_conditional_edges("guard", route_guard, {"model": "model", "blocked_fallback": "blocked_fallback"})
    builder.add_conditional_edges("model", route_after_model_to_end, {"tools": "tools", "fallback": "error_fallback", END: END})
    builder.add_edge("tools", "model")
    builder.add_edge("blocked_fallback", END)
    builder.add_edge("error_fallback", END)
    return builder.compile()


resilient_agent = build_resilient_agent()
print("nodes:", list(resilient_agent.get_graph().nodes))
````

```
START → guard ──blocked──► blocked_fallback → END
          │
          └──ok──► model ──tools──► tools ──► model …
                    │ error/limit
                    └──► error_fallback → END
```

## 9. LangChain `with_retry` on the Model

LangChain Runnables support **`.with_retry()`** — alternative to hand-written retry loops:

````python
from langchain_core.runnables import RunnableLambda

llm_with_retry = llm.bind_tools(TOOLS).with_retry(
    stop_after_attempt=3,
    wait_exponential_jitter=True,
)


def model_with_runnable_retry(state: ResilientState) -> dict:
    system = {"role": "system", "content": "Notes API assistant."}
    try:
        response = llm_with_retry.invoke([system] + state["messages"])
        return {
            "messages": [response],
            "step_count": state.get("step_count", 0) + 1,
            "error": None,
            "trace": state.get("trace", []) + ["model:with_retry:ok"],
        }
    except Exception as exc:
        return {
            "error": str(exc),
            "trace": state.get("trace", []) + [f"model:with_retry:failed:{exc}"],
            "step_count": state.get("step_count", 0) + 1,
        }

print("with_retry configured:", llm_with_retry)
````

Use **`with_retry`** when the LLM is a Runnable in an LCEL chain inside a node. Use a **retry node** when you need custom logging or state updates per attempt.

## 10. Circuit Breaker Pattern (Sketch)

After repeated failures, **stop calling** the primary model and serve fallback immediately:

````python
class CircuitBreaker:
  """In-memory circuit breaker — use Redis in production."""

  def __init__(self, failure_threshold: int = 3, cooldown_seconds: float = 30.0):
      self.failure_threshold = failure_threshold
      self.cooldown_seconds = cooldown_seconds
      self.failures = 0
      self.opened_at: float | None = None

  @property
  def is_open(self) -> bool:
      if self.opened_at is None:
          return False
      if time.time() - self.opened_at > self.cooldown_seconds:
          self.failures = 0
          self.opened_at = None
          return False
      return True

  def record_success(self) -> None:
      self.failures = 0
      self.opened_at = None

  def record_failure(self) -> None:
      self.failures += 1
      if self.failures >= self.failure_threshold:
          self.opened_at = time.time()


breaker = CircuitBreaker(failure_threshold=2, cooldown_seconds=5.0)


def model_with_breaker(state: ResilientState) -> dict:
    trace = state.get("trace", [])
    if breaker.is_open:
        return {
            "error": "circuit_open",
            "trace": trace + ["circuit:open"],
            "step_count": state.get("step_count", 0) + 1,
        }
    try:
        response = llm.bind_tools(TOOLS).invoke(
            [{"role": "system", "content": "Notes API assistant."}] + state["messages"]
        )
        breaker.record_success()
        return {"messages": [response], "error": None, "trace": trace + ["circuit:closed"], "step_count": state.get("step_count", 0) + 1}
    except Exception as exc:
        breaker.record_failure()
        return {"error": str(exc), "trace": trace + [f"circuit:failure:{exc}"], "step_count": state.get("step_count", 0) + 1}

# Simulate failures opening the circuit
breaker.record_failure()
breaker.record_failure()
print("circuit open:", breaker.is_open)
````

Wire **`circuit_open`** into **`route_after_model`** → **`error_fallback`**. Share breaker state across requests via Redis when deploying (**16**).

## 11. Logging Errors to `trace` in State

Append structured strings every step — stream them with **`stream_mode="updates"`** (**13**) or export to **15**:

````python
def print_result_trace(result: dict) -> None:
    print("trace:")
    for entry in result.get("trace", []):
        prefix = "  ⚠ " if "error" in entry or "failed" in entry or "blocked" in entry else "  · "
        print(f"{prefix}{entry}")
    last = result["messages"][-1]
    print(f"\nfinal ({last.type}):", str(last.content)[:120], "...")


INITIAL = {
    "messages": [],
    "step_count": 0,
    "tool_calls_count": 0,
    "error": None,
    "blocked": False,
    "trace": [],
}
````

## 13. Combining Patterns — Reference Architecture

| Layer | Pattern | State keys |
|-------|---------|------------|
| **Input** | Guard node | `blocked`, `trace` |
| **Model** | Retry / `with_retry` / circuit breaker | `error`, `step_count` |
| **Tools** | try/except → `ToolMessage` | `tool_calls_count`, `trace` |
| **Router** | `route_after_model` | reads `error`, limits, `tool_calls` |
| **Output** | Fallback nodes | final `AIMessage` |

All layers compose in one **`StateGraph`** — no single middleware replaces graph-level routing.

---

## 14. Design Guidelines

| Guideline | Rationale |
|-----------|----------|
| **Errors → `ToolMessage`, not raises** | Model can self-correct |
| **Set `error` flag for graph routing** | Don't rely on exceptions crossing nodes |
| **Separate `step_count` and `tool_calls_count`** | Different limit semantics |
| **Static fallback for policy blocks** | Predictable, auditable |
| **Log to `trace`, not stdout** | Streamable, checkpointable (**08**) |
| **Circuit breaker at service scope** | Protects provider quota |
| **Test each branch** | guard, tool error, retry exhaust (**05** matrix) |

---

## 15. Common Mistakes

| Mistake | Consequence | Fix |
|---------|-------------|-----|
| Tool raises uncaught | Whole graph fails | try/except → `ToolMessage` |
| Retry without backoff | Rate-limit storm | Exponential delay or `with_retry` |
| Only `step_count`, no tool cap | 1 step, 50 parallel tools | Track `tool_calls_count` |
| Guard after model | Tokens spent on blocked input | Guard as first node |
| Fallback still calls main LLM | Cost on failure path | Static message for hard errors |
| `error` flag never cleared | Stale route to fallback | Set `error=None` on success |
| Circuit breaker per request | Never opens globally | Shared store (Redis) |
| No trace on errors | Undebuggable production | Append to `trace` every branch |

---

## 16. Summary

Resilient LangGraph agents **catch failures inside nodes**, **record them in state**, and **route** to fallbacks instead of crashing. Tool errors become **`ToolMessage`** content; LLM failures set **`error`** for conditional edges; guards block policy violations before any model call.

Key takeaways:

- **Tool try/except → `ToolMessage`** lets the model recover (**07**).
- **Retry loops** (manual or **`with_retry`**) handle transient LLM failures.
- **`MAX_STEPS`** and **`tool_calls_count`** prevent runaway agent loops.
- **Guard nodes** + **fallback nodes** enforce policy without extra LLM cost.
- **Circuit breakers** protect upstream providers after repeated outages.
- **`trace`** in state feeds streaming UIs and observability (**13**, **15**).
