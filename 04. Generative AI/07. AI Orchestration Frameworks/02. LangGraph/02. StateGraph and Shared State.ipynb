# StateGraph and Shared State

Every LangGraph workflow centers on **shared state** — a typed object that flows through **nodes**. Each node **reads** the current state and returns a **partial update**; LangGraph **merges** those updates before routing to the next node. **01. Introduction to LangGraph** built a minimal `PipelineState`; this notebook goes deep: schemas, merge semantics, **reducers**, **`add_messages`**, **`MessagesState`**, and designing state for agents and RAG.

Prerequisites: **01. Introduction to LangGraph**. Node wiring and compilation details continue in **03. Nodes, Edges, and Graph Compilation**.

## 1. Shared State — The Graph's Memory

Unlike LCEL, where data passes **along a pipe**, LangGraph keeps one **central state object**:

```
invoke(initial_state)
       │
       ▼
   ┌─────────┐   patch    ┌─────────┐   patch    ┌─────────┐
   │ node A  │ ────────► │  state  │ ◄──────── │ node B  │
   └─────────┘           └─────────┘           └─────────┘
                             ▲
                             │ merge
                         partial update
```

| Concept | Meaning |
|---------|--------|
| **State schema** | Declares keys and types (`TypedDict`, `MessagesState`) |
| **Node input** | Full current state (read any key) |
| **Node output** | **Partial** dict — only keys that changed |
| **Merge** | LangGraph applies updates key-by-key |
| **Reducer** | Custom merge function for a key (e.g. append messages) |

Good state design makes graphs **testable** — nodes are pure functions of state.

---

## 2. State Schema with TypedDict

Define allowed keys explicitly — same Notes API teaching context as **01**:

````python
from typing_extensions import TypedDict

from langgraph.graph import END, START, StateGraph


class NotesState(TypedDict):
    question: str
    context: str
    answer: str
    step_count: int


def ingest(state: NotesState) -> dict:
    return {
        "context": "FastAPI Notes API under /notes; Alembic for migrations.",
        "step_count": state.get("step_count", 0) + 1,
    }


def draft_answer(state: NotesState) -> dict:
    return {
        "answer": f"Based on: {state['context'][:50]}... → re: {state['question']}",
        "step_count": state["step_count"] + 1,
    }


builder = StateGraph(NotesState)
builder.add_node("ingest", ingest)
builder.add_node("draft", draft_answer)
builder.add_edge(START, "ingest")
builder.add_edge("ingest", "draft")
builder.add_edge("draft", END)

graph = builder.compile()

out = graph.invoke({
    "question": "What handles migrations?",
    "context": "",
    "answer": "",
    "step_count": 0,
})
print(json.dumps(out, indent=2))
````

**Rule:** nodes return **only changed keys** — `ingest` does not repeat `question`; it stays from the initial invoke.

## 3. Default Merge — Replace Scalars

Without a custom reducer, returning a key **replaces** its previous value:

````python
class CounterState(TypedDict):
    value: int
    label: str


def set_value(state: CounterState) -> dict:
    return {"value": state["value"] + 1}


def set_label(state: CounterState) -> dict:
    return {"label": f"count={state['value']}"}


cb = StateGraph(CounterState)
cb.add_node("inc", set_value)
cb.add_node("label", set_label)
cb.add_edge(START, "inc")
cb.add_edge("inc", "label")
cb.add_edge("label", END)

counter_graph = cb.compile()
print(counter_graph.invoke({"value": 0, "label": ""}))
````

`label` node sees `value=1` because **`inc`** ran first and merged before **`label`** executed.

## 4. The Problem with Naive Lists

Appending to lists **without reducers** requires reading old state and returning the full new list:

````python
class LogState(TypedDict):
    log: list[str]


def step_a(state: LogState) -> dict:
    return {"log": state.get("log", []) + ["step_a"]}


def step_b(state: LogState) -> dict:
    return {"log": state.get("log", []) + ["step_b"]}


lb = StateGraph(LogState)
lb.add_node("a", step_a)
lb.add_node("b", step_b)
lb.add_edge(START, "a")
lb.add_edge("a", "b")
lb.add_edge("b", END)

log_graph = lb.compile()
print(log_graph.invoke({"log": []}))
````

This works for **sequential** nodes but breaks down when:

- Multiple nodes update the same list in **parallel** (future map-reduce shapes)
- You want to return **only new items** (`["step_a"]`) instead of the full list
- Message lists need **deduplication by id** (tool loops)

**Reducers** solve this.

---

## 5. Reducers — Annotated Keys

Attach a **reducer function** to a state key with **`Annotated[type, reducer]`**:

````python
class ReducerLogState(TypedDict):
    log: Annotated[list[str], operator.add]


def node_a(state: ReducerLogState) -> dict:
    return {"log": ["step_a"]}  # only the NEW items


def node_b(state: ReducerLogState) -> dict:
    return {"log": ["step_b"]}


rb = StateGraph(ReducerLogState)
rb.add_node("a", node_a)
rb.add_node("b", node_b)
rb.add_edge(START, "a")
rb.add_edge("a", "b")
rb.add_edge("b", END)

reducer_graph = rb.compile()
print(reducer_graph.invoke({"log": []}))
````

**`operator.add`** on lists concatenates: `[] + ["step_a"] + ["step_b"]`. Nodes return **deltas**, not full history — cleaner and parallel-safe.

### 5.1 Common Reducers

| Reducer | Use case |
|---------|----------|
| **`operator.add`** | Append to lists (logs, tags, simple accumulators) |
| **`add_messages`** | Merge LangChain message lists (agents) |
| **Custom callable** | Max, set-union, last-wins-with-timestamp |
| **(none)** | Replace scalar or whole value |

## 6. add_messages — The Agent State Key

Agent graphs store **`list[BaseMessage]`**. The **`add_messages`** reducer appends new messages and updates existing ones by **`id`**:

````python
from langchain_core.messages import AIMessage, HumanMessage
from langgraph.graph.message import add_messages


class MessageState(TypedDict):
    messages: Annotated[list, add_messages]


def user_node(state: MessageState) -> dict:
    return {"messages": [HumanMessage(content="What is Alembic?")]}


def assistant_node(state: MessageState) -> dict:
    return {"messages": [AIMessage(content="Alembic manages SQLAlchemy migrations.")]}


mb = StateGraph(MessageState)
mb.add_node("user", user_node)
mb.add_node("assistant", assistant_node)
mb.add_edge(START, "user")
mb.add_edge("user", "assistant")
mb.add_edge("assistant", END)

msg_graph = mb.compile()
result = msg_graph.invoke({"messages": []})

for m in result["messages"]:
    print(f"[{m.type}] {m.content}")
````

This matches how **`create_agent`** (**01. LangChain/13**) accumulates `HumanMessage` → `AIMessage` → `ToolMessage` across loop iterations. Tool loops: **06. Agent Loops and the ReAct Pattern**.

## 7. MessagesState — Built-in Schema

LangGraph provides **`MessagesState`** with `messages` pre-configured:

````python
from langgraph.graph import MessagesState


class AgentState(MessagesState):
    """Extend MessagesState with extra fields."""
    context: str
    tool_calls_count: int


def retrieve(state: AgentState) -> dict:
    return {
        "context": "[c2] Run alembic upgrade head for migrations.",
        "tool_calls_count": 0,
    }


def model_node(state: AgentState) -> dict:
    from langchain_openai import ChatOpenAI

    llm = ChatOpenAI(model="gpt-4o-mini", api_key=OPENAI_API_KEY, temperature=0)
    prior = state.get("messages", [])
    user_q = prior[-1].content if prior else "Explain migrations."
    prompt = f"Context:\n{state['context']}\n\nQuestion: {user_q}"
    ai = llm.invoke([HumanMessage(content=prompt)])
    return {"messages": [ai]}


ab = StateGraph(AgentState)
ab.add_node("retrieve", retrieve)
ab.add_node("model", model_node)
ab.add_edge(START, "retrieve")
ab.add_edge("retrieve", "model")
ab.add_edge("model", END)

agent_graph = ab.compile()
final = agent_graph.invoke({
    "messages": [HumanMessage(content="What command applies migrations?")],
    "context": "",
    "tool_calls_count": 0,
})
print(final["messages"][-1].content)
````

Pattern: **`MessagesState`** for conversation + extra keys for **RAG context**, **citations**, **counters**, **route labels**.

## 8. RAG-Oriented State Shape

Agentic RAG graphs (**12. Agentic RAG with LangGraph**) often track:

````python
class RagGraphState(TypedDict):
    question: str
    documents: Annotated[list[str], operator.add]
    context: str
    answer: str
    retrieval_attempts: int
    grade: str  # e.g. "relevant" | "irrelevant"


def search(state: RagGraphState) -> dict:
    docs = ["[c2] alembic upgrade head", "[c3] JWT Bearer header"]
    return {
        "documents": docs,
        "context": "\n".join(docs),
        "retrieval_attempts": state.get("retrieval_attempts", 0) + 1,
    }


def grade_docs(state: RagGraphState) -> dict:
    grade = "relevant" if "alembic" in state["question"].lower() else "irrelevant"
    return {"grade": grade}


rg = StateGraph(RagGraphState)
rg.add_node("search", search)
rg.add_node("grade", grade_docs)
rg.add_edge(START, "search")
rg.add_edge("search", "grade")
rg.add_edge("grade", END)

rag_state_graph = rg.compile()
print(rag_state_graph.invoke({
    "question": "Alembic command?",
    "documents": [],
    "context": "",
    "answer": "",
    "retrieval_attempts": 0,
    "grade": "",
}))
````

Conditional routing on **`grade`** is **04. Conditional Routing and Branching** — state holds the router's input.

## 9. Multiple Keys in One Update

A single node may update **several keys** at once:

````python
def enrich(state: NotesState) -> dict:
    return {
        "context": "JWT uses Authorization: Bearer header.",
        "answer": "See security docs.",
        "step_count": state.get("step_count", 0) + 1,
    }


eb = StateGraph(NotesState)
eb.add_node("enrich", enrich)
eb.add_edge(START, "enrich")
eb.add_edge("enrich", END)

print(eb.compile().invoke({
    "question": "JWT header?",
    "context": "",
    "answer": "",
    "step_count": 0,
}))
````

Prefer **one coherent patch** per node over scattering related updates across multiple nodes.

## 11. State Design Guidelines

| Guideline | Rationale |
|-----------|----------|
| **Minimal keys** | Easier testing and checkpoint payloads |
| **Serializable values** | Checkpointers persist JSON-friendly data |
| **Reducers for append-only lists** | `add_messages`, `operator.add` |
| **Scalars for replace** | `answer`, `grade`, `route` |
| **Don't duplicate data** | Store `messages` OR `answer`, not both unless needed |
| **Counters for limits** | `step_count`, `tool_calls_count` for guards (**14**) |
| **Separate input question from messages** | Routers read `question` before chat history grows |

---

## 12. State vs LCEL Data Flow

| | **LCEL** | **LangGraph state** |
|--|----------|---------------------|
| Data shape | Changes type each step | One evolving dict |
| Access | Next step receives prior output | Any node reads full state |
| Branching | `RunnableBranch` on last value | Router reads any state key |
| Loops | Manual | Cycles return to same state |

Wrap LCEL **inside** nodes; use state for **what persists** across the graph.

## 14. Common Mistakes

| Mistake | Consequence | Fix |
|---------|-------------|-----|
| Returning full state dict | Overwrites unrelated keys | Return partial patch only |
| List append without reducer | Verbose; parallel-unsafe | `Annotated[list, operator.add]` |
| Plain list for messages | Broken tool loops | `add_messages` / `MessagesState` |
| Huge state (full docs × N) | Slow checkpoints | Store ids; re-fetch in nodes |
| Non-serializable objects in state | Checkpoint failures | IDs/strings only |
| Typo in key name | Silent missing data | TypedDict + tests |
| Mutable in-place mutation | Race conditions | Return new patch, don't mutate `state` |

## 15. Summary

**Shared state** is the contract between LangGraph nodes: a **schema** (`TypedDict`, **`MessagesState`**), **partial updates** per node, and **reducers** (`add_messages`, **`operator.add`**) for append semantics.

Key takeaways:

- Nodes **read** full state, **write** patches.
- Scalars **replace** by default; lists need **reducers** for delta merges.
- **`add_messages`** is required for agent message history.
- **`MessagesState`** + extra fields is the standard agent pattern.
- RAG graphs add **`context`**, **`documents`**, **`grade`**, attempt counters.
- Design **minimal, serializable** state before writing nodes.
