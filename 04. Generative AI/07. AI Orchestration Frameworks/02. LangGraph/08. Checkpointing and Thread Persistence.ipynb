# Checkpointing and Thread Persistence

Stateless **`invoke`** calls forget everything when the function returns. Real assistants need **multi-turn memory**, **human-in-the-loop resume**, and **crash recovery**. LangGraph **checkpointing** snapshots graph state after each super-step, keyed by **`thread_id`**.

This notebook covers **`MemorySaver`**, **`compile(checkpointer=...)`**, **`configurable.thread_id`**, **`get_state`**, checkpoint listing, message accumulation with **`add_messages`**, production backends (**PostgresSaver**, **SqliteSaver**), metadata, and per-user thread patterns.

**What you'll learn:**

- Why checkpointing matters for conversations, HITL, and resume-after-failure
- How **`thread_id`** selects a persisted conversation thread
- How **`get_state`** inspects live checkpointed state between turns

## 1. Why Checkpointing?

| Need | Without checkpointing | With checkpointing |
|------|----------------------|-------------------|
| **Multi-turn chat** | Resend full history every call | Append one message; state merges |
| **HITL / interrupts** | Manual pause/resume logic | Resume from exact node (**09**) |
| **Crash recovery** | Lost in-flight work | Replay from last checkpoint |
| **Time travel / audit** | None | List and inspect past checkpoints |

```
Turn 1: invoke({messages: [HumanMessage("JWT?")]}, thread_id=A)
              │
              ▼ checkpoint saved
Turn 2: invoke({messages: [HumanMessage("And pytest?")]}, thread_id=A)
              │
              ▼ merges with Turn 1 state (add_messages)
```

Checkpointing is how **`create_agent`** threads work in **01. LangChain/13** — LangGraph exposes the same mechanism explicitly.

---

## 2. Checkpoint Model — Thread, Checkpoint, State

| Concept | Meaning |
|---------|--------|
| **`thread_id`** | Stable id for one conversation (user session, ticket, run) |
| **Checkpoint** | Immutable snapshot of state after a super-step |
| **Checkpointer** | Storage backend (memory, SQLite, Postgres, …) |
| **`config.configurable`** | Per-invoke settings — includes **`thread_id`** |

```
checkpointer
    └── thread_id: "user-42"
            ├── checkpoint-1  (after node A)
            ├── checkpoint-2  (after node B)
            └── checkpoint-3  (latest — current state)
```

Pass the **same `thread_id`** on every turn for the same conversation.

## 3. Teaching Corpus and Minimal Agent

Build a small ReAct agent to demonstrate multi-turn memory — same Notes API stack as **06–07**:

````python
from typing import Annotated
from typing_extensions import TypedDict

from langchain_core.messages import AIMessage, HumanMessage, ToolMessage
from langchain_core.tools import tool
from langchain_openai import ChatOpenAI
from langgraph.graph import END, START, StateGraph
from langgraph.graph.message import add_messages
from langgraph.prebuilt import ToolNode, tools_condition

llm = ChatOpenAI(model="gpt-4o-mini", api_key=OPENAI_API_KEY, temperature=0)

NOTES = {
    "n1": "The Notes API is built with FastAPI. Routes live under /notes.",
    "n2": "Run alembic upgrade head after pulling database migrations.",
    "n3": "JWT bearer tokens use Authorization: Bearer header.",
    "n4": "Pytest fixtures for DB setup belong in conftest.py.",
}

DOC_CHUNKS = [
    {"id": "c1", "text": NOTES["n1"]},
    {"id": "c2", "text": NOTES["n2"]},
    {"id": "c3", "text": NOTES["n3"]},
    {"id": "c4", "text": NOTES["n4"]},
    {"id": "c5", "text": "Use httpx.AsyncClient in FastAPI tests for async endpoints."},
]


@tool
def get_note(note_id: str) -> str:
    """Fetch a note by id (n1-n4)."""
    return NOTES.get(note_id, f"Unknown note id: {note_id}")


@tool
def search_docs(query: str) -> str:
    """Keyword search over documentation chunks."""
    tokens = set(query.lower().split())
    hits = [f"[{c['id']}] {c['text']}" for c in DOC_CHUNKS if tokens & set(c["text"].lower().split())]
    return "\n".join(hits) if hits else "No matching chunks."


tools = [get_note, search_docs]
llm_with_tools = llm.bind_tools(tools)

SYSTEM = "You are a Notes API assistant. Use tools for factual answers. Cite [c#] ids."


class ChatState(TypedDict):
    messages: Annotated[list, add_messages]


def call_model(state: ChatState) -> dict:
    msgs = [{"role": "system", "content": SYSTEM}] + state["messages"]
    return {"messages": [llm_with_tools.invoke(msgs)]}


def build_chat_agent(checkpointer):
    builder = StateGraph(ChatState)
    builder.add_node("model", call_model)
    builder.add_node("tools", ToolNode(tools))
    builder.add_edge(START, "model")
    builder.add_conditional_edges("model", tools_condition)
    builder.add_edge("tools", "model")
    return builder.compile(checkpointer=checkpointer)
````

## 4. `MemorySaver` Checkpointer

**`MemorySaver`** stores checkpoints in process memory — ideal for notebooks and unit tests:

````python
from langgraph.checkpoint.memory import MemorySaver

memory = MemorySaver()
chat_agent = build_chat_agent(memory)
print("compiled with MemorySaver:", chat_agent is not None)
````

Data is **lost on process restart** — use Postgres or SQLite in production (**16**).

## 6. `configurable.thread_id` in Config

Every checkpointed call passes **`config`** with a **`thread_id`**:

````python
THREAD = "notes-user-alice-001"
config = {"configurable": {"thread_id": THREAD}}

print("thread config:", config)
````

Different **`thread_id`** values = isolated conversations. Same id = shared history.

## 7. Invoke with Same `thread_id` Resumes State

Turn 1 asks about JWT; turn 2 references prior context without resending turn 1 manually:

````python
turn1 = chat_agent.invoke(
    {"messages": [HumanMessage(content="What note covers JWT? Use get_note or search_docs.")]},
    config=config,
)
print("turn1 messages:", len(turn1["messages"]))
print("turn1 answer:", turn1["messages"][-1].content[:100], "...")

turn2 = chat_agent.invoke(
    {"messages": [HumanMessage(content="What was my previous question about?")]},
    config=config,
)
print("\nturn2 messages (accumulated):", len(turn2["messages"]))
print("turn2 answer:", turn2["messages"][-1].content[:120], "...")
````

Only **new** input is passed each turn — the checkpointer merges with prior state.

## 8. `get_state` — Inspect Live Thread

**`get_state(config)`** returns the latest checkpoint for a thread without running the graph:

````python
snapshot = chat_agent.get_state(config)
values = snapshot.values
msgs = values.get("messages", [])

print(f"thread {THREAD}: {len(msgs)} messages in checkpoint")
print("message types:", [m.type for m in msgs])
print("next nodes:", snapshot.next)
print("checkpoint id:", snapshot.config["configurable"].get("checkpoint_id", "(latest)"))
````

Use **`get_state`** in admin UIs, debug panels, and before HITL resume (**09**).

## 9. List Checkpoints Preview

Walk checkpoint history with **`get_state_history`** (newest first):

````python
print(f"Checkpoint history for {THREAD}:")
for i, snap in enumerate(chat_agent.get_state_history(config)):
    n_msgs = len(snap.values.get("messages", []))
    ckpt_id = snap.config["configurable"].get("checkpoint_id", "?")[:12]
    print(f"  {i}: checkpoint={ckpt_id}... messages={n_msgs} next={snap.next}")
    if i >= 4:
        print("  ...")
        break
````

Enables **time travel** debugging — replay or fork from an older checkpoint in advanced workflows.

## 10. Messages Accumulate with `add_messages`

The **`add_messages`** reducer (**02**) appends new messages and **updates by id** when the same message is resent:

````python
def message_summary(state_msgs: list) -> None:
    for i, m in enumerate(state_msgs):
        preview = str(m.content)[:50] if m.content else "(empty)"
        extra = ""
        if isinstance(m, AIMessage) and m.tool_calls:
            extra = f" calls={[tc['name'] for tc in m.tool_calls]}"
        print(f"  {i:02d} {m.type:6s} {preview}{extra}")


final_state = chat_agent.get_state(config)
print("Full thread transcript:")
message_summary(final_state.values["messages"])
````

Without **`add_messages`**, passing new messages would **replace** the list — breaking multi-turn chat.

## 11. Isolated Threads — Side-by-Side

Two users with different **`thread_id`** values do not share state:

````python
cfg_bob = {"configurable": {"thread_id": "notes-user-bob-002"}}

chat_agent.invoke(
    {"messages": [HumanMessage(content="Tell me about Alembic migrations.")]},
    config=cfg_bob,
)

alice_msgs = len(chat_agent.get_state(config).values["messages"])
bob_msgs = len(chat_agent.get_state(cfg_bob).values["messages"])

print(f"alice thread messages: {alice_msgs}")
print(f"bob thread messages:   {bob_msgs}")
````

## 14. Thread-per-User / Session Pattern

Map application identity to **`thread_id`** at the API boundary:

````python
import uuid


def make_thread_id(user_id: str, session_id: str | None = None) -> str:
    """Stable thread key for a user session."""
    sid = session_id or str(uuid.uuid4())
    return f"notes-{user_id}-{sid}"


def run_turn(graph, user_id: str, session_id: str, text: str) -> str:
    tid = make_thread_id(user_id, session_id)
    cfg = {"configurable": {"thread_id": tid}}
    result = graph.invoke({"messages": [HumanMessage(content=text)]}, config=cfg)
    last = result["messages"][-1]
    return last.content if isinstance(last.content, str) else str(last.content)


session = "sess-demo-99"
r1 = run_turn(chat_agent, "user-7", session, "Where do pytest DB fixtures go?")
r2 = run_turn(chat_agent, "user-7", session, "Summarize what we discussed.")
print("turn1:", r1[:80], "...")
print("turn2:", r2[:80], "...")
````

In FastAPI (**16**): derive **`user_id`** from JWT (**03. Python Backend/06. Security**), create or lookup **`session_id`**, pass **`thread_id`** in every graph call.

## 15. Stateless vs Checkpointed Comparison

| | **Stateless** | **Checkpointed** |
|--|---------------|------------------|
| **Input each turn** | Full message history | New messages only |
| **Server memory** | Client holds history | Server holds thread state |
| **HITL** | Custom pause store | Built-in resume (**09**) |
| **Compile** | `compile()` | `compile(checkpointer=...)` |
| **Config** | Optional | **`thread_id` required** |

---

## 16. Design Guidelines

| Guideline | Rationale |
|-----------|----------|
| **Compile with checkpointer once** | Expensive to rebuild per request |
| **Stable `thread_id` per conversation** | Random ids defeat persistence |
| **`add_messages` on `messages` key** | Required for append semantics |
| **`MemorySaver` only in dev** | Lost on restart |
| **Call `setup()` on Postgres once** | Creates checkpoint tables |
| **Scope threads per tenant** | Prefix `thread_id` with org/user id |
| **Inspect with `get_state`** | Debug without re-invoking LLM |

---

## 17. Common Mistakes

| Mistake | Consequence | Fix |
|---------|-------------|-----|
| No `checkpointer` at compile | Every invoke starts fresh | `compile(checkpointer=...)` |
| Missing `thread_id` in config | Error or ephemeral default | Always pass `configurable.thread_id` |
| New `thread_id` every request | No memory between turns | Stable id per session |
| Replace reducer with plain `list` | Messages overwrite | Use `add_messages` |
| Resending full history + checkpoint | Duplicate messages | Send delta only |
| `MemorySaver` in production | Lost threads on deploy | Postgres / SQLite |
| New graph compile per request | Slow, connection leaks | Singleton at startup (**16**) |
| Sharing one thread across users | Data leak | Per-user `thread_id` |

---

## 18. Summary

Checkpointing persists graph **state** across invocations, keyed by **`thread_id`**. Attach a **checkpointer** at **`compile()`**, pass **`config={"configurable": {"thread_id": ...}}`** on every call, and let **`add_messages`** merge new turns into the conversation.

Key takeaways:

- **`MemorySaver`** is for learning; **SqliteSaver** / **PostgresSaver** survive restarts.
- **`get_state`** and **`get_state_history`** inspect threads without running the LLM.
- Multi-turn ReAct agents only need **new `HumanMessage`** inputs when checkpointed.
- Map **user + session → `thread_id`** at the API layer for production chat.
