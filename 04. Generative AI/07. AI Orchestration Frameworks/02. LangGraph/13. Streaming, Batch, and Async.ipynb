## 1. Why Stream Graph Runs?

**`invoke`** blocks until the graph finishes — fine for scripts, poor for UIs and long agent loops.

```
invoke:   input ─────────────────────────────► final state

stream:   input ──► chunk₁ ──► chunk₂ ──► … ──► done
                    (per node / value / token)
```

| API | Sync | Async | Returns |
|-----|------|-------|--------|
| **`invoke` / `ainvoke`** | ✅ | ✅ | Final state dict |
| **`stream` / `astream`** | ✅ | ✅ | Iterator of chunks |
| **`batch` / `abatch`** | ✅ | ✅ | List of final states |
| **`astream_events`** | — | ✅ | Structured Runnable events |

Choose **`stream_mode`** based on what your UI or log pipeline needs.

## 2. Demo Graph — Notes API Pipeline

A compact three-node graph (retrieve → generate) for comparing stream modes — same teaching corpus as **05** and **06**:

````python
from typing_extensions import TypedDict

from langchain_core.messages import AIMessage, HumanMessage
from langchain_openai import ChatOpenAI
from langgraph.graph import END, START, StateGraph

llm = ChatOpenAI(model="gpt-4o-mini", api_key=OPENAI_API_KEY, temperature=0)

DOC_CHUNKS = [
    {"id": "c1", "text": "The Notes API is built with FastAPI. Routes live under /notes."},
    {"id": "c2", "text": "Run alembic upgrade head after pulling database migrations."},
    {"id": "c3", "text": "JWT bearer tokens use Authorization: Bearer header."},
    {"id": "c4", "text": "Pytest fixtures for DB setup belong in conftest.py."},
]


class PipelineState(TypedDict):
    question: str
    context: str
    answer: str
    trace: list[str]


def retrieve_node(state: PipelineState) -> dict:
    tokens = set(state["question"].lower().split())
    hits = [
        f"[{c['id']}] {c['text']}"
        for c in DOC_CHUNKS
        if tokens & set(c["text"].lower().split())
    ]
    context = "\n".join(hits) if hits else DOC_CHUNKS[0]["text"]
    return {"context": context, "trace": state.get("trace", []) + ["retrieve"]}


def generate_node(state: PipelineState) -> dict:
    prompt = [
        {"role": "system", "content": "Answer using ONLY the context. Cite chunk ids."},
        {"role": "user", "content": f"Context:\n{state['context']}\n\nQuestion: {state['question']}"},
    ]
    response = llm.invoke(prompt)
    text = response.content if isinstance(response.content, str) else str(response.content)
    return {"answer": text, "trace": state["trace"] + ["generate"]}


def build_pipeline():
    builder = StateGraph(PipelineState)
    builder.add_node("retrieve", retrieve_node)
    builder.add_node("generate", generate_node)
    builder.add_edge(START, "retrieve")
    builder.add_edge("retrieve", "generate")
    builder.add_edge("generate", END)
    return builder.compile()


pipeline = build_pipeline()
INITIAL = {"question": "", "context": "", "answer": "", "trace": []}
print("nodes:", list(pipeline.get_graph().nodes))
````

---

## 3. `stream_mode="updates"` — Per-Node Patches

Each chunk is **`{node_name: state_patch}`** — only keys the node returned. Best for **progress indicators** and agent step logs:

````python
print("stream_mode='updates':")
for chunk in pipeline.stream(
    {**INITIAL, "question": "Alembic upgrade command?"},
    stream_mode="updates",
):
    node, patch = next(iter(chunk.items()))
    keys = list(patch.keys())
    preview = patch.get("answer", patch.get("context", ""))[:60]
    print(f"  {node}: keys={keys} preview={preview!r}...")
````

**Pattern:** `list(chunk.keys())[0]` gives the node that just ran; `patch` is the partial update — not the full accumulated state.

---

## 4. `stream_mode="values"` — Full State Snapshots

Each chunk is the **entire state** after a node completes. Useful when the UI renders cumulative fields (`answer`, `trace`, `messages`):

````python
print("stream_mode='values':")
for i, state in enumerate(pipeline.stream(
    {**INITIAL, "question": "JWT header format?"},
    stream_mode="values",
)):
    print(f"  step {i}: trace={state['trace']} answer_len={len(state.get('answer', ''))}")
````

The **last** values chunk equals **`invoke`** output (modulo internal keys).

---

## 5. `stream_mode="messages"` — LLM Token Chunks

Streams **`(message_chunk, metadata)`** tuples for LLM nodes — token-by-token text as the model generates. Only nodes that call an LLM emit chunks:

````python
print("stream_mode='messages' (token stream from generate node):")
buffer = ""
for msg_chunk, meta in pipeline.stream(
    {**INITIAL, "question": "Where do pytest DB fixtures go?"},
    stream_mode="messages",
):
    if msg_chunk.content:
        buffer += msg_chunk.content
        print(msg_chunk.content, end="", flush=True)
    # meta often includes langgraph_node, ls_model_name, etc.
print(f"\n\nmetadata keys (last chunk): {list(meta.keys())[:5]}")
print(f"total chars streamed: {len(buffer)}")
````

Use **`messages`** for chat UIs that type out answers. Non-LLM nodes (e.g. `retrieve`) produce no message chunks.

---

## 6. `stream_mode="debug"` — Internal Execution Trace

**`debug`** emits low-level execution events (task start/finish, channel writes). Verbose — use during development, not in production SSE:

````python
print("stream_mode='debug' (first 6 events):")
for i, event in enumerate(pipeline.stream(
    {**INITIAL, "question": "FastAPI routes?"},
    stream_mode="debug",
)):
    # event is a dict with type, payload, timestamp, etc.
    etype = event.get("type", type(event).__name__)
    print(f"  [{i}] {etype}")
    if i >= 5:
        print("  ... (truncated)")
        break
````

## 7. Stream Mode Comparison Table

| `stream_mode` | Chunk shape | Best for | Notes |
|---------------|-------------|----------|-------|
| **`updates`** | `{node: patch}` | Step progress, agent tool/model logs | Smallest payload per step |
| **`values`** | Full state dict | Dashboards showing cumulative state | Grows with `messages` list |
| **`messages`** | `(AIMessageChunk, meta)` | Token-by-token chat UI | LLM nodes only |
| **`debug`** | Internal event dict | Deep debugging | Noisy; dev only |

Pass a **list** to stream multiple modes simultaneously: `stream_mode=["updates", "messages"]`.

````python
print("stream_mode=['updates', 'messages']:")
for mode, chunk in pipeline.stream(
    {**INITIAL, "question": "Alembic command?"},
    stream_mode=["updates", "messages"],
):
    if mode == "updates":
        node = list(chunk.keys())[0]
        print(f"  [updates] node={node}")
    elif mode == "messages":
        msg_chunk, _ = chunk
        if msg_chunk.content:
            print(f"  [messages] token={msg_chunk.content!r}")
````

Each tuple is **`(mode_name, chunk)`** when multiple modes are requested.

## 8. Async — `ainvoke` and `astream`

FastAPI and asyncio services should use **`ainvoke`** / **`astream`** — same signatures, awaitable:

````python
import asyncio


async def run_async_demo():
    # ainvoke — single final result
    result = await pipeline.ainvoke({**INITIAL, "question": "JWT bearer header?"})
    print("ainvoke answer:", result["answer"][:80], "...")

    # astream — incremental chunks
    print("\nastream updates:")
    async for chunk in pipeline.astream(
        {**INITIAL, "question": "pytest fixtures file?"},
        stream_mode="updates",
    ):
        node = list(chunk.keys())[0]
        print(f"  async step: {node}")


await run_async_demo()
````

**Rule:** never call blocking **`invoke`** inside an async route handler — it blocks the event loop. Use **`ainvoke`** / **`astream`**.

## 9. `batch()` — Multiple Inputs in One Call

**`batch`** runs the graph for each input dict and returns a **list of final states**. LangGraph may parallelize independent runs:

````python
questions = [
    {**INITIAL, "question": "Alembic upgrade command?"},
    {**INITIAL, "question": "JWT header format?"},
    {**INITIAL, "question": "pytest conftest location?"},
]

results = pipeline.batch(questions)
for q, r in zip(questions, results):
    print(f"Q: {q['question']}")
    print(f"  trace={r['trace']} answer={r['answer'][:60]}...\n")
````

Use **`batch`** for offline eval datasets (**12**) or bulk doc Q&A. For live UIs, prefer **`stream`** per user.

## 10. Streaming the ReAct Agent Loop

Rebuild the **06** agent and watch **`model`** / **`tools`** iterations via **`updates`**:

````python
from langchain_core.messages import ToolMessage
from langchain_core.tools import tool
from langgraph.graph.message import add_messages


NOTES = {
    "n1": "The Notes API is built with FastAPI. Routes live under /notes.",
    "n2": "Run alembic upgrade head after pulling database migrations.",
    "n3": "JWT bearer tokens use Authorization: Bearer header.",
    "n4": "Pytest fixtures for DB setup belong in conftest.py.",
}


@tool
def search_docs(query: str) -> str:
    """Keyword search over Notes API documentation chunks."""
    tokens = set(query.lower().split())
    hits = [f"[{k}] {v}" for k, v in NOTES.items() if tokens & set(v.lower().split())]
    return "\n".join(hits) if hits else "No matching chunks."


@tool
def get_note(note_id: str) -> str:
    """Fetch a note by id (n1–n4)."""
    return NOTES.get(note_id, f"Unknown note id: {note_id}")


agent_tools = [search_docs, get_note]
TOOLS_BY_NAME = {t.name: t for t in agent_tools}
llm_with_tools = llm.bind_tools(agent_tools)

SYSTEM = """You are a Notes API assistant. Use search_docs and get_note before answering."""


class AgentState(TypedDict):
    messages: Annotated[list, add_messages]
    step_count: int


def call_model(state: AgentState) -> dict:
    msgs = [{"role": "system", "content": SYSTEM}] + state["messages"]
    response = llm_with_tools.invoke(msgs)
    return {"messages": [response], "step_count": state.get("step_count", 0) + 1}


def run_tools(state: AgentState) -> dict:
    last = state["messages"][-1]
    results = []
    for tc in last.tool_calls:
        tool_fn = TOOLS_BY_NAME[tc["name"]]
        results.append(ToolMessage(content=str(tool_fn.invoke(tc["args"])), tool_call_id=tc["id"]))
    return {"messages": results}


def should_continue(state: AgentState):
    last = state["messages"][-1]
    if isinstance(last, AIMessage) and last.tool_calls:
        return "tools"
    return END


def build_stream_agent():
    b = StateGraph(AgentState)
    b.add_node("model", call_model)
    b.add_node("tools", run_tools)
    b.add_edge(START, "model")
    b.add_conditional_edges("model", should_continue)
    b.add_edge("tools", "model")
    return b.compile()


stream_agent = build_stream_agent()
````

````python
print("ReAct agent stream (updates):")
for chunk in stream_agent.stream(
    {"messages": [HumanMessage(content="Search docs for alembic migration.")], "step_count": 0},
    stream_mode="updates",
):
    node, patch = next(iter(chunk.items()))
    if "messages" in patch:
        last = patch["messages"][-1]
        extra = ""
        if isinstance(last, AIMessage) and last.tool_calls:
            extra = f" tool_calls={[tc['name'] for tc in last.tool_calls]}"
        elif isinstance(last, ToolMessage):
            extra = f" tool_result={str(last.content)[:50]}..."
        print(f"  {node}: {last.type}{extra}")
    else:
        print(f"  {node}: {list(patch.keys())}")
````

Typical trace: **`model` (tool_calls) → `tools` (ToolMessage) → `model` (final answer)** — visible as separate update chunks.

## 11. `astream_events` — Token and Step Events

For **fine-grained** visibility (every Runnable step, tool start/end, tokens), use **`astream_events`** — same API as LCEL chains (**01. LangChain/15**). Works on compiled graphs:

````python
async def print_graph_events():
    print("astream_events (selected, version='v2'):")
    count = 0
    async for event in pipeline.astream_events(
        {**INITIAL, "question": "Alembic upgrade?"},
        version="v2",
    ):
        kind = event["event"]
        if kind in ("on_chat_model_stream", "on_chain_end", "on_tool_end"):
            name = event.get("name", "")
            print(f"  {kind}: {name}")
            count += 1
        if count >= 8:
            print("  ... (truncated)")
            break


await print_graph_events()
````

| Use case | Prefer |
|----------|--------|
| Node-level progress bar | **`stream_mode="updates"`** |
| Chat token typing | **`stream_mode="messages"`** or **`astream_events`** `on_chat_model_stream` |
| Full observability / LangSmith-style UI | **`astream_events(version="v2")`** |

See **01. LangChain/15** for custom callback handlers and LangSmith integration.

## 12. SSE-Friendly Chunk Shape (FastAPI Preview)

**16. Production Patterns** wraps the graph in a FastAPI route. Server-Sent Events need **JSON-serializable** dicts per chunk:

````python
def to_sse_chunk(mode: str, chunk: dict | tuple) -> dict:
    """Normalize a LangGraph stream chunk for EventSourceResponse."""
    if mode == "updates":
        node, patch = next(iter(chunk.items()))
        payload = {k: (v if not hasattr(v, "type") else v.type) for k, v in patch.items()}
        return {"event": "node_update", "node": node, "data": payload}
    if mode == "values":
        return {"event": "state", "data": {"trace": chunk.get("trace", []), "answer": chunk.get("answer", "")[:200]}}
    if mode == "messages":
        msg_chunk, meta = chunk
        return {"event": "token", "data": {"content": msg_chunk.content, "node": meta.get("langgraph_node")}}
    return {"event": mode, "data": str(chunk)[:200]}


# Simulate what a FastAPI generator would yield
sse_lines = []
for chunk in pipeline.stream({**INITIAL, "question": "JWT format?"}, stream_mode="updates"):
    sse = to_sse_chunk("updates", chunk)
    sse_lines.append(f"data: {json.dumps(sse)}\n\n")

print("Sample SSE payloads:")
for line in sse_lines:
    print(line.strip())
````

```python
# FastAPI sketch (see 16 for full app)
@app.post("/ask/stream")
async def ask_stream(body: AskRequest):
    async def event_gen():
        async for chunk in graph.astream({"question": body.q, ...}, stream_mode="updates"):
            yield f"data: {json.dumps(to_sse_chunk('updates', chunk))}\n\n"
    return EventSourceResponse(event_gen())
```

Keep payloads small — strip large `messages` lists or send only the delta.

## 13. Choosing a Stream Mode — Decision Guide

```
Need node progress only?     ──► updates
Need full state each step?   ──► values
Need token typing in chat?   ──► messages  (or astream_events)
Debugging graph internals?   ──► debug
Building observability UI?   ──► astream_events v2
```

| Layer | API | Consumer |
|-------|-----|----------|
| **Graph** | `stream` / `astream` | Your app / FastAPI |
| **Runnable** | `astream_events` | Debug panels, LangSmith |
| **LLM** | `stream_mode="messages"` | Chat frontend |

---

## 14. Design Guidelines

| Guideline | Rationale |
|-----------|----------|
| **Compile graph once** | Reuse across `stream` / `batch` calls |
| **`ainvoke` in async apps** | Avoid blocking the event loop |
| **`updates` for agents** | Shows model/tools steps without huge payloads |
| **Cap SSE payload size** | `messages` arrays grow quickly |
| **`version="v2"` for events** | Stable `astream_events` schema (**15**) |
| **`batch` for eval, not live chat** | Batch waits for all inputs |
| **Thread config for memory** | Pass `config={"configurable": {"thread_id": ...}}` with streams (**08**) |

---

## 15. Common Mistakes

| Mistake | Consequence | Fix |
|---------|-------------|-----|
| `invoke` in FastAPI async route | Blocks other requests | `ainvoke` / `astream` |
| Expecting tokens from `updates` | No token stream | Use `messages` or `astream_events` |
| Assuming `updates` is full state | Missing prior keys | Use `values` or merge patches |
| `debug` in production SSE | Huge noisy payloads | `updates` or `messages` only |
| Not iterating all stream chunks | Miss final answer | Consume full iterator |
| `batch` for single user chat | Higher latency | `stream` per session |
| Serializing `AIMessage` raw to JSON | `TypeError` | Map to `{type, content}` dicts |
| `astream_events` v1 schema | Broken event parsing | `version="v2"` |

---

## 16. Summary

LangGraph graphs support **`invoke`**, **`stream`**, **`batch`**, and their **async** counterparts. Pick **`stream_mode`** based on what consumers need: **`updates`** for node progress, **`values`** for full state, **`messages`** for LLM tokens, **`debug`** for internals.

Key takeaways:

- **`stream_mode="updates"`** emits `{node: patch}` — ideal for ReAct step logs.
- **`ainvoke` / `astream`** are mandatory in async web servers.
- **`batch()`** runs many inputs; use for eval, not interactive chat.
- **`astream_events(version="v2")`** links to **01. LangChain/15** for deep observability.
- **SSE helpers** normalize chunks to JSON for FastAPI **`EventSourceResponse`** (**16**).
