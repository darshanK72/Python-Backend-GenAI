# Observability and Debugging

LangGraph workflows are **multi-step** — guardrails, routers, retrieval, agents, and tools can each change the path. When answers are wrong or slow, you need **visibility**: which nodes ran, what state changed, and how nested LLM/tool calls behaved.

This notebook covers **`compile(debug=True)`**, **`stream_mode`** (`updates`, `debug`), the **`trace` list in state** pattern (**05**), **Mermaid** exports, **LangSmith** tracing, **`astream_events` v2**, **message traces** (**06**), **config tags/metadata**, and techniques for **debugging routing mistakes**.

Prerequisites: **05. Building Your First Graph**, **06. Agent Loops and the ReAct Pattern**, **13. Streaming, Batch, and Async**.

## 1. Observability Layers

Use **multiple layers** — each answers a different question:

```
Request
   │
   ├─► stream_mode="updates"     Which node fired? What patch?
   ├─► stream_mode="debug"       Scheduler / task metadata
   ├─► state["trace"]            Durable path in final output
   ├─► get_graph().draw_mermaid() Static topology for reviews
   ├─► LangSmith (env tracing)   Cross-run traces in UI
   └─► astream_events v2         Nested LLM + tool child runs
```

| Layer | Best for | Cost |
|-------|----------|------|
| **`trace` in state** | Unit tests, API response `path` field | Free |
| **`stream_mode="updates"`** | Progress UI, live dashboards | Low |
| **`stream_mode="debug"`** | Scheduler issues, parallel branches | Low |
| **LangSmith** | Staging/prod incident replay | SaaS |
| **`astream_events` v2** | Token streams + tool args inside nodes | Medium |

## 2. Notes API Graph — Teaching Corpus

We reuse the **Notes API documentation graph** from **05** (chunks **c1–c5**, same as **01. LangChain/11**). Each node appends to **`trace`** so routing is visible in the final state:

````python
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from langgraph.graph import END, START, StateGraph


class NotesGraphState(TypedDict):
    question: str
    intent: str
    blocked: bool
    context: str
    source_ids: list[str]
    grade: str
    answer: str
    attempts: int
    trace: Annotated[list[str], operator.add]


CORPUS = [
    {"id": "c1", "text": "The Notes API is built with FastAPI. Routes live under /notes with GET, POST, PUT, DELETE."},
    {"id": "c2", "text": "Alembic applies SQLAlchemy schema migrations. Run alembic upgrade head after pulling revisions."},
    {"id": "c3", "text": "JWT bearer tokens authenticate API requests. Send Authorization: Bearer token header."},
    {"id": "c4", "text": "Pytest fixtures share database setup. Use conftest.py for session-scoped engines."},
    {"id": "c5", "text": "Alembic autogenerate compares SQLAlchemy models to the live schema and drafts revision files."},
]

BLOCKED_TERMS = {"password", "secret", "api key", "private key"}
DOCS_KEYWORDS = {"jwt", "alembic", "pytest", "fastapi", "migration", "token", "route", "conftest"}
MAX_ATTEMPTS = 2


def keyword_retrieve(query: str, k: int = 2) -> tuple[str, list[str]]:
    tokens = set(query.lower().split())
    scored = [(len(tokens & set(d["text"].lower().split())), d) for d in CORPUS]
    scored.sort(key=lambda x: x[0], reverse=True)
    top = [d for s, d in scored if s > 0][:k] or [CORPUS[0]]
    context = "\n".join(f"[{d['id']}] {d['text']}" for d in top)
    return context, [d["id"] for d in top]


def guard_node(state: NotesGraphState) -> dict:
    blocked = any(t in state["question"].lower() for t in BLOCKED_TERMS)
    patch: dict = {"blocked": blocked, "trace": ["guard"]}
    if blocked:
        patch["answer"] = "I cannot help with credentials or secrets."
    return patch


def classify_node(state: NotesGraphState) -> dict:
    intent = "docs" if any(k in state["question"].lower() for k in DOCS_KEYWORDS) else "general"
    return {"intent": intent, "trace": ["classify"]}


def retrieve_node(state: NotesGraphState) -> dict:
    context, ids = keyword_retrieve(state["question"])
    return {"context": context, "source_ids": ids, "attempts": state.get("attempts", 0) + 1, "trace": ["retrieve"]}


def grade_node(state: NotesGraphState) -> dict:
    q, ctx = state["question"].lower(), state.get("context", "").lower()
    relevant = any(k in ctx for k in q.split() if len(k) > 3) or any(k in q and k in ctx for k in DOCS_KEYWORDS)
    return {"grade": "relevant" if relevant else "irrelevant", "trace": ["grade"]}


def rewrite_node(state: NotesGraphState) -> dict:
    return {"question": state["question"] + " (FastAPI Notes API alembic jwt pytest)", "trace": ["rewrite"]}


def chitchat_node(state: NotesGraphState) -> dict:
    return {"answer": "Ask me about the Notes API — migrations, JWT, pytest, or FastAPI routes.", "trace": ["chitchat"]}


llm = ChatOpenAI(model="gpt-4o-mini", api_key=OPENAI_API_KEY, temperature=0)
generate_chain = (
    ChatPromptTemplate.from_messages([
        ("system", "Answer using ONLY the context. Cite chunk ids like [c2]."),
        ("human", "Context:\n{context}\n\nQuestion: {question}"),
    ])
    | llm
)


def generate_node(state: NotesGraphState) -> dict:
    msg = generate_chain.invoke({"context": state["context"], "question": state["question"]})
    return {"answer": msg.content, "trace": ["generate"]}


def route_guard(state: NotesGraphState):
    return END if state["blocked"] else "classify"


def route_intent(state: NotesGraphState) -> Literal["docs", "general"]:
    return "docs" if state["intent"] == "docs" else "general"


def route_grade(state: NotesGraphState) -> Literal["relevant", "irrelevant"]:
    return "relevant" if state["grade"] == "relevant" else "irrelevant"


def route_rewrite(state: NotesGraphState):
    return "retrieve" if state.get("attempts", 0) < MAX_ATTEMPTS else END


def build_notes_graph(*, debug: bool = False):
    builder = StateGraph(NotesGraphState)
    builder.add_node("guard", guard_node)
    builder.add_node("classify", classify_node)
    builder.add_node("retrieve", retrieve_node)
    builder.add_node("grade", grade_node)
    builder.add_node("rewrite", rewrite_node)
    builder.add_node("generate", generate_node)
    builder.add_node("chitchat", chitchat_node)
    builder.add_edge(START, "guard")
    builder.add_edge("retrieve", "grade")
    builder.add_edge("generate", END)
    builder.add_edge("chitchat", END)
    builder.add_conditional_edges("guard", route_guard)
    builder.add_conditional_edges("classify", route_intent, {"docs": "retrieve", "general": "chitchat"})
    builder.add_conditional_edges("grade", route_grade, {"relevant": "generate", "irrelevant": "rewrite"})
    builder.add_conditional_edges("rewrite", route_rewrite)
    return builder.compile(debug=debug)


INITIAL = {
    "intent": "", "blocked": False, "context": "", "source_ids": [],
    "grade": "", "answer": "", "attempts": 0, "trace": [],
}

notes_graph = build_notes_graph()
print("nodes:", [n for n in notes_graph.get_graph().nodes if n not in ("__start__", "__end__")])
````

**Convention:** every node returns **`{"trace": ["node_name"]}`** — the **`operator.add`** reducer appends without overwriting prior steps.

## 3. `compile(debug=True)`

Pass **`debug=True`** at compile time to enable richer runtime diagnostics from the LangGraph engine (extra validation, verbose task scheduling). Use in **development and staging** — not required in production once graphs are stable:

````python
debug_graph = build_notes_graph(debug=True)

out = debug_graph.invoke({**INITIAL, "question": "Alembic upgrade command?"})
print("trace:", out["trace"])
print("answer preview:", out["answer"][:120])
````

Pair **`debug=True`** with **`stream_mode="debug"`** (next section) when a node never fires or a router sends traffic to the wrong branch.

## 4. `stream_mode="updates"` — Node-Level Visibility

**`updates`** yields **`{node_name: partial_state_patch}`** after each node completes — ideal for progress bars and quick routing checks (**03**, **13**):

````python
print("stream_mode=updates — JWT question")
for step in notes_graph.stream(
    {**INITIAL, "question": "How do I send a JWT bearer token?"},
    stream_mode="updates",
):
    node_name = next(iter(step))
    patch = step[node_name]
    keys = [k for k in patch if k != "trace"]
    print(f"  {node_name}: keys={keys} trace+={patch.get('trace', [])}")
````

Expected path for a docs question: **`guard → classify → retrieve → grade → generate`**.

## 5. `stream_mode="debug"` — Scheduler Detail

**`debug`** streams lower-level execution events (task ids, channel writes). Use when **`updates`** shows unexpected ordering or missing nodes:

````python
print("stream_mode=debug — first 6 events")
for i, event in enumerate(debug_graph.stream(
    {**INITIAL, "question": "pytest fixtures location?"},
    stream_mode="debug",
)):
    # Events are (namespace, mode, payload) tuples in recent LangGraph versions
    label = event if isinstance(event, str) else str(event)[:120]
    print(f"  [{i}] {label}")
    if i >= 5:
        break
````

Combine modes when needed: **`stream_mode=["updates", "debug"]`** returns tagged tuples — useful in custom log sinks.

## 6. Trace List in State — Durable Path

Unlike ephemeral streams, **`trace`** survives in the **final invoke result** — return it from your API as **`path`** (**05**, **16**):

````python
MATRIX = [
    "What is my api key password?",          # guard → END
    "Hello there",                            # guard → classify → chitchat
    "Alembic upgrade command?",               # full RAG path
]

for q in MATRIX:
    result = notes_graph.invoke({**INITIAL, "question": q})
    print(f"Q: {q}")
    print(f"  path: {' → '.join(result['trace'])}")
    print(f"  blocked: {result.get('blocked')}  sources: {result.get('source_ids', [])}\n")
````

**Test matrices** like this catch routing regressions before you add LLM graders or live models.

## 9. `astream_events` v2 — Nested Runs

Inside nodes that call LLMs or tools, **`astream_events(..., version="v2")`** exposes token chunks, tool starts, and retriever calls. Essential for agent loops (**06**, **13**):

````python
import asyncio


async def print_graph_events(question: str, limit: int = 12) -> None:
  count = 0
  async for event in notes_graph.astream_events(
      {**INITIAL, "question": question},
      version="v2",
      config={"tags": ["astream-events-demo"]},
  ):
      kind = event.get("event")
      name = event.get("name", "")
      if kind in ("on_chain_start", "on_chain_end", "on_chat_model_start", "on_chat_model_end"):
          print(f"  {kind:22s} {name}")
          count += 1
      if count >= limit:
          break


await print_graph_events("Alembic autogenerate workflow?")
````

Filter on **`event["event"]`** — common values: `on_chain_start`, `on_chat_model_stream`, `on_tool_end`. Pair with **`include_names`** / **`include_types`** to reduce noise in production log pipelines.

## 10. Printing Message Traces — From **06**

Agent graphs use **`MessagesState`** instead of flat keys. Reuse **`print_trace`** to inspect tool loops:

````python
from langchain_core.messages import AIMessage, HumanMessage, ToolMessage
from langchain_core.tools import tool
from langgraph.graph.message import add_messages


class AgentState(TypedDict):
    messages: Annotated[list, add_messages]


@tool
def search_docs(query: str) -> str:
    """Search Notes API documentation chunks."""
    ctx, _ = keyword_retrieve(query)
    return ctx


TOOLS = {"search_docs": search_docs}


def mini_model(state: AgentState) -> dict:
    last = state["messages"][-1].content.lower()
    if any(isinstance(m, ToolMessage) for m in state["messages"]):
        return {"messages": [AIMessage(content="See [c4] — pytest fixtures live in conftest.py.")]}
    if "pytest" in last or "fixture" in last:
        return {"messages": [AIMessage(content="", tool_calls=[{
            "name": "search_docs", "args": {"query": "pytest conftest"},
            "id": "call_obs_1", "type": "tool_call",
        }])]}
    return {"messages": [AIMessage(content="Ask about Notes API docs.")]}


def mini_tools(state: AgentState) -> dict:
    call = state["messages"][-1].tool_calls[0]
    fn = TOOLS[call["name"]]
    return {"messages": [ToolMessage(content=fn.invoke(call["args"]), tool_call_id=call["id"])]}


def route_agent(state: AgentState):
    last = state["messages"][-1]
    return "tools" if isinstance(last, AIMessage) and last.tool_calls else END


ab = StateGraph(AgentState)
ab.add_node("model", mini_model)
ab.add_node("tools", mini_tools)
ab.add_edge(START, "model")
ab.add_conditional_edges("model", route_agent)
ab.add_edge("tools", "model")
mini_agent = ab.compile()


def print_trace(result: dict) -> None:
    for i, msg in enumerate(result["messages"]):
        if isinstance(msg, AIMessage) and msg.tool_calls:
            names = [tc["name"] for tc in msg.tool_calls]
            print(f"{i:02d} ai     tool_calls={names}")
        elif isinstance(msg, ToolMessage):
            print(f"{i:02d} tool   {str(msg.content)[:70]}...")
        else:
            print(f"{i:02d} {msg.type:6s} {str(msg.content)[:70]}...")


agent_out = mini_agent.invoke({"messages": [HumanMessage(content="pytest fixtures file?")]})
print_trace(agent_out)
````

When **`print_trace`** shows repeated `tool_calls` with no final answer, check **`step_count`** limits (**06**, **14**) and router logic.

## 11. Configurable Tags in `config` Metadata

Pass **`tags`** and **`metadata`** on every invoke — they propagate to LangSmith and structured logs. Use **`configurable`** for per-tenant or per-thread values (**08**):

````python
def make_run_config(*, thread_id: str, user_tier: str = "free", env: str = "dev") -> dict:
    return {
        "configurable": {"thread_id": thread_id},
        "tags": ["notes-api", user_tier, env],
        "metadata": {
            "user_tier": user_tier,
            "env": env,
            "feature": "docs_qa",
        },
        "run_name": f"notes_qa_{user_tier}",
    }


pro_config = make_run_config(thread_id="thread-abc", user_tier="pro", env="staging")
print(json.dumps({k: pro_config[k] for k in ("tags", "metadata", "run_name")}, indent=2))

sample = notes_graph.invoke({**INITIAL, "question": "FastAPI route prefix?"}, config=pro_config)
print("path:", sample["trace"])
````

Filter LangSmith traces by **`tags:pro`** or **`metadata.env:staging`** during incident triage.

## 12. Debugging Routing Mistakes

The most common bug: **router reads a key the upstream node never set**. Compare a **buggy** router with the fix:

````python
class BuggyState(TypedDict):
    question: str
    intent: str
    answer: str
    trace: Annotated[list[str], operator.add]


def classify_buggy(state: BuggyState) -> dict:
    # BUG: writes "category" but router reads "intent"
    q = state["question"].lower()
    category = "docs" if "jwt" in q else "general"
    return {"intent": "", "trace": [f"classify:{category}"]}


def route_buggy(state: BuggyState):
    return "docs" if state.get("intent") == "docs" else "general"


def docs_answer(state: BuggyState) -> dict:
    return {"answer": "JWT uses Authorization Bearer [c3].", "trace": ["docs"]}


def general_answer(state: BuggyState) -> dict:
    return {"answer": "General chitchat.", "trace": ["general"]}


bb = StateGraph(BuggyState)
bb.add_node("classify", classify_buggy)
bb.add_node("docs", docs_answer)
bb.add_node("general", general_answer)
bb.add_edge(START, "classify")
bb.add_conditional_edges("classify", route_buggy, {"docs": "docs", "general": "general"})
bb.add_edge("docs", END)
bb.add_edge("general", END)
buggy_graph = bb.compile(debug=True)

print("=== Buggy graph (stream updates) ===")
for step in buggy_graph.stream(
    {"question": "JWT header?", "intent": "", "answer": "", "trace": []},
    stream_mode="updates",
):
    print(" ", step)

wrong = buggy_graph.invoke({"question": "JWT header?", "intent": "", "answer": "", "trace": []})
print("\nWRONG branch — trace:", wrong["trace"], "answer:", wrong["answer"])
print("FIX: classify must return intent='docs', or router must read the key you write")
````

**Debug playbook:**

1. Run **`stream_mode="updates"`** — which nodes fire?
2. Compare **`trace`** to expected path from Mermaid diagram.
3. Print **router input state** at the conditional edge.
4. Unit-test routers in isolation (**04**).
5. Enable LangSmith to see if LLM output inside a node is wrong vs routing.

## 13. Observability Checklist

| Step | Tool | When |
|------|------|------|
| Append node name to **`trace`** | State reducer | Always |
| **`stream_mode="updates"`** | `.stream()` | Dev + streaming UI |
| **`compile(debug=True)`** | Factory flag | Staging |
| **Mermaid export** | `get_graph().draw_mermaid()` | Design reviews |
| **LangSmith** | `LANGCHAIN_TRACING_V2` | Staging/prod |
| **`astream_events` v2** | Async pipelines | Token/tool visibility |
| **`print_trace`** | Agent graphs | Tool-loop bugs |
| **Config tags/metadata** | `invoke(..., config=)` | Tenant/tier filtering |
| **Routing test matrix** | Multiple `invoke` inputs | Before deploy |

---

## 14. Common Mistakes

| Mistake | Consequence | Fix |
|---------|-------------|-----|
| Only `print(final_state)` | Miss intermediate routing | `stream_mode="updates"` |
| No `trace` in state | Opaque API responses | Append per node (**05**) |
| Router reads wrong key | Silent wrong branch | Stream + test matrix |
| Tracing only in prod | Can't reproduce locally | Enable LangSmith in staging |
| `debug=True` in prod always | Extra overhead | Toggle via settings (**16**) |
| Ignoring `astream_events` for agents | Can't see tool args | v2 event filter |
| Same `thread_id` for all users | Cross-talk in checkpoints | Per-user ids (**08**, **16**) |

## 15. Summary

Observability turns LangGraph from a black box into a **debuggable pipeline**: **`trace`** for durable paths, **`stream_mode`** for live node visibility, **Mermaid** for static topology, **LangSmith** and **`astream_events`** for nested LLM/tool runs, and **config tags** for filtering traces by tenant or tier.

Key takeaways:

- **`compile(debug=True)`** + **`stream_mode="debug"`** diagnose scheduler and routing issues.
- **`stream_mode="updates"`** is the fastest way to see which nodes ran.
- **`trace` in state** belongs in your API response contract.
- **`print_trace`** from **06** is essential for agent tool loops.
- Test routers with a **matrix** before blaming the LLM.
