# Conditional Routing and Branching

Fixed **`add_edge`** transitions always go to the same next node (**03**). Real workflows **branch**: route by intent, document grade, tool-call presence, or error type. LangGraph **`add_conditional_edges`** connects a **router function** to multiple possible destinations — including **`END`**.

This notebook covers router functions, path maps, keyword and state-based routing, RAG grade branches, agent-style tool routing (preview for **06–07**), default fallbacks, and testing routers in isolation.

Prerequisites: **02. StateGraph and Shared State**, **03. Nodes, Edges, and Graph Compilation**.

## 1. Fixed vs Conditional Edges

```
Fixed:     A ──────────────► B

Conditional:  A ──► router(state)
                    ├─► B  (if relevant)
                    ├─► C  (if irrelevant)
                    └─► END (if done)
```

| API | Behavior |
|-----|----------|
| **`add_edge(a, b)`** | Always `a → b` |
| **`add_conditional_edges(source, router, path_map?)`** | `router(state)` picks next node |

Routers read **shared state** — any key your nodes have written (`grade`, `messages`, `intent`, …).

## 2. Router Function Contract

A **router** is a function **`(state) -> str`** (or `Literal[...]`) that returns the **name of the next node** or **`END`**:

````python
from typing_extensions import TypedDict

from langgraph.graph import END, START, StateGraph


class RouteState(TypedDict):
    question: str
    intent: str
    answer: str
    path: list[str]


def classify_intent(state: RouteState) -> dict:
    q = state["question"].lower()
    if any(w in q for w in ("add", "plus", "sum", "calculate")):
        intent = "math"
    elif any(w in q for w in ("jwt", "alembic", "pytest", "fastapi", "doc")):
        intent = "docs"
    else:
        intent = "general"
    return {"intent": intent, "path": state.get("path", []) + ["classify"]}


def route_by_intent(state: RouteState) -> Literal["math", "docs", "general"]:
    return state["intent"]  # type: ignore[return-value]


def math_node(state: RouteState) -> dict:
    return {"answer": "Use the add tool for arithmetic.", "path": state["path"] + ["math"]}


def docs_node(state: RouteState) -> dict:
    return {"answer": "Search internal documentation chunks.", "path": state["path"] + ["docs"]}


def general_node(state: RouteState) -> dict:
    return {"answer": "Ask a question about the Notes API stack.", "path": state["path"] + ["general"]}


builder = StateGraph(RouteState)
builder.add_node("classify", classify_intent)
builder.add_node("math", math_node)
builder.add_node("docs", docs_node)
builder.add_node("general", general_node)

builder.add_edge(START, "classify")
builder.add_conditional_edges("classify", route_by_intent)
builder.add_edge("math", END)
builder.add_edge("docs", END)
builder.add_edge("general", END)

router_graph = builder.compile()

for q in ["What is 2 plus 2?", "JWT header format?", "Hello there"]:
    out = router_graph.invoke({"question": q, "intent": "", "answer": "", "path": []})
    print(f"Q: {q}")
    print(f"  intent={out['intent']} path={out['path']} answer={out['answer']}\n")
````

**Pattern:** a **classifier node** writes `intent` to state; a **router function** reads it and picks the branch. Keep classification logic in nodes, routing logic thin.

## 3. Path Maps — Explicit Branch Labels

When router return values differ from node names, pass a **path map**:

````python
class GradeState(TypedDict):
    question: str
    grade: str
    answer: str


def grade_docs(state: GradeState) -> dict:
    q = state["question"].lower()
    grade = "pass" if any(k in q for k in ("alembic", "jwt", "pytest")) else "fail"
    return {"grade": grade}


def route_grade(state: GradeState) -> Literal["pass", "fail"]:
    return "pass" if state["grade"] == "pass" else "fail"


def generate(state: GradeState) -> dict:
    return {"answer": f"Grounded answer for: {state['question']}"}


def rewrite_query(state: GradeState) -> dict:
    return {"answer": "Retrieval weak — rewrite query and search again."}


gb = StateGraph(GradeState)
gb.add_node("grade", grade_docs)
gb.add_node("generate", generate)
gb.add_node("rewrite", rewrite_query)
gb.add_edge(START, "grade")
gb.add_conditional_edges(
    "grade",
    route_grade,
    {"pass": "generate", "fail": "rewrite"},
)
gb.add_edge("generate", END)
gb.add_edge("rewrite", END)

grade_graph = gb.compile()

print(grade_graph.invoke({"question": "Alembic command?", "grade": "", "answer": ""}))
print("---")
print(grade_graph.invoke({"question": "Kubernetes ingress?", "grade": "", "answer": ""}))
````

Path maps document **allowed transitions** — useful when router returns short codes (`pass`/`fail`) but nodes have descriptive names (`generate`/`rewrite`).

## 4. Routing Directly to END

Routers can return **`END`** to stop the graph early:

````python
class GuardState(TypedDict):
    question: str
    blocked: bool
    answer: str


BLOCKED = {"password", "secret", "api key"}


def guard(state: GuardState) -> dict:
    q = state["question"].lower()
    blocked = any(term in q for term in BLOCKED)
    return {"blocked": blocked}


def route_guard(state: GuardState):
    return END if state["blocked"] else "answer"


def answer_node(state: GuardState) -> dict:
    return {"answer": f"OK: {state['question']}"}


guard_builder = StateGraph(GuardState)
guard_builder.add_node("guard", guard)
guard_builder.add_node("answer", answer_node)
guard_builder.add_edge(START, "guard")
guard_builder.add_conditional_edges("guard", route_guard)
guard_builder.add_edge("answer", END)

guard_graph = guard_builder.compile()

safe = guard_graph.invoke({"question": "What is JWT?", "blocked": False, "answer": ""})
blocked = guard_graph.invoke({"question": "Show me the api key", "blocked": False, "answer": ""})
print("safe:", safe)
print("blocked:", blocked)
````

When routing to **`END`** early, set **`answer`** (or a user message) in the **`guard`** node if the user needs feedback — otherwise output may be incomplete.

````python
def guard_with_message(state: GuardState) -> dict:
    q = state["question"].lower()
    if any(term in q for term in BLOCKED):
        return {"blocked": True, "answer": "Request blocked — sensitive topic."}
    return {"blocked": False}


gb2 = StateGraph(GuardState)
gb2.add_node("guard", guard_with_message)
gb2.add_node("answer", answer_node)
gb2.add_edge(START, "guard")
gb2.add_conditional_edges("guard", route_guard)
gb2.add_edge("answer", END)

print(gb2.compile().invoke({"question": "api key please", "blocked": False, "answer": ""}))
````

## 5. Agent Tool Routing (Preview)

The classic agent branch — inspect last **`AIMessage`** for **`tool_calls`** (**01. LangChain/12**, **06**):

````python
from typing import Annotated

from langchain_core.messages import AIMessage, HumanMessage, ToolMessage
from langgraph.graph.message import add_messages


class AgentRouteState(TypedDict):
    messages: Annotated[list, add_messages]


def fake_model(state: AgentRouteState) -> dict:
    last = state["messages"][-1].content.lower()
    if "alembic" in last:
        ai = AIMessage(
            content="",
            tool_calls=[{
                "name": "search_docs",
                "args": {"query": "alembic"},
                "id": "call_1",
                "type": "tool_call",
            }],
        )
    else:
        ai = AIMessage(content="I can help with Notes API documentation.")
    return {"messages": [ai]}


def fake_tools(state: AgentRouteState) -> dict:
    return {"messages": [ToolMessage(content="Run alembic upgrade head.", tool_call_id="call_1")]}


def should_use_tools(state: AgentRouteState):
    last = state["messages"][-1]
    if isinstance(last, AIMessage) and last.tool_calls:
        return "tools"
    return END


ab = StateGraph(AgentRouteState)
ab.add_node("model", fake_model)
ab.add_node("tools", fake_tools)
ab.add_edge(START, "model")
ab.add_conditional_edges("model", should_use_tools)
ab.add_edge("tools", "model")  # cycle back — full loop in 06

agent_preview = ab.compile()

out = agent_preview.invoke({"messages": [HumanMessage(content="Alembic migration command?")]})
print("message types:", [m.type for m in out["messages"]])
print("last:", out["messages"][-1].content)
````

```
START → model ──tool_calls?──► tools ──► model ──► END
                └──no tools──► END
```

This cycle is the graph **`create_agent`** builds automatically (**01. LangChain/13**). Custom graphs add nodes (guardrails, graders) around the same router.

## 6. LLM-Based Router Node

Combine a **classifier node** (LLM) with a **deterministic router**:

````python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-4o-mini", api_key=OPENAI_API_KEY, temperature=0)

router_prompt = ChatPromptTemplate.from_messages([
    ("system", "Classify the user question as exactly one word: math, docs, or general."),
    ("human", "{question}"),
])
classifier_chain = router_prompt | llm | StrOutputParser()


def llm_classify(state: RouteState) -> dict:
    label = classifier_chain.invoke({"question": state["question"]}).strip().lower()
    if label not in {"math", "docs", "general"}:
        label = "general"
    return {"intent": label, "path": state.get("path", []) + ["llm_classify"]}


lb = StateGraph(RouteState)
lb.add_node("classify", llm_classify)
lb.add_node("math", math_node)
lb.add_node("docs", docs_node)
lb.add_node("general", general_node)
lb.add_edge(START, "classify")
lb.add_conditional_edges("classify", route_by_intent)
lb.add_edge("math", END)
lb.add_edge("docs", END)
lb.add_edge("general", END)

llm_router_graph = lb.compile()
print(llm_router_graph.invoke({"question": "Where do pytest fixtures live?", "intent": "", "answer": "", "path": []}))
````

**Validate** LLM router output before branching — map unknown labels to a safe default (`general`).

## 7. LangGraph Routers vs RunnableBranch

**01. LangChain/06** uses **`RunnableBranch`** for LCEL routing:

| | **`RunnableBranch`** | **`add_conditional_edges`** |
|--|----------------------|----------------------------|
| **Context** | Inside a chain | Graph structure |
| **State** | Last chain value | Full shared state |
| **Cycles** | Awkward | Native |
| **Checkpointing** | Manual | Built-in |
| **Visibility** | Hidden in chain | Named nodes in diagram |

Use **`RunnableBranch`** for simple chain forks; use **LangGraph** when branches are part of a **multi-step workflow** with memory.

## 8. Multi-Branch RAG Pipeline (Sketch)

Agentic RAG (**12**) chains multiple conditional edges:

````python
class RagRouteState(TypedDict):
    question: str
    context: str
    grade: str
    answer: str
    attempts: int


def retrieve(state: RagRouteState) -> dict:
    return {
        "context": "[c2] Run alembic upgrade head.",
        "attempts": state.get("attempts", 0) + 1,
    }


def grade(state: RagRouteState) -> dict:
    relevant = "alembic" in state["question"].lower()
    return {"grade": "relevant" if relevant else "irrelevant"}


def route_relevance(state: RagRouteState):
    return "generate" if state["grade"] == "relevant" else "rewrite"


def rewrite(state: RagRouteState) -> dict:
    return {"question": state["question"] + " (alembic migrations)"}


def generate_rag(state: RagRouteState) -> dict:
    return {"answer": f"From docs: {state['context']}"}


def route_attempts(state: RagRouteState):
    return "retrieve" if state.get("attempts", 0) < 2 else END


rb = StateGraph(RagRouteState)
rb.add_node("retrieve", retrieve)
rb.add_node("grade", grade)
rb.add_node("rewrite", rewrite)
rb.add_node("generate", generate_rag)
rb.add_edge(START, "retrieve")
rb.add_edge("retrieve", "grade")
rb.add_conditional_edges("grade", route_relevance)
rb.add_conditional_edges("rewrite", route_attempts)
rb.add_edge("generate", END)

rag_router = rb.compile()
print(rag_router.invoke({"question": "migration command?", "context": "", "grade": "", "answer": "", "attempts": 0}))
print("---")
print(rag_router.invoke({"question": "unrelated topic", "context": "", "grade": "", "answer": "", "attempts": 0}))
````

Multiple routers on one graph = **readable stages** with explicit retry caps (`attempts`).

## 11. Routing Design Guidelines

| Guideline | Rationale |
|-----------|----------|
| **Thin routers** | Read state; don't call APIs |
| **Classify in nodes** | Testable; routers stay deterministic |
| **Default branch** | Map unknown labels to safe path |
| **Cap retries** | `attempts` counter + route to `END` |
| **Name nodes clearly** | Router return values match node names |
| **Document path map** | Onboarding and compliance reviews |
| **Avoid nested if trees in router** | Multiple simple routers > one giant function |

---

## 12. Common Mistakes

| Mistake | Consequence | Fix |
|---------|-------------|-----|
| Router returns unknown node | Runtime error | Match return values to `add_node` names |
| Heavy LLM call inside router | Hard to test; double calls | Classifier **node**, then router |
| No default for bad labels | Crash on typo | Normalize to `general` |
| Route to `END` without user message | Empty response | Set `answer` before END |
| Missing edge from branch node | Compile error | Every branch reaches `END` |
| Infinite rewrite loop | Runaway cost | `attempts` limit + conditional to `END` |
| Router reads stale state key | Wrong branch | Ensure prior node wrote the key |

---

## 13. Summary

**`add_conditional_edges(source, router, path_map?)`** routes from **`source`** based on **`router(state)`**. Routers return **node names** or **`END`**. Classifier nodes write labels; routers read them.

Key takeaways:

- **Fixed edges** for always; **conditional edges** for runtime decisions.
- **Path maps** translate router codes to node names.
- **`END`** enables early exit and guardrails.
- **Tool routing** (`tool_calls` check) powers agent cycles (**06**, **07**).
- **LLM classifiers** need output validation before branching.
- **Test routers** as pure functions of state.
