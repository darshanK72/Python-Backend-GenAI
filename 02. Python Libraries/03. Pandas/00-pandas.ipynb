## 02 — DataFrame


```python
df = pd.DataFrame({
    "name": ["Asha", "Ravi", "Meera", "Kiran"],
    "age": [22, 23, 21, 24],
    "city": ["Nashik", "Pune", "Mumbai", "Nashik"],
    "score": [88, 76, 92, 85],
})
df
```

## 05 — groupby


```python
sales = pd.DataFrame({"city": ["A", "A", "B", "B"], "amount": [100, 150, 80, 120]})
sales.groupby("city")["amount"].sum()
```

## 06 — merge


```python
left = pd.DataFrame({"id": [1, 2], "name": ["Asha", "Ravi"]})
right = pd.DataFrame({"id": [1, 2], "math": [90, 76]})
pd.merge(left, right, on="id")
```
