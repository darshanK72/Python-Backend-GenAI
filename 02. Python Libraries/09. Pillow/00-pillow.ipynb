# Pillow — images inline

Companion: `01_image_basics.py`, `02_image_operations.py`.


```python
from PIL import Image, ImageDraw
from IPython.display import display

img = Image.new('RGB', (200, 120), (70, 130, 180))
draw = ImageDraw.Draw(img)
draw.text((50, 50), 'Hello', fill='white')
display(img)
```

```python
thumb = img.copy()
thumb.thumbnail((80, 80))
display(thumb)
```
