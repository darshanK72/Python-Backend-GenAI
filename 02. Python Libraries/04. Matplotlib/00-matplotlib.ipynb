# Matplotlib

`%matplotlib inline` shows plots under each cell.

`.py` scripts in this folder save PNG when run from terminal.


```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

## 01 — Line plot


```python
plt.figure(figsize=(6, 4))
plt.plot([1, 2, 3, 4, 5], [2, 4, 3, 7, 6], marker="o")
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Units")
plt.grid(True, alpha=0.3)
plt.show()
```

## 02 — Bar chart


```python
plt.figure(figsize=(6, 4))
plt.bar(["Python", "SQL", "ML"], [90, 75, 60])
plt.ylabel("Score")
plt.show()
```

## 03 — Scatter


```python
plt.scatter([1, 2, 3, 4, 5, 6], [55, 62, 68, 74, 80, 88], s=80)
plt.xlabel("Study hours")
plt.ylabel("Exam score")
plt.show()
```

## 04 — Subplots


```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
x = np.linspace(0, 6.28, 100)
axes[0].plot(x, np.sin(x))
axes[1].plot(x, np.cos(x), color="orange")
plt.tight_layout()
plt.show()
```

## 05 — Histogram


```python
data = np.random.default_rng(0).normal(170, 8, 200)
plt.hist(data, bins=15, edgecolor="white")
plt.title("Height distribution")
plt.show()
```
