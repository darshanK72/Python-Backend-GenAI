# Seaborn

Statistical plots on top of Matplotlib. Companion: `01_seaborn_intro.py` …


```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
sns.set_theme(style="whitegrid")
```

## 01 — Scatter with hue


```python
df = pd.DataFrame({
    "hours": [1, 2, 3, 4, 5, 2, 3, 4, 5, 6],
    "score": [55, 60, 65, 70, 80, 58, 68, 72, 85, 90],
    "course": ["A"]*5 + ["B"]*5,
})
sns.scatterplot(data=df, x="hours", y="score", hue="course")
plt.show()
```

## 02 — Histogram + KDE


```python
vals = np.random.default_rng(0).normal(100, 15, 80)
sns.histplot(vals, kde=True)
plt.show()
```

## 03 — Box plot


```python
tips = sns.load_dataset("tips")
sns.boxplot(data=tips, x="day", y="total_bill")
plt.show()
```

## 04 — Correlation heatmap


```python
nums = pd.DataFrame({"math": [90, 76, 88], "science": [85, 70, 92], "english": [88, 82, 80]})
sns.heatmap(nums.corr(), annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.show()
```
