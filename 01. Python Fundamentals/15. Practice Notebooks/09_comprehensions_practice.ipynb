````python
# Run this cell once per notebook — provides test validation helpers.

def check_equal(name, actual, expected):
    """Compare actual vs expected; print pass/fail."""
    if actual == expected:
        print(f"✓ {name}")
        return True
    print(f"✗ {name}")
    print(f"  Expected: {expected!r}")
    print(f"  Got:      {actual!r}")
    return False


def run_tests(cases):
    """Run a list of (name, callable) tuples. Callable should return the answer."""
    passed = 0
    for name, fn in cases:
        try:
            if check_equal(name, fn(), True):
                passed += 1
        except Exception as exc:
            print(f"✗ {name} — Error: {type(exc).__name__}: {exc}")
    total = len(cases)
    print(f"\nResult: {passed}/{total} passed")
    if passed == total:
        print("🎉 All tests passed!")
    return passed == total
````

## Problem 1: Squares list comprehension `Easy`

Build `squares` = squares of 1..5 using a list comprehension.

**Examples**

```
[1, 4, 9, 16, 25]
```


````python
# YOUR CODE
squares = [s**2 for s in range(1,6)]
````

````python
check_equal('squares', squares, [1, 4, 9, 16, 25])
````

## Problem 2: Evens only `Easy`

List comprehension: even numbers from 1..10 into `evens`.

**Examples**

```
[2, 4, 6, 8, 10]
```


````python
# YOUR CODE
evens = [i for i in range(1,11) if i % 2 == 0]
````

````python
check_equal('evens', evens, [2, 4, 6, 8, 10])
````

## Problem 3: Uppercase names `Easy`

Uppercase each name in `names` using a list comprehension.

**Examples**

```
["ada", "guido"] → ["ADA", "GUIDO"]
```


````python
names = ["ada", "guido"]

# YOUR CODE
upper = [word.upper() for word in names]
````

````python
check_equal("upper", upper, ["ADA", "GUIDO"])
````

## Problem 4: Dict from pairs `Medium`

Build `scores` dict from parallel lists `students` and `marks` using dict comprehension or zip.

**Examples**

```
students = ["A", "B"], marks = [90, 85]
```


````python
students = ["A", "B"]
marks = [90, 85]

# YOUR CODE
scores = {s:m for s,m in zip(students,marks)}
````

````python
check_equal("scores", scores, {"A": 90, "B": 85})
````

## Problem 5: Set of lengths `Medium`

Set comprehension: unique word lengths from `words`.

**Examples**

```
["hi", "hey", "yo"] → {2, 3}
```


````python
words = ["hi", "hey", "yo"]

# YOUR CODE
lengths = {len(word) for word in words}
````

````python
check_equal('lengths', lengths, {2, 3})
````

## Problem 6: Flatten matrix `Hard`

Use nested list comprehension to flatten `matrix` into `flat`.

**Examples**

```
[[1, 2], [3]] → [1, 2, 3]
```


````python
matrix = [[1, 2], [3]]

# YOUR CODE
flat = [i for sub_list in matrix for i in sub_list]
````

````python
check_equal('flat', flat, [1, 2, 3])
````
