````python
# Run this cell once per notebook — provides test validation helpers.

def check_equal(name, actual, expected):
    """Compare actual vs expected; print pass/fail."""
    if actual == expected:
        print(f"✓ {name}")
        return True
    print(f"✗ {name}")
    print(f"  Expected: {expected!r}")
    print(f"  Got:      {actual!r}")
    return False


def run_tests(cases):
    """Run a list of (name, callable) tuples. Callable should return the answer."""
    passed = 0
    for name, fn in cases:
        try:
            if check_equal(name, fn(), True):
                passed += 1
        except Exception as exc:
            print(f"✗ {name} — Error: {type(exc).__name__}: {exc}")
    total = len(cases)
    print(f"\nResult: {passed}/{total} passed")
    if passed == total:
        print("🎉 All tests passed!")
    return passed == total
````

## Problem 1: Simple Dog class `Easy`

Define class `Dog` with `__init__(self, name)` setting `self.name`. Method `speak()` returns `f"{self.name} says woof!"`.

**Examples**

```
Dog("Rex").speak() → "Rex says woof!"
```


````python
# YOUR CODE
class Dog:
    def __init__(self,name):
        self.name = name
    
    def speak(self):
        return f"{self.name} says woof!"
````

## Problem 2: Rectangle attributes `Easy`

Class `Rectangle` with `width`, `height` in `__init__`. Method `area()` returns width × height.

**Examples**

```
Rectangle(3, 4).area() → 12
```


````python
# YOUR CODE
class Rectangle:
    def __init__(self,width,height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height
````

````python
check_equal('area', Rectangle(3, 4).area(), 12)
````

## Problem 3: BankAccount deposit `Medium`

Class `BankAccount` starts with `balance=0`. Method `deposit(amount)` adds to balance and returns new balance.

**Examples**

```
acc = BankAccount(); acc.deposit(100) → 100
```


````python
# YOUR CODE
class BankAccount:
    def __init__(self):
        self.balance = 0

    def deposit(self,amount):
        self.balance += amount
        return self.balance
````

````python
acc = BankAccount()
check_equal('deposit', acc.deposit(100), 100)
check_equal('balance', acc.balance, 100)
````

## Problem 4: Person full_name `Easy`

Class `Person` with `first` and `last`. Method `full_name()` returns `"first last"`.

**Examples**

```
Person("Ada", "Lovelace").full_name() → "Ada Lovelace"
```


````python
# YOUR CODE
class Person:
    def __init__(self,first_name,last_name):
        self.first_name = first_name
        self.last_name = last_name 

    def full_name(self):
        return self.first_name + " " + self.last_name
````

````python
check_equal("full", Person("Ada", "Lovelace").full_name(), "Ada Lovelace")
````

## Problem 5: Counter increment `Medium`

Class `Counter` with `count=0`. Methods `increment()` (+1) and `reset()` (back to 0).

**Examples**

```
c = Counter(); c.increment(); c.increment(); c.count → 2
```


````python
# YOUR CODE
class Counter:
    def __init__(self):
        self.count = 0;

    def increment(self):
        self.count += 1

    def reset(self):
        self.count = 0;
````

````python
c = Counter()
c.increment()
c.increment()
check_equal('count', c.count, 2)
c.reset()
check_equal('after reset', c.count, 0)
````

## Problem 6: Inheritance — Animal/Cat `Medium`

`Animal` has `name`. `Cat` inherits `Animal`, `sound()` returns `f"{self.name} says meow"`.

**Examples**

```
Cat("Luna").sound() → "Luna says meow"
```


````python
# YOUR CODE
class Animal:
    def __init__(self,name):
        self.name = name

class Cat(Animal):
    def __init__(self,name):
        super().__init__(name)

    def sound(self):
        return f"{self.name} says meow"
````

````python
check_equal("sound", Cat("Luna").sound(), "Luna says meow")
````
