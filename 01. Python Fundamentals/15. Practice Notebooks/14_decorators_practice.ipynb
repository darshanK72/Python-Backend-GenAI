````python
# Run this cell once per notebook — provides test validation helpers.

def check_equal(name, actual, expected):
    """Compare actual vs expected; print pass/fail."""
    if actual == expected:
        print(f"✓ {name}")
        return True
    print(f"✗ {name}")
    print(f"  Expected: {expected!r}")
    print(f"  Got:      {actual!r}")
    return False


def run_tests(cases):
    """Run a list of (name, callable) tuples. Callable should return the answer."""
    passed = 0
    for name, fn in cases:
        try:
            if check_equal(name, fn(), True):
                passed += 1
        except Exception as exc:
            print(f"✗ {name} — Error: {type(exc).__name__}: {exc}")
    total = len(cases)
    print(f"\nResult: {passed}/{total} passed")
    if passed == total:
        print("🎉 All tests passed!")
    return passed == total
````

## Problem 1: shout `Easy`

Define a decorator `shout` that uppercases the string returned by the wrapped function.

**Examples**

```
@shout
def greet(): return 'hi'
greet() → 'HI'
```


````python
# YOUR CODE
def shout(func):
    def wrapper(*args,**kwargs):
        res = func(*args,**kwargs)
        return res.upper()
    return wrapper
````

````python
@shout
def greet():
    return 'hi'
check_equal('shout', greet(), 'HI')
````

## Problem 2: double_result `Easy`

Define a decorator `double_result` that multiplies the numeric return value by 2.

**Examples**

```
@double_result
def add(a, b): return a + b
add(2, 3) → 10
```


````python
# YOUR CODE
def double_result(func):
    def wrapper(*args,**kwargs):
        res = func(*args,**kwargs)
        res *= 2
        return res
    return wrapper
````

````python
@double_result
def add(a, b):
    return a + b
check_equal('double', add(2, 3), 10)
````

## Problem 3: count_calls `Medium`

Define a decorator `count_calls` that adds a `.calls` attribute to the wrapper, counting how many times it was called.

**Examples**

```
after 3 calls → fn.calls == 3
```


````python
# YOUR CODE
def count_calls(func):
    calls = 0;
    def wrapper(*args,**kwargs):
        wrapper.calls += 1
        return func(*args,**kwargs)
    wrapper.calls = 0 
    return wrapper
        
         
````

````python
@count_calls
def ping():
    return 'pong'
ping(); ping(); ping()
check_equal('calls', ping.calls, 3)
````

## Problem 4: repeat `Medium`

Define a parameterized decorator `repeat(times)` that calls the function `times` times and returns a list of the results.

**Examples**

```
@repeat(3)
def f(): return 7
f() → [7, 7, 7]
```


````python
# YOUR CODE
def repeat(times):
    def decorator(func):
        def wrapper(*args,**kwargs):
            res = []
            for _ in range(times):
                res.append(func(*args,**kwargs))
            return res
        return wrapper
    return decorator
````

````python
@repeat(3)
def seven():
    return 7
check_equal('repeat', seven(), [7, 7, 7])
````

## Problem 5: memoize `Medium`

Define a decorator `memoize` that caches results in a dict keyed by the positional args.

**Examples**

```
repeated calls with same arg return cached value
```


````python
# YOUR CODE
def memoize(func):
    def wrapper(n):
        cache = wrapper.dict.get(n,None)
        if cache != None:
            return cache
        else:
            res = func(n)
            wrapper.dict[n] = res
            return res
    wrapper.dict = {}
    return wrapper
        
````

````python
_log = []
@memoize
def square(n):
    _log.append(n)
    return n * n
square(4); square(4); square(5)
check_equal('value', square(4), 16)
check_equal('computed_once_per_arg', _log, [4, 5])
````

## Problem 6: ensure_positive `Medium`

Define a decorator `ensure_positive` that raises `ValueError('must be positive')` if the single argument is `<= 0`, else runs the function.

**Examples**

```
f(-1) → ValueError, f(4) → normal result
```


````python
# YOUR CODE
def ensure_positive(func):
    def wrapper(n):
        if n < 0:
            raise ValueError('must be positive')
        return func(n)
    return wrapper
````

````python
@ensure_positive
def root(n):
    return n ** 0.5
check_equal('ok', root(16), 4.0)
try:
    root(-1)
    print('✗ should raise')
except ValueError as e:
    check_equal('message', str(e), 'must be positive')
````
