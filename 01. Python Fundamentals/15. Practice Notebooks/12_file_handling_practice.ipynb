````python
# Run this cell once per notebook — provides test validation helpers.

def check_equal(name, actual, expected):
    """Compare actual vs expected; print pass/fail."""
    if actual == expected:
        print(f"✓ {name}")
        return True
    print(f"✗ {name}")
    print(f"  Expected: {expected!r}")
    print(f"  Got:      {actual!r}")
    return False


def run_tests(cases):
    """Run a list of (name, callable) tuples. Callable should return the answer."""
    passed = 0
    for name, fn in cases:
        try:
            if check_equal(name, fn(), True):
                passed += 1
        except Exception as exc:
            print(f"✗ {name} — Error: {type(exc).__name__}: {exc}")
    total = len(cases)
    print(f"\nResult: {passed}/{total} passed")
    if passed == total:
        print("🎉 All tests passed!")
    return passed == total
````

## Problem 1: write_text `Easy`

Define `write_text(path, text)` — write `text` to the file at `path` (UTF-8). Return `None`.

**Examples**

```
write_text('a.txt', 'hi') then reading a.txt → 'hi'
```


````python
# YOUR CODE
def write_text(path, text):
    with open(path,'w') as f:
        f.write(text)
````

````python
import tempfile, os
_d = tempfile.mkdtemp()
_p = os.path.join(_d, 'a.txt')
write_text(_p, 'hello')
with open(_p, encoding='utf-8') as _f:
    check_equal('written', _f.read(), 'hello')
````

## Problem 2: read_text `Easy`

Define `read_text(path)` — return the entire contents of the file at `path` as a string.

**Examples**

```
file contains 'data' → read_text(path) → 'data'
```


````python
# YOUR CODE
def read_text(path):
    with open(path,'r') as f:
        content = f.read()
    return content
````

````python
import tempfile, os
_d = tempfile.mkdtemp()
_p = os.path.join(_d, 'b.txt')
with open(_p, 'w', encoding='utf-8') as _f:
    _f.write('data')
check_equal('read', read_text(_p), 'data')
````

## Problem 3: count_lines `Easy`

Define `count_lines(path)` — return the number of lines in the file at `path`.

**Examples**

```
file with 3 newline-separated lines → 3
```


````python
# YOUR CODE
def count_lines(path):
    with open(path,'r') as f:
        l = f.readlines()
    return len(l)


````

````python
import tempfile, os
_d = tempfile.mkdtemp()
_p = os.path.join(_d, 'c.txt')
with open(_p, 'w', encoding='utf-8') as _f:
    _f.write('one\ntwo\nthree\n')
check_equal('lines', count_lines(_p), 3)
````

## Problem 4: append_line `Medium`

Define `append_line(path, line)` — append `line` plus a newline to the end of the file (create it if missing).

**Examples**

```
start 'a\n', append_line(path, 'b') → file now 'a\nb\n'
```


````python
# YOUR CODE
def append_line(path, line):
    with open(path,'a',encoding='utf-8') as f:
        f.write(line + '\n')
````

````python
import tempfile, os
_d = tempfile.mkdtemp()
_p = os.path.join(_d, 'd.txt')
with open(_p, 'w', encoding='utf-8') as _f:
    _f.write('a\n')
append_line(_p, 'b')
with open(_p, encoding='utf-8') as _f:
    check_equal('appended', _f.read(), 'a\nb\n')
````

## Problem 5: json round-trip `Medium`

Define `save_json(path, data)` and `load_json(path)` — write `data` as JSON, then read it back as a Python object.

**Examples**

```
save_json(path, {'x': 1}); load_json(path) → {'x': 1}
```


````python
# YOUR CODE
import json

def save_json(path, data):
    with open(path,'w',encoding='utf-8') as f:
        json.dump(data,f)

def load_json(path):
    with open(path,'r',encoding='utf-8') as f:
        return json.load(f)
````

````python
import tempfile, os
_d = tempfile.mkdtemp()
_p = os.path.join(_d, 'e.json')
save_json(_p, {'x': 1, 'items': [1, 2, 3]})
check_equal('roundtrip', load_json(_p), {'x': 1, 'items': [1, 2, 3]})
````

## Problem 6: word_count `Medium`

Define `word_count(path)` — return a dict mapping each whitespace-separated word to how many times it appears.

**Examples**

```
file 'a b a' → {'a': 2, 'b': 1}
```


````python
# YOUR CODE
def word_count(path):
    with open(path,'r',encoding='utf-8') as f:
        chars = f.readline().split(' ')
        out = {}
        for c in chars:
            if out.get(c,0) == 0:
                out[c] = 0
            out[c] += 1
    return out

````

````python
import tempfile, os
_d = tempfile.mkdtemp()
_p = os.path.join(_d, 'f.txt')
with open(_p, 'w', encoding='utf-8') as _f:
    _f.write('a b a c b a')
check_equal('counts', word_count(_p), {'a': 3, 'b': 2, 'c': 1})
````
