````python
# Run this cell once per notebook — provides test validation helpers.

def check_equal(name, actual, expected):
    """Compare actual vs expected; print pass/fail."""
    if actual == expected:
        print(f"✓ {name}")
        return True
    print(f"✗ {name}")
    print(f"  Expected: {expected!r}")
    print(f"  Got:      {actual!r}")
    return False


def run_tests(cases):
    """Run a list of (name, callable) tuples. Callable should return the answer."""
    passed = 0
    for name, fn in cases:
        try:
            if check_equal(name, fn(), True):
                passed += 1
        except Exception as exc:
            print(f"✗ {name} — Error: {type(exc).__name__}: {exc}")
    total = len(cases)
    print(f"\nResult: {passed}/{total} passed")
    if passed == total:
        print("🎉 All tests passed!")
    return passed == total
````

## Problem 5: While countdown list `Medium`

Use a `while` loop to build `countdown` as `[3, 2, 1]` starting from `start = 3`.

**Examples**

```
countdown → [3, 2, 1]
```


````python
start = 3

# YOUR CODE
countdown = []
while(start >= 1):
    countdown.append(start)
    start -= 1
````

## Problem 10: Nested loop — pairs `Hard`

Build `pairs`: all `(i, j)` where `i` in 1..2 and `j` in 1..3 as tuples in a list.

**Examples**

```
pairs → [(1,1), (1,2), (1,3), (2,1), (2,2), (2,3)]
```


````python
# YOUR CODE
pairs = []
for i in range(1,3):
    for j in range(1,4):
        pairs.append((i,j))
````

````python
check_equal('pairs', pairs, [(1, 1), (1, 2), (1, 3), (2, 1), (2, 2), (2, 3)])
````
