````python
# Run this cell once per notebook — provides test validation helpers.

def check_equal(name, actual, expected):
    """Compare actual vs expected; print pass/fail."""
    if actual == expected:
        print(f"✓ {name}")
        return True
    print(f"✗ {name}")
    print(f"  Expected: {expected!r}")
    print(f"  Got:      {actual!r}")
    return False


def run_tests(cases):
    """Run a list of (name, callable) tuples. Callable should return the answer."""
    passed = 0
    for name, fn in cases:
        try:
            if check_equal(name, fn(), True):
                passed += 1
        except Exception as exc:
            print(f"✗ {name} — Error: {type(exc).__name__}: {exc}")
    total = len(cases)
    print(f"\nResult: {passed}/{total} passed")
    if passed == total:
        print("🎉 All tests passed!")
    return passed == total
````

## Problem 1: List sum `Easy`

Sum all numbers in `nums` using a loop (not `sum()`).

**Examples**

```
nums = [4, 5, 6] → total = 15
```


````python
nums = [4, 5, 6]

# YOUR CODE
total = 0
for ele in nums:
    total += ele
````

````python
check_equal('total', total, 15)
````

## Problem 5: Second largest `Medium`

Return the second largest value in `nums` (assume at least 2 distinct values).

**Examples**

```
nums = [10, 5, 8, 20, 15] → second = 15
```


````python
nums = [10, 5, 8, 20, 15]

# YOUR CODE
first = nums[0]
second = -1

for ele in nums:
    if ele > first:
        second = first
        first = ele
    elif ele > second:
        second = ele
````

````python
check_equal('second', second, 15)
````

## Problem 7: Count occurrences `Easy`

Count how many times `target` appears in `items`.

**Examples**



````python
items = ["a", "b", "a", "c"]
target = "a"

# YOUR CODE
count = 0
for c in items:
    if c == target:
        count += 1
````

## Problem 8: Rotate right by 1 `Hard`

Rotate `nums` right by 1: last element moves to front. New list in `rotated`.

**Examples**

```
[1, 2, 3, 4] → [4, 1, 2, 3]
```


````python
nums = [1, 2, 3, 4]

# YOUR CODE
rotated = []
rotated.append(nums[-1])
for i in range(0,len(nums)-1):
    rotated.append(nums[i])
````

````python
check_equal('rotated', rotated, [4, 1, 2, 3])
````

## Problem 9: Filter positives `Medium`

Build `positives` with only values > 0 from `mixed`.

**Examples**

```
[-1, 2, 0, 5, -3] → [2, 5]
```


````python
mixed = [-1, 2, 0, 5, -3]

# YOUR CODE
positives = []
for ele in mixed:
    if(ele > 0):
        positives.append(ele)
````
